```python
from heapq import heappush, heappop
import io, os, sys
input = io.BytesIO(os.read(0, os.fstat(0).st_size)).readline

inf = 1000000000000000000000000000000

n, k = map(int, input().split())

li = [-inf] + [int(input()) for _ in range(n)] + [inf]
n += 2

heap = []
dists = []
for i in range(n - 1):
    heappush(heap, (li[i + 1] - li[i], i, i + 1))
    dists.append(li[i + 1] - li[i])
    
prefix_even = [0] * (n + 1)
prefix_odd = [0] * (n + 1)

cumul = 0
for i in range(0, n - 1, 2):
    cumul += dists[i]
    prefix_even[i + 1] = cumul
    
cumul = 0
for i in range(1, n - 1, 2):
    cumul += dists[i]
    prefix_odd[i + 1] = cumul
    
to_rights = list(range(n))
to_lefts = list(range(n))

c = 0
total = 0
vt = [False] * n
while c < k:
    dist, s, e = heappop(heap)
    
    if vt[s] or vt[e]:
        continue
    vt[s] = vt[e] = True
        
    c += 1
    total += dist
    
    nl = s
    if to_lefts[s - 1] != s - 1:
        nl = to_lefts[s - 1]
    
    nr = e
    if to_rights[e + 1] != e + 1:
        nr = to_rights[e + 1]
        
    to_rights[nl] = nr
    to_lefts[nr] = nl
    
    if nl % 2 == 1:
        cur = prefix_odd[nr] - prefix_odd[nl - 1]
        extended = prefix_even[nr + 1] - prefix_even[nl - 2]
    else:
        cur = prefix_even[nr] - prefix_even[nl - 1]
        extended = prefix_odd[nr + 1] - prefix_odd[nl - 2]
    
    heappush(heap, (extended - cur, nl - 1, nr + 1))
    
print(total)
```

```python
import io, os, sys
# input = io.BytesIO(os.read(0, os.fstat(0).st_size)).readline

inf = float('inf')

n = int(input())
li = [list(map(int, input().split())) for _ in range(n)]

to_original = []
for s, e in li:
    e -= 1
    
    to_original.append(s)
    to_original.append(e)
    to_original.append(e + 1)
    
to_original = sorted(set(to_original))
to_reduce = dict()

to_ori_len = len(to_original)
dists = [0] * to_ori_len
for i in range(to_ori_len):
    to_reduce[str(to_original[i])] = i
    
    if i < to_ori_len - 1:
        dists[i] = to_original[i + 1] - to_original[i]
    
deltas = [0] * (to_ori_len + 1)
for s, e in li:
    e -= 1
    
    s = to_reduce[str(s)]
    e = to_reduce[str(e)]
    
    deltas[s] += 1
    deltas[e + 1] -= 1
    
prefix_ones = [0] * (to_ori_len + 1)
cur = 0
one_cumul = 0
all_cumul = 0
for i in range(to_ori_len):
    cur += deltas[i]
    
    if cur == 1:
        one_cumul += dists[i]
    if cur >= 1:
        all_cumul += dists[i] 
        
    prefix_ones[i + 1] = one_cumul
    
min_lost = inf
for s, e in li:
    e -= 1
    
    s = to_reduce[str(s)]
    e = to_reduce[str(e)]
    
    lost = prefix_ones[e + 1] - prefix_ones[s]
    min_lost = min(min_lost, lost)
    
result = all_cumul - min_lost

print(result)
```

```python
from collections import defaultdict as dd
import random
import math

# https://gist.github.com/thomdixon/dd1e280681f16535fbf1
# https://www.geeksforgeeks.org/primality-test-set-3-miller-rabin/

# Python 3 program to find a prime factor of composite using
# Pollard's Rho algorithm
# Python3 program Miller-Rabin primality test
 
# Utility function to do
# modular exponentiation.
# It returns (x^y) % p
def power(x, y, p):
     
    # Initialize result
    res = 1;
     
    # Update x if it is more than or
    # equal to p
    x = x % p;
    while (y > 0):
         
        # If y is odd, multiply
        # x with result
        if (y & 1):
            res = (res * x) % p;
 
        # y must be even now
        y = y>>1; # y = y/2
        x = (x * x) % p;
     
    return res;
 
# This function is called
# for all k trials. It returns
# false if n is composite and
# returns false if n is
# probably prime. d is an odd
# number such that d*2<sup>r</sup> = n-1
# for some r >= 1
def miillerTest(d, n):
     
    # Pick a random number in [2..n-2]
    # Corner cases make sure that n > 4
    a = 2 + random.randint(1, n - 4);
 
    # Compute a^d % n
    x = power(a, d, n);
 
    if (x == 1 or x == n - 1):
        return True;
 
    # Keep squaring x while one
    # of the following doesn't
    # happen
    # (i) d does not reach n-1
    # (ii) (x^2) % n is not 1
    # (iii) (x^2) % n is not n-1
    while (d != n - 1):
        x = (x * x) % n;
        d *= 2;
 
        if (x == 1):
            return False;
        if (x == n - 1):
            return True;
 
    # Return composite
    return False;
 
# It returns false if n is
# composite and returns true if n
# is probably prime. k is an
# input parameter that determines
# accuracy level. Higher value of
# k indicates more accuracy.
def isPrime(n):
     
    # Corner cases
    if (n <= 1 or n == 4):
        return False;
    if (n <= 3):
        return True;
 
    # Find r such that n =
    # 2^d * r + 1 for some r >= 1
    d = n - 1;
    while (d % 2 == 0):
        d //= 2;
 
    # Iterate given number of 'k' times
    for i in range(k):
        if (miillerTest(d, n) == False):
            return False;
 
    return True;
 
# Function to calculate (base^exponent)%modulus
def modular_pow(base, exponent,modulus):
 
    # initialize result
    result = 1
 
    while (exponent > 0):
     
        # if y is odd, multiply base with result
        if (exponent & 1):
            result = (result * base) % modulus
 
        # exponent = exponent/2
        exponent = exponent >> 1
 
        # base = base * base
        base = (base * base) % modulus
     
    return result
 
def pollard_rho(n):
    if n%2 == 0: return 2
    if is_prime(n): return n # define it somehow, e.g. return False, then it infinite loops on primes
    while True:
        c = random.randint(2, n-1)
        f = lambda x: x**2 + c 
        x = y = 2 
        d = 1 
        while d == 1:
            x = f(x) % n 
            y = f(f(y)) % n 
            d = gcd((x - y) % n, n)
        if d != n: return d

k = 3

n = int(input())
li = list(map(int, input().split()))

cts = dd(set)
ps = [2, 3, 5, 7, 11]
for i in range(n):
    num = li[i]
    
    for p in ps:
        while num % p == 0:
            num //= p
            cts[p].add(i)
    
    while num >= 2 and not isPrime(num):
        div = pollard_rho(num)
        cts[div].add(i)
        num //= div
        
    if num != 1:
        cts[num].add(i)
        
for key in cts:
    if len(cts[key]) >= (n + 1) // 2:
        print('YES')
        print(*[li[index] for index in cts[key]][:(n + 1) // 2])
        break
else:
    print('NO')
```

```python
from collections import defaultdict as dd
from math import log2, ceil
import io, os, sys
input = io.BytesIO(os.read(0, os.fstat(0).st_size)).readline

def find(v):
    while v != pa[v]:
        v = pa[v]
    return v

def update(node, l, r):
    if ul <= l and r <= ur:
        seg[node].append((a, b))
    elif ur < l or r < ul:
        return
    else:
        mid = (l + r) // 2
        
        update(node * 2, l, mid)
        update(node * 2 + 1, mid + 1, r)
        
def query(node, l, r):
    unioned_c = 0
    for a, b in seg[node]:
        ra = find(a)
        rb = find(b)
     
        if ra != rb:
            if rank[ra] < rank[rb]:
                ra, rb = rb, ra

            pa[rb] = ra
            is_rank_added = False
            if rank[ra] == rank[rb]:
                rank[ra] += 1
                is_rank_added = True
                
            unioned_c += 1
            unioned_info.append((ra, rb, is_rank_added))
    if l == r:
        a, b = check_queries[l]
        ra = find(a)
        rb = find(b)
        
        if ra == rb:
            result[l] = 1
    else:
        mid = (l + r) // 2
        
        query(node * 2, l, mid)
        query(node * 2 + 1, mid + 1, r)
        
    for _ in range(unioned_c):
        ra, rb, is_rank_added = unioned_info.pop()
       
        pa[rb] = rb
        if is_rank_added:
            rank[ra] -= 1

n, m = map(int, input().split())

pendings = dd(int)
cur_check_used = -1
check_queries = []
edge_history = []
for qi in range(m):
    tp, a, b = map(int, input().split())
    
    if a > b:
        a, b = b, a
        
    if tp == 3:
        cur_check_used += 1
        check_queries.append((a, b))
    else:
        if (a, b) not in pendings:
            pendings[(a, b)] = cur_check_used
        else:
            f_check_used = pendings[(a, b)]
    
            if f_check_used < cur_check_used:
                edge_history.append((a, b, f_check_used + 1, cur_check_used))
                
            del pendings[(a, b)]
            
for key in pendings:
    f_check_used = pendings[key]
    edge_history.append((key[0], key[1], f_check_used + 1, cur_check_used))
    
size = 2 ** (ceil(log2(cur_check_used + 2)) + 1) + 1
seg = [[] for _ in range(size)]

for a, b, ul, ur in edge_history:
    update(1, 0, cur_check_used)
    
result = [0] * (cur_check_used + 1)
rank = [0] * (n + 1)
pa = list(range(n + 1))
unioned_info = []
query(1, 0, cur_check_used)

print('\n'.join(map(str, result)))
```

```python
# 0, 1, 2, 3
# u  d  l  r

from collections import deque
import sys
input = lambda: sys.stdin.readline().rstrip()

## PYRIVAL BOOTSTRAP
# https://github.com/cheran-senthil/PyRival/blob/master/pyrival/misc/bootstrap.py
# This decorator allows for recursion without actually doing recursion
## @bootstrap, yield when getting and returning value in recursive functions, end of functions
from types import GeneratorType

def bootstrap(f, stack=[]):
    def wrappedfunc(*args, **kwargs):
        if stack:
            return f(*args, **kwargs)
        else:
            to = f(*args, **kwargs)
            while True:
                if type(to) is GeneratorType:
                    stack.append(to)
                    to = next(to)
                else:
                    stack.pop()
                    if not stack:
                        break
                    to = stack[-1].send(to)
            return to
    return wrappedfunc

def get_2sat_num(v):
    return v * 2

def get_near_2sat_nums(two_sat_v):
    v = two_sat_v // 2
    
    start = (v // 4) * 4

    if v % 4 == 0 or v % 4 == 1:
        return get_2sat_num(start + 2), get_2sat_num(start + 3)
    else:
        return get_2sat_num(start), get_2sat_num(start + 1)
    
def get_oppo_2sat_num(two_sat_v):
    v = two_sat_v // 2
    
    start = (v // 4) * 4

    if v % 4 == 0:
        return get_2sat_num(start + 1)
    elif v % 4 == 1:
        return get_2sat_num(start)
    elif v % 4 == 2:
        return get_2sat_num(start + 3)
    elif v % 4 == 3:
        return get_2sat_num(start + 2)
    
def except_2sat_num(ori):
    n1, n2 = get_near_2sat_nums(ori)
    oppo = get_oppo_2sat_num(ori)

    li[n1].append(invert(ori))
    li[n2].append(invert(ori))
    li[oppo].append(invert(ori))
    li[n1].append(oppo)
    li[n2].append(oppo)

    li[ori].append(invert(n1))
    li[ori].append(invert(n2))
    li[ori].append(invert(oppo))         

    rli[invert(ori)].append(n1)
    rli[invert(ori)].append(n2)
    rli[invert(ori)].append(oppo)
    rli[oppo].append(n1)
    rli[oppo].append(n2)

    rli[invert(n1)].append(ori)
    rli[invert(n2)].append(ori)
    rli[invert(oppo)].append(ori)       

def invert(num):
    if num % 2 == 0:
        num += 1
    else:
        num -= 1
    
    return num

@bootstrap
def dfs(value):
    for v in li[value]:
        if vt[v]:
            continue
        vt[v] = True
        yield dfs(v)
        
    stack.append(value)
    yield

@bootstrap 
def dfs2(value):
    for v in rli[value]:
        if vt[v]:
            continue
        vt[v] = True
        temp.append(v)
        yield dfs2(v)
    yield

n, m = map(int, input().split())
grid = [list(input()) for _ in range(n)]

# n, m = 8, 13
# s = '''......n....n.
# ..........#..
# .............
# ..#n..n....T.
# .............
# .n.T..T....n.
# .............
# ......n......'''
# grid = [list(v) for v in s.split('\n')]

cur_to_start = 0
t_c2si = dict() # t_coord_to_start_index
t_si2c = []

clones = []
nxt = 0
for i in range(n):
    for j in range(m):
        if grid[i][j] == 'T':
            t_c2si[(i, j)] = cur_to_start
            t_si2c.append((i, j))
            cur_to_start += 4
        elif grid[i][j] == 'n':
            clones.append((i, j))
    
k = cur_to_start
k *= 2
result = [-1] * k

li = [[] for _ in range(k)]
rli = [[] for _ in range(k)]

for i in range(0, cur_to_start, 4):
    i1 = get_2sat_num(i)
    i2 = get_2sat_num(i + 1)
    i3 = get_2sat_num(i + 2)
    i4 = get_2sat_num(i + 3)
    
#     print(i1, i2, i3, i4)
    
    li[i1].append(invert(i2))
    li[i2].append(invert(i1))
    li[invert(i1)].append(i2)
    li[invert(i2)].append(i1)
    
    li[i3].append(invert(i4))
    li[i4].append(invert(i3))
    li[invert(i3)].append(i4)
    li[invert(i4)].append(i3)

    rli[i1].append(invert(i2))
    rli[i2].append(invert(i1))
    rli[invert(i1)].append(i2)
    rli[invert(i2)].append(i1)
    
    rli[i3].append(invert(i4))
    rli[i4].append(invert(i3))
    rli[invert(i3)].append(i4)
    rli[invert(i4)].append(i3)    
    
# print(li)

for i, (x, y) in enumerate(t_si2c):
    met_ti_0 = -1
    ny = y
    for nx in range(x - 1, -1, -1):
        if grid[nx][ny] == '#':
            break
        if grid[nx][ny] == 'T':
            met_ti_0 = t_c2si[(nx, ny)] + 1
            break
            
    met_ti_1 = -1
    ny = y
    for nx in range(x + 1, n):
        if grid[nx][ny] == '#':
            break
        if grid[nx][ny] == 'T':
            met_ti_1 = t_c2si[(nx, ny)]
            break
            
    met_ti_2 = -1
    nx = x
    for ny in range(y - 1, -1, -1):
        if grid[nx][ny] == '#':
            break
        if grid[nx][ny] == 'T':
            met_ti_2 = t_c2si[(nx, ny)] + 3
            break
            
    met_ti_3 = -1
    nx = x
    for ny in range(y + 1, m):
        if grid[nx][ny] == '#':
            break
        if grid[nx][ny] == 'T':
            met_ti_3 = t_c2si[(nx, ny)] + 2
            break
            
    n_met_ti_0 = get_2sat_num(met_ti_0)
    n_met_ti_1 = get_2sat_num(met_ti_1)
    n_met_ti_2 = get_2sat_num(met_ti_2)
    n_met_ti_3 = get_2sat_num(met_ti_3)
    
    start = i * 4
    if met_ti_0 != -1:
        to_except_ori1 = get_2sat_num(start)
        to_except_ori2 = n_met_ti_0
        
        except_2sat_num(to_except_ori1)
        except_2sat_num(to_except_ori2)
    if met_ti_1 != -1:
        to_except_ori1 = get_2sat_num(start + 1)
        to_except_ori2 = n_met_ti_1

        except_2sat_num(to_except_ori1)
        except_2sat_num(to_except_ori2)
    if met_ti_2 != -1:
        to_except_ori1 = get_2sat_num(start + 2)
        to_except_ori2 = n_met_ti_2

        except_2sat_num(to_except_ori1)
        except_2sat_num(to_except_ori2)
    if met_ti_3 != -1:
        to_except_ori1 = get_2sat_num(start + 3)
        to_except_ori2 = n_met_ti_3

        except_2sat_num(to_except_ori1)
        except_2sat_num(to_except_ori2)
    
for x, y in clones:
    met_ti_0 = -1
    ny = y
    for nx in range(x - 1, -1, -1):
        if grid[nx][ny] == '#':
            break
        if grid[nx][ny] == 'T':
            met_ti_0 = t_c2si[(nx, ny)] + 1
            break
            
    met_ti_1 = -1
    ny = y
    for nx in range(x + 1, n):
        if grid[nx][ny] == '#':
            break
        if grid[nx][ny] == 'T':
            met_ti_1 = t_c2si[(nx, ny)]
            break
            
    met_ti_2 = -1
    nx = x
    for ny in range(y - 1, -1, -1):
        if grid[nx][ny] == '#':
            break
        if grid[nx][ny] == 'T':
            met_ti_2 = t_c2si[(nx, ny)] + 3
            break
            
    met_ti_3 = -1
    nx = x
    for ny in range(y + 1, m):
        if grid[nx][ny] == '#':
            break
        if grid[nx][ny] == 'T':
            met_ti_3 = t_c2si[(nx, ny)] + 2
            break
            
    met_ti_s = [met_ti_0, met_ti_1, met_ti_2, met_ti_3]
        
    n_met_ti_0 = get_2sat_num(met_ti_0)
    n_met_ti_1 = get_2sat_num(met_ti_1)
    n_met_ti_2 = get_2sat_num(met_ti_2)
    n_met_ti_3 = get_2sat_num(met_ti_3)
        
    ori = -1
    to_except_ori1 = -1
    to_except_ori2 = -1
    if met_ti_0 != -1 and met_ti_1 != -1:
        if met_ti_2 != -1:
            ori = n_met_ti_2
        else:
            ori = n_met_ti_3
        to_except_ori1 = n_met_ti_0
        to_except_ori2 = n_met_ti_1
        
    elif met_ti_2 != -1 and met_ti_3 != -1:
        if met_ti_0 != -1:
            ori = n_met_ti_0
        else:
            ori = n_met_ti_1
        
        to_except_ori1 = n_met_ti_2
        to_except_ori2 = n_met_ti_3
            
    elif met_ti_s.count(-1) == 3:
        if met_ti_0 != -1:
            ori = n_met_ti_0
        elif met_ti_1 != -1:
            ori = n_met_ti_1
        elif met_ti_2 != -1:
            ori = n_met_ti_2
        elif met_ti_3 != -1:
            ori = n_met_ti_3
            
#     print(111, met_ti_s)
            
    if ori != -1:
        n1, n2 = get_near_2sat_nums(ori)
        oppo = get_oppo_2sat_num(ori)
    
        li[n1].append(ori)
        li[n2].append(ori)
        li[oppo].append(ori)
        li[n1].append(invert(oppo))
        li[n2].append(invert(oppo))
        
        li[invert(ori)].append(invert(n1))
        li[invert(ori)].append(invert(n2))
        li[invert(ori)].append(invert(oppo))
        
        rli[ori].append(n1)
        rli[ori].append(n2)
        rli[ori].append(oppo)
        rli[invert(oppo)].append(n1)
        rli[invert(oppo)].append(n2)       
        
        rli[invert(n1)].append(invert(ori))
        rli[invert(n2)].append(invert(ori))
        rli[invert(oppo)].append(invert(ori))
        
        if to_except_ori1 != -1:
            ori = to_except_ori1
            except_2sat_num(to_except_ori1)
            except_2sat_num(to_except_ori2)
    else:
        if met_ti_0 != -1 and met_ti_2 != -1:
            li[invert(n_met_ti_0)].append(n_met_ti_2)
            li[invert(n_met_ti_2)].append(n_met_ti_0)
            
            rli[n_met_ti_2].append(invert(n_met_ti_0))
            rli[n_met_ti_0].append(invert(n_met_ti_2))
        elif met_ti_0 != -1 and met_ti_3 != -1:
            li[invert(n_met_ti_0)].append(n_met_ti_3)
            li[invert(n_met_ti_3)].append(n_met_ti_0)
            
            rli[n_met_ti_3].append(invert(n_met_ti_0))
            rli[n_met_ti_0].append(invert(n_met_ti_3))
        elif met_ti_1 != -1 and met_ti_2 != -1:
            li[invert(n_met_ti_1)].append(n_met_ti_2)
            li[invert(n_met_ti_2)].append(n_met_ti_1)
            
            rli[n_met_ti_2].append(invert(n_met_ti_1))
            rli[n_met_ti_1].append(invert(n_met_ti_2))
        elif met_ti_1 != -1 and met_ti_3 != -1:
            li[invert(n_met_ti_1)].append(n_met_ti_3)
            li[invert(n_met_ti_3)].append(n_met_ti_1)
            
            rli[n_met_ti_3].append(invert(n_met_ti_1))
            rli[n_met_ti_1].append(invert(n_met_ti_3))
            
# print(li)
    
stack = []
vt = [False] * k
for i in range(k):
    if vt[i]:
        continue
    vt[i] = True
    dfs(i)
    
vt = [False] * k
scc = []
while stack:
    value = stack.pop()
    if vt[value]:
        continue
    vt[value] = True
        
    temp = [value]
    dfs2(value)
    scc.append(temp.copy())
    
ip = True
for v in scc:
    v_st = set(v)
    
    for w in v_st:
        if w % 2 == 0 and (w + 1) in v_st:
            ip = False
            break
            
if not ip:
    assert False
else:
    result = [-1] * (k // 2)
    vt = [False] * k
    for i in range(len(scc)):
        for v in scc[i]:
            vt[v] = True
            inv_v = invert(v)
            
            if vt[inv_v]:
                continue
            else:
                if v % 2 == 0:
                    result[v // 2] = 0
                else:
                    result[v // 2] = 1
                      
    nresult = []
    for i in range(k // 2):
        if result[i]:
            nresult.append(i)
            
#     print(*nresult)
    
for i in range(0, len(t_si2c) * 2, 2):
    a = nresult[i]
    b = nresult[i + 1]
    
    a_mod4 = a % 4
    b_mod4 = b % 4
    
    x, y = t_si2c[i // 2]
    if a_mod4 == 0 and b_mod4 == 2:
        grid[x][y] = '4'
    elif a_mod4 == 0 and b_mod4 == 3:
        grid[x][y] = '3'
    elif a_mod4 == 1 and b_mod4 == 2:
        grid[x][y] = '1'
    elif a_mod4 == 1 and b_mod4 == 3:
        grid[x][y] = '2'
    else:
        assert False
        
for gr in grid:
    print(''.join(gr))
```
