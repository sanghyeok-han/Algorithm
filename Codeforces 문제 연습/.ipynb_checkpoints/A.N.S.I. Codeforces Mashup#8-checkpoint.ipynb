```python
# 틀렸음
def find(v):
    if v == parent[v]:
        return v
    parent[v] = find(parent[v])
    return parent[v]

def union(a, b):
    if len(sets[a]) >= len(sets[b]):
        parent[b] = a
        sets[a] |= sets[b]
        sets[b] = set()
    else:
        parent[a] = b
        sets[b] |= sets[a]
        sets[a] = set()

get_value = lambda x: int(x) - 1
n = int(input())
li = list(map(get_value, input().split()))
        
parent = list(range(n))
sets = [{i} for i in range(n)]

cands = set(range(n))
for i in range(n):
    if li[i] == -1:
        continue
    cands.discard(li[i])
    
    t1 = i
    t2 = li[i]
    rt1 = find(t1)
    rt2 = find(t2)
    
    if rt1 != rt2:
        union(rt1, rt2)

cands = list(cands)
for i in range(n):
    if li[i] != -1:
        continue
    if len(cands) == 1:
        li[i] = cands.pop()
        break
        
    t1 = cands.pop()
    t2 = cands.pop()
    ri = find(i)
    
    if t1 not in sets[ri]:
        li[i] = t1
        cands.append(t2)
    elif t2 not in sets[ri]:
        li[i] = t2
        cands.append(t1)
    else:
        assert False
    
li = [v + 1 for v in li]

print(*li)
```

```python
n = int(input())
li = list(map(int, input().split()))

to_fills = set(range(1, n + 1))
zero_locs = set()

for i in range(n):
    if li[i] == 0:
        zero_locs.add(i + 1)
    else:
        to_fills.discard(li[i])
        
sames = []
for v in to_fills:
    if v in zero_locs:
        sames.append(v)
        
if len(sames) == 1:
    loc = 0
    for v in zero_locs:
        if v != sames[0]:
            loc = v
            break
    
    li[loc - 1] = sames[0]
    zero_locs.discard(loc)
    to_fills.discard(sames[0])
    
    while to_fills:
        num = to_fills.pop()
        loc = zero_locs.pop()
        li[loc - 1] = num
elif len(sames) % 2 == 0:
    for i in range(0, len(sames), 2):
        t1 = sames[i]
        t2 = sames[i + 1]
        
        li[t1 - 1] = t2
        li[t2 - 1] = t1
        
        zero_locs.discard(t1)
        zero_locs.discard(t2)
        to_fills.discard(t1)
        to_fills.discard(t2)
        
    while to_fills:
        num = to_fills.pop()
        loc = zero_locs.pop()
        li[loc - 1] = num    
else:
    t1 = sames[0]
    t2 = sames[1]
    t3 = sames[2]
    
    li[t1 - 1] = t2
    li[t2 - 1] = t3
    li[t3 - 1] = t1
    
    zero_locs.discard(t1)
    zero_locs.discard(t2)
    zero_locs.discard(t3)
    to_fills.discard(t1)
    to_fills.discard(t2)
    to_fills.discard(t3)

    for i in range(3, len(sames), 2):
        t1 = sames[i]
        t2 = sames[i + 1]
        
        li[t1 - 1] = t2
        li[t2 - 1] = t1
        
        zero_locs.discard(t1)
        zero_locs.discard(t2)
        to_fills.discard(t1)
        to_fills.discard(t2)    
        
    while to_fills:
        num = to_fills.pop()
        loc = zero_locs.pop()
        li[loc - 1] = num
        
print(*li)
```

```python
from heapq import heappush, heappop

n = int(input())
li = list(map(int, input().split()))

cur = 0
count = 0
heap = []
for v in li:
    if v >= 0:
        cur += v
        count += 1
    else:
        if cur + v >= 0:
            cur += v
            count += 1
            heappush(heap, v)
        else:
            if not heap or heap[0] >= v:
                continue
            cur -= heappop(heap)
            cur += v
            heappush(heap, v)
            
print(count)
```

```python
import sys
input = lambda: sys.stdin.readline().rstrip()

# https://github.com/tatyam-prime/SortedSet/blob/main/SortedMultiset.py
import math
from bisect import bisect_left, bisect_right, insort
from typing import Generic, Iterable, Iterator, TypeVar, Union, List
T = TypeVar('T')

class SortedMultiset(Generic[T]):
    BUCKET_RATIO = 50
    REBUILD_RATIO = 170

    def _build(self, a=None) -> None:
        "Evenly divide `a` into buckets."
        if a is None: a = list(self)
        size = self.size = len(a)
        bucket_size = int(math.ceil(math.sqrt(size / self.BUCKET_RATIO)))
        self.a = [a[size * i // bucket_size : size * (i + 1) // bucket_size] for i in range(bucket_size)]
    
    def __init__(self, a: Iterable[T] = []) -> None:
        "Make a new SortedMultiset from iterable. / O(N) if sorted / O(N log N)"
        a = list(a)
        if not all(a[i] <= a[i + 1] for i in range(len(a) - 1)):
            a = sorted(a)
        self._build(a)

    def __iter__(self) -> Iterator[T]:
        for i in self.a:
            for j in i: yield j

    def __reversed__(self) -> Iterator[T]:
        for i in reversed(self.a):
            for j in reversed(i): yield j
    
    def __len__(self) -> int:
        return self.size
    
    def __repr__(self) -> str:
        return "SortedMultiset" + str(self.a)
    
    def __str__(self) -> str:
        s = str(list(self))
        return "{" + s[1 : len(s) - 1] + "}"

    def _find_bucket(self, x: T) -> List[T]:
        "Find the bucket which should contain x. self must not be empty."
        for a in self.a:
            if x <= a[-1]: return a
        return a

    def __contains__(self, x: T) -> bool:
        if self.size == 0: return False
        a = self._find_bucket(x)
        i = bisect_left(a, x)
        return i != len(a) and a[i] == x

    def count(self, x: T) -> int:
        "Count the number of x."
        return self.index_right(x) - self.index(x)

    def add(self, x: T) -> None:
        "Add an element. / O(√N)"
        if self.size == 0:
            self.a = [[x]]
            self.size = 1
            return
        a = self._find_bucket(x)
        insort(a, x)
        self.size += 1
        if len(a) > len(self.a) * self.REBUILD_RATIO:
            self._build()

    def discard(self, x: T) -> bool:
        "Remove an element and return True if removed. / O(√N)"
        if self.size == 0: return False
        a = self._find_bucket(x)
        i = bisect_left(a, x)
        if i == len(a) or a[i] != x: return False
        a.pop(i)
        self.size -= 1
        if len(a) == 0: self._build()
        return True

    def lt(self, x: T) -> Union[T, None]:
        "Find the largest element < x, or None if it doesn't exist."
        for a in reversed(self.a):
            if a[0] < x:
                return a[bisect_left(a, x) - 1]

    def le(self, x: T) -> Union[T, None]:
        "Find the largest element <= x, or None if it doesn't exist."
        for a in reversed(self.a):
            if a[0] <= x:
                return a[bisect_right(a, x) - 1]

    def gt(self, x: T) -> Union[T, None]:
        "Find the smallest element > x, or None if it doesn't exist."
        for a in self.a:
            if a[-1] > x:
                return a[bisect_right(a, x)]

    def ge(self, x: T) -> Union[T, None]:
        "Find the smallest element >= x, or None if it doesn't exist."
        for a in self.a:
            if a[-1] >= x:
                return a[bisect_left(a, x)]
    
    def __getitem__(self, x: int) -> T:
        "Return the x-th element, or IndexError if it doesn't exist."
        if x < 0: x += self.size
        if x < 0: raise IndexError
        for a in self.a:
            if x < len(a): return a[x]
            x -= len(a)
        raise IndexError

    def index(self, x: T) -> int:
        "Count the number of elements < x."
        ans = 0
        for a in self.a:
            if a[-1] >= x:
                return ans + bisect_left(a, x)
            ans += len(a)
        return ans

    def index_right(self, x: T) -> int:
        "Count the number of elements <= x."
        ans = 0
        for a in self.a:
            if a[-1] > x:
                return ans + bisect_right(a, x)
            ans += len(a)
        return ans
    
for _ in range(int(input())):
    n = int(input())
    li = [list(map(int, input().split())) for _ in range(n)]

    starts = SortedMultiset()
    ends = SortedMultiset()
    for s, e in li:
        starts.add(s)
        ends.add(e)

    min_deleted = float('inf')
    for s, e in li:
        c1 = ends.index(s)
        c2 = n - starts.index_right(e)
        deleted = c1 + c2

        min_deleted = min(min_deleted, deleted)

    print(min_deleted)
```

```python
from collections import defaultdict as dd
import sys
input = lambda: sys.stdin.readline().rstrip()

for _ in range(int(input())):
    n = int(input())
    li = list(map(int, input().split()))

    d = dd(list)
    for i in range(n):
        d[li[i]].append(i)

    r = 0
    for key in d:
        if len(d[key]) == 1:
            continue
        i_cumul = d[key][0] + 1
        for j in range(1, len(d[key])):
            j_cur = d[key][j]
            r += i_cumul * (n - j_cur)
            i_cumul += j_cur + 1

    print(r)
```

```python
import sys
# input = lambda: sys.stdin.readline().rstrip()

def check(l, m, r):
    if l < 0 or r >= n:
        return False
    if li[m] < li[l] and li[m] < li[r]:
        return True
    if li[m] > li[l] and li[m] > li[r]:
        return True
    
    return False

def check_triple(mid):
    r = check(mid - 2, mid - 1, mid)
    r += check(mid - 1, mid, mid + 1)
    r += check(mid, mid + 1, mid + 2)
    
    return r
   
for _ in range(int(input())):
    n = int(input())
    li = list(map(int, input().split()))

    max_saved = 0
    all_c = 0
    for mid in range(1, n - 1):
        all_c += check(mid - 1, mid, mid + 1)

        fvalue = li[mid]

        c = check_triple(mid)

        li[mid] = li[mid - 1]
        c1 = check_triple(mid)

        saved = c - c1
        if saved > max_saved:
            max_saved = saved

        li[mid] = li[mid + 1]
        c2 = check_triple(mid)

        saved = c - c2
        if saved > max_saved:
            max_saved = saved
            
        li[mid] = fvalue


    print(all_c - max_saved)
```

```python
from collections import Counter
import sys
# input = lambda: sys.stdin.readline().rstrip()

def nC3(num):
    return num * (num - 1) * (num - 2) // 6
          
n, k = map(int, input().split())
li = [input() for _ in range(n)]

ct = Counter(li)

r = 0
for i in range(n):
    for j in range(i + 1, n):
        to_find = []
        same_c = 0
        for ki in range(k):
            if li[i][ki] == li[j][ki]:
                same_c += 1
                to_find.append(li[i][ki])
            else:
                a = li[i][ki]
                b = li[j][ki]
                
                if a > b:
                    a, b = b, a
                
                if a == 'E' and b == 'S':
                    to_find.append('T')
                elif a == 'E' and b == 'T':
                    to_find.append('S')
                else:
                    to_find.append('E')
                    
        if same_c == k:
            continue
                
        r += ct[''.join(to_find)]
        
r //= 3

for key in ct:
    if ct[key] >= 3:
        r += nC3(ct[key])
        
print(r)
```
