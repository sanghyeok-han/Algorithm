```python
def get_divisors(num):
    divisors = set()
    for mod in range(2, int(num ** 0.5) + 1):
        if num % mod == 0:
            divisors.add(mod)
            divisors.add(num // mod)
            
    return divisors

get_divisors(12)
```

```python
def recur(num):
    if dp[num] != -1:
        return dp[num]
    
    is_other_win = True
    for to_sub in get_divisors(num):
        nnum = num - to_sub
        
        temp = recur(nnum)
        
        if not temp:
            is_other_win = False
            
    is_win = not is_other_win
    
    dp[num] = is_win
    return is_win

dp = [-1] * 100000

fv = ''
for num in range(1, 10000):
    if fv == recur(num):
        print(num, recur(num))
#     print(num, recur(num))
    fv = recur(num)
```

```python
from collections import deque

n, k = map(int, input().split())
li = [set() for _ in range(n + 1)]
for _ in range(n - 1):
    a, b = map(int, input().split())
    li[a].add(b)
    li[b].add(a)

deleted = set()
queue = deque()
for i in range(1, n + 1):
    if len(li[i]) <= 1:
        deleted.add(i)
        queue.append(i)

for _ in range(k - 1):
    for _ in range(len(queue)):
        value = queue.popleft()
        
        for v in li[value]:
            if v in deleted:
                continue
            
            li[v].discard(value)
            if len(li[v]) <= 1:
                queue.append(v)
                deleted.add(value)
                
        if len(deleted) == n:
            break

print(n - len(deleted))
```

```python
from itertools import permutations as perm, product as prod
from collections import deque

# s = input()
# n = len(s) + 1

n = 8

for games in prod(*[range(2)] * (n - 1)):
    st = set()
    for case in perm(range(1, n + 1)):
        queue = deque(case)
        gi = 0
        while len(queue) >= 2:
            a = queue.popleft()
            b = queue.popleft()
            
            if a > b:
                a, b = b, a
            if games[gi] == 0:
                queue.append(a)
            else:
                queue.append(b)
            gi += 1
            
        st.add(queue[0])
        
    print(games, st)
```
