```python
from math import log2, ceil
import sys
# input = lambda: sys.stdin.readline().rstrip()

inf = float('inf')

def construct(segtree, li, node, start, end):
    if start == end:
        segtree[node] = li[start]
        return segtree[node]
    mid = (start + end) // 2
    
    t1 = construct(segtree, li, node * 2, start, mid)
    t2 = construct(segtree, li, node * 2 + 1, mid + 1, end)
    
    segtree[node] = max(t1, t2)
    
    return segtree[node]

def query(segtree, node, s, e, qs, qe):
    if e < qs or s > qe:
        return -inf
    if qs <= s and e <= qe:
        return segtree[node]
    else:
        mid = (s + e) // 2
        
        t1 = query(segtree, node * 2, s, mid, qs, qe)
        t2 = query(segtree, node * 2 + 1, mid + 1, e, qs, qe)
        
        return max(t1, t2)

for _ in range(int(input())):
    n = int(input())
    li = list(map(int, input().split()))

    stack = []
    itvs = [[-1, -1] for _ in range(n)]
    for i in range(n):
        while stack and stack[-1][0] < li[i]:
            fv, fis = stack.pop()
            for fi in fis:
                itvs[fi][1] = i - 1
        if stack:
            if stack[-1][0] == li[i]:
                stack[-1][1].append(i)
                if len(stack) >= 2:
                    itvs[i][0] = stack[-2][1][-1] + 1
                else:
                    itvs[i][0] = 0
            else:
                itvs[i][0] = stack[-1][1][-1] + 1
                stack.append([li[i], [i]])
        else:
            itvs[i][0] = 0
            stack.append([li[i], [i]])

    while stack:
        fv, fis = stack.pop()
        for fi in fis:
            itvs[fi][1] = n - 1

    to_right_sum = [0] * n
    cumul = 0
    for i in range(n):
        cumul += li[i]
        to_right_sum[i] = cumul

    to_left_sum = [0] * n
    cumul = 0
    for i in range(n - 1, -1, -1):
        cumul += li[i]
        to_left_sum[i] = cumul

    size = 2 ** (ceil(log2(n)) + 1) + 1

    to_right_tree = [0] * size
    construct(to_right_tree, to_right_sum, 1, 0, n - 1)

    to_left_tree = [0] * size
    construct(to_left_tree, to_left_sum, 1, 0, n - 1)

    ip = True
    for i in range(n):
        l = itvs[i][0]
        r = itvs[i][1]

        if 0 < i < n - 1:
            left_max = query(to_left_tree, 1, 0, n - 1, l, i) - to_left_sum[i] + li[i]
            right_max = query(to_right_tree, 1, 0, n - 1, i, r) - to_right_sum[i] + li[i] 
            if max(left_max, right_max) > li[i]:
                ip = False
                break
        elif i == 0:
            right_max = query(to_right_tree, 1, 0, n - 1, i, r) - to_right_sum[i] + li[i] 
            if right_max > li[i]:
                ip = False
                break
        else:
            left_max = query(to_left_tree, 1, 0, n - 1, l, i) - to_left_sum[i] + li[i]
            if left_max > li[i]:
                ip = False
                break

    if ip:
        print('YES')
    else:
        print('NO')
```
