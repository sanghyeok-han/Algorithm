```python
from collections import deque

li = [list(map(int, input().split())) for _ in range(3)]

target = 123456789

to_coord = lambda index: (index // 3, index % 3)
to_index = lambda x, y: x * 3 + y

def get_w(x, y):
    index = to_index(x, y)
    return 10 ** (8 - index)
    
cur = 0
for i in range(3):
    for j in range(3):
        if li[i][j] == 0:
            li[i][j] = 9
        cur += get_w(i, j) * li[i][j]
        
dx = [-1, 1, 0, 0]
dy = [0, 0, -1, 1]
        
result = -1

vt = {cur}

queue = deque()
queue.append((cur, 0))
while queue:
    cur, c = queue.popleft()
    
    if cur == target:
        result = c
        break
    
    scur = str(cur)
    index = scur.index('9')
    x, y = to_coord(index)
    
    temp_ncur = cur - get_w(x, y) * 9
    for d in range(4):
        nx = x + dx[d]
        ny = y + dy[d]
        
        if nx < 0 or nx >= 3 or ny < 0 or ny >= 3:
            continue
        nindex = to_index(nx, ny)
        
        ncur = temp_ncur
        num = int(scur[nindex])
        ncur -= num * get_w(nx, ny)
        
        ncur += 9 * get_w(nx, ny)
        ncur += num * get_w(x, y)
        
        if ncur not in vt:
            vt.add(ncur)
            queue.append((ncur, c + 1))
            
print(result)
```

```python
서로 알고 있으면 반드시 같은 위원회 -> union find
위원회 수의 최대 -> 위의 조건이 아니면 전부 다른 위원회로 구분

```

```python
from collections import defaultdict as dd, deque
import sys
input = lambda: sys.stdin.readline().rstrip()

def find(v):
    if parent[v] != v:
        parent[v] = find(parent[v])
    return parent[v]

def union(a, b):
    if a > b:
        a, b = b, a
    parent[b] = a

n = int(input())
m = int(input())

parent = list(range(n + 1))
li = [[] for _ in range(n + 1)]
for _ in range(m):
    u, v = map(int, input().split())
    li[u].append(v)
    li[v].append(u)
    
    ru = find(u)
    rv = find(v)
    
    if ru != rv:
        union(ru, rv)
        
d = dd(list)
for value in range(1, n + 1):
    d[find(value)].append(value)    
r_num = len(d)

min_c_starts = []
for key in d:
    min_c = 1000
    min_c_start = -1
    for start in d[key]:
        vt = [False] * (n + 1)
        vt[start] = True
        
        queue = deque()
        queue.append((start, 0))
        
        max_temp_c = 0
        while queue:
            value, temp_c = queue.popleft()
            max_temp_c = max(max_temp_c, temp_c)
            
            for v in li[value]:
                if not vt[v]:
                    vt[v] = True
                    queue.append((v, temp_c + 1))
        if max_temp_c < min_c:
            min_c = max_temp_c
            min_c_start = start
    min_c_starts.append(min_c_start)
    
min_c_starts.sort()

print(r_num)
for v in min_c_starts:
    print(v)
```

```python
from collections import deque
import sys
# input = lambda: sys.stdin.readline().rstrip()

get_value = lambda x: int(x) - 1

n, m = map(int, input().split())

info = [[[] for _ in range(n)] for _ in range(n)]
for _ in range(m):
    x, y, a, b = map(get_value, input().split())
    info[x][y].append((a, b))
    
dx = [-1, 1, 0, 0]
dy = [0, 0, -1, 1]
    
cands = set()
vt_on = [[False] * n for _ in range(n)]
vt_on[0][0] = True

is_on = [[False] * n for _ in range(n)]
is_on[0][0] = True

queue = deque()
queue.append((0, 0))
while queue:
    x, y = queue.popleft()
    
    for a, b in info[x][y]:
        if not is_on[a][b]:
            is_on[a][b] = True
            if (a, b) in cands:
                vt_on[a][b] = True
                queue.append((a, b))
                
    for d in range(4):
        nx = x + dx[d]
        ny = y + dy[d]
        
        if nx < 0 or nx >= n or ny < 0 or ny >= n:
            continue
        if not is_on[nx][ny]:
            cands.add((nx, ny))
            continue
        if not vt_on[nx][ny]:
            vt_on[nx][ny] = True
            queue.append((nx, ny))
            
r = 0
for i in range(n):
    for j in range(n):
        if is_on[i][j]:
            r += 1
            
print(r)
```

```python
from collections import defaultdict as dd
import sys
# input = lambda: sys.stdin.readline().rstrip()

n = int(input())
an, bn = map(int, input().split())
a = [int(input()) for _ in range(an)]
b = [int(input()) for _ in range(bn)]

ca = dd(int)

cumul = a[an - 1]
ca[cumul] += 1
to_add_i = an - 2
to_del_i = an - 1
for _ in range(an - 1):
    for i in range(an):
        cumul += a[to_add_i]
        cumul -= a[to_del_i]
        if i < an - 1:
            ca[cumul] += 1
        to_add_i -= 1
        to_del_i -= 1
    
    to_add_i += an
    to_del_i += an
    
    cumul += a[to_add_i]
    ca[cumul] += 1
    to_add_i -= 1

cb = dd(int)

cumul = b[bn - 1]
cb[cumul] += 1
to_add_i = bn - 2
to_del_i = bn - 1
for _ in range(bn - 1):
    for i in range(bn):
        cumul += b[to_add_i]
        cumul -= b[to_del_i]
        if i < bn - 1:
            cb[cumul] += 1
        to_add_i -= 1
        to_del_i -= 1
    
    to_add_i += bn
    to_del_i += bn
    
    cumul += b[to_add_i]
    cb[cumul] += 1
    to_add_i -= 1
    
total = ca[n] + cb[n]

for a_key in ca:
    to_find = n - a_key
    total += ca[a_key] * cb[to_find]
    
print(total)
```

```python
from collections import deque
import sys
input = lambda: sys.stdin.readline().rstrip()

n, m = map(int, input().split())
init = [input() for _ in range(n)]

li = [[0] * m for _ in range(n)]
for i in range(n):
    for j in range(m):
        if init[i][j] != '.':
            li[i][j] = int(init[i][j])

deleted = [[False] * m for _ in range(n)]
queue = deque()
for i in range(n):
    for j in range(m):
        if li[i][j] == 0:
            deleted[i][j] = True
            queue.append((i, j))

dx = [-1, -1, -1, 0, 0, 1, 1, 1]
dy = [-1, 0, 1, -1, 1, -1, 0, 1]

tm = 0            
while True:
    for _ in range(len(queue)):
        x, y = queue.popleft()
        
        for d in range(8):
            nx = x + dx[d]
            ny = y + dy[d]

            if nx < 0 or nx >= n or ny < 0 or ny >= m:
                continue
            li[nx][ny] -= 1
            if not deleted[nx][ny] and li[nx][ny] <= 0:
                deleted[nx][ny] = True
                queue.append((nx, ny))
    
    if not queue:
        break
    tm += 1
    
print(tm)
```

```python
from math import gcd

n = int(input())
li = list(map(int, input().split()))

g = li[0]
for i in range(1, n):
    g = gcd(g, li[i])

nli = [li[i] // g for i in range(n)]
    
cands = []
cur = nli[0]
for i in range(1, n):
    cur = gcd(cur, nli[i])
    
    if cur == 1:
        cands.append(i)
        break

cur = nli[cands[0]]
for i in range(n):
    if cur == cands[0]:
        continue
    cur = gcd(cur, nli[i])
    
    if cur == 1:
        cands.append(i)
        break
    
max_r = -1
r_to_ex = -1
for to_ex in cands: 
    g2 = -1
    for i in range(n):
        if i == to_ex:
            continue
        if g2 == -1:
            g2 = nli[i]
        else:
            g2 = gcd(g2, nli[i])

    if g2 == 1:
        continue
    
    temp = g * g2
    if temp > max_r:
        max_r = temp
        r_to_ex = to_ex
        
if max_r == -1:
    print(-1)
else:
    print(max_r, li[r_to_ex])
```

```python
from collections import deque
from math import ceil
import sys
input = lambda: sys.stdin.readline().rstrip()

def gnf(x1, y1, x2, y2):
    '''get needed fuel'''
    
    dist_s = (x1 - x2) ** 2 + (y1 - y2) ** 2
    dist = dist_s ** 0.5
    
    fuel = int(dist // 10)
    if dist % 10:
        fuel += 1
        
    return fuel

n, k = map(int, input().split())
li = [[0, 0]] + [list(map(int, input().split())) for _ in range(n)] + [[10000, 10000]]
n += 2

start = 1
end = 1415
min_r = 1416
while start <= end:
    mid = (start + end) // 2
    
    not_vt = set(range(n))
    not_vt.discard(0)
    
    queue = deque()
    queue.append(0)
    for _ in range(k + 1):
        for _ in range(len(queue)):
            value = queue.popleft()
            
            td = set() # to delete
            for v in not_vt:
                if gnf(li[value][0], li[value][1], li[v][0], li[v][1]) <= mid:
                    td.add(v)
                    queue.append(v)
                    
            not_vt -= td
            
        if not queue:
            break
            
    if n - 1 in not_vt:
        start = mid + 1
    else:
        min_r = mid
        end = mid - 1
        
print(min_r)
```

### 등산 1386

```python
from collections import deque
import sys
input = lambda: sys.stdin.readline().rstrip()

inf = float('inf')

dx = [-1, 1, 0, 0]
dy = [0, 0, -1, 1]

def bfs(sx, sy):
    spents = [[inf] * m for _ in range(n)]
    spents[sx][sy] = 0
    
    queue = deque()
    queue.append((sx, sy))
    while queue:
        x, y = queue.popleft()
        
        for di in range(4):
            nx = x + dx[di]
            ny = y + dy[di]
            
            if nx < 0 or nx >= n or ny < 0 or ny >= m:
                continue
            diff = abs(li[x][y] - li[nx][ny])
            if diff > t:
                continue
            nspent = spents[x][y]
            if li[x][y] >= li[nx][ny]:
                nspent += 1
            else:
                nspent += diff ** 2
                
            if nspent < spents[nx][ny]:
                spents[nx][ny] = nspent
                queue.append((nx, ny))
                
    return spents
            
        

n, m, t, d = map(int, input().split())
init = [input() for _ in range(n)]

li = [[0] * m for _ in range(n)]
for i in range(n):
    for j in range(m):
        num = ord(init[i][j])
        
        if 65 <= num <= 90:
            h = num - 65
        else:
            h = num - 97 + 26
        
        li[i][j] = h
    
to_forward = bfs(0, 0)
max_r = li[0][0]
for tx in range(n):
    for ty in range(m):
        if li[tx][ty] <= max_r:
            continue
        spent = to_forward[tx][ty] + bfs(tx, ty)[0][0]
        
        if spent <= d:
            max_r = li[tx][ty]
            
print(max_r)
```

### 일요일 아침의 데이트 1445

```python
from collections import deque
import sys
input = lambda: sys.stdin.readline().rstrip()

inf = float('inf')

n, m = map(int, input().split())
li = [input() for _ in range(n)]

sx = sy = -1
fx = fy = -1
for i in range(n):
    for j in range(m):
        if li[i][j] == 'S':
            sx = i
            sy = j
        elif li[i][j] == 'F':
            fx = i
            fy = j
            
dx = [-1, 1, 0, 0]
dy = [0, 0, -1, 1]
gbs = [[[inf] * 2 for _ in range(m)] for _ in range(n)]
gbs[sx][sy] = [0, 0]

queue = deque()
queue.append((sx, sy, 0, 0))
while queue:
    x, y, th, adj = queue.popleft()
    
    for d in range(4):
        nx = x + dx[d]
        ny = y + dy[d]
        if nx < 0 or nx >= n or ny < 0 or ny >= m:
            continue
        
        nth = th
        nadj = adj
        if li[nx][ny] != 'F':
            if li[nx][ny] == 'g':
                nth += 1
            else:
                is_adj = False
                for dd in range(4):
                    nnx = nx + dx[dd]
                    nny = ny + dy[dd]

                    if nnx < 0 or nnx >= n or nny < 0 or nny >= m:
                        continue
                    if li[nnx][nny] == 'g':
                        is_adj = True
                        break
                if is_adj:
                    nadj += 1
        
        if nth > gbs[nx][ny][0]:
            continue
        if nth == gbs[nx][ny][0] and nadj >= gbs[nx][ny][1]:
            continue
        gbs[nx][ny] = [nth, nadj]
        queue.append((nx, ny, nth, nadj))
        
print(*gbs[fx][fy])
```
