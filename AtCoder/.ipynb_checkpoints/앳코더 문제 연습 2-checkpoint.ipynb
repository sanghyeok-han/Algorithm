```python
n = int(input())
li = list(map(int, input().split()))

ct = [0] * (n + 1)

for i in range(n):   
    for v in [(li[i] - 1) % n, li[i], (li[i] + 1) % n]:
        if i <= v:
            diff = v - i
        else:
            diff = n - i + v    
        ct[diff] += 1     
        
r = max(ct)

print(r)
```

```python
from collections import defaultdict as dd
from itertools import permutations as perm
import sys
input = lambda: sys.stdin.readline().rstrip()

def make_case(cur_i, cur, free):
    global cases_dash
    
    if cur_i == n - 1:
        total_used = char_len + n - 1 + init_free - free
        if total_used >= 3:
            cases_dash.append(tuple(cur))
        return
    
    for used in range(free + 1):
        cur.append(1 + used)
        make_case(cur_i + 1, cur, free - used)
        cur.pop()

n, m = map(int, input().split())
ss = [input() for _ in range(n)]
ts = [input() for _ in range(m)]

to_index = dict()
for i in range(n):
    to_index[ss[i]] = i
    
t_info = dd(set)
for i in range(m):
    included = []
    dashes = []
    temp = []
    c_dash = 0
    is_matched = True
    length = len(ts[i])
    st = set()
    for j in range(length):
        if ts[i][j] == '_':
            if temp:
                temp = ''.join(temp)
                if temp not in to_index:
                    is_matched = False
                    break
                else:
                    st.add(to_index[temp])
                    included.append(to_index[temp])
                    temp = []
            c_dash += 1
        else:
            if c_dash:
                dashes.append(c_dash)
                c_dash = 0
            temp.append(ts[i][j])
                
    if temp:
        temp = ''.join(temp)
        if temp not in to_index:
            is_matched = False
        else:
            st.add(to_index[temp])
            included.append(to_index[temp])
    if c_dash:
        dashes.append(c_dash)
        temp.append(ts[i][j])   
        
    if is_matched and len(st) == n:
        t_info[tuple(included)].add(tuple(dashes))
    
char_len = 0
for i in range(n):
    char_len += len(ss[i])

init_free = 16 - char_len - (n - 1)
cases_dash = []
make_case(0, [], init_free)
    
rs = []
rd = []
for case_str in perm(range(n)):
    for case_dash in cases_dash:
        if case_dash not in t_info[case_str]:
            rs = case_str
            rd = case_dash
            break
    if rs:
        break
    
if not rs:
    print(-1)
else:
    result = []
    for i in range(n - 1):
        result.append(ss[rs[i]])
        result.append('_' * rd[i])
    result.append(ss[rs[n - 1]])
    
    result = ''.join(result)
    print(result)
```

https://atcoder.jp/contests/abc266/tasks/abc266_c C - Convex Quadrilateral

```python
def ccw(x1, y1, x2, y2, x3, y3):
    temp = x1 * y2 + x2 * y3 + x3 * y1 - x2 * y1 - x3 * y2 - x1 * y3
    
    if temp < 0:
        return -1
    elif temp > 0:
        return 1
    else:
        return 0

ax, ay = map(int, input().split())
bx, by = map(int, input().split())
cx, cy = map(int, input().split())
dx, dy = map(int, input().split())

t1 = ccw(ax, ay, bx, by, cx, cy)
t2 = ccw(bx, by, cx, cy, dx, dy)
t3 = ccw(cx, cy, dx, dy, ax, ay)
t4 = ccw(dx, dy, ax, ay, bx, by)

if t1 == t2 == t3 == t4 == 1:
    print('Yes')
else:
    print('No')
```

https://atcoder.jp/contests/abc269/tasks/abc269_d D - Do use hexagon grid

```python
from collections import deque
import sys
input = lambda: sys.stdin.readline().rstrip()

dx = [0, 1, -1, 1, -1, 0]
dy = [1, 1, 0, 0, -1, -1]

def is_adj(a, b):
    ax, ay = a
    bx, by = b
    
    is_same = False
    for d in range(6):
        nax = ax + dx[d]
        nay = ay + dy[d]
        
        if nax == bx and nay == by:
            is_same = True
            break
            
    return is_same
    

n = int(input())
li = [list(map(int, input().split())) for _ in range(n)]

adj = [[] for _ in range(n)]
for i in range(n):
    for j in range(i + 1, n):
        if is_adj(li[i], li[j]):
            adj[i].append(j)
            adj[j].append(i)
            
vt = [False] * n
r = 0
for i in range(n):
    if vt[i]:
        continue
    vt[i] = True
    r += 1
    
    queue = deque()
    queue.append(i)
    while queue:
        value = queue.popleft()
        
        for v in adj[value]:
            if not vt[v]:
                vt[v] = True
                queue.append(v)

print(r)
```

https://atcoder.jp/contests/abc269/tasks/abc269_f F - Numbered Checker

```python
import sys
input = lambda: sys.stdin.readline().rstrip()

mod = 998244353

def get_sum(num):
    return num * (num + 1) // 2 % mod

def get_ari_sum(a, d, n):
    return (n * (2 * a + (n - 1) * d) // 2) % mod

def get_point(x, y):
    if x % 2 != y % 2:
        return 0
    else:
        temp = (y - 1) * m + x
        return temp % mod

def get_x_len_1(sy, ey, x):
    y_len = (ey - sy + 1) 
    
    if sy % 2 != x % 2:
        sy += 1
        y_len -= 1
    if ey % 2 != x % 2:
        ey -= 1
        y_len -= 1
        
    a = get_point(x, sy)
    d = m * 2
    n = y_len // 2 + 1
    
    return get_ari_sum(a, d, n) % mod

def get_y_len_1(sx, ex, y):
    x_len = ex - sx + 1    
    
    if sx % 2 != y % 2:
        sx += 1
        x_len -= 1
    if ex % 2 != y % 2:
        ex -= 1
        x_len -= 1
    
    a = get_point(sx, y)
    d = 2
    n = x_len // 2 + 1
    
    return get_ari_sum(a, d, n) % mod

def get_22(sx, ex, sy, ey):
    x_len = ex - sx + 1
    y_len = ey - sy + 1
    
    t1 = m * get_sum(y_len - 1) % mod * x_len % mod
    t2 = (get_sum(ex) - get_sum(sx - 1) + m * (sy - 1) * x_len) % mod * y_len % mod
    
    temp = t1 + t2
    temp = temp * pow(2, mod - 2, mod) % mod
    
    return temp

def get_xy(sx, ex, sy, ey):
    x_len = ex - sx + 1
    y_len = ey - sy + 1
    
    if x_len == 1 and y_len == 1:
        return get_point(sx, sy)
    elif x_len == 1:
        return get_x_len_1(sy, ey, sx)
    elif y_len == 1:
        return get_y_len_1(sx, ex, sy)
    else:
        if x_len % 2 == 1 and y_len % 2 == 1:
            t1 = get_x_len_1(sy, ey, ex) + get_y_len_1(sx, ex, ey) - get_point(ex, ey)
            t2 = get_22(sx, ex - 1, sy, ey - 1)
            return (t1 + t2) % mod
        elif x_len % 2 == 0 and y_len % 2 == 1:
            t1 = get_y_len_1(sx, ex, ey)
            t2 = get_22(sx, ex, sy, ey - 1)
            return (t1 + t2) % mod
        elif x_len % 2 == 1 and y_len % 2 == 0:
            t1 = get_x_len_1(sy, ey, ex)
            t2 = get_22(sx, ex - 1, sy, ey)
            return (t1 + t2) % mod
        else:
            return get_22(sx, ex, sy, ey)

n, m = map(int, input().split())
q = int(input())
for _ in range(q):
    sy, ey, sx, ex = map(int, input().split())
    
    result = get_xy(sx, ex, sy, ey)
    print(result)
```

https://atcoder.jp/contests/abc268/tasks/abc268_f abc268_f

```python
import functools
import sys
input = lambda: sys.stdin.readline().rstrip()

def get_value(s):
    xc = 0
    r = 0
    for v in s:
        if v == 'X':
            xc += 1
        else:
            r += int(v) * xc
    return r

def compare(a, b):
    s1 = a + b
    s2 = b + a
    
    if get_value(s1) >= get_value(s2):
        return -1
    else:
        return 1
    
n = int(input())
li = [input() for _ in range(n)]
li.sort(key=functools.cmp_to_key(compare))

r = get_value(''.join(li))

print(r)
```

https://atcoder.jp/contests/abc267/tasks/abc267_e E - Erasing Vertices 2

```python
from heapq import heappush, heappop
import sys
input = lambda: sys.stdin.readline().rstrip()

n, m = map(int, input().split())
costs = list(map(int, input().split()))

li = [[] for _ in range(n)]
to_cost = [0] * n
for _ in range(m):
    u, v = map(int, input().split())
    u -= 1
    v -= 1
    
    li[u].append(v)
    li[v].append(u)
    
    to_cost[u] += costs[v]
    to_cost[v] += costs[u]

heap = []
for i in range(n):
    heappush(heap, (to_cost[i], i))
    
max_v = -1
is_removeds = [False] * n
while heap:
    cost, value = heappop(heap)
    if is_removeds[value]:
        continue
    is_removeds[value] = True
    
    max_v = max(max_v, cost)
    
    for v in li[value]:
        if not is_removeds[v]:
            to_cost[v] -= costs[value]
            heappush(heap, (to_cost[v], v))
            
print(max_v)
```

```python
import sys
input = lambda: sys.stdin.readline().rstrip()

inf = float('inf')

n = int(input())

d = {}
for _ in range(n):
    t, x, a = map(int, input().split())
    
    d[str(t)] = (x, a)

dp = [[-inf] * 5 for _ in range(100001)]
dp[0][0] = 0

for cur_t in range(1, 100001):
    x, a = -1, -1
    if str(cur_t) in d:
        x, a = d[str(cur_t)]
    
    for cur_i in range(5):
        temp = -inf
        for f_i in range(cur_i - 1, cur_i + 2):
            if f_i < 0 or f_i >= 5:
                continue

            temp = max(temp, dp[cur_t - 1][f_i])

        dp[cur_t][cur_i] = temp
        if cur_i == x:
            dp[cur_t][cur_i] += a
            
max_r = max(dp[t])

print(max_r)
```

```python
import sys
input = lambda: sys.stdin.readline().rstrip()

## PYRIVAL BOOTSTRAP
# https://github.com/cheran-senthil/PyRival/blob/master/pyrival/misc/bootstrap.py
# This decorator allows for recursion without actually doing recursion
## @bootstrap, yield when getting and returning value in recursive functions, end of functions
from types import GeneratorType

def bootstrap(f, stack=[]):
    def wrappedfunc(*args, **kwargs):
        if stack:
            return f(*args, **kwargs)
        else:
            to = f(*args, **kwargs)
            while True:
                if type(to) is GeneratorType:
                    stack.append(to)
                    to = next(to)
                else:
                    stack.pop()
                    if not stack:
                        break
                    to = stack[-1].send(to)
            return to
    return wrappedfunc

@bootstrap
def find(v):
    if v != parent[v]:
        parent[v] = yield find(parent[v])
    
    yield parent[v]

def union(a, b):
    if a > b:
        a, b = b, a
    
    parent[b] = a

@bootstrap 
def dfs_to_find_route(value):
    global min_st
    
    vt[value] = True
    route.append(value)
    
    if value == needless[1]:
        min_st = set(route)
        
    for v in li[value]:
        if not vt[v]:
            yield dfs_to_find_route(v)
    
    route.pop()
    
    yield

@bootstrap
def dfs_to_group(value):
    vt[value] = True
    
    for v in li[value]:
        if not vt[v]:
            temp = (value in min_st) + (v in min_st)
            if temp != 1:
                union((yield find(value)), (yield find(v)))
            else:
                if value in min_st:
                    adj_mid[str(v)] = value
                else:
                    adj_mid[str(value)] = v
            yield dfs_to_group(v)

    yield
    
n = int(input())

li = [[] for _ in range(n + 1)]
needless = [-1, -1]

parent = list(range(n + 1))
for _ in range(n):
    u, v = map(int, input().split())
    
    ru = find(u)
    rv = find(v)
    
    if ru != rv:
        li[u].append(v)
        li[v].append(u)
        union(ru, rv)
    else:
        needless = [u, v]

route = []
min_st = set()
vt = [False] * (n + 1)
dfs_to_find_route(needless[0])

parent = list(range(n + 1))
adj_mid = dict()
vt = [False] * (n + 1)
dfs_to_group(1)

for i in range(1, n + 1):
    find(i)
    
gn_adj_mid = dict() # group num adj mid
for key in adj_mid:
    gn_adj_mid[str(parent[int(key)])] = adj_mid[key]
    
for _ in range(int(input())):
    x, y = map(int, input().split())
    
    temp = (x in min_st) + (y in min_st)
    
    if temp == 2:
        r = 0
    elif temp == 0:
        if gn_adj_mid[str(parent[x])] == gn_adj_mid[str(parent[y])]:
            r = 1
        else:
            r = 0
    else:
        if x in min_st:
            x, y = y, x
        
        rx = parent[x]
        if str(rx) in gn_adj_mid and gn_adj_mid[str(rx)] == y:
            r = 1
        else:
            r = 0
    
    if r:
        print('Yes')
    else:
        print('No')
```

```python
n, p, q, r = map(int, input().split())
li = list(map(int, input().split()))

sli = [0] * (n + 1)
cumul = 0
for i in range(n):
    cumul += li[i]
    sli[i + 1] = cumul
    
start = 1
end = 1
st = {sli[1]}
pur = p + q + r
ip = False
while True:
    sv = sli[end] - sli[start - 1]
    
    if sv < pur:
        end += 1
        if end >= n + 1:
            break
        st.add(sli[end])
    elif sv > pur:
        st.discard(sli[start])
        start += 1
        
        if start > end:
            end = start
            if end >= n + 1:
                break
            st = {sli[end]}
    else:
        if sli[end] - r in st and sli[end] - r - q in st:
            ip = True
            break
        end += 1
        if end >= n + 1:
            break
        st.add(sli[end])
        
if ip:
    print('Yes')
else:
    print('No')
```

```python
mod = 998244353

def check(i, j, k):
    x = a * i + c * j + e * k
    y = b * i + d * j + f * k

    if (x, y) in obs:
        return False
    return True
    
n, m = map(int, input().split())
a, b, c, d, e, f = map(int, input().split())
obs = {tuple(map(int, input().split())) for _ in range(m)}

dp = [[[0] * (n + 1) for _ in range(n + 1)] for _ in range(n + 1)]
dp[0][0][0] = 1

for i in range(n + 1):
    for j in range(n + 1 - i):
        for k in range(n + 1 - i - j):
            if i == j == k == 0 or not check(i, j, k):
                continue
            
            temp = 0
            if i - 1 >= 0:
                temp += dp[i - 1][j][k]
                temp %= mod
            if j - 1 >= 0:
                temp += dp[i][j - 1][k]
                temp %= mod
            if k - 1 >= 0:
                temp += dp[i][j][k - 1]
                temp %= mod
            
            dp[i][j][k] = temp
            
r = 0
for i in range(n + 1):
    for j in range(n + 1 - i):
        for k in range(n + 1 - i - j):
            if i + j + k == n:
                r += dp[i][j][k]
                r %= mod
                
print(r)
```

```python
import sys
input = lambda: sys.stdin.readline().rstrip()

def num_to_bli(num):
    bli = [0] * 30
    bnum = bin(num)
    cur = 0
    for i in range(len(bnum) - 1, 1, -1):
        if bnum[i] == '1':
            bli[cur] = 1
            
        cur += 1
    return bli

def bli_to_num(bli):
    w = 1
    num = 0
    for cur in range(30):
        num += bli[cur] * w
        w *= 2
        
    return num
        
n, c = map(int, input().split())

to_change = [-1] * 30
cur_bli = num_to_bli(c)
for _ in range(n):
    t, a = map(int, input().split())
    
    ch_bli = num_to_bli(a)
    if t == 1:
        for i in range(30):
            if ch_bli[i] == 0:
                to_change[i] = 0
    elif t == 2:
        for i in range(30):
            if ch_bli[i] ==1:
                to_change[i] = 1
    else:
        for i in range(30):
            if ch_bli[i]== 1:
                if to_change[i] in [0, 1]:
                    to_change[i] ^= 1
                elif to_change[i] == -2:
                    to_change[i] = -1
                elif to_change[i] == -1:
                    to_change[i] = -2
                    
    for i in range(30):
        if to_change[i] in [0, 1]:
            cur_bli[i] = to_change[i]
        elif to_change[i] == -2:
            cur_bli[i] ^= 1
            
    num = bli_to_num(cur_bli)
    
    print(num)
```

https://atcoder.jp/contests/abc261/tasks/abc261_f F - Sorting Color Balls

```python
from collections import defaultdict as dd

def merge_sort(li, colors):
    global count
    
    if len(li) == 1:
        return li, colors
    
    left, l_colors = merge_sort(li[:len(li) // 2], colors[:len(li) // 2])
    right, r_colors = merge_sort(li[len(li) // 2:], colors[len(li) // 2:])
    
    nl = len(left)
    nr = len(right)
    
    nli = []
    ncolors = []
    l_i = r_i = 0
    
    total_color_c = 0
    color_cs = dd(int)
    while True:
        if left[l_i] <= right[r_i]:
            nli.append(left[l_i])
            ncolors.append(l_colors[l_i])
            count += total_color_c - color_cs[str(l_colors[l_i])]
            l_i += 1
            
            if l_i == nl:
                while r_i < nr:
                    nli.append(right[r_i])
                    ncolors.append(r_colors[r_i])
                    r_i += 1
                break    
        else:
            nli.append(right[r_i])
            ncolors.append(r_colors[r_i])
            total_color_c += 1
            color_cs[str(r_colors[r_i])] += 1
            r_i += 1
            
            if r_i == nr:
                while l_i < nl:
                    nli.append(left[l_i])
                    ncolors.append(l_colors[l_i])
                    count += total_color_c - color_cs[str(l_colors[l_i])]
                    l_i += 1
                break
                
    return nli, ncolors
    
n = int(input())
colors = list(map(int, input().split()))
li = list(map(int, input().split()))
    
count = 0
merge_sort(li, colors)

print(count)
```

```python
# https://github.com/tatyam-prime/SortedSet/blob/main/SortedMultiset.py
import math
from bisect import bisect_left, bisect_right, insort
from typing import Generic, Iterable, Iterator, TypeVar, Union, List
T = TypeVar('T')

class SortedMultiset(Generic[T]):
    BUCKET_RATIO = 50
    REBUILD_RATIO = 170

    def _build(self, a=None) -> None:
        "Evenly divide `a` into buckets."
        if a is None: a = list(self)
        size = self.size = len(a)
        bucket_size = int(math.ceil(math.sqrt(size / self.BUCKET_RATIO)))
        self.a = [a[size * i // bucket_size : size * (i + 1) // bucket_size] for i in range(bucket_size)]
    
    def __init__(self, a: Iterable[T] = []) -> None:
        "Make a new SortedMultiset from iterable. / O(N) if sorted / O(N log N)"
        a = list(a)
        if not all(a[i] <= a[i + 1] for i in range(len(a) - 1)):
            a = sorted(a)
        self._build(a)

    def __iter__(self) -> Iterator[T]:
        for i in self.a:
            for j in i: yield j

    def __reversed__(self) -> Iterator[T]:
        for i in reversed(self.a):
            for j in reversed(i): yield j
    
    def __len__(self) -> int:
        return self.size
    
    def __repr__(self) -> str:
        return "SortedMultiset" + str(self.a)
    
    def __str__(self) -> str:
        s = str(list(self))
        return "{" + s[1 : len(s) - 1] + "}"

    def _find_bucket(self, x: T) -> List[T]:
        "Find the bucket which should contain x. self must not be empty."
        for a in self.a:
            if x <= a[-1]: return a
        return a

    def __contains__(self, x: T) -> bool:
        if self.size == 0: return False
        a = self._find_bucket(x)
        i = bisect_left(a, x)
        return i != len(a) and a[i] == x

    def count(self, x: T) -> int:
        "Count the number of x."
        return self.index_right(x) - self.index(x)

    def add(self, x: T) -> None:
        "Add an element. / O(√N)"
        if self.size == 0:
            self.a = [[x]]
            self.size = 1
            return
        a = self._find_bucket(x)
        insort(a, x)
        self.size += 1
        if len(a) > len(self.a) * self.REBUILD_RATIO:
            self._build()

    def discard(self, x: T) -> bool:
        "Remove an element and return True if removed. / O(√N)"
        if self.size == 0: return False
        a = self._find_bucket(x)
        i = bisect_left(a, x)
        if i == len(a) or a[i] != x: return False
        a.pop(i)
        self.size -= 1
        if len(a) == 0: self._build()
        return True

    def lt(self, x: T) -> Union[T, None]:
        "Find the largest element < x, or None if it doesn't exist."
        for a in reversed(self.a):
            if a[0] < x:
                return a[bisect_left(a, x) - 1]

    def le(self, x: T) -> Union[T, None]:
        "Find the largest element <= x, or None if it doesn't exist."
        for a in reversed(self.a):
            if a[0] <= x:
                return a[bisect_right(a, x) - 1]

    def gt(self, x: T) -> Union[T, None]:
        "Find the smallest element > x, or None if it doesn't exist."
        for a in self.a:
            if a[-1] > x:
                return a[bisect_right(a, x)]

    def ge(self, x: T) -> Union[T, None]:
        "Find the smallest element >= x, or None if it doesn't exist."
        for a in self.a:
            if a[-1] >= x:
                return a[bisect_left(a, x)]
    
    def __getitem__(self, x: int) -> T:
        "Return the x-th element, or IndexError if it doesn't exist."
        if x < 0: x += self.size
        if x < 0: raise IndexError
        for a in self.a:
            if x < len(a): return a[x]
            x -= len(a)
        raise IndexError

    def index(self, x: T) -> int:
        "Count the number of elements < x."
        ans = 0
        for a in self.a:
            if a[-1] >= x:
                return ans + bisect_left(a, x)
            ans += len(a)
        return ans

    def index_right(self, x: T) -> int:
        "Count the number of elements <= x."
        ans = 0
        for a in self.a:
            if a[-1] > x:
                return ans + bisect_right(a, x)
            ans += len(a)
        return ans

get_value = lambda x: int(x) - 1
    
n, k = map(int, input().split())
li = list(map(get_value, input().split()))

ms = SortedMultiset()
nums = [0] * n
ct_by_nums = [0] * (n + 1)
erased_seq = {}

nxt_num = 1
for i in range(n):
    v = li[i]
    index = ms.index(v)
    num = 0
    if index == len(ms):
        num = nxt_num
        nxt_num += 1
        nums[v] = num
        ct_by_nums[num] = 1
    else:
        nv = ms[index]
        ms.discard(nv)
        num = nums[nv]
        nums[v] = num
        ct_by_nums[num] += 1
        
    if ct_by_nums[num] == k:
        erased_seq[num] = i + 1
    else:
        ms.add(v)
    
for i in range(n):
    num = nums[i]
    if num in erased_seq:
        print(erased_seq[num])
    else:
        print(-1)
```

https://atcoder.jp/contests/abc260/tasks/abc260_e E - At Least One

```python
# 예제 통과를 못함
n, m = map(int, input().split())

included = [[] for _ in range(m + 1)]
for i in range(n):
    a, b = map(int, input().split())
    included[a].append(i)
    included[b].append(i)

deltas = [0] * (m + 2) # deltas for possible length
ct = {}
start = 1
end = 1
for v in included[1]:
    if v not in ct:
        ct[v] = 1
    else:
        ct[v] += 1
        
max_right = -1
while True:
    if not included[end]:
        start = end + 1
        end = start
        if end >= m + 1:
            break
        ct = {}
        for v in included[end]:
            if v not in ct:
                ct[v] = 1
            else:
                ct[v] += 1   
        continue
    
    if len(ct) == n:
        if max_right <= end:
            for max_right in range(end + 1, m + 1):
                if not included[max_right]:
                    max_right -= 1
                    break
        min_len = end - start + 1
        max_len = max_right - start + 1
        deltas[min_len] += 1
        deltas[max_len + 1] -= 1
        
        for nstart in range(start + 1, end + 2):
            for v in included[nstart - 1]:
                ct[v] -= 1
                if ct[v] == 0:
                    del ct[v]
            if len(ct) == n:
                min_len = end - nstart + 1
                max_len = max_right - nstart + 1
                deltas[min_len] += 1
                deltas[max_len + 1] -= 1
            else:
                start = nstart
                break
        if start > end:
            end = start
            if end >= m + 1:
                break
            for v in included[end]:
                if v not in ct:
                    ct[v] = 1
                else:
                    ct[v] += 1
        
    else:
        end += 1
        if end >= m + 1:
            break
        for v in included[end]:
            if v not in ct:
                ct[v] = 1
            else:
                ct[v] += 1
                
    print(start, end, ct)
        
cur = 0
result = [0] * (m + 1)
for i in range(1, m + 1):
    cur += deltas[i]
    result[i] = cur
    
print(*result[1:])
```

https://atcoder.jp/contests/abc259/tasks/abc259_c XX to XXX

```python
def get_ct(s):
    r = []
    cont = 1
    fv = s[0]
    for i in range(1, len(s)):
        if s[i] == fv:
            cont += 1
        else:
            r.append([fv, cont])
            cont = 1
            fv = s[i]
            
    r.append([fv, cont])
    
    return r

s = input()
t = input()

ct_s = get_ct(s)
ct_t = get_ct(t)

ip = True
if len(ct_s) != len(ct_t):
    ip = False
    
if ip:
    for i in range(len(ct_s)):
        if ct_s[i][0] != ct_t[i][0]:
            ip = False
            break
        if ct_s[i][1] == 1 and ct_t[i][1] >= 2:
            ip = False
            break
        if ct_s[i][1] > ct_t[i][1]:
            ip = False
            break
            
if ip:
    print('Yes')
else:
    print('No')
```
