```python
mod = 998244353
max_num = int(3e6)

factos = [0] * (max_num + 1)
factos[0] = 1
invs = [0] * (max_num + 1)
cur = 1
for i in range(1, max_num + 1):
    cur *= i
    cur %= mod
    factos[i] = cur
    
invs[max_num] = pow(factos[max_num], mod - 2, mod)
for i in range(max_num - 1, -1, -1):
    invs[i] = (i + 1) * invs[i + 1] % mod

def nCk(n, k):
    if n < k:
        return 0
    if k < 0:
        return 0
    return factos[n] * invs[k] * invs[n - k] % mod

r, g, b, rg = map(int, input().split())

result = nCk(r, rg)

result *= nCk(r + b, b)
result %= mod

rsr = r - rg # remained solo r
result *= nCk(r + b - rsr + g - rg, g - rg)
result %= mod

print(result)
```

```python
import sys
# input = lambda: sys.stdin.readline().rstrip()

n, q, x = map(int, input().split())
ws = list(map(int, input().split()))
ks = [int(input()) for _ in range(q)]

cumul = 0
sws = [0]
for i in range(n):
    cumul += ws[i]
    sws.append(cumul)
    
sv = sws[n]

temp = max(0, (x - 1) // sv)
base_can_acm = temp * n
nx = x - temp * sv

can_acms = [0] * n # can accumulate
nxt_points = [0] * n
for start in range(n):
    delta_s = 0
    delta_e = n - 1
    result_delta = -1
    
    while delta_s <= delta_e:
        delta_mid = (delta_s + delta_e) // 2
        
        end = start + delta_mid
        end %= n
        
        if start <= end:
            cumul = sws[end + 1] - sws[start]
        else:
            cumul = sws[end + 1] + (sws[n] - sws[start])
            
        if cumul >= nx:
            result_delta = delta_mid
            delta_e = delta_mid - 1
        else:
            delta_s = delta_mid + 1
            
    end_point = (start + result_delta) % n
    
    added_can_acm = result_delta + 1
    can_acm = base_can_acm + added_can_acm
    
    can_acms[start] = can_acm
    nxt_points[start] = (end_point + 1) % n
    
seq = [0]
seq_st = {0}
cur = 0
while True:
    cur = nxt_points[cur]
    
    if cur in seq_st:
        cri = seq.index(cur)
        pre = seq[:cri]
        cycle = seq[cri:]
        break
        
    seq_st.add(cur)
    seq.append(cur)
    
for k in ks:
    k -= 1
    
    if k < cri:
        cur = seq[k]
        r = can_acms[cur]
    else:
        k -= cri
        k %= len(cycle)
        
        cur = cycle[k]
        r = can_acms[cur]
    
    print(r)
```

https://atcoder.jp/contests/abc257/tasks/abc257_f Teleporter Setting

```python
from collections import deque
import sys
input = lambda: sys.stdin.readline().rstrip()

inf = float('inf')

n, m = map(int, input().split())
li = [[] for _ in range(n + 1)]

for _ in range(m):
    u, v = map(int, input().split())
    li[u].append(v)
    li[v].append(u)
    
def get_dists(start):
    dists = [inf] * (n + 1)
    dists[start] = 0
    
    queue = deque()
    queue.append(start)
    while queue:
        value = queue.popleft()
        
        for v in li[value]:
            if dists[v] == inf:
                dists[v] = dists[value] + 1
                queue.append(v)
        
    return dists

dists_1 = get_dists(1)
dists_n = get_dists(n)

result = [0] * n
for target in range(1, n + 1):
    min_r = dists_1[n]
    
    temp_r = min(dists_1[0], dists_1[target]) + min(dists_n[0], dists_n[target])
    min_r = min(min_r, temp_r)
    
    if min_r == inf:
        min_r = -1
        
    result[target - 1] = min_r
        
print(*result)
```

```python
import sys
input = lambda: sys.stdin.readline().rstrip()

## PYRIVAL BOOTSTRAP
# https://github.com/cheran-senthil/PyRival/blob/master/pyrival/misc/bootstrap.py
# This decorator allows for recursion without actually doing recursion
## @bootstrap, yield when getting and returning value in recursive functions, end of functions
from types import GeneratorType

def bootstrap(f, stack=[]):
    def wrappedfunc(*args, **kwargs):
        if stack:
            return f(*args, **kwargs)
        else:
            to = f(*args, **kwargs)
            while True:
                if type(to) is GeneratorType:
                    stack.append(to)
                    to = next(to)
                else:
                    stack.pop()
                    if not stack:
                        break
                    to = stack[-1].send(to)
            return to
    return wrappedfunc

# https://www.techiedelight.com/construct-binary-tree-from-inorder-preorder-traversal/

# A class to store a binary tree node
class Node:
    # Constructor
    def __init__(self, data, left=None, right=None):
        self.data = data
        self.left = left
        self.right = right
 
 
# Recursive function to perform inorder traversal on a given binary tree
@bootstrap
def inorderTraversal(root):
    if root is None:
        yield
 
    yield inorderTraversal(root.left)
    temp.append(root.data)
    yield inorderTraversal(root.right)
    
    yield
 
 
# Recursive function to perform postorder traversal on a given binary tree
@bootstrap
def preorderTraversal(root):
    if root is None:
        yield
 
    temp.append(root.data)
    yield preorderTraversal(root.left)
    yield preorderTraversal(root.right)
    
    yield
 
 
# Recursive function to construct a binary tree from a given
# inorder and preorder sequence
@bootstrap
def construct(start, end, preorder, pIndex, d):
    # base case
    if start > end:
        yield None, pIndex
 
    # The next element in `preorder[]` will be the root node of subtree
    # formed by sequence represented by `inorder[start, end]`
    root = Node(preorder[pIndex])
    pIndex = pIndex + 1
 
    # get the index of the root node in inorder to determine the
    # left and right subtree boundary
    index = d[root.data]
 
    # recursively construct the left subtree
    root.left, pIndex = yield construct(start, index - 1, preorder, pIndex, d)
 
    # recursively construct the right subtree
    root.right, pIndex = yield construct(index + 1, end, preorder, pIndex, d)
 
    # return current node
    yield root, pIndex
    

# Construct a binary tree from inorder and preorder traversals.
# This function assumes that the input is valid
# i.e., given inorder and preorder sequence forms a binary tree
def constructTree(inorder, preorder):
 
    # create a dictionary to efficiently find the index of any element in
    # a given inorder sequence
    d = {}
    for i, e in enumerate(inorder):
        d[e] = i
 
    # `pIndex` stores the index of the next unprocessed node in a preorder sequence;
    # start with the root node (present at 0th index)
    pIndex = 0
 
    return construct(0, len(inorder) - 1, preorder, pIndex, d)[0]
 
@bootstrap
def dfs(value):
    if value.left:
        result[value.data - 1][0] = value.left.data
        yield dfs(value.left)
    
    if value.right:
        result[value.data - 1][1] = value.right.data
        yield dfs(value.right)
        
    yield

n = int(input())
preorder = list(map(int, input().split()))
inorder = list(map(int, input().split()))

ip = True
try:
    root = constructTree(inorder, preorder)
except:
    ip = False

if ip:
    temp = []
    preorderTraversal(root)
    if temp != preorder:
        ip = False
if ip:
    temp = []
    inorderTraversal(root)
    if temp != inorder:
        ip = False
if ip and root.data != 1:
    ip = False

if not ip:
    print(-1)
else:
    result = [[0, 0] for _ in range(n)]
    dfs(root)
    for v in result:
        print(*v)
```

```python
import sys
input = lambda: sys.stdin.readline().rstrip()

## PYRIVAL BOOTSTRAP
# https://github.com/cheran-senthil/PyRival/blob/master/pyrival/misc/bootstrap.py
# This decorator allows for recursion without actually doing recursion
## @bootstrap, yield when getting and returning value in recursive functions, end of functions
from types import GeneratorType

def bootstrap(f, stack=[]):
    def wrappedfunc(*args, **kwargs):
        if stack:
            return f(*args, **kwargs)
        else:
            to = f(*args, **kwargs)
            while True:
                if type(to) is GeneratorType:
                    stack.append(to)
                    to = next(to)
                else:
                    stack.pop()
                    if not stack:
                        break
                    to = stack[-1].send(to)
            return to
    return wrappedfunc

# https://www.techiedelight.com/construct-binary-tree-from-inorder-preorder-traversal/

# Recursive function to perform inorder traversal on a given binary tree
@bootstrap
def inorderTraversal(root):
    if root == 0:
        yield
 
    yield inorderTraversal(li[root][0])
    temp.append(root)
    yield inorderTraversal(li[root][1])
 
    yield

# Recursive function to perform postorder traversal on a given binary tree
@bootstrap
def preorderTraversal(root):
    if root == 0:
        yield
 
    temp.append(root)
    yield preorderTraversal(li[root][0])
    yield preorderTraversal(li[root][1])
 
    yield

# Recursive function to construct a binary tree from a given
# inorder and preorder sequence
@bootstrap
def construct(start, end, preorder, pIndex, d):
    # base case
    if start > end:
        yield (0, pIndex)
 
    # The next element in `preorder[]` will be the root node of subtree
    # formed by sequence represented by `inorder[start, end]`
    root = preorder[pIndex]
    pIndex = pIndex + 1
 
    # get the index of the root node in inorder to determine the
    # left and right subtree boundary
    index = d[root]
 
    # recursively construct the left subtree
    li[root][0], pIndex = yield construct(start, index - 1, preorder, pIndex, d)
 
    # recursively construct the right subtree
    li[root][1], pIndex = yield construct(index + 1, end, preorder, pIndex, d)
 
    # return current node
    yield (root, pIndex)
 

# Construct a binary tree from inorder and preorder traversals.
# This function assumes that the input is valid
# i.e., given inorder and preorder sequence forms a binary tree
def constructTree(inorder, preorder):
 
    # create a dictionary to efficiently find the index of any element in
    # a given inorder sequence
    d = {}
    for i, e in enumerate(inorder):
        d[e] = i
 
    # `pIndex` stores the index of the next unprocessed node in a preorder sequence;
    # start with the root node (present at 0th index)
    pIndex = 0
 
    return construct(0, len(inorder) - 1, preorder, pIndex, d)[0]

n = int(input())
preorder = list(map(int, input().split()))
inorder = list(map(int, input().split()))

li = [[0, 0] for _ in range(n + 1)]

ip = True
try:
    root = constructTree(inorder, preorder)
except:
    ip = False

if ip:
    temp = []
    preorderTraversal(root)
    if temp != preorder:
        ip = False
if ip:
    temp = []
    inorderTraversal(root)
    if temp != inorder:
        ip = False
if ip and root != 1:
    ip = False

if not ip:
    print(-1)
else:
    for v in li[1:]:
        print(*v)
```

```python
from collections import defaultdict as dd

class SegmentTree:
    def __init__(self, data, default=10**15, func=lambda a, b: max(a,b)):
        """initialize the segment tree with data"""
        self._default = default
        self._func = func
        self._len = len(data)
        self._size = _size = 1 << (self._len - 1).bit_length()
 
        self.data = [default] * (2 * _size)
        self.data[_size:_size + self._len] = data
        for i in reversed(range(_size)):
            self.data[i] = func(self.data[i + i], self.data[i + i + 1])
 
    def __delitem__(self, idx):
        self[idx] = self._default
 
    def __getitem__(self, idx):
        return self.data[idx + self._size]
 
    def __setitem__(self, idx, value):
        idx += self._size
        self.data[idx] = value
        idx >>= 1
        while idx:
            self.data[idx] = self._func(self.data[2 * idx], self.data[2 * idx + 1])
            idx >>= 1
 
    def __len__(self):
        return self._len
 
    def query(self, start, stop):
        if start == stop:
            return self.__getitem__(start)
        stop += 1
        start += self._size
        stop += self._size
 
        res = self._default
        while start < stop:
            if start & 1:
                res = self._func(res, self.data[start])
                start += 1
            if stop & 1:
                stop -= 1
                res = self._func(res, self.data[stop])
            start >>= 1
            stop >>= 1
        return res
 
    def __repr__(self):
        return "SegmentTree({0})".format(self.data)
    
func = lambda a, b: a + b

n, m, q = map(int, input().split())
qus = [list(map(int, input().split())) for _ in range(q)]

segtree = SegmentTree([0] * (m + 2), 0, func=func)
waits = dd(list)
result = [0] * q 
for qi in range(q - 1, -1, -1):
    tp = qus[qi][0]
    
    if tp == 1:
        l, r, x = qus[qi][1:]
        segtree[l] += x
        segtree[r + 1] -= x
    elif tp == 3:
        i, j = qus[qi][1:]
        
        sv = segtree.query(0, j) - segtree.query(0, j - 1)
        waits[str(i)].append((sv, j, qi))
    elif tp == 2:
        i, x = qus[qi][1:]
        for sv, j, qi in waits[str(i)]:
            fsv = segtree.query(0, j) - segtree.query(0, j - 1)
            result[qi] = x + sv - fsv
        waits[str(i)].clear()
        
for key in waits:
    if waits[key]:
        for sv, j, qi in waits[key]:
            fsv = segtree.query(0, j) - segtree.query(0, j - 1)
            result[qi] = sv - fsv

for v in result:
    if v:
        print(v)
```

```python
import sys
input = lambda: sys.stdin.readline().rstrip()

class Node():
    def __init__(self, parent, value):
        self.value = value
        self.parent = parent
    
    def append_and_move_to_node(self, value):
        node = Node(self, value)
        return node

q = int(input())
qus = [input().split() for _ in range(q)]

pages = {}
cur = Node(None, -1)
result = []
for qi in range(q):
    if qus[qi][0] == 'DELETE':
        parent = cur.parent
        
        if not parent:
            r = -1
        else:
            cur = parent
            r = cur.value
    else:
        tp = qus[qi][0]
        value = qus[qi][1]
        
        if tp == 'ADD':
            cur = cur.append_and_move_to_node(value)
            r = cur.value
        elif tp == 'SAVE':
            pages[value] = cur
            r = cur.value
        elif tp == 'LOAD':
            if value not in pages:
                cur = Node(None, -1)
            else:
                cur = pages[value]
            r = cur.value
            
    result.append(r)
    
print(*result)
```
