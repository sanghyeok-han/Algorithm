```python
n, m = map(int, input().split())

if m == 1:
    print('9' * n)
else:
    st = set()
    cur = m
    for _ in range(100):
        st.add(int(str(cur)[-1]))
        cur *= m

    st.discard(0)

    max_r = -1
    for num in st:
        if num % m == 0:
            max_r = max(max_r, num)

        nnum = num
        target = -1
        for _ in range(n - 1):
            nnum *= 10
            nnum += num

            if nnum % m == 0:
                target = nnum
                break

        if target != -1:
            s_target = str(target)
            length = len(s_target)

            itered = n // length

            temp_r = str(target) * itered
            max_r = max(max_r, int(temp_r))

    print(max_r)
```

```python
from bisect import bisect_left

def get_lis(li):
    nli = [li[0]]
    for i in range(1, len(li)):
        index = bisect_left(nli, li[i])

        if index == len(nli):
            nli.append(li[i])
        else:
            nli[index] = li[i]

    return len(nli)

n = int(input())

a = list(map(int, input().split()))
b = list(map(int, input().split()))

li = [(a[i], b[i]) for i in range(n)]

li.sort(key=lambda x: x[0])

a = [li[i][0] for i in range(n)]
b = [li[i][1] for i in range(n)]

t1 = n + get_lis(b)

li.sort(key=lambda x: x[1])

a = [li[i][0] for i in range(n)]
b = [li[i][1] for i in range(n)]

t2 = get_lis(a) + n

r = max(t1, t2)

print(r)
```

```python
# 문제를 잘못 해석하였음
max_num = 1000001
ip = [True] * (max_num + 1)
ip[0] = ip[1] = False
for mod in range(2, int((max_num + 1) ** 0.5) + 1):
    if ip[mod]:
        for v in range(mod * mod, max_num + 1, mod):
            ip[v] = False
            
primes = {i for i, v in enumerate(ip) if v}

n = int(input())

pa = [[0] * n for _ in range(n)]

c0 = c1 = (n * n) // 2
if n % 2 == 1:
    c1 += 1
    
for i in range(0, n, 2):
    for j in range(0, n, 2):
        if i + 1 < n and j + 1 < n:
            if c1 >= 4:
                pa[i][j] = pa[i + 1][j] = pa[i][j + 1] = pa[i + 1][j + 1] = 1
                c1 -= 4
            elif c0 >= 4:
                pa[i][j] = pa[i + 1][j] = pa[i][j + 1] = pa[i + 1][j + 1] = 2
                c0 -= 4

# print(c0, c1)
# for v in pa:
#     print(v)

r = [[0] * n for _ in range(n)]
if n % 4 == 0:
    cur0 = 2
    cur1 = 1
    for i in range(n):
        for j in range(n):
            if pa[i][j] == 2:
                r[i][j] = cur0
                cur0 += 2
            else:
                r[i][j] = cur1
                cur1 += 2
elif n % 2 == 1:
    for i in range(n - 1):
        pa[i][n - 1] = 2
        pa[n - 1][i] = 2
        c0 -= 2
        
    pa[n - 1][n - 1] = 1
    c1 -= 1
   
    cur0 = 2
    for i in range(n):
        for j in range(n):
            if pa[i][j] == 2:
                r[i][j] = cur0
                cur0 += 2
                
    t1 = sum(r[n - 1])
    t2 = 0
    for i in range(n):
        t2 += r[i][n - 1]
        
    to_use1 = set(range(1, n * n + 1, 2))
    for cand in to_use1:
        if t1 + cand not in primes and t2 + cand not in primes:
            r[n - 1][n - 1] = cand
            to_use1.discard(cand)
            break       
    
    for i in range(n):
        for j in range(n):
            if r[i][j] == 0:
                r[i][j] = to_use1.pop()
else:
    pa[n - 2][n - 2] = 2
    pa[n - 2][n - 1] = 2
    c0 -= 2
    
    cur0 = 2
    for i in range(n):
        for j in range(n):
            if pa[i][j] == 2:
                r[i][j] = cur0
                cur0 += 2
                
    cur1 = 1
    for i in range(n - 1):
        for j in range(n - 2, n):
            if pa[i][j] == 1:
                r[i][j] = cur1
                cur1 += 2
                
    t1 = 0
    t2 = 0
    for i in range(n):
        t1 += r[i][n - 2]
        t2 += r[i][n - 1]
        
    to_use1 = set(range(cur1, n * n + 1, 2))
    for cand in to_use1:
        if t1 + cand not in primes:
            r[n - 1][n - 2] = cand
            to_use1.discard(cand)
            break    
            
    for cand in to_use1:
        if t1 + cand not in primes:
            r[n - 1][n - 1] = cand
            to_use1.discard(cand)
            break
            
    for i in range(n):
        for j in range(n):
            if r[i][j] == 0:
                r[i][j] = to_use1.pop()
                
for v in r:
    print(*v)
```

```python
n = int(input())

pa = [[0] * n for _ in range(n)]

c1 = n * n // 2
if n % 2 == 1:
    c1 += 1

for i in range(n):
    for j in range(n):
        if c1:
            pa[i][j] = 1
            c1 -= 1
        else:
            pa[i][j] = 0

r = [[0] * n for _ in range(n)]
to_use1 = set(range(1, n * n + 1, 2))
to_use0 = set(range(2, n * n + 1, 2))

if n == 3:
    print(1, 3, 9)
    print(5, 7, 8)
    print(4, 2, 6)
elif n % 2 == 1:
    for i in range(1, n):
        for j in range(1, n):
            if pa[i][j] == 0:
                if pa[i - 1][j] == 1 and pa[i][j - 1] == 1:
                    r[i - 1][j] = n * 3
                    r[i][j - 1] = n
                    r[i][j] = n * 2

                    to_use1.discard(n * 3)
                    to_use1.discard(n)
                    to_use0.discard(n * 2)

    if n % 2 == 1 and n * n in to_use1:
        r[0][0] = n * n
        to_use1.discard(n * n)
        
    for i in range(n):
        for j in range(n):
            if r[i][j] == 0:
                if pa[i][j] == 1:
                    r[i][j] = to_use1.pop()
                else:
                    if pa[i - 1][j] == 1:
                        r[i][j] = n * n - r[i - 1][j]
                        to_use0.discard(n * n - r[i - 1][j])
                    else:
                        r[i][j] = to_use0.pop()
                        
    for v in r:
        print(*v)
else:
    r[0][0] = n * n - 1
    to_use1.discard(n * n - 1)
    
    r[n - 1][0] = n * n
    to_use0.discard(n * n)
    
    for i in range(n):
        for j in range(n):
            if r[i][j] == 0:
                if pa[i][j] == 1:
                    r[i][j] = to_use1.pop()
                else:
                    if pa[i - 1][j] == 1:
                        r[i][j] = n * n - 1 - r[i - 1][j]
                        to_use0.discard(n * n - 1 - r[i - 1][j])
                    else:
                        r[i][j] = to_use0.pop()    
                    
    for v in r:
        print(*v)
```
