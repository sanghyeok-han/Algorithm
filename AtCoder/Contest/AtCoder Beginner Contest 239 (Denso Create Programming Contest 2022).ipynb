```python
deltas = [(-2, 1), (-2, -1), (-1, 2), (-1, -2), (1, 2), (1, -2), (2, 1), (2, -1)]

x1, y1, x2, y2 = map(int, input().split())

st1 = set()
for dx, dy in deltas:
    nx = x1 + dx
    ny = y1 + dy
    
    st1.add((nx, ny))
    
st2 = set()
for dx, dy in deltas:
    nx = x2 + dx
    ny = y2 + dy
    
    st2.add((nx, ny))
    
if len(st1 & st2) >= 1:
    print('Yes')
else:
    print('No')
```

```python
Takahashi가 부를 수 있는 모든 수에 Aoki가 prime을 만들 수 있는 수가 있다면 Aoke가 이기고 그렇지 않으면 Takahashi가 이김
```

```python
n = 300
li = [True] * (n + 1)
li[0] = li[1] = False

for i in range(2, int(n ** (1/2) + 1)):
    if li[i]:
        for j in range(i * i, n + 1, i):
            li[j] = False
            
primes = {i for i, v in enumerate(li) if v}

a, b, c, d = map(int, input().split())

all_is_prime = True
for i in range(a, b + 1):
    can_make_prime = False
    for j in range(c, d + 1):
        if i + j in primes:
            can_make_prime = True
            break
    if not can_make_prime:
        all_is_prime = False
        break
        
if not all_is_prime:
    print('Takahashi')
else:
    print('Aoki')
```
