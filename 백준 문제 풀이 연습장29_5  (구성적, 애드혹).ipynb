### 네모네모 시력검사 18242

```python
n, m = map(int, input().split())
li = [input() for _ in range(n)]

xs, xe = -1, -1
for i in range(n):
    for j in range(m):
        if li[i][j] == '#':
            if xs == -1:
                xs = i
            xe = i
            
ys, ye = -1, -1
for j in range(m):
    for i in range(n):
        if li[i][j] == '#':
            if ys == -1:
                ys = j
            ye = j
    
is_end = False
for j in range(1, m - 1):
    if li[xs][j - 1] == '#' and li[xs][j] == '.' and li[xs][j + 1] == '#':
        print('UP')
        is_end = True
        break
        
if not is_end:
    for j in range(1, m - 1):
        if li[xe][j - 1] == '#' and li[xe][j] == '.' and li[xe][j + 1] == '#':
            print('DOWN')
            is_end = True
            break
            
if not is_end:
    for i in range(1, n - 1):
        if li[i - 1][ys] == '#' and li[i][ys] == '.' and li[i + 1][ys] == '#':
            print('LEFT')
            is_end = True
            break
            
if not is_end:
    for i in range(1, n - 1):
        if li[i - 1][ye] == '#' and li[i][ye] == '.' and li[i + 1][ye] == '#':
            print('RIGHT')
            is_end = True
            break
```
