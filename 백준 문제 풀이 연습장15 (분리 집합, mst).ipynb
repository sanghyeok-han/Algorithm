### 군사 이동 11085
- 그래프 이론, 분리 집합

```python
from collections import deque
import sys
# input = sys.stdin.readline

p, w = map(int, input().strip().split())
start, target = map(int, input().strip().split())

li = [[] for _ in range(p)]
for _ in range(w):
    a, b, w = map(int, input().strip().split())
    li[a].append((b, w))
    li[b].append((a, w))

maximum_arrivals = [0] * p
maximum_arrivals[start] = float('inf')

queue = deque()
queue.append((start, float('inf')))

while queue:
    node, max_possible = queue.popleft()
    
    for v in li[node]:
        n, mp = v
        
        mp = min(max_possible, mp)
        if mp > maximum_arrivals[n]:
            maximum_arrivals[n] = mp
            queue.append((n, mp))
            
print(maximum_arrivals[target])
```

```python
만약 x 또는 y가 1~n 범위를 벗어나면 무시함

eaten_by와 same_type 두 가지를 관리함

1번 정보가 들어올시 same_type로 묶어줘야하는데 이때 둘 사이에 eaten_by 관계가 있다면 안됨

2번 정보가 들어올시 eaten_by로 묶어줘야하는데 이때 잡아먹히는 쪽에 잡아먹는 쪽의 same_type parent를 설정해줌.
또한 이때 서로 eaten_by 관계에 있거나 둘이 same_type 관계에 있으면 안됨

eaten_by를 싹 다 갱신해준 후 종류가 3개를 넘으면 안됨
```

```python
import sys
sys.setrecursionlimit(500000)
# input = lambda: sys.stdin.readline().rstrip()

n, k = map(int, input().split())

def find(v):
    if v == 0:
        return 0
    if parent[v] == v:
        return v
    parent[v] = find(parent[v])
    return parent[v]

def union(rx, ry):
    if rx == 0 and ry == 0:
        return 0
    elif rx == 0 and ry != 0:
        return 'right'
    elif rx != 0 and ry == 0:
        return 'left'
    else: 
        if rank[rx] >= rank[ry]:
            if rank[rx] == rank[ry]:
                rank[rx] += 1
            parent[ry] = rx
            return 'left'
        else:
            parent[rx] = ry
            return 'right'

def update(rx, ry):
    can_eat[rx], can_eat[ry] = find(can_eat[rx]), find(can_eat[ry])
    eaten_by[rx], eaten_by[ry] = find(eaten_by[rx]), find(eaten_by[ry])
    
    can_eat[can_eat[rx]], can_eat[can_eat[ry]] = find(can_eat[can_eat[rx]]), find(can_eat[can_eat[ry]])
    can_eat[eaten_by[rx]], can_eat[eaten_by[ry]] = find(can_eat[eaten_by[rx]]), find(can_eat[eaten_by[ry]])
    eaten_by[can_eat[rx]], eaten_by[can_eat[ry]] = find(eaten_by[can_eat[rx]]), find(eaten_by[can_eat[ry]])
    eaten_by[eaten_by[rx]], eaten_by[eaten_by[ry]] = find(eaten_by[eaten_by[rx]]), find(eaten_by[eaten_by[ry]])
    
        
def same_check(rx, ry):
    if can_eat[rx] == ry or can_eat[ry] == rx:
        return False
    if eaten_by[rx] == ry or eaten_by[ry] == rx:
        return False
    return True
    
def same(rx, ry):
    union(rx, ry)
    
    r = union(can_eat[rx], can_eat[ry])
    if r == 'left':
        can_eat[ry] = can_eat[rx]
    elif r == 'right':
        can_eat[rx] = can_eat[ry]
        
    r = union(eaten_by[rx], eaten_by[ry])
    if r == 'left':
        eaten_by[ry] = eaten_by[rx]
    elif r == 'right':
        eaten_by[rx] = eaten_by[ry]
        
    eat(can_eat[ry], eaten_by[rx])
    eat(can_eat[rx], eaten_by[ry])         
    
def eat_check(rx, ry):
    if rx == ry or can_eat[ry] == rx or eaten_by[rx] == ry:
        return False
    if eaten_by[rx] != 0 and can_eat[ry] != 0:
        nx, ny = eaten_by[rx], can_eat[ry]
        if eaten_by[nx] == ny or nx == can_eat[ny]:
            return False
    return True
    
def eat(rx, ry):
    r = union(can_eat[rx], ry)
    if r == 'right':
        can_eat[rx] = ry
    r = union(eaten_by[ry], rx)
    if r == 'right':
        eaten_by[ry] = rx
    r = union(eaten_by[rx], can_eat[ry])
    if r == 'left':
        can_eat[ry] = eaten_by[rx]
    elif r == 'right':
        eaten_by[rx] = can_eat[ry]
    
parent = list(range(n + 1))
rank = [0] * (n + 1)
can_eat = [0] * (n + 1)
eaten_by = [0] * (n + 1)

result = 0
for _ in range(k):
    t, x, y = map(int, input().split())
    if x < 1 or x > n or y < 1 or y > n:
        result += 1
        continue
    rx, ry = find(x), find(y)
    update(rx, ry)
    if t == 1:
        if same_check(rx, ry):
            same(rx, ry)
        else:
            result += 1
    else:
        if eat_check(rx, ry):
            eat(rx, ry)
        else:
            result += 1
print(result)
```

```python
from collections import deque

n, k = map(int, input().split())

edge = []
for i in range(k):
    a, b, c = map(int, input().split())
    edge.append((c, a, b))
    
edge.sort()

parent = [0] * n
for i in range(n):
    parent[i] = i
    

def find_parent(x):
    if parent[x] != x:
        parent[x] = find_parent(parent[x])
        
    return parent[x]

def union(x, y):
    x = find_parent(x)
    y = find_parent(y)
    
    if x < y:
        parent[y] = x
    else:
        parent[x] = y
      
    
    

lst = []
answer = 0
for cost, a, b in edge:
    if find_parent(a) != find_parent(b):
        union(a, b)
        answer += cost
        
        lst.append((cost, a, b))

course = [[] for _ in range(n)]
for i in lst:
    course[i[1]].append((i[2], i[0]))
    course[i[2]].append((i[1], i[0]))
    

node = [i for i in range(n)]
maxi = 0
for i in node:
    queue = deque([i])
    visited = [0] * n
    dist = [0] * n
    
    while queue:
        x = queue.popleft()
        
        for nxt, cost in course[x]:
            if visited[nxt] == 0:
                visited[nxt] = 1
                dist[nxt] = cost
                queue.append(nxt)
                
                
    maxi = max(maxi, max(dist))
            
        
print(answer)
print(maxi)
```

### 안정적인 네트워크 2406

```python
import sys
# input = sys.stdin.readline

def find(v):
    if parent[v] == v:
        return v
    parent[v] = find(parent[v])
    return parent[v]

def union(a, b):
    if rank[a] >= rank[b]:
        if rank[a] == rank[b]:
            rank[a] += 1
        parent[b] = a
    else:
        parent[a] = b

n, m = map(int, input().rstrip().split())

parent = list(range(n))
rank = [0] * n

for _ in range(m):
    a, b = map(int, input().rstrip().split())
    a -= 1
    b -= 1
    
    a = find(a)
    b = find(b)  
    if a != b:
        union(a, b)
    
costs = [list(map(int, input().rstrip().split())) for _ in range(n)]
edges = []
for i in range(1, n):
    for j in range(i + 1, n):
        edges.append((i, j, costs[i][j]))
edges.sort(key=lambda x: x[2])

total = 0
count = 0
added = []
for a, b, c in edges:
    ra = find(a)
    rb = find(b)  
    if ra != rb:
        union(ra, rb)
        total += c
        count += 1
        added.append(f'{a + 1} {b + 1}')
        
print(total, count)
print('\n'.join(added))
```
