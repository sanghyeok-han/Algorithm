https://www.acmicpc.net/problem/6515 Frequent values

```python
import io, os, sys
input = io.BytesIO(os.read(0,os.fstat(0).st_size)).readline
print = sys.stdout.write

inf = float('inf')

def add(i):
    global max_c
    
    v = li[i]
    ct_v = ct[v]
    
    m_ct[ct_v] -= 1

    ct_v += 1
    m_ct[ct_v] += 1
    if ct_v > max_c:
        max_c = ct_v
        
    ct[v] += 1
    
def sub(i):
    global max_c
    
    v = li[i]
    ct_v = ct[v]
    
    m_ct[ct_v] -= 1
    if ct_v == max_c and m_ct[ct_v] == 0:
        max_c -= 1
    
    ct_v -= 1
    m_ct[ct_v] += 1
    
    ct[v] -= 1

get_added_value = lambda x: int(x) + 100000
get_value = lambda x: int(x) - 1

while True:
    inp = list(map(int, input().split()))
    if inp[0] == 0:
        break
        
    n, m = inp
    li = list(map(get_added_value, input().split()))
    qu = [list(map(get_value, input().split())) + [i] for i in range(m)]

    sn = int(n ** 0.5)
    qu.sort(key=lambda x: (x[0] // sn, x[1]))

    ct = [0] * 200001
    m_ct = [0] * 100001 # meta
    max_c = 0

    result = [-1] * m

    s, e, o_qi = qu[0]
    for i in range(s, e + 1):
        add(i)

    result[o_qi] = max_c

    fs, fe = s, e
    for qi in range(1, m):
        s, e, o_qi = qu[qi]

        if fs < s:
            for i in range(fs, s):
                sub(i)
        elif fs > s:
            for i in range(fs - 1, s - 1, -1):
                add(i)

        if fe < e:
            for i in range(fe + 1, e + 1):
                add(i)
        elif fe > e:
            for i in range(fe, e, -1):
                sub(i)

        result[o_qi] = max_c

        fs, fe = s, e

    print('\n'.join(map(str, result)))
    print('\n')
```
