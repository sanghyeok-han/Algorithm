### 신년 파티 1623

```python
import sys
input = lambda: sys.stdin.readline().rstrip()

## PYRIVAL BOOTSTRAP
# https://github.com/cheran-senthil/PyRival/blob/master/pyrival/misc/bootstrap.py
# This decorator allows for recursion without actually doing recursion
## @bootstrap, yield when getting and returning value in recursive functions, end of functions
from types import GeneratorType

def bootstrap(f, stack=[]):
    def wrappedfunc(*args, **kwargs):
        if stack:
            return f(*args, **kwargs)
        else:
            to = f(*args, **kwargs)
            while True:
                if type(to) is GeneratorType:
                    stack.append(to)
                    to = next(to)
                else:
                    stack.pop()
                    if not stack:
                        break
                    to = stack[-1].send(to)
            return to
    return wrappedfunc

@bootstrap
def dfs(value):
    global ic_max
    global nic_max
    
    ic = ws[value]# inlcluded
    nic = 0 # not included
    for v in li[value]:
        ic += (yield dfs(v))
        nic += dp[v]
            
    if value == 0:
        ic_max = ic
        nic_max = nic
            
    if ic >= nic:
        used[value] = True
        dp[value] = ic
    else:
        dp[value] = nic
        
    yield nic

@bootstrap
def retrace(value, can_use):
    global r
    
    if can_use and (used[value] or value == 0):
        r.append(value + 1)
        for v in li[value]:
            yield retrace(v, False)
    else:
        for v in li[value]:
            yield retrace(v, True)
            
    yield

n = int(input())
ws = list(map(int, input().split()))

get_value = lambda x: int(x) - 1
parents = [-1] + list(map(get_value, input().split()))

li = [[] for _ in range(n)]
for cn in range(1, n):
    pn = parents[cn]
    li[pn].append(cn)
    
used = [False] * n
dp = [0] * n
ic_max = 0
nic_max = 0
dfs(0)
          
print(ic_max, nic_max)
    
r = []
retrace(0, True)

r.sort()
r.append(-1)
print(*r)

r = []
retrace(0, False)

r.sort()
r.append(-1)
print(*r)
```
