```python
import sys
input = lambda: sys.stdin.readline().rstrip()

## PYRIVAL BOOTSTRAP
# https://github.com/cheran-senthil/PyRival/blob/master/pyrival/misc/bootstrap.py
# This decorator allows for recursion without actually doing recursion
## @bootstrap, yield when getting and returning value in recursive functions, end of functions
from types import GeneratorType

def bootstrap(f, stack=[]):
    def wrappedfunc(*args, **kwargs):
        if stack:
            return f(*args, **kwargs)
        else:
            to = f(*args, **kwargs)
            while True:
                if type(to) is GeneratorType:
                    stack.append(to)
                    to = next(to)
                else:
                    stack.pop()
                    if not stack:
                        break
                    to = stack[-1].send(to)
            return to
    return wrappedfunc

@bootstrap
def dfs(value):
    global ic_max
    global nic_max
    
    ic = ws[value]# inlcluded
    nic = 0 # not included
    for v in li[value]:
        ic += (yield dfs(v))
        nic += dp[v]
            
    if value == 0:
        ic_max = ic
        nic_max = nic
            
    if ic >= nic:
        used[value] = True
        dp[value] = ic
    else:
        dp[value] = nic
        
    yield nic

@bootstrap
def retrace(value, can_use):
    global r
    
    if can_use and (used[value] or value == 0):
        r.append(value + 1)
        for v in li[value]:
            yield retrace(v, False)
    else:
        for v in li[value]:
            yield retrace(v, True)
            
    yield

n = int(input())
ws = list(map(int, input().split()))

get_value = lambda x: int(x) - 1
parents = [-1] + list(map(get_value, input().split()))

li = [[] for _ in range(n)]
for cn in range(1, n):
    pn = parents[cn]
    li[pn].append(cn)
    
used = [False] * n
dp = [0] * n
ic_max = 0
nic_max = 0
dfs(0)
          
print(ic_max, nic_max)
    
r = []
retrace(0, True)

r.sort()
r.append(-1)
print(*r)

r = []
retrace(0, False)

r.sort()
r.append(-1)
print(*r)
```

```python
from heapq import heappush, heappop
import sys
sys.setrecursionlimit(110000)
input = lambda: sys.stdin.readline().rstrip()

inf = float('inf')

def recur(v):
    if dp[v] != -1:
        return dp[v]
    
    c = 0
    for value, _ in li[v]:
        if dists[value] > dists[v]:
            recur(value)
            c += dp[value]
            
    dp[v] = c

n, m = map(int, input().split())

li = [[] for _ in range(n + 1)]
for _ in range(m):
    a, b, c = map(int, input().split())
    li[a].append((b, c))
    li[b].append((a, c))
    
s, t = 1, 2
dists = [inf] * (n + 1)
dists[t] = 0

heap = []
heappush(heap, (0, t))

while heap:
    dist, value = heappop(heap)
    
    if dist > dists[value]:
        continue
        
    for v, d in li[value]:
        ndist = dist + d
        if ndist < dists[v]:
            dists[v] = ndist
            heappush(heap, (ndist, v))
            
dp = [-1] * (n + 1)
dp[s] = 1

recur(t)

r = dp[t]

print(r)
```

```python
from heapq import heappush, heappop
from collections import defaultdict as dd
import sys
input = lambda: sys.stdin.readline().rstrip()

inf = float('inf')

n, m = map(int, input().split())
li = [input() for _ in range(n)]

edges = dd(list)
for i in range(n):
    for j in range(m):
        c = int(li[i][j] == '/')
            
        edges[(i, j)].append((i + 1, j + 1, c))
        edges[(i + 1, j + 1)].append((i, j, c))
        
        c ^= 1
        
        edges[(i, j + 1)].append((i + 1, j, c))
        edges[(i + 1, j)].append((i, j + 1, c))        
        
costs = [[inf] * (m + 1) for _ in range(n + 1)]
costs[0][0] = 0

heap = []
heappush(heap, (0, 0, 0))

while heap:
    cost, x, y = heappop(heap)
    
    if cost > costs[x][y]:
        continue
        
    for nx, ny, c in edges[(x, y)]:
        ncost = cost + c
        if ncost < costs[nx][ny]:
            costs[nx][ny] = ncost
            heappush(heap, (ncost, nx, ny))
            
r = costs[n][m]

if r == inf:
    print('NO SOLUTION')
else:
    print(r)
```

```python
import sys
input = lambda: sys.stdin.readline().rstrip()

## PYRIVAL BOOTSTRAP
# https://github.com/cheran-senthil/PyRival/blob/master/pyrival/misc/bootstrap.py
# This decorator allows for recursion without actually doing recursion
## @bootstrap, yield when getting and returning value in recursive functions, end of functions
from types import GeneratorType

def bootstrap(f, stack=[]):
    def wrappedfunc(*args, **kwargs):
        if stack:
            return f(*args, **kwargs)
        else:
            to = f(*args, **kwargs)
            while True:
                if type(to) is GeneratorType:
                    stack.append(to)
                    to = next(to)
                else:
                    stack.pop()
                    if not stack:
                        break
                    to = stack[-1].send(to)
            return to
    return wrappedfunc

@bootstrap
def find(v):
    if v != parent[v]:
        parent[v] = yield find(parent[v])
    
    yield parent[v]

def union(a, b):
    if a > b:
        a, b = b, a
    
    parent[b] = a

@bootstrap 
def dfs_to_find_route(value):
    global min_st
    
    vt[value] = True
    route.append(value)
    
    if value == needless[1]:
        min_st = set(route)
        
    for v in li[value]:
        if not vt[v]:
            yield dfs_to_find_route(v)
    
    route.pop()
    
    yield

@bootstrap
def dfs_to_group(value):
    vt[value] = True
    
    for v in li[value]:
        if not vt[v]:
            temp = (value in min_st) + (v in min_st)
            if temp != 1:
                union((yield find(value)), (yield find(v)))
            else:
                if value in min_st:
                    adj_mid[str(v)] = value
                else:
                    adj_mid[str(value)] = v
            yield dfs_to_group(v)

    yield
    
n, q = map(int, input().split())

li = [[] for _ in range(n + 1)]
needless = [-1, -1]

parent = list(range(n + 1))
for _ in range(n):
    u, v = map(int, input().split())
    
    ru = find(u)
    rv = find(v)
    
    if ru != rv:
        li[u].append(v)
        li[v].append(u)
        union(ru, rv)
    else:
        needless = [u, v]

route = []
min_st = set()
vt = [False] * (n + 1)
dfs_to_find_route(needless[0])

parent = list(range(n + 1))
adj_mid = dict()
vt = [False] * (n + 1)
dfs_to_group(1)

for i in range(1, n + 1):
    find(i)
    
gn_adj_mid = dict() # group num adj mid
for key in adj_mid:
    gn_adj_mid[str(parent[int(key)])] = adj_mid[key]
    
for _ in range(q):
    x, y = map(int, input().split())
    
    temp = (x in min_st) + (y in min_st)
    
    if temp == 2:
        r = 0
    elif temp == 0:
        if gn_adj_mid[str(parent[x])] == gn_adj_mid[str(parent[y])]:
            r = 1
        else:
            r = 0
    else:
        if x in min_st:
            x, y = y, x
        
        rx = parent[x]
        if str(rx) in gn_adj_mid and gn_adj_mid[str(rx)] == y:
            r = 1
        else:
            r = 0
    
    if r:
        print(1)
    else:
        print(2)
```
