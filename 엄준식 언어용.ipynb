```python
def prime_factor(num):
    r = []
    for mod in range(2, int(num ** 0.5) + 1):
         while True:
            if num % mod == 0:
                num //= mod
                r.append(mod)
            else:
                break
    
    if num != 1:
        r.append(num)
    
    if not r:
        r.append(1)
        
    return r

def get_s_num(num):
    for_numbers = ['어,,', '어,', '어', '어.', '어..', '어...', '어....', '어.....', '어......', '어.......']
    
    if num == 0:
        s = '.,'
    elif num < 0:
        s = ',' * abs(num)
    elif 48 <= num <= 57:
        num -= 48
        s = for_numbers[num]
    else:
        r = []
        used = prime_factor(num)
        for v in used:
            r.append('.' * v)
            
        s = ' '.join(r)
        
    return s

def string_to_code(s):
    temp = []
    for v in s:
        code = f'식{get_s_num(ord(v))}ㅋ'
        temp.append(code)
    return temp

def get_s_num_o(num):
    if num == 0:
        s = '.,'
    elif num < 0:
        s = ',' * abs(num)
    else:
        r = []
        used = prime_factor(num)
        for v in used:
            r.append('.' * v)
            
        s = ' '.join(r)
        
    return s

def string_to_code_o(s):
    temp = []
    for v in s:
        code = f'식{get_s_num_o(ord(v))}ㅋ'
        temp.append(code)
    return temp
```

```python
dp = [0, 1, 1, 1, 2, 2, 3, 4, 5, 7, 9, 12, 16, 21]

for i in range(101):
    dp.append(dp[-1] + dp[-5])

outputs = [''] * 101 
for n in range(1, 101):
    outputs[n] = str(dp[n])
```

```python
codes = ['어떻게', '', '어어엄식?', '', '어엄식?', '엄..... ..... ..']
cur_line = len(codes) + 1
for n in range(1, 101):
    cond_code = f'동탄어어{"," * n}?준{get_s_num(cur_line + 2)}'
    codes.append(cond_code)
    cur_line += 1
    
    nxt = cur_line + len(outputs[n]) + 2
    nxt_code = f'준{get_s_num(nxt)}'
    codes.append(nxt_code)
    
    codes.extend(string_to_code(outputs[n]))
    codes.append('준end_line')
    
    cur_line = nxt
    
for i in range(len(codes)):
    codes[i] = codes[i].replace('end_line', f'{get_s_num(nxt)}')
    
codes.extend(['', '식ㅋ', '어어엄어어어,'])
cur_line = len(codes) + 1

to_end = f'동탄어어어?준{get_s_num(cur_line + 2)}'
codes.append(to_end)

to_init = f'준{get_s_num(codes.index("어엄식?") + 1)}'
codes.append(to_init)

codes.extend(['', '이 사람이름이냐ㅋㅋ'])

for code in codes:
    print(code)
```
