```python
from collections import deque


wheels = [[]]
for _ in range(4):
    wheels.append(deque(map(int, input())))

k = int(input())
commands = [tuple(map(int, input().split())) for _ in range(k)]


def convert(d):
    if d == 1:
        return -1
    else:
        return 1
    

def get_connection_info():
    connection_info = [None]
    for i in range(1, 4):
        if wheels[i][2] == wheels[i + 1][6]:
            connection_info.append('same')
        else:
            connection_info.append('diff')
    return connection_info


def turn(wheel_num, d):
    wheels[wheel_num].rotate(d)
    
    
def get_total_score():
    total_score = 0
    for i in range(1, 5):
        if wheels[i][0] == 1:
            total_score += 2 ** (i - 1)
    return total_score


for command in commands:
    connection_info = get_connection_info()
    o_wheel_num, o_d = command 
    turn(o_wheel_num, o_d)
        
    wheel_num, d = o_wheel_num, o_d
    # 오른쪽으로 확인할 여지가 있는 경우
    while wheel_num < 4:
        if connection_info[wheel_num] == 'diff':
            d = convert(d)
            turn(wheel_num + 1, d)
        else:
            break

        wheel_num += 1
            
    wheel_num, d = o_wheel_num, o_d
    # 왼쪽으로 확인할 여지가 있는 경우
    while wheel_num > 1:
        if connection_info[wheel_num - 1] == 'diff':
            d = convert(d)
            turn(wheel_num - 1, d)
        else:
            break

        wheel_num -= 1
        
print(get_total_score())
```

```python
from copy import deepcopy

n, m, t = map(int, input().split())
li = [list(map(int, input().split())) for _ in range(n)]

def find_device():
    for i in range(n):
        for j in range(m):
            if li[i][j] == -1:
                return (i, j)   

upper_device = find_device()
lower_device = upper_device[0] + 1, upper_device[1]

di = [0, 0, -1, 1]
dj = [-1, 1, 0, 0]

def diffuse(li):
    new_li = [[0] * m for _ in range(n)]
    new_li[upper_device[0]][upper_device[1]] = -1
    new_li[lower_device[0]][lower_device[1]] = -1
    
    for i in range(n):
        for j in range(m):
            if li[i][j] >= 1:
                diffused_count = 0
                base_amount = li[i][j] // 5
                for k in range(4):
                    ni = i + di[k]
                    nj = j + dj[k]

                    if 0 <= ni < n and 0 <= nj < m:
                        if li[ni][nj] != -1:
                            new_li[ni][nj] += int(base_amount)
                            diffused_count += 1

                new_li[i][j] += li[i][j] - base_amount * diffused_count
    
    return new_li

def wind(li):
    new_li = deepcopy(li)
    new_li[upper_device[0]][upper_device[1]] = -1
    new_li[lower_device[0]][lower_device[1]] = -1
    
    x, y = upper_device[0], upper_device[1] + 1
    while True:
        if y + 1 >= m:
            break
        new_li[x][y + 1] = li[x][y]
        y += 1
        
    new_li[x - 1][y] = li[x][y]
    x -= 1
    while True:
        if x - 1 < 0:
            break
        new_li[x - 1][y] = li[x][y]
        x -= 1
     
    new_li[x][y - 1] = li[x][y]
    y -= 1
    while True:
        if y - 1 < 0:
            break
        new_li[x][y - 1] = li[x][y]
        y -= 1

    new_li[x + 1][y] = li[x][y]
    x += 1
    while True:
        if li[x + 1][y] == -1:
            break
        new_li[x + 1][y] = li[x][y]
        x += 1
        
    x, y = lower_device[0], lower_device[1] + 1
    while True:
        if y + 1 >= m:
            break
        new_li[x][y + 1] = li[x][y]
        y += 1   
        
    new_li[x + 1][y] = li[x][y]
    x += 1
    while True:
        if x + 1 >= n:
            break
        new_li[x + 1][y] = li[x][y]
        x += 1    
    
    new_li[x][y - 1] = li[x][y]
    y -= 1
    while True:
        if y - 1 < 0:
            break
        new_li[x][y - 1] = li[x][y]
        y -= 1
        
    new_li[x - 1][y] = li[x][y]
    x -= 1        
    while True:
        if li[x - 1][y] == -1:
            break
        new_li[x - 1][y] = li[x][y]
        x -= 1
        
    new_li[upper_device[0]][upper_device[1] + 1] = 0
    new_li[lower_device[0]][lower_device[1] + 1] = 0
    return new_li
        
def get_total():
    total = 0
    for i in range(n):
        for j in range(m):
            if li[i][j] >= 1:
                total += li[i][j]
    return total

for _ in range(t):
    li = diffuse(li)
    li = wind(li)
    
print(get_total())
```

```python
from copy import deepcopy
import sys
input = sys.stdin.readline

n = int(input().rstrip())
board = [[0] * 10 for _ in range(10)] # 블록은 1, 빈 공간은 0
commands = [tuple(map(int, input().rstrip().split())) for _ in range(n)]

def move_cover(blocks):
    dx, dy, = 0, 1
    move(deepcopy(blocks), dx, dy)

    dx, dy = 1, 0
    move(deepcopy(blocks), dx, dy)

def move(blocks, dx, dy):
    while True:
        is_move_stopped = False
        for block in blocks:
            x, y = block
            nx, ny = x + dx, y + dy
            
            if nx >= 10 or ny >= 10:
                is_move_stopped = True
                break
            if board[nx][ny] == 1:
                is_move_stopped = True
                break
        
        if is_move_stopped:
            break
        else:
            for block in blocks:
                if dx == 1:
                    block[0] += 1
                else:
                    block[1] += 1
    for block in blocks:
        x, y = block
        board[x][y] = 1

def drag_down(bottom):
    for i in range(bottom - 1, 3, -1):
        for j in range(4):
            board[i + 1][j] = board[i][j]
    for j in range(4):
        board[4][j] = 0
    
def drag_right(rightest):
    for j in range(rightest - 1, 3, -1):
        for i in range(4):
            board[i][j + 1] = board[i][j]
    for i in range(4):
        board[i][4] = 0

def get_score_and_drag_down():
    score = 0
    i = 9
    while True:
        count = 0
        for j in range(0, 4):
            if board[i][j] == 1:
                count += 1
        if count == 4:
            score += 1
            for j in range(0, 4):
                board[i][j] = 0
            drag_down(i)
        else:
            i -= 1
        
        if i == 5:
            break
        
    return score
        
def get_score_and_drag_right():
    score = 0
    j = 9
    while True:
        count = 0
        for i in range(0, 4):
            if board[i][j] == 1:
                count += 1
        if count == 4:
            score += 1
            for i in range(0, 4):
                board[i][j] = 0
            drag_right(j)
        else:
            j -= 1
        
        if j == 5:
            break  
    return score

def reflect_light_green():
    occupied_row_count = 0
    for i in range(4, 6):
        for j in range(4):
            if board[i][j] == 1:
                occupied_row_count += 1
                break
    for _ in range(occupied_row_count):
        drag_down(9)
        
def reflect_light_blue():
    occupied_column_count = 0
    for j in range(4, 6):
        for i in range(4):
            if board[i][j] == 1:
                occupied_column_count += 1
                break
    for _ in range(occupied_column_count):
        drag_right(9)    

def get_tile_count():
    return sum([sum(v) for v in board])

total_score = 0
for command in commands:
    t, x, y = command
    
    if t == 1:
        blocks = [[x, y]]
    elif t == 2:
        blocks = [[x, y], [x, y + 1]]
    else:
        blocks = [[x, y], [x + 1, y]]
        
    move_cover(blocks)
    
    total_score += get_score_and_drag_down()
    total_score += get_score_and_drag_right()
    
    reflect_light_green()
    reflect_light_blue()
    
print(total_score)
print(get_tile_count())
```
