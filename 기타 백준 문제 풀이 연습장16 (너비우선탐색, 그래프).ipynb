```python
# 맞았음, sys 사용시 틀리다고 나옴
from itertools import combinations

def to_coord(x):
    if x == 0:
        return (0, 0)
    elif x == 1:
        return (0, 1)
    elif x == 2:
        return (0, 2)
    elif x == 3:
        return (1, 0)
    elif x == 4:
        return (1, 1)
    elif x == 5:
        return (1, 2)
    elif x == 6:
        return (2, 0)
    elif x == 7:
        return (2, 1)
    elif x == 8:
        return (2, 2)
    
def get_change_cases(x, y):
    dx = [0, 1, -1, 0, 0]
    dy = [0, 0, 0, 1, -1]
    
    for i in range(5):
        yield x + dx[i], y + dy[i]
        
def change_color(x):
    if x == '.':
        return '*'
    else:
        return '.'

T = int(input())
for _ in range(T):
    target_li = []
    for _ in range(3):
        target_li.append(list(input()))
        
    if target_li == [["."] * 3 for _ in range(3)]:
        print(0)
        continue

    cases = []
    for i in range(1, 10):
        cases.extend(list(combinations(range(0, 9), i)))

    min_available_count = float('inf')   
    for case in cases:
        li = [["."] * 3 for _ in range(3)]
        for v in case:
            x, y = to_coord(v)

            for a, b in get_change_cases(x, y):
                if 0 <= a < 3 and 0 <= b < 3:
                    li[a][b] = change_color(li[a][b])
        if li == target_li:
            count = len(case)

            if count < min_available_count:
                min_available_count = count

    print(min_available_count)
```

```python
# 맞았음, 뱀을 전략적으로 타야할 수도 있었음
from collections import defaultdict, deque
import sys
# input = sys.stdin.readline

ladders_or_snakes = defaultdict(int)

n, m = map(int, input().split())
for _ in range(n + m):
    _from, _to = map(int, input().split())
    ladders_or_snakes[_from] = _to
    
def bfs():    
    min_counts = [float('inf')] * 101
    
    queue = deque()
    queue.append((1, 0))
    min_counts[1] = 0
    
    while queue:
        num, count = queue.popleft()
        
        for i in range(1, 7):
            new_num = num + i
            if 1 <= new_num < 101:
                new_count = count + 1
                if new_count < min_counts[new_num]:
                    min_counts[new_num] = new_count
                    
                    if ladders_or_snakes[new_num]:
                        new_num = ladders_or_snakes[new_num]
                    
                    queue.append((new_num, new_count))

    return min_counts[100]

result = bfs()

print(result)
```

### 텔레포트 정거장 18232

```python
# 틀렸음, 일단 포기
from collections import defaultdict, deque
import sys
# input = sys.stdin.readline

n, m = map(int, input().split())
s, e = map(int, input().split())

teleport = defaultdict(list)
for _ in range(m):
    a, b = map(int, input().split())
    teleport[a].append(b)
    teleport[b].append(a)
    
def bfs():
    min_times = [float('inf')] * (n + 1)
    queue = deque()
    queue.append((s, 0))
    min_times[s] = 0
    while queue:
        value, time = queue.popleft()
        
        case_set = {value - 1, value + 1}
        for v in teleport[value]:
            case_set.add(v)
            
        for v in case_set:
            new_time = time + 1
            
            if 1 <= v <= n:
                if new_time < min_times[v]:
                    min_times[v] = new_time
                    queue.append((v, new_time))
    return min_times[e]

result = bfs()

print(result)
```

### 특정 거리의 도시 찾기 18352

```python
import heapq

n, m, k, x = map(int, input().split())
li = [[] for _ in range(n + 1)]
for _ in range(m):
    s, e = map(int, input().split())
    li[s].append(e)
    
def dij():
    distances = [float('inf')] * (n + 1)
    distances[x] = 0
    
    heap = []
    heapq.heappush(heap, (x, 0))
    
    while heap:
        value, distance = heapq.heappop(heap)
        
        for v in li[value]:
            d = distance + 1
            
            if d < distances[v]:
                distances[v] = d
                heapq.heappush(heap, (v, d))
                
    return distances

distances = dij()
distances = [i for i, v in enumerate(distances) if v == k]
distances.sort()

if distances:
    for d in distances:
        print(d)
else:
    print(-1)
```

```python
from collections import deque
import sys
# input = sys.stdin.readline

n = int(input())
li = [[] for _ in range(n + 1)]
while True:
    a, b = map(int, input().split())
    if a == b == -1:
        break
    li[a].append(b)
    li[b].append(a)

def bfs(s):
    distances = [float('inf')] * (n + 1) 
    distances[s] = 0
    queue = deque()
    queue.append(s)
    while queue:
        value = queue.popleft()
        
        for v in li[value]:
            if distances[v] == float('inf'):
                distances[v] = distances[value] + 1
                queue.append(v)
    return max(distances[1:])
                
min_score = float('inf')
min_score_people = []
for s in range(1, n + 1):
    score = bfs(s)
    if score < min_score:
        min_score = score
        min_score_people = [s]
    elif score == min_score:
        min_score_people.append(s)
        
min_score_people.sort()

print(min_score, len(min_score_people))
print(" ".join(map(str, min_score_people)))
```

### 빙산 2573
- 너비 우선 탐색, 깊이 우선 탐색

```python
from collections import deque
import sys
# input = sys.stdin.readline

n, m = map(int, input().strip().split())

li = []
for _ in range(n):
    li.append(list(map(int, input().strip().split())))
    
dx = [0, 0, -1, 1]
dy = [-1, 1, 0, 0]
    
count = 0
while True:
    pieces = 0
    to_decrease = [[0] * m for _ in range(n)]
    visited = [[False] * m for _ in range(n)]
    
    for i in range(n):
        for j in range(m):
            if li[i][j] == 0:
                continue
            if visited[i][j]:
                continue
                
            pieces += 1
            
            visited[i][j] = True
            
            queue = deque()
            queue.append((i, j))
            
            while queue:
                x, y = queue.popleft()
                
                for k in range(4):
                    nx = x + dx[k]
                    ny = y + dy[k]
                    
                    if 0 <= nx < n and 0 <= ny < m:
                        if li[nx][ny] == 0:
                            to_decrease[x][y] += 1
                        else:
                            if not visited[nx][ny]:
                                visited[nx][ny] = True
                                queue.append((nx, ny))
    
    for i in range(n):
        for j in range(m):
            if to_decrease[i][j] >= 1:
                li[i][j] = max(0, li[i][j] - to_decrease[i][j])
                                
    if pieces >= 2:
        print(count)
        break
    elif pieces == 0:
        print(0)
        break
        
    count += 1
```

```python
# 시간 초과
from collections import deque

n, m = map(int, input().split())

li = []
for _ in range(n):
    li.append(list(map(int, tuple(input()))))

walls = []
for i in range(n):
    for j in range(m):
        if li[i][j] == 1:
            walls.append((i, j))

dx = [0, 0, -1, 1]
dy = [-1, 1, 0, 0]
            
for wall in walls:
    visited = [[False] * m for _ in range(n)]
    visited[wall[0]][wall[1]] = True
    
    queue = deque()
    queue.append(wall)
    
    while queue:
        x, y = queue.popleft()
        
        for i in range(4):
            nx = x + dx[i]
            ny = y + dy[i]
            
            if 0 <= nx < n and 0 <= ny < m:
                if not visited[nx][ny]:
                    visited[nx][ny] = True
                    
                    if li[nx][ny] == 0:
                        li[wall[0]][wall[1]] += 1
                        queue.append((nx, ny))
    
li = ["".join(map(lambda x: str(x % 10), v)) for v in li]
print("\n".join(li))
```
