A

```python
import sys
input = lambda: sys.stdin.readline().rstrip()

Y1, Y2, Y3, Y4, M1, M2, D1, D2 = list(map(int, input()))

max_value = -1
max_value_li = []

n = int(input())
for _ in range(n):
    date = input()
    y1, y2, y3, y4, m1, m2, d1, d2 = list(map(int, date))
    
    temp1 = (Y1 - y1) ** 2 + (Y2 - y2) ** 2 + (Y3 - y3) ** 2 + (Y4 - y4) ** 2
    temp2 = (M1 - m1) ** 2 + (M2 - m2) ** 2
    temp3 = (D1 - d1) ** 2 + (D2 - d2) ** 2
    
    value = temp1 * temp2 * temp3
    if value >= max_value:
        if value == max_value:
            max_value_li.append(date)
        else:
            max_value = value
            max_value_li = [date]
            
max_value_li.sort()

print(max_value_li[0])
```

```python
from collections import defaultdict as dd
import sys
input = lambda: sys.stdin.readline().rstrip()

t, n = map(int, input().split())

cur_loc = [1] * (n + 1)
collected = [dd(int) for _ in range(n + 1)]
bad_logs = []
banned = set()
for _ in range(t):
    log = input().split()
    
    log_num, p_num, act_code = log[0], log[1], log[2]
    log_num = int(log_num)
    p_num = int(p_num)
    
    if act_code == 'C':
        action = [log[3], log[4]]
        action[0] = int(action[0])
        action[1] = int(action[1])
    else:
        action = log[3]
        action = int(action)
        
    if act_code == 'M':
        cur_loc[p_num] = action
    elif act_code == 'F':
        collected[p_num][action] += 1
        if action != cur_loc[p_num]:
            bad_logs.append(log_num)
    elif act_code == 'C':
        if collected[p_num][action[0]] == 0 or collected[p_num][action[1]] == 0:
            bad_logs.append(log_num)
            
        collected[p_num][action[0]] -= 1
        if collected[p_num][action[0]] < 0:
            collected[p_num][action[0]] = 0
        collected[p_num][action[1]] -= 1
        if collected[p_num][action[1]] < 0:
            collected[p_num][action[1]] = 0
    elif act_code == 'A':
        if cur_loc[p_num] != cur_loc[action]:
            bad_logs.append(log_num)
            banned.add(p_num)
            
print(len(bad_logs))
if len(bad_logs) >= 1:
    print(*bad_logs)
    
print(len(banned))
if len(banned) >= 1:
    print(*sorted(banned))
```

```python
from collections import deque
import sys
input = lambda: sys.stdin.readline().rstrip()

n, q = map(int, input().split())

cur_page = -1
go_back = deque()
go_front = deque()
for _ in range(q):
    query = input().split()
    
    if query[0] == 'B':
        if not go_back:
            continue
        go_front.appendleft(cur_page)
        cur_page = go_back.pop()
    elif query[0] == 'F':
        if not go_front:
            continue
        go_back.append(cur_page)
        cur_page = go_front.popleft()
    elif query[0] == 'A':
        next_page = int(query[1])
        go_front = deque()
        if cur_page == -1:
            cur_page = next_page
        else:
            go_back.append(cur_page)
            cur_page = next_page
    elif query[0] == 'C':
        f_value = -1
        n_go_back = deque()
        while go_back:
            value = go_back.pop()
            if value != f_value:
                n_go_back.appendleft(value)
                f_value = value
        go_back = n_go_back.copy()
        
print(cur_page)
if not go_back:
    print(-1)
else:
    print(*reversed(go_back))
    
if not go_front:
    print(-1)
else:
    print(*go_front)
```

```python
from collections import deque
import sys
# input = lambda: sys.stdin.readline().rstrip()

n, q, max_c = map(int, input().split())
caps = [0] + list(map(int, input().split()))

cur_page = -1
go_back = deque()
go_front = deque()
total_c = 0
back_c = 0
cur_c = 0
front_c = 0
for _ in range(q):
    query = input().split()
    
    if query[0] == 'B':
        if not go_back:
            continue
        go_front.appendleft(cur_page)
        front_c += caps[cur_page]
        cur_page = go_back.pop()
        cur_c = caps[cur_page]
        back_c -= cur_c
    elif query[0] == 'F':
        if not go_front:
            continue
        go_back.append(cur_page)
        back_c += caps[cur_page]
        cur_page = go_front.popleft()
        cur_c = caps[cur_page]
        front_c -= cur_c
    elif query[0] == 'A':
        next_page = int(query[1])
        total_c -= front_c
        front_c = 0
        go_front = deque()
        if cur_page == -1:
            cur_page = next_page
            cur_c = caps[cur_page]
        else:
            go_back.append(cur_page)
            back_c += caps[cur_page]
            cur_page = next_page
            cur_c = caps[cur_page]
            
        total_c = back_c + cur_c + front_c
        while total_c > max_c:
            value = caps[go_back.popleft()]
            back_c -= value
            total_c -= value
    
    elif query[0] == 'C':
        f_value = -1
        n_go_back = deque()
        c_to_remove = 0
        while go_back:
            value = go_back.pop()
            if value != f_value:
                n_go_back.appendleft(value)
                f_value = value
            else:
                c_to_remove += caps[value]
        go_back = n_go_back.copy()
        back_c -= c_to_remove
        
print(cur_page)
if not go_back:
    print(-1)
else:
    print(*reversed(go_back))
    
if not go_front:
    print(-1)
else:
    print(*go_front)
```

```python
import sys
input = lambda: sys.stdin.readline().rstrip()

n, t = map(int, input().split())

possibles = [0] * 100001 # possibles[0]은 0~1의 구간
for _ in range(n):
    k = int(input())
    for _ in range(k):
        s, e = map(int, input().split())
        for i in range(s, e):
            possibles[i] += 1
            
cur_sum = sum(possibles[:t])
sp = [cur_sum]

cur_end_index = t - 1
cur_end_index += 1
while cur_end_index <= 100000:
    cur_sum += possibles[cur_end_index]
    cur_sum -= possibles[cur_end_index - t]
    sp.append(cur_sum)
    cur_end_index += 1
    
max_value = -1
max_value_index = -1
for i in range(len(sp) - 1, -1, -1):
    if sp[i] >= max_value:
        max_value = sp[i]
        max_value_index = i
        
print(max_value_index, max_value_index + t)
```

```python
import sys
input = lambda: sys.stdin.readline().rstrip()

n, t = map(int, input().split())

delta = [0] * 100001
for _ in range(n):
    k = int(input())
    for _ in range(k):
        s, e = map(int, input().split())
        delta[s] += 1
        delta[e] -= 1
        
scores = [0] * 100001
cur = 0
for i in range(100001):
    cur += delta[i]
    scores[i] = cur
    
cur_sum = sum(scores[:t])
sp = [cur_sum]

cur_end_index = t - 1
cur_end_index += 1
while cur_end_index <= 100000:
    cur_sum += scores[cur_end_index]
    cur_sum -= scores[cur_end_index - t]
    sp.append(cur_sum)
    cur_end_index += 1
    
max_value = -1
max_value_index = -1
for i in range(len(sp) - 1, -1, -1):
    if sp[i] >= max_value:
        max_value = sp[i]
        max_value_index = i
        
print(max_value_index, max_value_index + t)
```

```python
import sys
input = lambda: sys.stdin.readline().rstrip()

n = int(input())

get_value = lambda x: int(x) - 1
li = list(map(get_value, input().split()))

vt = [False] * n
check_cands = set()
cycle_check_cands = set()
for ocur in range(n):
    if vt[ocur]:
        continue
    vt[ocur] = True
    
    cur = ocur
    while True:
        ncur = li[cur]
        
        if vt[ncur]:
            if ncur in check_cands:
                check_cands.discard(ncur)
            
            if ncur == ocur:
                cycle_check_cands.add(ocur)
            else:
                check_cands.add(ocur)
            
            break
        vt[ncur] = True
        cur = ncur

rts = []
vt = [False] * n
vt[0] = True

cur = 0
while True:
    ncur = li[cur]
    rts.append(ncur + 1)

    if vt[ncur]:
        break
    vt[ncur] = True
    
    cur = ncur
    
for cur in check_cands:
    if vt[cur]:
        continue
    vt[cur] = True
    rts.append(cur + 1)
    
    while True:
        ncur = li[cur]
        rts.append(ncur + 1)
        
        if vt[ncur]:
            break
        vt[ncur] = True

        cur = ncur
        
for cur in cycle_check_cands:
    if vt[cur]:
        continue
    vt[cur] = True
    rts.append(cur + 1)
    
    while True:
        ncur = li[cur]
        rts.append(ncur + 1)

        if vt[ncur]:
            break
        vt[ncur] = True

        cur = ncur

print(len(rts))
print(*rts)
```

```python
n, m = map(int, input().split())

li = [input().split() for _ in range(n)]

alpha_to_num = dict()
for i in range(26):
    alpha_to_num[chr(i + 65)] = i
    
cur = 26
for i in range(26):
    for j in range(26):
        alpha_to_num[chr(i + 65) + chr(j + 65)] = cur
        cur += 1
    
def get_value(s):
    al = ''
    num = ''
    for v in s:
        if v.isdigit():
            num += v
        else:
            al += v

    al = alpha_to_num[al]
    num = int(num) - 1

    final = num * m + al
    
    return final

nli = [set() for _ in range(n * m)]
for i in range(n):
    for j in range(m):
        temp = li[i][j]
        if temp == '.':
            continue
        elif '+' in temp:
            a, b = temp.split('+')
            nli[i * m + j].add(get_value(a))
            nli[i * m + j].add(get_value(b))
        else:
            nli[i * m + j].add(get_value(temp))

def dfs(value):
    global is_cycled
    
    if is_cycled:
        return
    
    for v in nli[value]:
        if vt[v]:
            if v in temp:
                is_cycled = True
                return
            continue
        vt[v] = True
        temp.add(v)
        dfs(v)
        temp.discard(v)
    
vt = [False] * (n * m)
for value in range(n * m):
    is_cycled = False
    if vt[value]:
        continue
    vt[value] = True
    
    temp = set()
    temp.add(value)
    dfs(value)
    temp.discard(value)
    if is_cycled:
        print('yes')
        break
else:
    print('no')
```
