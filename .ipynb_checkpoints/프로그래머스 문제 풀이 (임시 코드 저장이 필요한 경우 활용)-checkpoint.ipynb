### 도둑질
- 동적계획법

```python
def solution( money ):
    if len( money ) == 3:
        answer = max( money )
    else:
        money_length = len( money ) - 1
        
        # 첫번째 집을 선택하는 경우
        dp1 = [[0, 0] for _ in range(money_length)]
        dp1[0], dp1[1] = [money[0], money[0]], [0, money[0]]
        
        # 첫번째 집을 선택하지 않는 경우
        dp2 = [[0, 0] for _ in range(money_length)]
        dp2[0], dp2[1] = [money[1], 0], [money[2], money[1]]
        
        for i in range(2, money_length):
            max1 = max(dp1[i - 2])
            max2 = max(dp1[i - 2][0], dp1[i - 2][1], dp1[i - 1][0])
            
            max3 = max(dp2[i - 2])
            max4 = max(dp2[i - 2][0], dp2[i - 2][1], dp2[i - 1][0])
            
            dp1[i][0] = max1 + money[i]
            dp1[i][1] = max2
            
            dp2[i][0] = max3 + money[i + 1]
            dp2[i][1] = max4
            
        answer = max(max(dp1[-1]), max(dp2[-1]))
        
    return answer
```

### 방의 개수
- 그래프

```python
# 일단 포기
def get_opposite_dir(x):
    if x == 0:
        return 4
    if x == 1:
        return 5
    if x == 2:
        return 6
    if x == 3:
        return 7
    if x == 4:
        return 0
    if x == 5:
        return 1
    if x == 6:
        return 2
    if x == 7:
        return 3

def solution(arrows):
    from collections import defaultdict
    
    d = defaultdict(set)
    x = y = 0
    d[(0, 0)].add(-1) # -1: 원점, 0 이상: 0 이상에 해당하는 방향에서 온 경우
    
    dx = [-1, -1, 0, 1, 1, 1, 0, -1]
    dy = [0, 1, 1, 1, 0, -1, -1, -1]
    
    answer = 0
    for arrow in arrows:
        nx = x + dx[arrow]
        ny = y + dy[arrow]
        
        if d[(nx, ny)] and arrow not in d[(nx, ny)]:
            if get_opposite_dir(arrow) not in d[(x, y)]:
                answer += 1
                    
        x, y = nx, ny
        d[(x, y)].add(arrow)
    
        # 대각선으로 교차하는 경우들
        if arrow == 1:
            if 7 in d[(x, y - 1)] or 3 in d[(x + 1, y)]:
                answer += 1
        elif arrow == 7:
            if 1 in d[(x, y + 1)] or 5 in d[(x + 1, y)]:
                answer += 1
        elif arrow == 5:
            if 7 in d[(x - 1, y)] or 3 in d[(x, y + 1)]:
                answer += 1
        elif arrow == 3:
            if 1 in d[(x - 1, y)] or 5 in d[(x, y - 1)]:
                answer += 1
    
    return answer
```

### 섬 연결하기
- 탐욕법 level 3

```python
def find(v, parent):
    if v == parent[v]:
        return v
    else:
        parent[v] = find(parent[v], parent)
        return parent[v]

def union(a, b, parent, rank):
    if rank[a] >= rank[b]:
        parent[b] = a
        if rank[a] == rank[b]:
            rank[a] += 1
    else:
        parent[a] = b

def solution(n, costs):
    costs.sort(key=lambda x: x[2])
    parent = [i for i in range(n)]
    rank = [0] * n

    min_total_cost = 0
    for a, b, cost in costs:
        a = find(a, parent)
        b = find(b, parent)

        if a != b:
            union(a, b, parent, rank)
            min_total_cost += cost

    return min_total_cost
```

### 이중우선순위큐
- 힙 level 3

```python
def solution(operations):
    # min heap과 max heap을 별도로 만들어서 저장
    # min heap의 경우에는 (value, index), max heap의 경우에는 (-value, index)
    # 최솟값을 꺼낼 때는 min heap에서 꺼내고 최댓값을 꺼낼 때는 max heap에서 꺼냄
    # 단 꺼낸 인덱스는 defaultdict bool로 저장해놓음
    # 꺼낼 때 해당 값이 이미 다른 곳에서 꺼내진 값이라면 제거하고 다시 꺼냄

    import heapq
    from collections import defaultdict

    min_heap = []
    max_heap = []
    d = defaultdict(bool)

    current_index = 0
    for operation in operations:
        command, num = operation.split()
        num = int(num)

        if command == 'I':
            heapq.heappush(min_heap, (num, current_index))
            heapq.heappush(max_heap, (-num, current_index))

            current_index += 1
        else:
            if num == 1:
                while max_heap:
                    value, index = heapq.heappop(max_heap)
                    if not d[index]:
                        d[index] = True
                        break
            else:
                while min_heap:
                    value, index = heapq.heappop(min_heap)
                    if not d[index]:
                        d[index] = True
                        break

    max_result = -float('inf')
    while max_heap:
        value, index = heapq.heappop(max_heap)
        if not d[index]:
            max_result = -value
            break

    if max_result == -float('inf'):
        max_result = 0

    min_result = float('inf')
    while min_heap:
        value, index = heapq.heappop(min_heap)
        if not d[index]:
            min_result = value
            break

    if min_result == float('inf'):
        min_result = 0  

    answer = [max_result, min_result]

    return answer
```

### 블록 이동하기
- level 3

```python
def solution(board):
    from collections import deque

    n = len(board)
    answer = 0

    # 로봇 현재 위치 기준: 오른쪽, 밑
    queue = deque()
    queue.append((0, 1, 0, 0)) # (x, y, status, count), status의 경우 0: 가로, 1: 세로

    visited = [[[float('inf')] * n for _ in range(n)] for _ in range(2)] # visited[status][x][y]
    visited[0][0][1] = 0
    while queue:
        x, y, status, count = queue.popleft()

        if x == n - 1 and y == n - 1:
            print(visited)
            return count

        # 현재 가로인 경우
        if status == 0:
            if y - 2 >= 0:
                if board[x][y - 2] == 0:
                    if visited[status][x][y] + 1 < visited[0][x][y - 1]:
                        visited[0][x][y - 1] = visited[status][x][y] + 1
                        queue.append((x, y - 1, status, count + 1))

            if y + 1 < n:
                if board[x][y + 1] == 0:
                    if visited[status][x][y] + 1 < visited[0][x][y + 1]:
                        visited[0][x][y + 1] = visited[status][x][y] + 1
                        queue.append((x, y + 1, status, count + 1))

            if x + 1 < n:
                if board[x + 1][y - 1] == 0 and board[x + 1][y] == 0:
                    if visited[status][x][y] + 1 < visited[0][x + 1][y]:
                        visited[0][x + 1][y] = visited[status][x][y] + 1
                        queue.append((x + 1, y, status, count + 1))

                    if visited[status][x][y] + 1 < visited[1][x + 1][y - 1]:
                        visited[1][x + 1][y - 1] = visited[status][x][y] + 1
                        queue.append((x + 1, y - 1, 1, count + 1))

                    if visited[status][x][y] + 1 < visited[1][x + 1][y]:
                        visited[1][x + 1][y] = visited[status][x][y] + 1
                        queue.append((x + 1, y, 1, count + 1))                        

            if x - 1 >= 0:
                if board[x - 1][y - 1] == 0 and board[x - 1][y] == 0:
                    if visited[status][x][y] + 1 < visited[0][x - 1][y]:
                        visited[0][x - 1][y] = visited[status][x][y] + 1
                        queue.append((x - 1, y, status, count + 1))

                    if visited[status][x][y] + 1 < visited[1][x][y - 1]:
                        visited[1][x][y - 1] = visited[status][x][y] + 1
                        queue.append((x, y - 1, 1, count + 1))

                    if visited[status][x][y] + 1 < visited[1][x][y]:
                        visited[1][x][y] = visited[status][x][y] + 1
                        queue.append((x, y, 1, count + 1))

        # 현재 세로인 경우
        else:
            if x - 2 >= 0:
                if board[x - 2][y] == 0:
                    if visited[status][x][y] + 1 < visited[1][x - 1][y]:
                        visited[1][x - 1][y] = visited[status][x][y] + 1
                        queue.append((x - 1, y, status, count + 1))   

            if x + 1 < n:
                if board[x + 1][y] == 0:
                    if visited[status][x][y] + 1 < visited[1][x + 1][y]:
                        visited[1][x + 1][y] = visited[status][x][y] + 1
                        queue.append((x + 1, y, status, count + 1))

            if y - 1 >= 0:
                if board[x - 1][y - 1] == 0 and board[x][y - 1] == 0:
                    if visited[status][x][y] + 1 < visited[1][x][y - 1]:
                        visited[1][x][y - 1] = visited[status][x][y] + 1
                        queue.append((x, y - 1, status, count + 1))

                    if visited[status][x][y] + 1 < visited[0][x - 1][y]:
                        visited[0][x - 1][y] = visited[status][x][y] + 1
                        queue.append((x - 1, y, 0, count + 1))

                    if visited[status][x][y] + 1 < visited[0][x][y]:
                        visited[0][x][y] = visited[status][x][y] + 1
                        queue.append((x, y, 0, count + 1))

            if y + 1 < n:
                if board[x - 1][y + 1] == 0 and board[x][y + 1] == 0:
                    if visited[status][x][y] + 1 < visited[1][x][y + 1]:
                        visited[1][x][y + 1] = visited[status][x][y] + 1
                        queue.append((x, y + 1, status, count + 1))

                    if visited[status][x][y] + 1 < visited[0][x - 1][y + 1]:
                        visited[0][x - 1][y + 1] = visited[status][x][y] + 1
                        queue.append((x - 1, y + 1, 0, count + 1))

                    if visited[status][x][y] + 1 < visited[0][x][y + 1]:
                        visited[0][x][y + 1] = visited[status][x][y] + 1
                        queue.append((x, y + 1, 0, count + 1))
```

### 등굣길
- 동적계획법

```python
def solution(m, n, puddles):
    mod = 1000000007

    li = [[0] * n for _ in range(m)] # 0: 빈 칸, 1: 웅덩이
    for x, y in puddles:
        li[x - 1][y - 1] = 1

    dp = [[0] * (n + 1) for _ in range(m + 1)]
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if i == 1 and j == 1:
                dp[i][j] = 1
                continue
            if li[i - 1][j - 1] == 1:
                continue

            dp[i][j] = (dp[i - 1][j] + dp[i][j - 1]) % mod

    return dp[-1][-1]
```

### 숫자 블록
연습문제 level 4

```python
def solution(begin, end):
    n = 10000000
    is_prime = [False] * 2 + [True] * (n - 1)

    for mod in range(2, int(n ** 0.5) + 1):
        if is_prime[mod]:
            for i in range(mod * mod, n + 1, mod):
                is_prime[i] = False

    primes = [i for i, v in enumerate(is_prime) if v]
    primes2 = [v * v for v in primes]

    result = [1] * (end - begin + 1)
    if begin == 1:
        result[0] = 0

    for i in range(begin, end + 1):
        for j in range(len(primes)):
            if i % primes[j] == 0:
                result[i - begin] = i // primes[j]
                break
            if primes2[j] > i:
                break

    return result
```

### 징검다리
- 이분탐색 level 4

```python
def solution(distance, rocks, n):
    rocks.sort()
    marks = [0] + rocks
    marks.append(distance)

    start = 1
    end = distance
    result = -1
    while start <= end:
        mid = (start + end) // 2

        deleted = 0
        current = 0
        for mark in marks[1:]:
            if current + mid > mark:
                deleted += 1
            else:
                current = mark

        if deleted <= n:
            result = mid
            start = mid + 1
        else:
            end = mid - 1

    return result
```

```python
number = "44444"
k = 4

import heapq
from collections import deque

number = deque([(v, i) for i, v in enumerate(number)])
heap = []
heapq.heappush(heap, number.popleft())
while k and number:
    value, index = number.popleft()
    while k and heap:
        former_min_value, former_min_index = heapq.heappop(heap)
        if former_min_value < value:
            k -= 1
            continue

        heapq.heappush(heap, (former_min_value, former_min_index))
        break

    heapq.heappush(heap, (value, index))

while k:
    heapq.heappop(heap)

number.extend(heap)
number = list(number)
number.sort(key=lambda x: x[1])
result = [v[0] for v in number]
result = "".join(map(str, result))

print(result)
```
