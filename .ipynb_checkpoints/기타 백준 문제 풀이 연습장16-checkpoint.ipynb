https://www.acmicpc.net/problem/1533 길의 개수

```python
mod = 1000003

def recur(n, size):
    global init_mul_mat
    
    if n == 1:
        return init_mul_mat
    
    mat = recur(n // 2, size)
    
    mul_mat = [[0] * size for _ in range(size)]
    for i in range(size):
        for j in range(size):
            temp = 0
            for k in range(size):
                temp += mat[i][k] * mat[k][j]
                temp %= mod
            mul_mat[i][j] = temp
            
    if n % 2 == 1:
        nmul_mat = [[0] * size for _ in range(size)]
        for i in range(size):
            for j in range(size):
                temp = 0
                for k in range(size):
                    temp += mul_mat[i][k] * init_mul_mat[k][j]
                    temp %= mod
                nmul_mat[i][j] = temp
                
        return nmul_mat
    else:
        return mul_mat

n, s, e, t = map(int, input().split())
s -= 1
e -= 1

li = [list(map(int, input())) for _ in range(n)]

edges = []
for i in range(n):
    ni = i * 5

    for d in range(1, 5):
        edges.append((ni + d, ni + d - 1))
        
for i in range(n):
    for j in range(n):
        ni = i * 5
        nj = j * 5
        
        if li[i][j] >= 1:
            edges.append((ni, nj + li[i][j] - 1))

nn = n * 5
s *= 5
e *= 5

init_mul_mat = [[0] * nn for _ in range(nn)]
for a, b in edges:
    init_mul_mat[a][b] = 1       

init_mat = [[0] * nn for _ in range(nn)]
init_mat[s][s] = 1

init_mul_mat = [[0] * nn for _ in range(nn)]
for a, b in edges:
    init_mul_mat[a][b] = 1
    
mat = recur(t, nn)

r = [[0] * nn for _ in range(nn)]
for i in range(nn):
    for j in range(nn):
        temp = 0
        for k in range(nn):
            temp += init_mat[i][k] * mat[k][j]
            temp %= mod
        r[i][j] = temp
        
result = r[s][e]

print(result)
```

https://www.acmicpc.net/problem/26166 끝말잇기 하실 분!!

```python
from collections import deque, defaultdict as dd
import sys
input = lambda: sys.stdin.readline().rstrip()

n = int(input())
li = [input() for _ in range(n)]

nli = []
indegree = dd(int)
graph = dd(list)
status = dd(lambda: '')
for i in range(n):
    nli.append(li[i][0] + li[i][-1])
    
    f = li[i][0]
    e = li[i][-1]
    
    indegree[f]
    indegree[e]
    status[f]
    status[e]
    
    graph[e].append(f)
    indegree[f] += 1

queue = deque()
for key in indegree:
    if indegree[key] == 0:
        status[key] = 'w'
        queue.append(key)

while queue:
    key = queue.popleft()
    
    if status[key] == 'w':
        for nkey in graph[key]:
            if indegree[nkey] >= 1:
                indegree[nkey] = 0
                status[nkey] = 'l'
                queue.append(nkey)
    else:
        for nkey in graph[key]:
            indegree[nkey] -= 1
            if indegree[nkey] == 0:
                status[nkey] = 'w'
                queue.append(nkey)
                
result = []
for v in li:
    if status[v[-1]] == 'w':
        result.append(v)
        
result.sort()

print(len(result))
for v in result:
    print(v)
```

```python
inf = float('inf')

dx = [-1, 1, 0, 0]
dy = [0, 0, -1, 1]

def dfs(x, y, mc, ac):
    global r
    
    if ac == 3:
        r = min(r, mc)
    
    vt[x][y] = True
    
    for d in range(4):
        nx = x + dx[d]
        ny = y + dy[d]
        
        if nx < 0 or nx >= 5 or ny < 0 or ny >= 5:
            continue
        if li[nx][ny] == -1:
            continue
        if not vt[nx][ny]:
            dfs(nx, ny, mc + 1, ac + li[nx][ny])
            
    vt[x][y] = False
    
li = [list(map(int, input().split())) for _ in range(5)]
x, y = map(int, input().split())

assert li[x][y] != -1

r = inf
vt = [[False] * 5 for _ in range(5)]
dfs(x, y, 0, li[x][y])

if r == inf:
    print(-1)
else:
    print(r)
```

https://www.acmicpc.net/problem/26078 곰곰이와 토너먼트

```python
mod = 998244353

def nPk(n, k):
    if n == 0:
        return 0
    if n < k:
        return 0
    
    r = 1
    for i in range(n, n - k, -1):
        r *= i
        r %= mod
    
    return r

k = int(input())
li = list(map(int, input().split()))
prize = list(map(int, input().split()))

less_c = 0
cri = li[0]
for i in range(1, 2 ** k):
    if li[i] < cri:
        less_c += 1

factos = [1] * 300000
for num in range(1, 300000):
    factos[num] = factos[num - 1] * num
    factos[num] %= mod
        
denom = factos[2 ** k - 1]

total = 0
for cur_k in range(1, k + 1):
    num = nPk(less_c, 2 ** cur_k - 1) % mod
    
    num *= factos[2 ** k - 2 ** cur_k]
    
    num %= mod
    
    r = num * pow(denom, mod - 2, mod) % mod
        
    r *= prize[cur_k - 1]
    r %= mod
    
    total += r
    total %= mod
    
print(total)
```

https://www.acmicpc.net/problem/12920 평범한 배낭 2

```python
from heapq import heappush, heappop
import io, os, sys
input = io.BytesIO(os.read(0,os.fstat(0).st_size)).readline

inf = float('inf')

n, m = map(int, input().split())
li = [list(map(int, input().split())) for _ in range(n)]

dp = [0] * (m + 1)

for w, value, count in li:
    for end in range(m, m - w, -1):
        heap = []
        last_cand_index = -1
        for cur_c in range(1, count + 1):
            cand_index = end - w * cur_c
            if cand_index < 0:
                break
            last_cand_index = cand_index
            cand_nv = dp[cand_index] + cur_c * value
            heappush(heap, (-cand_nv, cand_index))
                     
        if heap:
            dp[end] = max(dp[end], -heap[0][0])
        
        last_cand_index -= w
        ww = value * (count + 1)
        added = 1
        for cur_i in range(end - w, -1, -w):
            while heap and heap[0][1] >= cur_i:
                heappop(heap)
            
            if last_cand_index >= 0:
                cand_nv = dp[last_cand_index] + ww
                heappush(heap, (-cand_nv, last_cand_index))
                
            if heap:
                dp[cur_i] = max(dp[cur_i], -heap[0][0] - added * value)
            
            ww += value
            added += 1
            last_cand_index -= w
            
result = dp[-1]

print(result)
```

```python
import io, os, sys
input = io.BytesIO(os.read(0,os.fstat(0).st_size)).readline

# https://github.com/tatyam-prime/SortedSet/blob/main/SortedSet.py
import math
from bisect import bisect_left, bisect_right
from typing import Generic, Iterable, Iterator, TypeVar, Union, List
T = TypeVar('T')

class SortedSet(Generic[T]):
    BUCKET_RATIO = 50
    REBUILD_RATIO = 170

    def _build(self, a=None) -> None:
        "Evenly divide `a` into buckets."
        if a is None: a = list(self)
        size = self.size = len(a)
        bucket_size = int(math.ceil(math.sqrt(size / self.BUCKET_RATIO)))
        self.a = [a[size * i // bucket_size : size * (i + 1) // bucket_size] for i in range(bucket_size)]
    
    def __init__(self, a: Iterable[T] = []) -> None:
        "Make a new SortedSet from iterable. / O(N) if sorted and unique / O(N log N)"
        a = list(a)
        if not all(a[i] < a[i + 1] for i in range(len(a) - 1)):
            a = sorted(set(a))
        self._build(a)

    def __iter__(self) -> Iterator[T]:
        for i in self.a:
            for j in i: yield j

    def __reversed__(self) -> Iterator[T]:
        for i in reversed(self.a):
            for j in reversed(i): yield j
    
    def __len__(self) -> int:
        return self.size
    
    def __repr__(self) -> str:
        return "SortedSet" + str(self.a)
    
    def __str__(self) -> str:
        s = str(list(self))
        return "{" + s[1 : len(s) - 1] + "}"

    def _find_bucket(self, x: T) -> List[T]:
        "Find the bucket which should contain x. self must not be empty."
        for a in self.a:
            if x <= a[-1]: return a
        return a

    def __contains__(self, x: T) -> bool:
        if self.size == 0: return False
        a = self._find_bucket(x)
        i = bisect_left(a, x)
        return i != len(a) and a[i] == x

    def add(self, x: T) -> bool:
        "Add an element and return True if added. / O(√N)"
        if self.size == 0:
            self.a = [[x]]
            self.size = 1
            return True
        a = self._find_bucket(x)
        i = bisect_left(a, x)
        if i != len(a) and a[i] == x: return False
        a.insert(i, x)
        self.size += 1
        if len(a) > len(self.a) * self.REBUILD_RATIO:
            self._build()
        return True

    def discard(self, x: T) -> bool:
        "Remove an element and return True if removed. / O(√N)"
        if self.size == 0: return False
        a = self._find_bucket(x)
        i = bisect_left(a, x)
        if i == len(a) or a[i] != x: return False
        a.pop(i)
        self.size -= 1
        if len(a) == 0: self._build()
        return True
    
    def lt(self, x: T) -> Union[T, None]:
        "Find the largest element < x, or None if it doesn't exist."
        for a in reversed(self.a):
            if a[0] < x:
                return a[bisect_left(a, x) - 1]

    def le(self, x: T) -> Union[T, None]:
        "Find the largest element <= x, or None if it doesn't exist."
        for a in reversed(self.a):
            if a[0] <= x:
                return a[bisect_right(a, x) - 1]

    def gt(self, x: T) -> Union[T, None]:
        "Find the smallest element > x, or None if it doesn't exist."
        for a in self.a:
            if a[-1] > x:
                return a[bisect_right(a, x)]

    def ge(self, x: T) -> Union[T, None]:
        "Find the smallest element >= x, or None if it doesn't exist."
        for a in self.a:
            if a[-1] >= x:
                return a[bisect_left(a, x)]
    
    def __getitem__(self, x: int) -> T:
        "Return the x-th element, or IndexError if it doesn't exist."
        if x < 0: x += self.size
        if x < 0: raise IndexError
        for a in self.a:
            if x < len(a): return a[x]
            x -= len(a)
        raise IndexError
    
    def index(self, x: T) -> int:
        "Count the number of elements < x."
        ans = 0
        for a in self.a:
            if a[-1] >= x:
                return ans + bisect_left(a, x)
            ans += len(a)
        return ans

    def index_right(self, x: T) -> int:
        "Count the number of elements <= x."
        ans = 0
        for a in self.a:
            if a[-1] > x:
                return ans + bisect_right(a, x)
            ans += len(a)
        return 

n, k = map(int, input().split())
li = [int(input()) for _ in range(n)]

sst = SortedSet()
for i in range(k):
    sst.add((li[i], i))
    
total = sst[(len(sst) - 1) // 2][0]

for i in range(k, n):
    sst.discard((li[i - k], i - k))
    sst.add((li[i], i))
    
    total += sst[(len(sst) - 1) // 2][0]
    
print(total)
```

```python
from collections import deque
import io, os, sys
input = io.BytesIO(os.read(0,os.fstat(0).st_size)).readline

n, m = map(int, input().split())

f_to_es = [[] for _ in range(n + 1)]
e_to_fs = [[] for _ in range(m)]

for e in range(m):
    x, y = map(int, input().split())
    
    for f in range(x, n + 1, y):
        f_to_es[f].append(e)
        e_to_fs[e].append(f)
        
a, b = map(int, input().split())

from_info = [[-1] * 3 for _ in range(n + 1)]
from_info[a] = [-1, -1, 0]

e_vt = [False] * m

queue_e = deque()
for e in f_to_es[a]:
    queue_e.append((e, a))
    e_vt[e] = True

dist_e = 1
while True:
    queue_f = deque()
    while queue_e:
        e, ff = queue_e.popleft()
        
        for f in e_to_fs[e]:
            if from_info[f][2] == -1:
                from_info[f] = [e, ff, dist_e]
                queue_f.append(f)
                
    queue_e = deque()
    while queue_f:
        f = queue_f.popleft()
        
        for e in f_to_es[f]:
            if not e_vt[e]:
                e_vt[e] = True
                queue_e.append((e, f))
                
    if not queue_e:
        break
        
    dist_e += 1
        
if from_info[b][2] == -1:
    print(-1)
else:
    print(from_info[b][2])
    
    result = []
    cur = b
    while True:
        e, fcur = from_info[cur][0], from_info[cur][1]
        result.append(e + 1)
        
        cur = fcur
        if cur == a:
            break
            
    for i in range(len(result) - 1, -1, -1):
        print(result[i])
```

https://www.acmicpc.net/problem/9463 순열 그래프

```python
import io, os, sys
input = io.BytesIO(os.read(0,os.fstat(0).st_size)).readline

def merge_sort(li):
    global count
    
    if len(li) == 1:
        return li
    
    left = merge_sort(li[:len(li) // 2])
    right = merge_sort(li[len(li) // 2:])
    
    nl = len(left)
    nr = len(right)
    
    nli = []
    l_i = r_i = 0
    while True:
        if left[l_i] <= right[r_i]:
            nli.append(left[l_i])
            count += r_i
            l_i += 1
            
            if l_i == nl:
                while r_i < nr:
                    nli.append(right[r_i])
                    r_i += 1
                break    
        else:
            nli.append(right[r_i])
            r_i += 1
            
            if r_i == nr:
                while l_i < nl:
                    nli.append(left[l_i])
                    count += r_i
                    l_i += 1
                break
    return nli 

for _ in range(int(input())):
    n = int(input())    
    a = list(map(int, input().split()))
    b = list(map(int, input().split()))

    d = {b[i]: i + 1 for i in range(n)}
    li = [d[v] for v in a]

    count = 0
    merge_sort(li)

    print(count)
```

https://www.acmicpc.net/problem/10917 Your life

```python
from collections import deque
import io, os, sys
input = io.BytesIO(os.read(0,os.fstat(0).st_size)).readline

inf = float('inf')

n, m = map(int, input().split())
li = [[] for _ in range(n + 1)]
for _ in range(m):
    x, y = map(int, input().split())
    
    li[x].append(y)
    
queue = deque()
dists = [-1] * (n + 1)

queue.append(1)
dists[1] = 0

while queue:
    value = queue.popleft()
    
    for v in li[value]:
        if dists[v] == -1:
            dists[v] = dists[value] + 1
            queue.append(v)
            
print(dists[n])
```

https://www.acmicpc.net/problem/14452 Cow Dance Show

```python
from heapq import heappush, heappop
import io, os, sys
input = io.BytesIO(os.read(0,os.fstat(0).st_size)).readline

n, t = map(int, input().split())
li = [int(input()) for _ in range(n)]

start = 1
end = n
r = n
while start <= end:
    mid = (start + end) // 2
    
    heap = []
    cur_i = 0
    for cur_t in range(t + 1):
        while heap and heap[0] == cur_t:
            heappop(heap)
            
        while len(heap) < mid and cur_i < n:
            heappush(heap, cur_t + li[cur_i])
            
            cur_i += 1
            
    if not heap:
        r = mid
        end = mid - 1
    else:
        start = mid + 1
        
print(r)
```

https://www.acmicpc.net/problem/17455 kdh9949

```python
import io, os, sys
sys.setrecursionlimit(200100)
input = io.BytesIO(os.read(0, os.fstat(0).st_size)).readline

def dfs(value):
    global is_cycle
    global total_max_len
    
    vt[value] = True
    on_route[value] = True
    
    num = nums[value]
    to_find = num + 1
    if to_find == 3:
        to_find = 0
    
    max_len = 0
    for v in li[value]:
        if nums[v] == to_find:
            if vt[v]:
                if on_route[v]:
                    is_cycle = True
                else:
                    max_len = max(max_len, maxes[v])
            else:
                temp_len = dfs(v)
                max_len = max(max_len, temp_len)
                
    if nums[value] == 2:
        max_len += 1
        
    maxes[value] = max_len
    total_max_len = max(total_max_len, max_len)
        
    on_route[value] = False
    return max_len
    
n, m = map(int, input().split())
s = input().decode().rstrip()

nums = [0] * n
for i, v in enumerate(s):
    if v == 'K':
        nums[i] = 0
    elif v == 'D':
        nums[i] = 1
    else:
        nums[i] = 2

get_value = lambda x: int(x) - 1
li = [[] for _ in range(n)]
for _ in range(m):
    u, v = map(get_value, input().split())
    
    li[u].append(v)
    li[v].append(u)
    
maxes = [0] * n

total_max_len = 0
is_cycle = False
vt = [False] * n
on_route = [False] * n
for value in range(n):
    if nums[value] == 0 and not vt[value]:
        dfs(value)
        
if is_cycle or total_max_len == 0:
    result = -1
else:
    result = total_max_len * 3

print(result)
```

https://www.acmicpc.net/problem/25402 트리와 쿼리

```python
from collections import deque
import io, os, sys
sys.setrecursionlimit(260000)
input = io.BytesIO(os.read(0,os.fstat(0).st_size)).readline

def nC2(n):
    if n <= 1:
        return 0
    return n * (n - 1) // 2

def find(v):
    if v != pa[v]:
        pa[v] = find(pa[v])
        
    return pa[v]

def union(a, b):
    if rank[a] < rank[b]:
        a, b = b, a
        
    pa[b] = a
    if rank[a] == rank[b]:
        rank[a] += 1
        
    delta = 0
    
    delta -= nC2(ct[a])
    delta -= nC2(ct[b])
    ct[a] += ct[b]
    
    delta += nC2(ct[a])
    
    return delta  

n = int(input())

li = [[] for _ in range(n + 1)]
for _ in range(n - 1):
    a, b = map(int, input().split())
    
    li[a].append(b)
    li[b].append(a)
    
parents = [-1] * (n + 1)

vt = [False] * (n + 1)
queue = deque()

vt[1] = True
queue.append(1)
while queue:
    value = queue.popleft()
    
    for v in li[value]:
        if not vt[v]:
            vt[v] = True
            parents[v] = value
            queue.append(v)
        
pa = list(range(n + 1))
rank = [0] * (n + 1)
ct = [1] * (n + 1)
    
q = int(input())
for _ in range(q):
    nn, *temp = map(int, input().split())

    st = set(temp)
    
    total = 0
    for value in st:
        v = parents[value]
        
        if v in st:
            rvalue = find(value)
            rv = find(v)        
            
            if rvalue != rv:
                total += union(rvalue, rv)
                    
    print(total)

    for v in st:
        pa[v] = v
        rank[v] = 0
        ct[v] = 1
```

https://www.acmicpc.net/problem/6890 Simple Encryption

```python
key = input()
s = input()

key_len = len(key)

li = []
for v in s:
    if v.isupper():
        li.append(v)

for i in range(len(li)):
    le = key[i % key_len]
    
    diff = ord(le) - 65
    
    temp = ord(li[i]) + diff
    if temp > 90:
        temp -= 26
        
    li[i] = chr(temp)
    
result = ''.join(li)

print(result)
```

https://www.acmicpc.net/problem/6918 The Students' Council Breakfast

```python
li = [int(input()) for _ in range(4)]
amount = int(input())

result = []
min_c = amount + 5
for ac in range(amount + 1):
    for bc in range(amount + 1):
        for cc in range(amount + 1):
            for dc in range(amount + 1):
                cs = [ac, bc, cc, dc]
                
                total = 0
                for i in range(4):
                    total += li[i] * cs[i]
                
                if total == amount:
                    min_c = min(min_c, sum(cs))
                    result.append(cs)
 
if len(result) == 0:
    min_c = 0
                    
for ac, bc, cc, dc in result:
    print(f'# of PINK is {ac} # of GREEN is {bc} # of RED is {cc} # of ORANGE is {dc}')
    
print(f'Total combinations is {len(result)}.')
print(f'Minimum number of tickets to print is {min_c}.')
```

```python
from math import gcd

num = int(input())
denom = int(input())

if num < denom:
    g = gcd(num, denom)
    
    num //= g
    denom //= g
    
    if num == 0:
        print(0)
    else:
        print(f'{num}/{denom}')
else:
    front = num // denom
    num %= denom
    
    g = gcd(num, denom)
    
    num //= g
    denom //= g
    
    if num == 0:
        print(front)
    else:
        print(f'{front} {num}/{denom}')
```

https://www.acmicpc.net/problem/17130 토끼가 정보섬에 올라온 이유

```python
inf = float('inf')

n, m = map(int, input().split())
li = [input() for _ in range(n)]

dp = [[-inf] * m for _ in range(n)]

sx = sy = -1
for i in range(n):
    for j in range(m):
        if li[i][j] == 'R':
            sx = i
            sy = j
            break
    if sx != -1:
        break

dp[sx][sy] = 0

for j in range(sy + 1, m):
    for i in range(n):
        if li[i][j] == '#':
            continue
        
        max_earn = -inf
        for fi in range(i - 1, i + 2):
            if fi < 0 or fi >= n:
                continue
            
            max_earn = max(max_earn, li[fi][j - 1])
            
        if li[i][j] == 'C':
            max_earn += 1
            
        dp[i][j] = max_earn
        
result = -inf
for i in range(n):
    for j in range(m):
        if li[i][j] == 'O':
            result = max(result, dp[i][j])
            
if result == -inf:
    print(-1)
else:
    print(result)
```
