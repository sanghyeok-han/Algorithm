### N-Queen 9663

2:54 ~ 3.07, 3.15 ~ 3.47

```python
# 시간 초과 예상
n = int(input())

def nqueen(level, li):
    global count
    
    possible_i_s = set(range(n)) ^ {former_i for former_level, former_i in li}
    for i in possible_i_s:  
        is_suitable = True     
        for former_level, former_i in li:
            if level - former_level == abs(i - former_i):
                is_suitable = False
                break
                
        if is_suitable:          
            if level + 1 == n:
                count += 1
            else:
                nqueen(level + 1, li + ((level, i),))
    
count = 0
for i in range(n):
    nqueen(1, ((0, i),))
    
print(count)
```

```python
# 시간 초과 예상
import time

s = time.time()
n = int(input())


def nqueen(level, li):
    count = 0

    for i in range(n):
        is_suitable = True
        for former_level, former_i in enumerate(li):
            if i == former_i or level - former_level == abs(i - former_i):
                is_suitable = False
                break
        if is_suitable:
            if level + 1 == n:
                count += 1
            else:
                li.append(i)
                count += nqueen(level + 1, li)
                li.pop()
    return count
    
count = 0
li = []
for i in range(n):
    li.append(i)
    count += nqueen(1, li)
    li.pop()
    
print(count)
e = time.time()
print(e - s)
```

```python
from collections import deque
import time

s = time.time()
n = int(input())

def nqueen(level, d):
    global count
    
    for i in range(n):
        if not [1 for former_level, former_i in d if i == former_i or level - former_level == abs(i - former_i)]:
            if level + 1 == n:
                count += 1
            else:
                d.append((level, i))
                nqueen(level + 1, d)
                d.pop()
    
count = 0

d = deque()
for i in range(n):
    d.append((0, i))
    nqueen(1, d)
    d.pop()
    
    
print(count)
e = time.time()
print(e - s)
```

### 알파벳 1987

```python
# Key Error
import string
import sys
# input = sys.stdin.readline

r, c = map(int, input().split())

d = {v: i for i, v in enumerate(string.ascii_uppercase)}

def alpha_to_num(alpha):
    return d[alpha]

li = []
for _ in range(r):
    li.append(tuple(map(alpha_to_num, input())))

visited = [False] * 26
max_count = 0
stack = []
visited[li[0][0]] = True
stack.append(((0, 0), visited, 1))

dx = [-1, 1, 0, 0]
dy = [0, 0, -1, 1]

    
while stack:
    (x, y), visited, count = stack.pop()
    
    for i in range(4):
        a = x + dx[i]
        b = y + dy[i]
        
        if 0 <= a < r and 0 <= b < c:
            if not visited[li[a][b]]:
                new_visited = visited.copy()
                new_visited[li[a][b]] = True
                stack.append(((a, b), new_visited, count + 1))
            else:
                if count > max_count:
                    max_count = count
                    
print(max_count)
```

```python
# 다른 사람 코드 (gkgg123)

# 1987 알파벳


R,C = map(int,input().split())

arr = [list(input()) for _ in range(R)]
check = [['']*C for _ in range(R)]

stack = [(0,0,1,arr[0][0])]
result = 0
dx = [-1,1,0,0]
dy = [0,0,1,-1]
while stack:
    x,y,cnt,total = stack.pop()
    if result < cnt:
        result = cnt
    if result == 26:
        break
    for i in range(4):
        nx = x + dx[i]
        ny = y + dy[i]
        if 0 <= nx < R and 0<= ny <C:
            if arr[nx][ny] not in total:
                temp = total + arr[nx][ny]
                if check[nx][ny] != temp:
                    check[nx][ny] = temp
                    stack.append((nx,ny,cnt+1,temp))
print(result)
```

```python
# 다른 사람 코드 (wider93)

# https://www.acmicpc.net/problem/1987
from collections import defaultdict as dd

def main():
    m = 26  # number of upper alphabets
    r, c = map(int, input().split())
    board = [list(map(lambda s: 1 << (ord(s)-65), input())) for i in range(r)]  # save char is int 0 ~ 25

    if r > c:
        board = [[board[i][j] for i in range(r)] for j in range(c)]  # Transpose operation. now r <= c. maybe unnecessary
        r, c = c, r
    x = dd(lambda: set())  # x[k] represents (k//c, k%c)

    x[0] = {board[0][0]}  # mark as visited]
    z = {}
    def adj(k):
        i, j = k//c, k%c
        if i:
            yield k-c
        if j:
            yield k-1
        if c-j-1:
            yield k+1
        if r-i-1:
            yield k+c

    que = {0}
    for k in range(1, m):
        new_que = set()
        y = dd(lambda: set())
        if not que:
            break
        while que:
            p = que.pop()
            for q in adj(p):
                bean = board[q//c][q%c]
                flag = True
                for b in x[p]:
                    if not bean & b:
                        y[q].add(b | bean)
                        if flag:
                            new_que.add(q)
                            flag = False
        que = new_que
        x = y
        for key in x.keys():
            z[key] = k
    if z:
        #print(z)
        return max(z.values())+1
    return 1

print(main())
```

### 암호 만들기 1759

```python
def alpha_to_num(alpha):
    return ord(alpha) -  97

def num_to_alpha(num):
    return chr(num + 97) 

l, c = map(int, input().split())
li = list(map(alpha_to_num, input().split()))

vowels = {alpha_to_num(v) for v in "aeiou"}
result_li = []

for v in li:
    remain_set = set(li)
    remain_set.remove(v)
    case = []
    case.append(v)

    stack = []
    stack.append((case[:], remain_set))
    while stack:
        case, remain_set = stack.pop()
        
        for w in remain_set:
            if case[-1] < w:
                new_case = case + [w]
                
                if len(new_case) == l:
                    if vowels & set(new_case) and len(set(new_case) - vowels) >= 2:
                        result_li.append("".join(map(num_to_alpha, new_case)))
                else:   
                    stack.append((new_case, remain_set - set([w])))
            
result_li.sort()
print("\n".join(result_li))
```
