```python
from collections import deque
import sys, os, io
input = io.BytesIO(os.read(0,os.fstat(0).st_size)).readline

inf = float('inf')

def bfs():
    levels = [-1] * (n + m + 2)
    levels[source] = 0
    
    queue = deque()
    queue.append(source)
    
    while queue:
        value = queue.popleft()
        
        for v in graph[value]:
            rem = c[value][v] - f[value][v]
            if levels[v] == -1 and rem > 0:
                levels[v] = levels[value] + 1
                
                if v != sink:
                    queue.append(v)
                
    return levels
    
def dfs(value, can_f):
    if value == sink:
        return can_f
    
    total_out_flow = 0
    for v in graph[value]:
        if levels[v] == levels[value] + 1 and f[value][v] < c[value][v]:
            rem = c[value][v] - f[value][v]
            n_can_f = min(can_f, rem)
            
            to_flow = dfs(v, n_can_f)
            f[value][v] += to_flow
            f[v][value] -= to_flow
            
            total_out_flow += to_flow
            can_f -= to_flow
            
    return total_out_flow

n, m = map(int, input().split())
A = list(map(int, input().split()))
B = list(map(int, input().split()))

c = [[0] * (n + m + 2) for _ in range(n + m + 2)]
f = [[0] * (n + m + 2) for _ in range(n + m + 2)]
graph = [[] for _ in range(n + m + 2)]
for b_num in range(m):
    temp = list(map(int, input().split()))
    
    for i in range(n):
        a_num = i + m
            
        graph[a_num].append(b_num)
        graph[b_num].append(a_num)
        c[b_num][a_num] = temp[i]
        
source = n + m
sink = n + m + 1

for b_num in range(m):
    graph[source].append(b_num)
    graph[b_num].append(source)
    c[source][b_num] = B[b_num]
    
for a_num in range(m, m + n):
    graph[a_num].append(sink)
    graph[sink].append(a_num)
    c[a_num][sink] = A[a_num - m]

total = 0
while True:
    levels = bfs()
    
    if levels[sink] == -1:
        break
    
    to_flow = dfs(source, inf)
    total += to_flow
    
print(total)
```

```python
from collections import deque

inf = float('inf')

def bfs():
    queue = deque()
    dists[0] = inf
    
    for x in range(1, n + 1):
        if xs[x] == 0:
            dists[x] = 0
            queue.append(x)
        else:
            dists[x] = inf
            
    while queue:
        x = queue.popleft()
        
        for y in wants[x]:
            nx = ys[y]
            
            if dists[nx] == inf:
                dists[nx] = dists[x] + 1
                queue.append(nx)
                
    return dists[0] != inf
            
def dfs(x):
    for y in wants[x]:
        nx = ys[y]
        
        if dists[nx] == dists[x] + 1 and (ys[y] == 0 or dfs(nx)):
            xs[x] = y
            ys[y] = x
            
            return True
    
    dists[x] = inf
    
    return False

n, m = map(int, input().split())

wants = [[] for _ in range(n + 1)]
for cow_i in range(1, n + 1):
    _, *temp = map(int, input().split())
    wants[cow_i] = temp
    
xs = [0] * (n + 1)
ys = [0] * (m + 1)
dists = [inf] * (n + 1)

total = 0
while True:
    if not bfs():
        break
        
    for x in range(1, n + 1):
        if xs[x] == 0:
            if dfs(x):
                total += 1
                
print(total)
```

```python
from collections import deque
import sys, os, io
input = io.BytesIO(os.read(0,os.fstat(0).st_size)).readline

inf = float('inf')

def bfs():
    dists = [inf] * (n + 2)
    queue = deque()
    
    dists[sc] = 0
    queue.append(sc)
    while queue:
        value = queue.popleft()
        
        for v in G[value]:
            if F[value][v] < C[value][v] and dists[v] == inf:
                dists[v] = dists[value] + 1
                
                if v != sk:
                    queue.append(v)
    
    return dists

def dfs(value, to_flow):
    if value == sk:
        return to_flow
    
    total_can_flow = 0
    
    for v in G[value]:
        rem = C[value][v] - F[value][v]
        if dists[value] + 1 == dists[v] and rem > 0:
            n_to_flow = min(to_flow, rem)
            
            can_flow = dfs(v, n_to_flow)
                
            total_can_flow += can_flow
            to_flow -= can_flow
            F[value][v] += can_flow
            F[v][value] -= can_flow
            
    return total_can_flow

n, m = map(int, input().split())
ps = list(map(int, input().split()))

for i in range(n):
    ps[i] = -ps[i]

sc = n
sk = n + 1

F = [[0] * (n + 2) for _ in range(n + 2)]
C = [[0] * (n + 2) for _ in range(n + 2)]
G = [[] for _ in range(n + 2)]

out_sum = 0
in_sum = 0
    
edges = []
demands = ps.copy()
for _ in range(m):
    u, v, s, e = map(int, input().split())
    u -= 1 
    v -= 1

    edges.append((u, v, s, e))
    G[u].append(v)
    G[v].append(u)
    C[u][v] = e - s

    demands[u] += s
    demands[v] -= s

out_sum = 0
in_sum = 0
for i in range(n):
    if demands[i] > 0:
        in_sum += demands[i]
    elif demands[i] < 0:
        out_sum += -demands[i]

if in_sum != out_sum:
    print(-1)
else:
    for i in range(n):
        if demands[i] < 0:
            G[sc].append(i)
            C[sc][i] = -demands[i]
        elif demands[i] > 0:
            G[i].append(sk)
            C[i][sk] = demands[i]

    total_flow = 0
    while True:
        dists = bfs()

        if dists[sk] == inf:
            break

        to_flow = dfs(sc, inf)
        total_flow += to_flow
    
    if total_flow != out_sum:
        print(-1)
    else:
        print(1)
        for u, v, s, e in edges:
            result = F[u][v] + s
            total_flow += s

            print(result)
```
