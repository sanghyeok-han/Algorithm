### 숨바꼭질 5 17071

```python
## 71%에서 시간 초과

n, k = map(int, input().split())
o_set = set()

def bfs(n, k):
    num = 500000
    second = 0
    
    st = set()
    st.add(n)
    new_st = set()
    
    while True:
        for value in st:
            if value == k:
                return second
            
            v = value * 2
            w = value + 1
            if v <= num:
                new_st.add(v)
                new_st.add(w)
            else:
                if w <= num:
                    new_st.add(w)

            v = value - 1
            if 0 <= v:
                new_st.add(v)
        
        st = new_st.copy()
        new_st = o_set.copy()
            
        k += second + 1
        
        if k > num:
            return -1
        second += 1

    return -1
        
result = bfs(n, k)

print(result)
```

### 통나무 옮기기  1938

```python
from collections import deque

n = int(input())
li = [input() for _ in range(n)]

start_center = ()
start_status = 0 # 0: 가로, 1: 세로
end_center = ()
end_status = 0
for i in range(n):
    for j in range(n):
        if li[i][j] == 'B':
            if not start_center:
                if i + 1 >= n:
                    start_center = (i, j + 1)
                    start_status = 0
                else:
                    if li[i + 1][j] == 'B':
                        start_center = (i + 1, j)
                        start_status = 1
                    else:
                        start_center = (i, j + 1)
                        start_status = 0
        elif li[i][j] == 'E':
            if not end_center:
                if i + 1 >= n:
                    end_center = (i, j + 1)
                    end_status = 0
                else:
                    if li[i + 1][j] == 'E':
                        end_center = (i + 1, j)
                        end_status = 1
                    else:
                        end_center = (i, j + 1)
                        end_status = 0
        
distance = [[[float('inf')] * n for _ in range(n)] for _ in range(2)] # distance[status][x][y]
distance[start_status][start_center[0]][start_center[1]] = 0
queue = deque()
queue.append((start_center[0], start_center[1], start_status))

while queue:
    x, y, status = queue.popleft()

    # 위쪽으로 이동하는 경우
    nx, ny = x - 1, y
    if status == 0:
        if 0 <= nx:
            if not (li[nx][ny - 1] == '1' or li[nx][ny] == '1' or li[nx][ny + 1] == '1'):         
                if distance[status][x][y] + 1 < distance[status][nx][ny]:
                    distance[status][nx][ny] = distance[status][x][y] + 1
                    queue.append((nx, ny, status))
    else:
        if 0 <= nx - 1:
            if li[nx - 1][ny] != '1':
                if distance[status][x][y] + 1 < distance[status][nx][ny]:
                    distance[status][nx][ny] = distance[status][x][y] + 1
                    queue.append((nx, ny, status))
                
    # 아래쪽으로 이동하는 경우
    nx, ny = x + 1, y
    if status == 0:
        if nx <= n - 1:
            if not (li[nx][ny - 1] == '1' or li[nx][ny] == '1' or li[nx][ny + 1] == '1'):
                if distance[status][x][y] + 1 < distance[status][nx][ny]:
                    distance[status][nx][ny] = distance[status][x][y] + 1
                    queue.append((nx, ny, status))
    else:
        if nx + 1 <= n - 1:
            if li[nx + 1][ny] != '1':
                if distance[status][x][y] + 1 < distance[status][nx][ny]:
                    distance[status][nx][ny] = distance[status][x][y] + 1
                    queue.append((nx, ny, status))
                
    # 왼쪽으로 이동하는 경우
    nx, ny = x, y - 1
    if status == 0:
        if 0 <= ny - 1:
            if li[nx][ny - 1] != '1':
                if distance[status][x][y] + 1 < distance[status][nx][ny]:
                    distance[status][nx][ny] = distance[status][x][y] + 1
                    queue.append((nx, ny, status))
    else:
        if 0 <= ny:
            if not (li[nx - 1][ny] == '1' or li[nx][ny] == '1' or li[nx + 1][ny] == '1'):    
                if distance[status][x][y] + 1 < distance[status][nx][ny]:
                    distance[status][nx][ny] = distance[status][x][y] + 1
                    queue.append((nx, ny, status))    
                
    # 아래쪽으로 이동하는 경우
    nx, ny = x, y + 1
    if status == 0:
        if ny + 1 <= n - 1:
            if li[nx][ny + 1] != '1':
                if distance[status][x][y] + 1 < distance[status][nx][ny]:
                    distance[status][nx][ny] = distance[status][x][y] + 1
                    queue.append((nx, ny, status))
    else:
        if ny <= n - 1:
            if not (li[nx - 1][ny] == '1' or li[nx][ny] == '1' or li[nx + 1][ny] == '1'):
                if distance[status][x][y] + 1 < distance[status][nx][ny]:
                    distance[status][nx][ny] = distance[status][x][y] + 1
                    queue.append((nx, ny, status))       
        
    # 회전하는 경우 (status를 바꾸는 경우)
    if status == 0:
        if 0 <= x - 1 and x + 1 <= n - 1:
            if '1' not in (li[x - 1][y - 1], li[x - 1][y], li[x - 1][y + 1], li[x + 1][y - 1], li[x + 1][y], li[x + 1][y + 1]):
                if distance[status][x][y] + 1 < distance[1 ^ status][x][y]:
                    distance[1 ^ status][x][y] = distance[status][x][y] + 1
                    queue.append((x, y, 1 ^ status))
    else:
        if 0 <= y - 1 and y + 1 <= n - 1:
            if '1' not in (li[x - 1][y - 1], li[x][y - 1], li[x + 1][y - 1], li[x - 1][y + 1], li[x][y + 1], li[x + 1][y + 1]):
                if distance[status][x][y] + 1 < distance[1 ^ status][x][y]:
                    distance[1 ^ status][x][y] = distance[status][x][y] + 1
                    queue.append((x, y, 1 ^ status))        
    
if distance[end_status][end_center[0]][end_center[1]] == float('inf'):
    print(0)
else:
    print(distance[end_status][end_center[0]][end_center[1]])
```
