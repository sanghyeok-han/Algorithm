### 로봇 청소기 14503
- 시뮬레이션

12:18 1:2

```python
# 맞았음, =을 ==로 쓴 것을 디버깅하는 것에 오래 걸렸음

# 0: 위쪽, 1: 오른쪽, 2: 아래쪽, 3: 왼쪽
# 0: 청소할 곳, 1: 벽, 2: 청소된 상태

n, m = map(int, input().split())
x, y, d = map(int, input().split())
li = [list(map(int, input().split())) for _ in range(n)]

# 위쪽, 오른쪽, 아래쪽, 왼쪽
dx = [-1, 0, 1, 0]
dy = [0, 1, 0, -1]

li[x][y] = 2
count = 1
while True:
    not_availabe_count = 0
    for i in range(4):
        nx = x + dx[i]
        ny = y + dy[i]
        if li[nx][ny] != 0:
            not_availabe_count += 1
            
    if not_availabe_count == 4:
        # 현재 위쪽을 보고 있는 경우
        if d == 0:
            if li[x + 1][y] == 1:
                break
            else:
                x += 1
        # '' 오른쪽
        elif d == 1:
            if li[x][y - 1] == 1:
                break
            else:
                y -= 1
        # '' 아래쪽
        elif d == 2:
            if li[x - 1][y] == 1:
                break
            else:
                x -= 1    
        # '' 왼쪽
        else:
            if li[x][y + 1] == 1:
                break
            else:
                y += 1
    else:
        # 현재 위쪽
        if d == 0:
            d = 3
            if li[x][y - 1] == 0:   
                y -= 1
                li[x][y] = 2 
                count += 1
        # 현재 오른쪽
        elif d == 1:
            d = 0
            if li[x - 1][y] == 0:   
                x -= 1
                li[x][y] = 2
                count += 1
        # 현재 아래쪽
        elif d == 2:
            d = 1
            if li[x][y + 1] == 0:   
                y += 1
                li[x][y] = 2
                count += 1
       # 현재 왼쪽
        else:
            d = 2
            if li[x + 1][y] == 0:
                x += 1
                li[x][y] = 2
                count += 1
                
print(count)
```

### 뱀 3190
- 자료 구조, 시뮬레이션, 덱, 큐

1:16

```python
from collections import deque

n = int(input())
li = [[0] * n for _ in range(n)] # 0: 빈칸, 1: 뱀이 차지하는 칸, 2: 사과 

k = int(input())
for _ in range(k):
    x, y = map(int, input().split())
    x, y = x - 1, y - 1
    li[x][y] = 2
    
l = int(input())
snake_infos = deque()
for _ in range(l):
    sec, turn = input().split()
    sec = int(sec)
    snake_infos.append((sec, turn))

li[0][0] = 1
snake_body = deque()
snake_body.append((0, 0))

# 0: 상, 1: 하, 2: 좌, 3: 우
dx = [-1, 1, 0, 0]
dy = [0, 0, -1, 1]

def change_dir(current_dir, turn):
    if current_dir == 0:
        if turn == 'L':
            current_dir = 2
        else:
            current_dir = 3
    elif current_dir == 1:
        if turn == 'L':
            current_dir = 3
        else:
            current_dir = 2                
    elif current_dir == 2:
        if turn == 'L':
            current_dir = 1
        else:
            current_dir = 0            
    else:
        if turn == 'L':
            current_dir = 0
        else:
            current_dir = 1
    return current_dir

current_snake_info = snake_infos.popleft()

current_dir = 3
current_sec = 1
x, y = 0, 0
while True:
    x += dx[current_dir]
    y += dy[current_dir]
    
    if x < 0 or x >= n or y < 0 or y >= n:
        break
    if li[x][y] == 1:
        break
        
    if li[x][y] == 2:
        snake_body.append((x, y))
        li[x][y] = 1
    else:
        snake_body.append((x, y))
        li[x][y] = 1
        tail_x, tail_y = snake_body.popleft()
        li[tail_x][tail_y] = 0
        
    if current_snake_info:
        if current_sec == current_snake_info[0]:
            current_dir = change_dir(current_dir, current_snake_info[1])
                    
            if snake_infos:
                current_snake_info = snake_infos.popleft()
            else:
                current_snake_info = ""
            
    current_sec += 1
            
print(current_sec)
```

### 로봇 시뮬레이션 2174
- 시뮬레이션

```python
a, b = map(int, input().split())
n, m = map(int, input().split())

board = [[0] * a for _ in range(b)] # 0은 빈칸, 1 이상은 각 번호에 해당하는 로봇

# 방향은 0: 위쪽, 1: 아래쪽, 2: 왼쪽, 3: 오른쪽
robots = [[]]
for i in range(1, n + 1):
    x, y, d = input().split()
    x, y = int(x), int(y)
    
    x, y = b - y, x - 1
    d = "NSWE".index(d)
    
    robots.append([x, y, d])
    board[x][y] = i
    
commands = []
for _ in range(m):
    robot_num, command_type, recur_count = input().split()
    robot_num, recur_count = int(robot_num), int(recur_count)
    commands.append((robot_num, command_type, recur_count))
    
def turn(command_type, d):
    if d == 0:
        if command_type == 'L':
            d = 2
        else:
            d = 3
    elif d == 1:
        if command_type == 'L':
            d = 3
        else:
            d = 2
    elif d == 2:
        if command_type == 'L':
            d = 1
        else:
            d = 0
    else:
        if command_type == 'L':
            d = 0
        else:
            d = 1
            
    return d

def move(x, y, d):
    if d == 0:
        return (x - 1, y)
    elif d == 1:
        return (x + 1, y)
    elif d == 2:
        return (x, y - 1)
    else:
        return (x, y + 1)
    
def execute_command(robot_num, command_type, recur_count):
    current_robot = robots[robot_num]
    x, y, d = current_robot
    
    for _ in range(recur_count):
        if command_type == 'F':
            board[x][y] = 0
            x, y = move(x, y, d)
            if x < 0 or x >= b or y < 0 or y >= a:
                return f'Robot {robot_num} crashes into the wall'
            if board[x][y] >= 1:
                return f'Robot {robot_num} crashes into robot {board[x][y]}'
            board[x][y] = robot_num
        else:
            d = turn(command_type, d)
        robots[robot_num] = [x, y, d]
            
    return 'OK'
    
for robot_num, command_type, recur_count in commands:
    result = execute_command(robot_num, command_type, recur_count)
    
    if result != 'OK':
        print(result)
        break
else:
    print(result)
```

```python
from collections import deque


wheels = [[]]
for _ in range(4):
    wheels.append(deque(map(int, input())))

k = int(input())
commands = [tuple(map(int, input().split())) for _ in range(k)]


def convert(d):
    if d == 1:
        return -1
    else:
        return 1
    

def get_connection_info():
    connection_info = [None]
    for i in range(1, 4):
        if wheels[i][2] == wheels[i + 1][6]:
            connection_info.append('same')
        else:
            connection_info.append('diff')
    return connection_info


def turn(wheel_num, d):
    wheels[wheel_num].rotate(d)
    
    
def get_total_score():
    total_score = 0
    for i in range(1, 5):
        if wheels[i][0] == 1:
            total_score += 2 ** (i - 1)
    return total_score


for command in commands:
    connection_info = get_connection_info()
    o_wheel_num, o_d = command 
    turn(o_wheel_num, o_d)
        
    wheel_num, d = o_wheel_num, o_d
    # 오른쪽으로 확인할 여지가 있는 경우
    while wheel_num < 4:
        if connection_info[wheel_num] == 'diff':
            d = convert(d)
            turn(wheel_num + 1, d)
        else:
            break

        wheel_num += 1
            
    wheel_num, d = o_wheel_num, o_d
    # 왼쪽으로 확인할 여지가 있는 경우
    while wheel_num > 1:
        if connection_info[wheel_num - 1] == 'diff':
            d = convert(d)
            turn(wheel_num - 1, d)
        else:
            break

        wheel_num -= 1
        
print(get_total_score())
```

### 14499 주사위 굴리기
- 시뮬레이션

```python
n, m, x, y, k = map(int, input().split())
li = [list(map(int, input().split())) for _ in range(n)]
commands = tuple(map(int, input().split())) # 1: 동, 2: 서, 3: 북, 4: 남

current = [0] * 6

dx = [0, 0, -1, 1]
dy = [1, -1, 0, 0]


def turn(current, command):    
    if command == 1:
        change_index = [3, 1, 2, 5, 0, 4] # i 위치에 있던 것을 change_index[i] 위치로 옮기기
    elif command == 2:
        change_index = [4, 1, 2, 0, 5, 3]
    elif command == 3:
        change_index = [2, 0, 5, 3, 4, 1]
    else:
        change_index = [1, 5, 0, 3, 4, 2]
        
    new_current = [0] * 6
    for i in range(6):
        new_current[i] = current[change_index[i]]
        
    return new_current
        

def move(command, x, y):
    nx = x + dx[command - 1]
    ny = y + dy[command - 1]
    
    if nx < 0 or nx >= n or ny < 0 or ny >= m:
        return False
    
    return (nx, ny)
    
        
for command in commands:
    check_result = move(command, x, y)
    
    if check_result:
        x, y = check_result
        current = turn(current, command)
        
        if li[x][y] == 0:
            li[x][y] = current[-1]
        else:
            current[-1] = li[x][y]
            li[x][y] = 0
            
        print(current[0])
```
