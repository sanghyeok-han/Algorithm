### 안티 팰린드롬 1545

```python
가장 많이 남은 문자의 개수가 (남은 자리의 수 + 1) // 2 + 1 이상이면 안됨
```

```python
from collections import Counter

def check(li):
    n = len(li)
    
    for i in range(n // 2):
        if li[i] == li[n - 1 - i]:
            return False
    return True

s = input()
n = len(s)

ct = Counter(s)
if ct.most_common()[0][1] >= (n + 1) // 2 + 1:
    print(-1)
else:
    li = sorted(list(v) for v in ct.items())
    
    r = []
    used = 0
    for i in range(len(li)):
        while used < n // 2 and li[i][1] >= 1:
            li[i][1] -= 1
            used += 1
            r.append(li[i][0])
            
    if n % 2 == 1:
        if ct.most_common()[0][1] == (n + 1) // 2:
            for i in range(len(li)):
                if li[i][0] == ct.most_common()[0][0]:
                    li[i][1] -= 1
                    r.append(li[i][0])
                    break
        else:
            for i in range(len(li)):
                if li[i][1] >= 1:
                    li[i][1] -= 1
                    r.append(li[i][0])
                    break
    
    right_r = []
    cur_check = n // 2 - 1
    for i in range(n // 2):
        for j in range(len(li)):
            if li[j][1] and li[j][0] != r[cur_check]:
                right_r.append(li[j][0])
                li[j][1] -= 1
                break
        cur_check -= 1
            
    r += right_r
    
    print(''.join(r))
```

```python
# https://github.com/tatyam-prime/SortedSet/blob/main/SortedMultiset.py
import math
from bisect import bisect_left, bisect_right, insort
from typing import Generic, Iterable, Iterator, TypeVar, Union, List
T = TypeVar('T')

class SortedMultiset(Generic[T]):
    BUCKET_RATIO = 50
    REBUILD_RATIO = 170

    def _build(self, a=None) -> None:
        "Evenly divide `a` into buckets."
        if a is None: a = list(self)
        size = self.size = len(a)
        bucket_size = int(math.ceil(math.sqrt(size / self.BUCKET_RATIO)))
        self.a = [a[size * i // bucket_size : size * (i + 1) // bucket_size] for i in range(bucket_size)]
    
    def __init__(self, a: Iterable[T] = []) -> None:
        "Make a new SortedMultiset from iterable. / O(N) if sorted / O(N log N)"
        a = list(a)
        if not all(a[i] <= a[i + 1] for i in range(len(a) - 1)):
            a = sorted(a)
        self._build(a)

    def __iter__(self) -> Iterator[T]:
        for i in self.a:
            for j in i: yield j

    def __reversed__(self) -> Iterator[T]:
        for i in reversed(self.a):
            for j in reversed(i): yield j
    
    def __len__(self) -> int:
        return self.size
    
    def __repr__(self) -> str:
        return "SortedMultiset" + str(self.a)
    
    def __str__(self) -> str:
        s = str(list(self))
        return "{" + s[1 : len(s) - 1] + "}"

    def _find_bucket(self, x: T) -> List[T]:
        "Find the bucket which should contain x. self must not be empty."
        for a in self.a:
            if x <= a[-1]: return a
        return a

    def __contains__(self, x: T) -> bool:
        if self.size == 0: return False
        a = self._find_bucket(x)
        i = bisect_left(a, x)
        return i != len(a) and a[i] == x

    def count(self, x: T) -> int:
        "Count the number of x."
        return self.index_right(x) - self.index(x)

    def add(self, x: T) -> None:
        "Add an element. / O(√N)"
        if self.size == 0:
            self.a = [[x]]
            self.size = 1
            return
        a = self._find_bucket(x)
        insort(a, x)
        self.size += 1
        if len(a) > len(self.a) * self.REBUILD_RATIO:
            self._build()

    def discard(self, x: T) -> bool:
        "Remove an element and return True if removed. / O(√N)"
        if self.size == 0: return False
        a = self._find_bucket(x)
        i = bisect_left(a, x)
        if i == len(a) or a[i] != x: return False
        a.pop(i)
        self.size -= 1
        if len(a) == 0: self._build()
        return True

    def lt(self, x: T) -> Union[T, None]:
        "Find the largest element < x, or None if it doesn't exist."
        for a in reversed(self.a):
            if a[0] < x:
                return a[bisect_left(a, x) - 1]

    def le(self, x: T) -> Union[T, None]:
        "Find the largest element <= x, or None if it doesn't exist."
        for a in reversed(self.a):
            if a[0] <= x:
                return a[bisect_right(a, x) - 1]

    def gt(self, x: T) -> Union[T, None]:
        "Find the smallest element > x, or None if it doesn't exist."
        for a in self.a:
            if a[-1] > x:
                return a[bisect_right(a, x)]

    def ge(self, x: T) -> Union[T, None]:
        "Find the smallest element >= x, or None if it doesn't exist."
        for a in self.a:
            if a[-1] >= x:
                return a[bisect_left(a, x)]
    
    def __getitem__(self, x: int) -> T:
        "Return the x-th element, or IndexError if it doesn't exist."
        if x < 0: x += self.size
        if x < 0: raise IndexError
        for a in self.a:
            if x < len(a): return a[x]
            x -= len(a)
        raise IndexError

    def index(self, x: T) -> int:
        "Count the number of elements < x."
        ans = 0
        for a in self.a:
            if a[-1] >= x:
                return ans + bisect_left(a, x)
            ans += len(a)
        return ans

    def index_right(self, x: T) -> int:
        "Count the number of elements <= x."
        ans = 0
        for a in self.a:
            if a[-1] > x:
                return ans + bisect_right(a, x)
            ans += len(a)
        return ans

# n = int(input())
# li = [int(input()) for _ in range(n)]
needs = []
needs_st = set()
for i in range(n):
    v = li[i]
    temp = []
    for w in [v - 1, v, v + 1]:
        if 1 <= w <= n:
            temp.append(w)
    
    temp = tuple(temp)
    needs.append(temp)
    needs_st.add(temp)
    
used = SortedMultiset()
r = []
for i in range(n):
    for v in needs[i]:
        if v in used:
            continue
        used.add(v)
        index = used.index(v)
        
        if v - 1 >= 1 and index - 1 >= 0: 
            to_check = (used[index - 1], used[index])

            if to_check != needs[i]:
                if to_check in needs_st:
                    used.discard(v)
                    continue
                        
        if v - 2 >= 1 and index - 2 >= 0:
            to_check = (used[index - 2], used[index - 1], used[index])

            if to_check != needs[i]:
                if to_check in needs_st:
                    used.discard(v)
                    continue
            
        if v - 1 >= 1 and v + 1 <= n and index - 1 >= 0 and index + 1 < len(used):
            to_check = (used[index - 1], used[index], used[index + 1])

            if to_check != needs[i]:
                if to_check in needs_st:
                    used.discard(v)
                    continue

        if v + 1 <= n and index + 1 < len(used):
            to_check = (used[index], used[index + 1])

            if to_check != needs[i]:
                if to_check in needs_st:
                    used.discard(v)
                    continue
                    
        if v + 2 <= n and index + 2 < len(used):
            to_check = (used[index], used[index + 1], used[index + 2])

            if to_check != needs[i]:
                if to_check in needs_st:
                    used.discard(v)
                    continue   
                    
        r.append(v)
        break
    else:
        assert False
        
print(r)
```

### 공주님의 정원 2457

```python
import sys
# input = lambda: sys.stdin.readline().rstrip()

added = [0, 31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]

def convert(month, date):
    total = date
    for i in range(month):
        total += added[i]
        
    return total

s = convert(3, 1)
e = convert(11, 30)

n = int(input())

li = []
for _ in range(n):
    sm, sd, em, ed = map(int, input().split())
    
    li.append([convert(sm, sd), convert(em, ed) - 1])

li.sort(key=lambda x: x[0])

cur_find = s
cur_e = 0
i = 0
r = 0
ip = True
while i < n:
    while True:
        if i < n and li[i][0] <= cur_find:
            cur_e = max(cur_e, li[i][1])
        else:
            break
        i += 1
        
    if cur_e < cur_find:
        ip = False
        break
    r += 1
    cur_find = cur_e + 1
    if cur_find > e:
        break
        
if cur_find <= e:
    ip = False

if not ip:
    print(0)
else:
    print(r)
```

### Passport Control 16288

```python
from bisect import bisect_left

n, k = map(int, input().split())
li = list(map(int, input().split()))

c = 0
while li:
    c += 1
    memory = [li[0]]
    trace = [(li[0], 0)]
    for i in range(1, len(li)):
        to_check = bisect_left(memory, li[i])
        if to_check == len(memory):
            memory.append(li[i])
        else:
            memory[to_check] = li[i]
        trace.append((li[i], to_check))

    result = set()
    current = len(memory) - 1
    for i in range(len(li) - 1, -1, -1):
        if trace[i][1] == current:
            result.add(trace[i][0])
            current -= 1

            if current == -1:
                break

    nli = []
    for v in li:
        if v not in result:
            nli.append(v)
            
    if not nli:
        break
        
    li = nli[:]
    
if c <= k:
    print('YES')
else:
    print('NO')
```

```python
import sys
input = lambda: sys.stdin.readline().rstrip()

# https://github.com/tatyam-prime/SortedSet/blob/main/SortedMultiset.py
import math
from bisect import bisect_left, bisect_right, insort
from typing import Generic, Iterable, Iterator, TypeVar, Union, List
T = TypeVar('T')

class SortedMultiset(Generic[T]):
    BUCKET_RATIO = 50
    REBUILD_RATIO = 170

    def _build(self, a=None) -> None:
        "Evenly divide `a` into buckets."
        if a is None: a = list(self)
        size = self.size = len(a)
        bucket_size = int(math.ceil(math.sqrt(size / self.BUCKET_RATIO)))
        self.a = [a[size * i // bucket_size : size * (i + 1) // bucket_size] for i in range(bucket_size)]
    
    def __init__(self, a: Iterable[T] = []) -> None:
        "Make a new SortedMultiset from iterable. / O(N) if sorted / O(N log N)"
        a = list(a)
        if not all(a[i] <= a[i + 1] for i in range(len(a) - 1)):
            a = sorted(a)
        self._build(a)

    def __iter__(self) -> Iterator[T]:
        for i in self.a:
            for j in i: yield j

    def __reversed__(self) -> Iterator[T]:
        for i in reversed(self.a):
            for j in reversed(i): yield j
    
    def __len__(self) -> int:
        return self.size
    
    def __repr__(self) -> str:
        return "SortedMultiset" + str(self.a)
    
    def __str__(self) -> str:
        s = str(list(self))
        return "{" + s[1 : len(s) - 1] + "}"

    def _find_bucket(self, x: T) -> List[T]:
        "Find the bucket which should contain x. self must not be empty."
        for a in self.a:
            if x <= a[-1]: return a
        return a

    def __contains__(self, x: T) -> bool:
        if self.size == 0: return False
        a = self._find_bucket(x)
        i = bisect_left(a, x)
        return i != len(a) and a[i] == x

    def count(self, x: T) -> int:
        "Count the number of x."
        return self.index_right(x) - self.index(x)

    def add(self, x: T) -> None:
        "Add an element. / O(√N)"
        if self.size == 0:
            self.a = [[x]]
            self.size = 1
            return
        a = self._find_bucket(x)
        insort(a, x)
        self.size += 1
        if len(a) > len(self.a) * self.REBUILD_RATIO:
            self._build()

    def discard(self, x: T) -> bool:
        "Remove an element and return True if removed. / O(√N)"
        if self.size == 0: return False
        a = self._find_bucket(x)
        i = bisect_left(a, x)
        if i == len(a) or a[i] != x: return False
        a.pop(i)
        self.size -= 1
        if len(a) == 0: self._build()
        return True

    def lt(self, x: T) -> Union[T, None]:
        "Find the largest element < x, or None if it doesn't exist."
        for a in reversed(self.a):
            if a[0] < x:
                return a[bisect_left(a, x) - 1]

    def le(self, x: T) -> Union[T, None]:
        "Find the largest element <= x, or None if it doesn't exist."
        for a in reversed(self.a):
            if a[0] <= x:
                return a[bisect_right(a, x) - 1]

    def gt(self, x: T) -> Union[T, None]:
        "Find the smallest element > x, or None if it doesn't exist."
        for a in self.a:
            if a[-1] > x:
                return a[bisect_right(a, x)]

    def ge(self, x: T) -> Union[T, None]:
        "Find the smallest element >= x, or None if it doesn't exist."
        for a in self.a:
            if a[-1] >= x:
                return a[bisect_left(a, x)]
    
    def __getitem__(self, x: int) -> T:
        "Return the x-th element, or IndexError if it doesn't exist."
        if x < 0: x += self.size
        if x < 0: raise IndexError
        for a in self.a:
            if x < len(a): return a[x]
            x -= len(a)
        raise IndexError

    def index(self, x: T) -> int:
        "Count the number of elements < x."
        ans = 0
        for a in self.a:
            if a[-1] >= x:
                return ans + bisect_left(a, x)
            ans += len(a)
        return ans

    def index_right(self, x: T) -> int:
        "Count the number of elements <= x."
        ans = 0
        for a in self.a:
            if a[-1] > x:
                return ans + bisect_right(a, x)
            ans += len(a)
        return ans
    
c, n = map(int, input().split())

ms = SortedMultiset()
for _ in range(c):
    v = int(input())
    ms.add(v)
    
li = []
for _ in range(n):
    a, b = map(int, input().split())
    li.append((a, b))
    
li.sort(key=lambda x: x[1])
for s, e in li:
    index = ms.index(s)
    
    if index >= len(ms) or ms[index] > e:
        continue
    ms.discard(ms[index])
    
r = c - len(ms)

print(r)
```
