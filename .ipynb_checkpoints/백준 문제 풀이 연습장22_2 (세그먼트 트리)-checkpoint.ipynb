```python
from math import log2, ceil
import sys
input = lambda: sys.stdin.readline().rstrip()

inf = float('inf')
          
def construct_min(node, s, e):
    if s == e:
        seg_min[node] = li[s]
        return seg_min[node]
    mid = (s + e) // 2
    
    t1 = construct_min(node * 2, s, mid)
    t2 = construct_min(node * 2 + 1, mid + 1, e)
    
    seg_min[node] = min(t1, t2)
    
    return seg_min[node]

def query_min(node, s, e, pending):
    seg_min[node] += pending
    pendings_min[node] += pending
    
    if e < qs or s > qe:
        return inf
    if qs <= s <= e <= qe:
        return seg_min[node]
    else:
        pending = pendings_min[node]
        pendings_min[node] = 0
        
        mid = (s + e) // 2
        
        t1 = query_min(node * 2, s, mid, pending)
        t2 = query_min(node * 2 + 1, mid + 1, e, pending)
        
        return min(t1, t2)
    
def update_range_min(node, s, e, pending):
    seg_min[node] += pending
    pendings_min[node] += pending
    
    if ts <= s <= e <= te:
        pendings_min[node] += to_add
        seg_min[node] += to_add
        
        return seg_min[node]
    elif e < ts or s > te:
        return seg_min[node]
    else:
        pending = pendings_min[node]
        pendings_min[node] = 0
        
        mid = (s + e) // 2
          
        t1 = update_range_min(node * 2, s, mid, pending)
        t2 = update_range_min(node * 2 + 1, mid + 1, e, pending)
        
        seg_min[node] = min(t1, t2)

        return seg_min[node]

    
def construct_max(node, s, e):
    if s == e:
        seg_max[node] = li[s]
        return seg_max[node]
    mid = (s + e) // 2
    
    t1 = construct_max(node * 2, s, mid)
    t2 = construct_max(node * 2 + 1, mid + 1, e)
    
    seg_max[node] = max(t1, t2)
    
    return seg_max[node]

def query_max(node, s, e, pending):
    seg_max[node] += pending
    pendings_max[node] += pending
    
    if e < qs or s > qe:
        return 0
    if qs <= s <= e <= qe:
        return seg_max[node]
    else:
        pending = pendings_max[node]
        pendings_max[node] = 0
        
        mid = (s + e) // 2
        
        t1 = query_max(node * 2, s, mid, pending)
        t2 = query_max(node * 2 + 1, mid + 1, e, pending)
        
        return max(t1, t2)
    
def update_range_max(node, s, e, pending):
    seg_max[node] += pending
    pendings_max[node] += pending
    
    if ts <= s <= e <= te:
        pendings_max[node] += to_add
        seg_max[node] += to_add
        
        return seg_max[node]
    elif e < ts or s > te:
        return seg_max[node]
    else:
        pending = pendings_max[node]
        pendings_max[node] = 0
        
        mid = (s + e) // 2
          
        t1 = update_range_max(node * 2, s, mid, pending)
        t2 = update_range_max(node * 2 + 1, mid + 1, e, pending)
        
        seg_max[node] = max(t1, t2)

        return seg_max[node]
    
n, ul, q = map(int, input().split())

li = [0] * n

size = 2 ** (ceil(log2(n)) + 1) + 1

seg_min = [0] * size
pendings_min = [0] * size
construct_min(1, 0, n - 1)

seg_max = [0] * size
pendings_max = [0] * size
construct_max(1, 0, n - 1)

for _ in range(q):
    tp, *qu = input().split()
    
    if tp[0] == 's':
        x = int(qu[0])
        qs = qe = x
        
        result = query_min(1, 0, n - 1, 0)
        print(result)
        
        continue
    
    if tp[0] == 'c':
        x, s = map(int, qu)
        qs = qe = ts = te = x
    else:
        a, b, s = map(int, qu)
        qs = ts = a
        qe = te = b
        
    if s < 0:
        min_v = query_min(1, 0, n - 1, 0)
        to_add = -min(min_v, -s)
    else:
        max_v = query_max(1, 0, n - 1, 0)
        to_add = min(ul - max_v, s)
     
    result = to_add
    print(result)
    
    update_range_min(1, 0, n - 1, 0)
    update_range_max(1, 0, n - 1, 0)
```

```python
from math import log2, ceil
import sys
input = lambda: sys.stdin.readline().rstrip()

## PYRIVAL BOOTSTRAP
# https://github.com/cheran-senthil/PyRival/blob/master/pyrival/misc/bootstrap.py
# This decorator allows for recursion without actually doing recursion
## @bootstrap, yield when getting and returning value in recursive functions, end of functions
from types import GeneratorType

def bootstrap(f, stack=[]):
    def wrappedfunc(*args, **kwargs):
        if stack:
            return f(*args, **kwargs)
        else:
            to = f(*args, **kwargs)
            while True:
                if type(to) is GeneratorType:
                    stack.append(to)
                    to = next(to)
                else:
                    stack.pop()
                    if not stack:
                        break
                    to = stack[-1].send(to)
            return to
    return wrappedfunc

def construct(node, s, e):
    if s == e:
        segtree[node] = earns[seq_to_node[s]]
        return segtree[node]
    mid = (s + e) // 2
    
    t1 = construct(node * 2, s, mid)
    t2 = construct(node * 2 + 1, mid + 1, e)
    
    segtree[node] = t1 + t2
    
    return segtree[node]

def update_range(node, s, e, pending):
    segtree[node] += pending * (e - s + 1)
    pendings[node] += pending
    
    if ts <= s <= e <= te:
        pendings[node] += to_add
        segtree[node] += to_add * (e - s + 1)
        
        return segtree[node]
    elif e < ts or s > te:
        return segtree[node]
    else:
        pending = pendings[node]
        pendings[node] = 0
        
        mid = (s + e) // 2
        
        t1 = update_range(node * 2, s, mid, pending)
        t2 = update_range(node * 2 + 1, mid + 1, e, pending)
        
        segtree[node] = t1 + t2
        
        return segtree[node]

def query(node, s, e, pending):
    segtree[node] += pending * (e - s + 1)
    pendings[node] += pending
    
    if e < qs or s > qe:
        return 0
    if qs <= s <= e <= qe:
        return segtree[node]
    else:
        pending = pendings[node]
        pendings[node] = 0
        
        mid = (s + e) // 2
        
        t1 = query(node * 2, s, mid, pending)
        t2 = query(node * 2 + 1, mid + 1, e, pending)
        
        return t1 + t2

@bootstrap
def dfs(value):
    global vt_num
    
    seq_to_node[vt_num] = value
    history[value][0] = vt_num
    
    for v in li[value]:
        vt_num += 1
        yield dfs(v)
    
    history[value][1] = vt_num
    
    yield
    
n, m = map(int, input().split())

size = 2 ** (ceil(log2(n)) + 1) + 1
segtree = [0] * size
pendings = [0] * size

li = [[] for _ in range(n)]
earns = [0] * n

earns[0] = int(input())

for c_num in range(1, n):
    earn, p_num = map(int, input().split())
    p_num -= 1
    
    li[p_num].append(c_num)
    earns[c_num] = earn

history = [[-1] * 2 for _ in range(n)]
seq_to_node = [-1] * n

vt_num = 0
dfs(0)

construct(1, 0, n - 1)

for _ in range(m):
    tp, *qu = input().split()
    
    if tp == 'p':
        a, x = map(int, qu)
        a -= 1
        
        ts, te = history[a]
        ts += 1
        to_add = x
        
        if ts <= te:
            update_range(1, 0, n - 1, 0)
    else:
        a = int(qu[0])
        a -= 1
        
        qs = qe = history[a][0]
        result = query(1, 0, n - 1, 0)
        
        print(result)
```

```python
from math import log2, ceil
import sys, os, io
sys.setrecursionlimit(110000)
input = io.BytesIO(os.read(0,os.fstat(0).st_size)).readline

class SegmentTree:
    def __init__(self, data, default=10**15, func=lambda a, b: max(a,b)):
        """initialize the segment tree with data"""
        self._default = default
        self._func = func
        self._len = len(data)
        self._size = _size = 1 << (self._len - 1).bit_length()
 
        self.data = [default] * (2 * _size)
        self.data[_size:_size + self._len] = data
        for i in reversed(range(_size)):
            self.data[i] = func(self.data[i + i], self.data[i + i + 1])
 
    def __delitem__(self, idx):
        self[idx] = self._default
 
    def __getitem__(self, idx):
        return self.data[idx + self._size]
 
    def __setitem__(self, idx, value):
        idx += self._size
        self.data[idx] = value
        idx >>= 1
        while idx:
            self.data[idx] = self._func(self.data[2 * idx], self.data[2 * idx + 1])
            idx >>= 1
 
    def __len__(self):
        return self._len
 
    def query(self, start, stop):
        if start == stop:
            return self.__getitem__(start)
        stop += 1
        start += self._size
        stop += self._size
 
        res = self._default
        while start < stop:
            if start & 1:
                res = self._func(res, self.data[start])
                start += 1
            if stop & 1:
                stop -= 1
                res = self._func(res, self.data[stop])
            start >>= 1
            stop >>= 1
        return res
 
    def __repr__(self):
        return "SegmentTree({0})".format(self.data)

def dfs(value):
    global vt_num
    
#     s_to_n[vt_num] = value
    n_to_s[value][0] = vt_num
    
    for v in li[value]:
        vt_num += 1
        dfs(v)
        
    n_to_s[value][1] = vt_num
    
get_value = lambda x: int(x) - 1
n, m = map(int, input().split())
pa = list(map(get_value, input().split()))

li = [[] for _ in range(n)]
for c_num in range(1, n):
    p_num = pa[c_num]
    
    li[p_num].append(c_num)
    
# s_to_n = [-1] * n # seq to node
n_to_s = [[-1] * 2 for _ in range(n)] # node to seq

vt_num = 0
dfs(0)
    
func = lambda a, b: a + b
seg = SegmentTree([0] * n, 0, func=func)

for _ in range(m):
    tp, *qu = map(int, input().split())
    
    if tp == 1:
        i, w = qu
        i -= 1
        
        seq_i = n_to_s[i][0]
        seg[seq_i] += w
    else:
        i = qu[0]
        i -= 1
        
        qs, qe = n_to_s[i]
        result = seg.query(qs, qe)
        
        print(result)
```

https://www.acmicpc.net/problem/10565 Salary Inequity

```python
from math import ceil, log2
import io, os, sys
input = lambda: sys.stdin.readline().rstrip()

inf = float('inf')

## PYRIVAL BOOTSTRAP
# https://github.com/cheran-senthil/PyRival/blob/master/pyrival/misc/bootstrap.py
# This decorator allows for recursion without actually doing recursion
## @bootstrap, yield when getting and returning value in recursive functions, end of functions
from types import GeneratorType

def bootstrap(f, stack=[]):
    def wrappedfunc(*args, **kwargs):
        if stack:
            return f(*args, **kwargs)
        else:
            to = f(*args, **kwargs)
            while True:
                if type(to) is GeneratorType:
                    stack.append(to)
                    to = next(to)
                else:
                    stack.pop()
                    if not stack:
                        break
                    to = stack[-1].send(to)
            return to
    return wrappedfunc

@bootstrap
def dfs(value):
    global vt_num
    
    vt[value] = True
    
    n_to_s[value][0] = vt_num
    s_to_n[vt_num] = value
    
    for v in li[value]:
        if not vt[v]:
            vt_num += 1
            yield dfs(v)
            
    n_to_s[value][1] = vt_num
    
    yield
    
def construct(node, s, e):
    if s == e:
        segtree[node] = [salaries[s_to_n[s]]] * 2
        return segtree[node]
    
    mid = (s + e) // 2
    
    t1 = construct(node * 2, s, mid)
    t2 = construct(node * 2 + 1, mid + 1, e)
    
    segtree[node][0] = min(t1[0], t2[0])
    segtree[node][1] = max(t1[1], t2[1])
    
    return segtree[node]
    
def update_range(node, s, e, pending):
    segtree[node][0] += pending
    segtree[node][1] += pending
    
    pendings[node] += pending

    if ts <= s <= e <= te:
        segtree[node][0] += to_add
        segtree[node][1] += to_add

        pendings[node] += to_add
        
        return segtree[node]
    elif e < ts or s > te:
        return segtree[node]
    else:
        pending = pendings[node]
        pendings[node] = 0
        
        mid = (s + e) // 2
        
        t1 = update_range(node * 2, s, mid, pending)
        t2 = update_range(node * 2 + 1, mid + 1, e, pending)
        
        segtree[node][0] = min(t1[0], t2[0])
        segtree[node][1] = max(t1[1], t2[1])
        
        return segtree[node]

def query(node, s, e, pending):
    segtree[node][0] += pending
    segtree[node][1] += pending
    
    pendings[node] += pending
    
    if e < qs or s > qe:
        return [inf, -inf]
    if qs <= s <= e <= qe:
        return segtree[node]
    else:
        pending = pendings[node]
        pendings[node] = 0
        
        mid = (s + e) // 2
        
        t1 = query(node * 2, s, mid, pending)
        t2 = query(node * 2 + 1, mid + 1, e, pending)
        
        return [min(t1[0], t2[0]), max(t1[1], t2[1])]

get_value = lambda x: int(x) - 1

for _ in range(int(input())):
    n = int(input())
    pa = list(map(get_value, input().split()))

    li = [[] for _ in range(n)]
    for i in range(n - 1):
        c_num = i + 1
        p_num = pa[i]

        li[p_num].append(c_num)

    salaries = list(map(int, input().split()))

    s_to_n = [-1] * n
    n_to_s = [[-1] * 2 for _ in range(n)]
    vt = [False] * n
    vt_num = 0
    dfs(0)

    size = 2 ** (ceil(log2(n)) + 1) + 1
    segtree = [[-1] * 2 for _ in range(size)] # [min, max]
    pendings = [0] * size

    construct(1, 0, n - 1)

    q = int(input())
    for _ in range(q):
        tp, *qu = input().split()

        if tp == 'Q':
            i = int(qu[0]) - 1

            qs, qe = n_to_s[i]
            min_v, max_v = query(1, 0, n - 1, 0)

            result = max_v - min_v
            print(result)
        else:
            i, to_add = map(int, qu)
            i -= 1

            ts, te = n_to_s[i]
            update_range(1, 0, n - 1, 0)
```

https://www.acmicpc.net/problem/25538 모험가 길드

```python
# 시간 초과
from math import ceil, log2
import io, os, sys
sys.setrecursionlimit(301000)
input = io.BytesIO(os.read(0,os.fstat(0).st_size)).readline
print = sys.stdout.write

class SegmentTree:
    def __init__(self, data, default=10**15, func=lambda a, b: max(a,b)):
        """initialize the segment tree with data"""
        self._default = default
        self._func = func
        self._len = len(data)
        self._size = _size = 1 << (self._len - 1).bit_length()
 
        self.data = [default] * (2 * _size)
        self.data[_size:_size + self._len] = data
        for i in reversed(range(_size)):
            self.data[i] = func(self.data[i + i], self.data[i + i + 1])
 
    def __delitem__(self, idx):
        self[idx] = self._default
 
    def __getitem__(self, idx):
        return self.data[idx + self._size]
 
    def __setitem__(self, idx, value):
        idx += self._size
        self.data[idx] = value
        idx >>= 1
        while idx:
            self.data[idx] = self._func(self.data[2 * idx], self.data[2 * idx + 1])
            idx >>= 1
 
    def __len__(self):
        return self._len
 
    def query(self, start, stop):
        if start == stop:
            return self.__getitem__(start)
        stop += 1
        start += self._size
        stop += self._size
 
        res = self._default
        while start < stop:
            if start & 1:
                res = self._func(res, self.data[start])
                start += 1
            if stop & 1:
                stop -= 1
                res = self._func(res, self.data[stop])
            start >>= 1
            stop >>= 1
        return res
 
    def __repr__(self):
        return "SegmentTree({0})".format(self.data)

def find(v):
    global vt_num
    
    if v != pa[v]:
        pa[v] = find(pa[v])
    return pa[v]

def dfs(value, depth):
    global vt_num
    
    s_to_n[vt_num] = value
    n_to_s[value][0] = vt_num
    depths[value] = depth
    
    for v in li[value]:
        vt_num += 1
        parents[v][0] = value
        dfs(v, depth + 1)
        
    n_to_s[value][1] = vt_num
    
def lca(a, b):
    if depths[a] < depths[b]:
        a, b = b, a
        
    for cur_tms in range(max_tms, -1, -1):
        if depths[a] - 2 ** cur_tms >= depths[b]:
            a = parents[a][cur_tms]
            
    for cur_tms in range(max_tms, -1, -1):
        if parents[a][cur_tms] != parents[b][cur_tms]:
            a = parents[a][cur_tms]
            b = parents[b][cur_tms]
            
    if a != b:
        a = parents[a][0]
        b = parents[b][0]
        
    return a

n, q = map(int, input().split())
populs = list(map(int, input().split()))
qs = [list(map(int, input().split())) for _ in range(q)]

max_range = n
for v in qs:
    if v[0] == 3:
        max_range += 1
        
max_range += 1

diff = max_range - n
populs += [0] * diff

pa = list(range(max_range))

nxt_num = n
li = [[] for _ in range(max_range)]

ips = [True] * q
for qi in range(q):
    tp, a, b = qs[qi]        
        
    if tp == 3:
        a -= 1
        b -= 1
        
        ra = find(a)
        rb = find(b)
        
        pa[ra] = nxt_num
        pa[rb] = nxt_num
        
        li[nxt_num].append(ra)
        li[nxt_num].append(rb)
        
        nxt_num += 1
    elif tp == 4:
        a -= 1
        b -= 1
        
        ra = find(a)
        rb = find(b)
        
        if ra != rb:
            ips[qi] = False
    
for i in range(max_range - 1):
    if find(i) == i:
        li[max_range - 1].append(i)

max_tms = 19
parents = [[0] * (max_tms + 1) for _ in range(max_range)]
depths = [-1] * max_range
        
root = max_range - 1
s_to_n = [-1] * max_range
n_to_s = [[-1] * 2 for _ in range(max_range)]
vt_num = 0

dfs(root, 0)

for cur_tms in range(1, max_tms + 1):
    for i in range(max_range):
        p1 = parents[i][cur_tms - 1]
        p2 = parents[p1][cur_tms - 1]
        parents[i][cur_tms] = p2

npopuls = [populs[s_to_n[vt_num]] for vt_num in range(max_range)]
        
func = lambda a, b: a + b
segtree = SegmentTree(npopuls, 0, func=func)

for qi in range(q):
    tp, a, b = qs[qi]
    
    if tp == 1 or tp == 2:
        i, x = a, b
        i -= 1
        seq_num = n_to_s[i][0]
        
        if tp == 2:
            x = -x
        
        segtree[seq_num] += x
    elif tp == 4:
        a -= 1
        b -= 1
        
        if ips[qi] == False:
            print('-1\n')
        else:
            common_p = lca(a, b)
            s, e = n_to_s[common_p]
            
            result = segtree.query(s, e)
            
            print(str(result) + '\n')
```
