```python
from collections import deque
import sys
input = lambda: sys.stdin.readline().rstrip()

inf = float('inf')

n, q = map(int, input().split())
s = input()

closest_right = [0] * (n + 1)
potato_num_left_cumul = [0] * n
dists = deque()
dist = 1
r_closest_right = [inf] * (n + 1)
r_potato_num_right_cumul = [0] * (n + 1)
r_dists = deque()
r_dist = 1
for i in range(n - 1, -1, -1):
    closest_right[i] = closest_right[i + 1] + 1
    r_closest_right[i] = r_closest_right[i + 1] + 1
    r_potato_num_right_cumul[i] = r_potato_num_right_cumul[i + 1]
    if s[i] == '.':
        r_dist += 1
        dist += 1
    else:
        r_dists.appendleft(r_dist)
        r_dist = 1
        r_closest_right[i] = 0
        if i != n - 1:
            dists.appendleft(dist)
            dist = 1
        closest_right[i] = 0

    if i + 1 < n and s[i + 1] == 'P':
        r_potato_num_right_cumul[i] += 1

dists.appendleft(dist)

dist_len = len(dists)
r_dist_len = len(r_dists)

rock_case_potato_cumul = []
rightest_rock_indices = [0] * n

next_rock_index = 0
cumul2 = 0
init_left_end_index = -1

cumul = 0
for i in range(n):
    potato_num_left_cumul[i] = cumul

    if s[i] == 'R':
        if init_left_end_index == -1:
            init_left_end_index = i
        rock_case_potato_cumul.append(cumul)
        cumul2 = 0
        next_rock_index += 1
    elif s[i] == 'P':
        cumul2 += 1
    
    if i - 1 >= 0 and s[i - 1] == 'P':
        potato_num_left_cumul[i] += 1
        cumul = potato_num_left_cumul[i]
        
    rightest_rock_indices[i] = next_rock_index
        
left_ws = list(range(1, dist_len * 2, 2))
right_ws = list(range(dist_len * 2, 0, -2))

dists_left_cumul = [0] * (dist_len + 1)
left_weighted_cumul = [0] * (dist_len + 1)
right_weighted_cumul = [0] * (dist_len + 1)

r_left_ws = list(range(2, r_dist_len * 2 + 1, 2))
r_right_ws = list(range(r_dist_len * 2 - 1, -1, -2))

r_dists_left_cumul = [0] * (r_dist_len + 1)
r_left_weighted_cumul = [0] * (r_dist_len + 1)
r_right_weighted_cumul = [0] * (r_dist_len + 1)

for i in range(r_dist_len - 1, -1, -1):
    r_right_weighted_cumul[i] = r_right_weighted_cumul[i + 1] + r_right_ws[i] * r_dists[i]
for i in range(r_dist_len):
    r_left_weighted_cumul[i + 1] = r_left_weighted_cumul[i] + r_left_ws[i] * r_dists[i] 
    r_dists_left_cumul[i + 1] = r_dists_left_cumul[i] + r_dists[i]

    right_weighted_cumul[i + 1] = right_weighted_cumul[i] + right_ws[i] * dists[i]
    left_weighted_cumul[i + 1] = left_weighted_cumul[i] + left_ws[i] * dists[i] 
    dists_left_cumul[i + 1] = dists_left_cumul[i] + dists[i]
    
if init_left_end_index == -1:
    for _ in range(q):
        start = int(input())
        start -= 1    
    
        if r_closest_right[start] == inf:
            p = 0
            t = n - start
        else:    
            # 왼쪽으로 나가는 경우
            if potato_num_left_cumul[start] <= r_potato_num_right_cumul[start]:
                t = closest_right[start]
                start += closest_right[start]
                p = 0
                if start != ln - 1:
                    p += 1

                t += left_weighted_cumul[potato_num_left_cumul[start] + 1]
                p += potato_num_left_cumul[start]

                right_weight_init = left_ws[potato_num_left_cumul[start]] - 1
                right_start_index = potato_num_left_cumul[start] + 1
                right_end_index = right_start_index + (right_weight_init // 2 - 1)

                if right_end_index < dist_len:
                    right_weight_last = 2
                else:
                    right_weight_last = 2 + (right_end_index - dist_len + 1) * 2
                    right_end_index = dist_len - 1

                if right_start_index <= right_end_index:
                    temp = right_weighted_cumul[right_end_index + 1] - right_weighted_cumul[right_start_index]
                    to_adjust = dists_left_cumul[right_end_index + 1] - dists_left_cumul[right_start_index]

                    temp += to_adjust * (right_weight_last - right_ws[right_end_index])
                    t += temp
                    p += right_end_index - right_start_index + 1
                    if right_end_index == dist_len - 1:
                        p -= 1
            else:
                p = 1
                t = r_closest_right[start]
                start += r_closest_right[start]

                t += r_right_weighted_cumul[r_dist_len - 1 - r_potato_num_right_cumul[start]]
                p += r_potato_num_right_cumul[start]

                r_left_weight_last = r_right_ws[r_dist_len - 1 - r_potato_num_right_cumul[start]] + 1
                r_left_end_index = r_dist_len - 1 - r_potato_num_right_cumul[start] - 1
                r_left_start_index = r_left_end_index - (r_left_weight_last // 2 - 1)

                if r_left_start_index >= 0:
                    r_left_weight_init = 2
                else:
                    r_left_weight_init = 2 + 2 * (-r_left_start_index)    
                    r_left_start_index = 0

                temp = r_left_weighted_cumul[r_left_end_index + 1] - r_left_weighted_cumul[r_left_start_index]
                to_adjust = r_dists_left_cumul[r_left_end_index + 1] - r_dists_left_cumul[r_left_start_index]

                temp += to_adjust * (r_left_weight_init - r_left_ws[r_left_start_index])
                t += temp
                p += r_left_end_index - r_left_start_index + 1
                if r_left_start_index == 0:
                    p -= 1
                
        print(p, t)
else:
    left = s[:init_left_end_index + 1]
    ln = len(left)
     
    closest_right = [0] * (ln + 1)
    potato_num_left_cumul = [0] * ln
    dists = deque()
    dist = 1
    for i in range(ln - 1, -1, -1):
        closest_right[i] = closest_right[i + 1] + 1
        if left[i] == '.':
            dist += 1
        else:
            if i != ln - 1:
                dists.appendleft(dist)
                dist = 1
            closest_right[i] = 0

    dists.appendleft(dist)

    cumul = 0
    for i in range(ln):
        potato_num_left_cumul[i] = cumul

        if i - 1 >= 0 and left[i - 1] == 'P':
            potato_num_left_cumul[i] += 1
            cumul = potato_num_left_cumul[i]

    dist_len = len(dists)

    left_ws = list(range(1, dist_len * 2, 2))
    right_ws = list(range(dist_len * 2, 0, -2))

    dists_left_cumul = [0] * (dist_len + 1)
    left_weighted_cumul = [0] * (dist_len + 1)
    right_weighted_cumul = [0] * (dist_len + 1)

    for i in range(dist_len):
        right_weighted_cumul[i + 1] = right_weighted_cumul[i] + right_ws[i] * dists[i]
        left_weighted_cumul[i + 1] = left_weighted_cumul[i] + left_ws[i] * dists[i] 
        dists_left_cumul[i + 1] = dists_left_cumul[i] + dists[i]
    
    
    init_right_start_index = s.rfind('R')
    right = s[init_right_start_index:]
    rn = len(right)
    
    r_closest_right = [inf] * (rn + 1)
    r_potato_num_right_cumul = [0] * (rn + 1)
    r_dists = deque()
    r_dist = 1
    for i in range(rn - 1, -1, -1):
        r_closest_right[i] = r_closest_right[i + 1] + 1
        r_potato_num_right_cumul[i] = r_potato_num_right_cumul[i + 1]
        if right[i] == '.':
            r_dist += 1
        else:
            r_dists.appendleft(r_dist)
            r_dist = 1
            r_closest_right[i] = 0

        if i + 1 < rn and right[i + 1] == 'P':
            r_potato_num_right_cumul[i] += 1

    r_dist_len = len(r_dists)

    r_left_ws = list(range(2, r_dist_len * 2 + 1, 2))
    r_right_ws = list(range(r_dist_len * 2 - 1, -1, -2))

    r_dists_left_cumul = [0] * (r_dist_len + 1)
    r_left_weighted_cumul = [0] * (r_dist_len + 1)
    r_right_weighted_cumul = [0] * (r_dist_len + 1)

    for i in range(r_dist_len - 1, -1, -1):
        r_right_weighted_cumul[i] = r_right_weighted_cumul[i + 1] + r_right_ws[i] * r_dists[i]
    for i in range(r_dist_len):
        r_left_weighted_cumul[i + 1] = r_left_weighted_cumul[i] + r_left_ws[i] * r_dists[i] 
        r_dists_left_cumul[i + 1] = r_dists_left_cumul[i] + r_dists[i]

    for _ in range(q):
        start = int(input())
        start -= 1
    
        # 같히는 경우
        if init_left_end_index <= start <= init_right_start_index:
            t = -1
            p = rock_case_potato_cumul[rightest_rock_indices[start]]
        # 가장 오른쪽 R의 오른쪽 구간에서 시작하는 경우
        elif start >= init_right_start_index:
            start -= init_right_start_index

            if r_closest_right[start] == inf:
                p = 0
                t = rn - start
            else:
                p = 1
                t = r_closest_right[start]
                start += r_closest_right[start]

                t += r_right_weighted_cumul[r_dist_len - 1 - r_potato_num_right_cumul[start]]
                p += r_potato_num_right_cumul[start]
                
                r_left_weight_last = r_right_ws[r_dist_len - 1 - r_potato_num_right_cumul[start]] + 1
                r_left_end_index = r_dist_len - 1 - r_potato_num_right_cumul[start] - 1
                r_left_start_index = r_left_end_index - (r_left_weight_last // 2 - 1)

                if r_left_start_index >= 0:
                    r_left_weight_init = 2
                else:
                    r_left_weight_init = 2 + 2 * (-r_left_start_index)    
                    r_left_start_index = 0

                temp = r_left_weighted_cumul[r_left_end_index + 1] - r_left_weighted_cumul[r_left_start_index]
                to_adjust = r_dists_left_cumul[r_left_end_index + 1] - r_dists_left_cumul[r_left_start_index]

                temp += to_adjust * (r_left_weight_init - r_left_ws[r_left_start_index])
                t += temp
                p += r_left_end_index - r_left_start_index + 1
                if r_left_start_index == 0:
                    p -= 1
        else:
            t = closest_right[start]
            start += closest_right[start]
            p = 0
            if start != ln - 1:
                p += 1

            t += left_weighted_cumul[potato_num_left_cumul[start] + 1]
            p += potato_num_left_cumul[start]

            right_weight_init = left_ws[potato_num_left_cumul[start]] - 1
            right_start_index = potato_num_left_cumul[start] + 1
            right_end_index = right_start_index + (right_weight_init // 2 - 1)

            if right_end_index < dist_len:
                right_weight_last = 2
            else:
                right_weight_last = 2 + (right_end_index - dist_len + 1) * 2
                right_end_index = dist_len - 1

            if right_start_index <= right_end_index:
                temp = right_weighted_cumul[right_end_index + 1] - right_weighted_cumul[right_start_index]
                to_adjust = dists_left_cumul[right_end_index + 1] - dists_left_cumul[right_start_index]
                    
                temp += to_adjust * (right_weight_last - right_ws[right_end_index])
                t += temp
                p += right_end_index - right_start_index + 1
                if right_end_index == dist_len - 1:
                    p -= 1
                    
        print(p, t)
```

```python
import sys
input = lambda: sys.stdin.readline().rstrip()

s = input()
n = len(s)

sli = [[0] * 26]
for i in range(n):
    sli.append(sli[-1].copy())
    sli[-1][ord(s[i]) - 97] += 1
    
q = int(input())
for _ in range(q):
    alpha, l, r = input().split()
    alnum = ord(alpha) - 97
    l = int(l)
    r = int(r)
    
    result = sli[r + 1][alnum] - sli[l][alnum]
    
    print(result)
```
