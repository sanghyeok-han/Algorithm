### Cartesian MST 21062

```python
# 틀렸음
import sys
sys.setrecursionlimit(110000)
# input = lambda: sys.stdin.readline().rstrip()

def find(v):
    if v == parent[v]:
        return v
    parent[v] = find(parent[v])
    return parent[v]

def union(a, b):
    if a > b:
        a, b = b, a
    
    parent[b] = a

n1, m1, n2, m2 = map(int, input().split())

edges1 = [list(map(int, input().split())) for _ in range(m1)]
edges1.sort(key=lambda x: x[2])
parent = list(range(n1))
tc1 = 0
for a, b, c in edges1:
    ra = find(a)
    rb = find(b)
    if ra != rb:
        union(ra, rb)
        tc1 += c
        
edges2 = [list(map(int, input().split())) for _ in range(m2)]
edges2.sort(key=lambda x: x[2])
parent = list(range(n2))
tc2 = 0
for a, b, c in edges2:
    ra = find(a)
    rb = find(b)
    if ra != rb:
        union(ra, rb)
        tc2 += c
        
t1 = tc1 * n2 + tc2
t2 = tc2 * n1 + tc1

r = min(t1, t2)

print(r)
```

### 레드 블루 스패닝 트리 2 16950

```python
import sys
# input = lambda: sys.stdin.readline().rstrip()

## PYRIVAL BOOTSTRAP
# https://github.com/cheran-senthil/PyRival/blob/master/pyrival/misc/bootstrap.py
# This decorator allows for recursion without actually doing recursion
## @bootstrap, yield when getting and returning value in recursive functions, end of functions
from types import GeneratorType

def bootstrap(f, stack=[]):
    def wrappedfunc(*args, **kwargs):
        if stack:
            return f(*args, **kwargs)
        else:
            to = f(*args, **kwargs)
            while True:
                if type(to) is GeneratorType:
                    stack.append(to)
                    to = next(to)
                else:
                    stack.pop()
                    if not stack:
                        break
                    to = stack[-1].send(to)
            return to
    return wrappedfunc

@bootstrap
def find(v):
    if v == parent[v]:
        yield v
    parent[v] = yield find(parent[v])
    yield parent[v]

def union(a, b):
    if a > b:
        a, b = b, a
    
    parent[b] = a

n, m, k = map(int, input().split())

reds = []
blues = []
for _ in range(m):
    tp, a, b = input().split()
    a = int(a)
    b = int(b)

    if tp == 'B':
        blues.append((a, b))
    else:
        reds.append((a, b))

parent = list(range(n + 1))
for a, b in reds:
    ra = find(a)
    rb = find(b)
    if ra != rb:
        union(ra, rb)

essential_blues = []
other_blues = []
for a, b in blues:
    ra = find(a)
    rb = find(b)
    if ra != rb:
        essential_blues.append((a, b))
        union(ra, rb)
    else:
        other_blues.append((a, b))

ip = True
if k < len(essential_blues):
    ip = False

if not ip:
    print(0)
else:
    
    parent = list(range(n + 1))
    used_other_blues = set()
    used_reds = []
    
    k -= len(essential_blues)
    for a, b in essential_blues:
        ra = find(a)
        rb = find(b)
        if ra != rb:
            union(ra, rb)

    for a, b in other_blues:
        if k == 0:
            break
        ra = find(a)
        rb = find(b)
        if ra != rb:
            k -= 1
            union(ra, rb)
            used_other_blues.add((a, b))

    for a, b in reds:
        ra = find(a)
        rb = find(b)
        if ra != rb:
            union(ra, rb)
            used_reds.append((a, b))

    ip = True
    if k != 0:
        ip = False

    st = set()
    for i in range(1, n + 1):
        st.add(find(i))
    if len(st) >= 2:
        ip = False

    if not ip:
        print(0)
    else:
        for a, b in essential_blues:
            print(a, b)
        for a, b in used_other_blues:
            print(a, b)
        for a, b in used_reds:
            print(a, b)
```

### The Great Revegetation (Silver) 17038

```python
import sys
# input = lambda: sys.stdin.readline().rstrip()
sys.setrecursionlimit(110000)

def find(v):
    if v == parent[v]:
        return v
    parent[v] = find(parent[v])
    return parent[v]

def union(a, b):
    if a > b:
        a, b = b, a
    parent[b] = a
    
n, m = map(int, input().split())
parent = list(range(n + 1))

diffs = [[] for _ in range(n + 1)]
diff_edges = []
for _ in range(m):
    tp, a, b = input().split()
    a = int(a)
    b = int(b)
    
    if tp == 'S':
        ra = find(a)
        rb = find(b)
        if ra != rb:
            union(ra, rb)
    else:
        diffs[a].append(b)
        diffs[b].append(a)
        diff_edges.append((a, b))

for i in range(1, n + 1):
    for j in range(len(diffs[i]) - 1):
        a = diffs[i][j]
        b = diffs[i][j + 1]
        ra = find(a)
        rb = find(b)
        if ra != rb:
            union(ra, rb)
            
ip = True
for a, b in diff_edges:
    ra = find(a)
    rb = find(b)
    if ra == rb:
        ip = False
        break
        
if not ip:
    print(0)
else:
    for a, b in diff_edges:
        ra = find(a)
        rb = find(b)
        if ra != rb:
            union(ra, rb)
            
    st = set()
    for i in range(1, n + 1):
        st.add(find(i))

    r = '1' + '0' * len(st)

    print(r)    
```
