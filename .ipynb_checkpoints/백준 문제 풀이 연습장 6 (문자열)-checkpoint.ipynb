### Java vs C++ 3613

```python
s = input()

def check(s):
    if s[0].isupper() or s[0] == '_' or s[-1] == '_' or '__' in s:
        return 'Error!'
    
    if '_' in s:
        for v in s:
            if v.isupper():
                return 'Error!'
        else:
            return 'C++'
    else:
        return 'Java'

s_type = check(s)

if s_type == 'Error!':
    print(s_type)
else:
    if s_type == 'C++':
        new_li = [s[0]]
        for i in range(1, len(s)):
            if s[i - 1] == '_':
                new_li.append(s[i].upper())
            else:
                new_li.append(s[i])
                
        s = "".join(new_li)
        s = s.replace('_', "")

    elif s_type == 'Java':
        new_li = []
        for i in range(len(s)):
            if s[i].isupper():
                new_li.append("_")
                new_li.append(s[i].lower())
            else:
                new_li.append(s[i])
                
        s = "".join(new_li)
        
    print(s)
```

### IOIOI 5525

```python
n = int(input())
m = int(input())
s = input()

ci_to_check = n + 1
cur_check = 'I'
count = 0
is_continued = False
for v in s:
    if v == cur_check:
        if not is_continued:
            if cur_check == 'I':
                ci_to_check -= 1
                cur_check = 'O'
    
                if ci_to_check == 0:
                    is_continued = True
                    count += 1
            else:
                cur_check = 'I'
        else:
            if cur_check == 'I':
                count += 1
                cur_check = 'O'
            else:
                cur_check = 'I'
    else:
        is_continued = False
        if v == 'I':
            ci_to_check = n
            cur_check = 'O'
        else:
            ci_to_check = n + 1
            cur_check = 'I'
            
print(count)
```

### Contact 1013

```python
from collections import deque
import sys
# input = lambda: sys.stdin.readline().rstrip()

def check(deq):
    while deq:
        if deq[0] == 0:
            if len(deq) < 2:
                break
            if deq[1] == 0:
                break
            deq.popleft()
            deq.popleft()
        else:
            if len(deq) < 4:
                break
            if deq[1] != 0 or deq[2] != 0:
                break
            deq.popleft()
            while deq and deq[0] == 0:
                deq.popleft()
            if not deq:
                deq.append(0)
                break
            deq.popleft()
            while deq and deq[0] == 1:
                if check(deq.copy()):
                    return True
                deq.popleft()
            
    if not deq:
        return True
    
for _ in range(int(input())):
    deq = deque(map(int, input()))
    
    if check(deq):
        print('YES')
    else:
        print('NO')
```

### 잠수함식별 2671

```python
from collections import deque

def check(deq):
    while deq:
        if deq[0] == 0:
            deq.popleft()
            if not deq or deq[0] != 1:
                return False
            deq.popleft()
        else:
            deq.popleft()
            if len(deq) < 3 or deq[0] != 0 or deq[1] != 0:
                return False
            deq.popleft()
            deq.popleft()
            
            while deq and deq[0] == 0:
                deq.popleft()
            if not deq or deq[0] != 1:
                return False
            deq.popleft()
            
            while deq and deq[0] == 1:
                if check(deq.copy()):
                    return True
                deq.popleft()
                
    return True

deq = deque(map(int, input()))

if check(deq):
    print('SUBMARINE')
else:
    print('NOISE')
```
