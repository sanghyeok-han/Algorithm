### 여러분의 다리가 되어 드리겠습니다!  17352

```python
import sys
# input = sys.stdin.readline

n = int(input().rstrip())
parent = list(range(n))
rank = [0] * n

def get_node_num(x):
    return int(x) - 1

def find(v):
    if v == parent[v]:
        return v
    parent[v] = find(parent[v])
    return parent[v]

def union(a, b):
    if rank[a] >= rank[b]:
        if rank[a] == rank[b]:
            rank[a] += 1
        parent[b] = a
    else:
        parent[a] = b

for _ in range(n - 2):
    a, b = map(get_node_num, input().rstrip().split())
    union(find(a), find(b))

for i in range(n):
    find(i)
    if i >= 1 and parent[i] != parent[i - 1]:
        print(i, i + 1)
        break
```

### 비용 2463

```python
import sys
# input = lambda: sys.stdin.readline().rstrip()

n, m = map(int, input().split())
li = [list(map(int, input().split())) for _ in range(m)]
li.sort(key=lambda x: -x[2])

parent = list(range(n + 1))
included = [1] * (n + 1)

def find(v):
    if v == parent[v]:
        return v
    parent[v] = find(parent[v])
    return parent[v]

def union(a, b):
    parent[b] = a
    included[a] += included[b]

comb_count = 0
total = 0
for x, y, w in li:
    rx = find(x)
    ry = find(y)
    if rx != ry:
        comb_count += included[rx] * included[ry]
        union(rx, ry)
    total += w * comb_count
    
print(total % 1000000000)
```

### 도로 4014

```python
만약 모든 도시가 전부 연결이 안되어 있거나 자갈 도로의 개수가 k개보다 적거나
정확히 k개의 자갈 도로만으로 모든 도시 연결이 불가능하면 안됨

콘크리트 간선을 먼저 전부 분리집합으로 합쳐줌.
그 후에 이용되는 자갈 간선은 필수 도로가 됨
만약 k가 필수 자갈 도로 개수 보다 적다면 no solution
k가 필수 자갈 도로 개수와 같다면 그대로 출력
k가 필수 자갈 도로 개수보다 많다면 안쓰였던 자갈 도로들을 추가해줌
만약 안쓰인 자갈 도로를 합쳐도 k가 안되면 no solution

-> 필수 자갈 도로 먼저 확인해놓고 그 다음에 다시 mst 한번 더 돌리는데
이번에는 자갈 도로를 먼저 합쳐줌
```

```python
import sys
sys.setrecursionlimit(110000)
# input = lambda: sys.stdin.readline().rstrip()

def find(v):
    if v == parent[v]:
        return v
    parent[v] = find(parent[v])
    return parent[v]

def union(a, b):
    if a > b:
        a, b = b, a
    
    parent[b] = a

n, m, k = map(int, input().split())

cons = []
cras = []
for _ in range(m):
    a, b, tp = map(int, input().split())
        
    if tp == 0:
        cras.append((a, b))
    else:
        cons.append((a, b))
        
parent = list(range(n + 1))
for a, b in cons:
    ra = find(a)
    rb = find(b)
    if ra != rb:
        union(ra, rb)
        
essential_cras = []
other_cras = []
for a, b in cras:
    ra = find(a)
    rb = find(b)
    if ra != rb:
        essential_cras.append((a, b))
        union(ra, rb)
    else:
        other_cras.append((a, b))
        
ip = True
if k < len(essential_cras):
    ip = False
    
if not ip:
    print('no solution')
else:
    parent = list(range(n + 1))
    used_other_cras = set()
    used_cons = []
    
    k -= len(essential_cras)
    for a, b in essential_cras:
        ra = find(a)
        rb = find(b)
        if ra != rb:
            union(ra, rb)
            
    for a, b in other_cras:
        if k == 0:
            break
        ra = find(a)
        rb = find(b)
        if ra != rb:
            k -= 1
            union(ra, rb)
            used_other_cras.add((a, b))
    
    for a, b in cons:
        ra = find(a)
        rb = find(b)
        if ra != rb:
            union(ra, rb)
            used_cons.append((a, b))
    
    ip = True
    if k != 0:
        ip = False
        
    st = set()
    for i in range(1, n + 1):
        st.add(find(i))
    if len(st) >= 2:
        ip = False
        
    if not ip:
        print('no solution')
    else:
        for a, b in essential_cras:
            print(a, b, 0)
        for a, b in used_other_cras:
            print(a, b, 0)
        for a, b in used_cons:
            print(a, b, 1)
```

### 레드 블루 스패닝 트리 4792

```python
import sys
# input = lambda: sys.stdin.readline().rstrip()

## PYRIVAL BOOTSTRAP
# https://github.com/cheran-senthil/PyRival/blob/master/pyrival/misc/bootstrap.py
# This decorator allows for recursion without actually doing recursion
## @bootstrap, yield when getting and returning value in recursive functions, end of functions
from types import GeneratorType

def bootstrap(f, stack=[]):
    def wrappedfunc(*args, **kwargs):
        if stack:
            return f(*args, **kwargs)
        else:
            to = f(*args, **kwargs)
            while True:
                if type(to) is GeneratorType:
                    stack.append(to)
                    to = next(to)
                else:
                    stack.pop()
                    if not stack:
                        break
                    to = stack[-1].send(to)
            return to
    return wrappedfunc

@bootstrap
def find(v):
    if v == parent[v]:
        yield v
    parent[v] = yield find(parent[v])
    yield parent[v]

def union(a, b):
    if a > b:
        a, b = b, a
    
    parent[b] = a

while True:
    n, m, k = map(int, input().split())
    if n == m == k == 0:
        break

    reds = []
    blues = []
    for _ in range(m):
        tp, a, b = input().split()
        a = int(a)
        b = int(b)

        if tp == 'B':
            blues.append((a, b))
        else:
            reds.append((a, b))

    parent = list(range(n + 1))
    for a, b in reds:
        ra = find(a)
        rb = find(b)
        if ra != rb:
            union(ra, rb)

    essential_blues = []
    other_blues = []
    for a, b in blues:
        ra = find(a)
        rb = find(b)
        if ra != rb:
            essential_blues.append((a, b))
            union(ra, rb)
        else:
            other_blues.append((a, b))

    ip = True
    if k < len(essential_blues):
        ip = False

    if ip:
        parent = list(range(n + 1))

        k -= len(essential_blues)
        for a, b in essential_blues:
            ra = find(a)
            rb = find(b)
            if ra != rb:
                union(ra, rb)

        for a, b in other_blues:
            if k == 0:
                break
            ra = find(a)
            rb = find(b)
            if ra != rb:
                k -= 1
                union(ra, rb)

        for a, b in reds:
            ra = find(a)
            rb = find(b)
            if ra != rb:
                union(ra, rb)

        if k != 0:
            ip = False

        st = set()
        for i in range(1, n + 1):
            st.add(find(i))
        if len(st) >= 2:
            ip = False

    if not ip:
        print(0)
    else:
        print(1)
```

### 달리기 12963

```python
import sys
sys.setrecursionlimit(10000)
input = lambda: sys.stdin.readline().rstrip()

mod = 1000000007

def find(v):
    if v == parent[v]:
        return v
    parent[v] = find(parent[v])
    return parent[v]

def union(a, b):
    if a == 0:
        parent[b] = a
    elif b == 0:
        parent[a] = b
    elif a == n - 1:
        parent[b] = a
    elif b == n - 1:
        parent[a] = b
    else:
        if a > b:
            a, b = b, a

        parent[a] = b
    
edges = []
n, m = map(int, input().split())
for i in range(m):
    a, b = map(int, input().split())
    edges.append((a, b, i))

edges.reverse()
parent = list(range(n))
r = []
for a, b, i in edges:
    ra = find(a)
    rb = find(b)
    
    if ra == n - 1 and rb == 0 or ra == 0 and rb == n - 1:
        r.append(i)
    else:
        if ra != rb:
            union(ra, rb)
            
result = 0
for v in r:
    result += pow(3, v, mod)
    result %= mod
    
print(result)
```

### Great GDP 17986

```python
# 일단 포기
def find(v):
    if v == parent[v]:
        return v
    parent[v] = find(parent[v])
    return parent[v]

def union(a, b):
    if a > b:
        a, b = b, a
    parent[b] = a

n = int(input())
gdps = list(map(int, input().split()))
populs = list(map(int, input().split()))

li = [[] for _ in range(n + 1)]
for _ in range(n - 1):
    a, b = map(int, input().split())
    li[a].append(b)
    li[b].append(a)
    
parent = list(range(n + 1))
selected = [False] * (n + 1)
selected[1] = True
gpc = [(gdps[i] / populs[i], i) for i in range(1, n + 1)]
gpc.sort(key=lambda x: -x[0])

for _, value in gpc:
    selected[value] = True
    
    for v in li[value]:
        rvalue = find(value)
        rv = find(v)
        
        if rvalue != rv:
            union(rvalue, rv)
            
    if find(value) == 1:
        break
        
total_gdps = 0
total_populs = 0
for i in range(1, n + 1):
    if find(i) == 0 and selected[i]:
        total_gdps += gdps[i]
        total_populs += populs[i]
        
r = total_populs / total_gdps
```
