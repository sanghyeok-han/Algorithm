### 도미노 4196

```python
import sys
sys.setrecursionlimit(int(1e5) + 100)
input = lambda: sys.stdin.readline().rstrip()

def dfs(value):
    for v in li[value]:
        if vt[v]:
            continue
        vt[v] = True
        dfs(v)

    stack.append(value)

def dfs2(value):
    for v in li[value]:
        if vt[v]:
            continue
        vt[v] = True
        dfs2(v)

for _ in range(int(input())):
    n, m = map(int, input().split())
    li = [[] for _ in range(n + 1)]
    for _ in range(m):
        a, b = map(int, input().split())

        li[a].append(b)

    stack = []
    vt = [False] * (n + 1)
    for i in range(1, n + 1):
        if vt[i]:
            continue
        vt[i] = True
        dfs(i)

    vt = [False] * (n + 1)
    count = 0
    while stack:
        value = stack.pop()
        if vt[value]:
            continue
        vt[value] = True
        count += 1

        dfs2(value)

    print(count)
```

### 축구 전술 3977

```python
import sys
sys.setrecursionlimit(int(1e5) + 100)
input = lambda: sys.stdin.readline().rstrip()

def dfs(value):
    for v in li[value]:
        if vt[v]:
            continue
        vt[v] = True
        dfs(v)
        
    stack.append(value)

def dfs2(value):
    for v in rli[value]:
        if vt[v]:
            continue
        vt[v] = True
        scc.append(v)
        dfs2(v)
        
def dfs3(value):
    for v in li[value]:
        if vt[v]:
            continue
        vt[v] = True
        dfs3(v)

t = int(input())
for ti in range(t):
    n, m = map(int, input().split())
    li = [[] for _ in range(n)]
    rli = [[] for _ in range(n)]

    for _ in range(m):
        a, b = map(int, input().split())

        li[a].append(b)
        rli[b].append(a)

    stack = []
    vt = [False] * n
    for i in range(n):
        if vt[i]:
            continue
        vt[i] = True
        dfs(i)

    value = stack.pop()
    vt = [False] * n
    vt[value] = True
    scc = [value]
    dfs2(value)

    vt = [False] * n
    vt[scc[0]] = True
    dfs3(scc[0])

    if sum(vt) == n:
        for v in sorted(scc):
            print(v)
    else:
        print('Confused')

    if ti < t - 1:
        input()
        print()
```

### 세진 바이러스 15783

```python
import sys
sys.setrecursionlimit(int(1e5) + 100)
input = lambda: sys.stdin.readline().rstrip()

def dfs(value):
    for v in li[value]:
        if vt[v]:
            continue
        vt[v] = True
        dfs(v)
    stack.append(value)
    
def dfs2(value):
    for v in li[value]:
        if vt[v]:
            continue
        vt[v] = True
        dfs2(v)
    
n, m = map(int, input().split())

li = [[] for _ in range(n)]
for _ in range(m):
    a, b = map(int, input().split())
    li[a].append(b)
    
stack = []
vt = [False] * n
for i in range(n):
    if vt[i]:
        continue
    vt[i] = True
    dfs(i)
    
count = 0
vt = [False]* n
while stack:
    value = stack.pop()
    if vt[value]:
        continue
    vt[value] = True
    dfs2(value)
    count += 1

print(count)
```

### 성격 진단 테스트 4305

```python
import sys
sys.setrecursionlimit(int(1e5) + 100)
# input = lambda: sys.stdin.readline().rstrip()

def dfs(value):
    vt[value] = True
    
    for v in li[value]:
        if vt[v]:
            continue
        dfs(v)
        
    stack.append(value)
    
def dfs2(value):
    vt[value] = True
    temp_r.append(value) 
    
    for v in rli[value]:
        if vt[v]:
            continue
        dfs2(v)
        
get_value = lambda x: ord(x) - 65

while True:
    n = int(input())
    if n == 0:
        break

    st = set()
    li = [[] for _ in range(26)]
    rli = [[] for _ in range(26)]
    for _ in range(n):
        temp = list(map(get_value, input().split()))
        a = temp.pop()
        st.add(a)
        for b in temp:
            if a == b:
                continue
            st.add(b)

            li[a].append(b)
            rli[b].append(a)

    stack = []
    vt = [False] * 26
    for i in range(26):
        if i not in st:
            continue
        if not vt[i]:
            dfs(i)

    vt = [False] * 26
    r = []
    while stack:
        value = stack.pop()
        if vt[value]:
            continue

        temp_r = []
        dfs2(value)
        r.append(sorted(temp_r))

    r.sort()

    for v in r:
        print(*[chr(w + 65) for w in v])

    print()
```

### 그래프의 싱크 6543

```python
import sys
sys.setrecursionlimit(int(1e5) + 100)
# input = lambda: sys.stdin.readline().rstrip()

def dfs(value):
    global next_id
    
    stack.append(value)
    on_stack[value] = True
    min_id = ids[value]
    
    for v in li[value]:
        if ids[v] != 0:
            if not on_stack[v]:
                continue
            min_id = min(min_id, low_ids[v])
        else:
            ids[v] = next_id
            low_ids[v] = next_id
            next_id += 1
            temp = dfs(v)
            min_id = min(min_id, temp)
        
    low_ids[value] = min_id
         
    if ids[value] == low_ids[value]:
        scc = set()
        while True:
            v = stack.pop()
            scc.add(v)
            on_stack[v] = False
            
            if value == v:
                break
        sccs.append(scc)
        
    return min_id

def dfs2(value, scc_id):
    vt[value] = True
    
    for v in li[value]:
        if v not in sccs[scc_id]:
            return False
        if not vt[v]:
            if not dfs2(v, scc_id):
                return False
        
    return True
    
while True:
    temp = list(map(int, input().split()))
    if temp[0] == 0:
        break
    n, m = temp
    
    li = [[] for _ in range(n + 1)]
    edges = list(map(int, input().split()))
    for i in range(0, m * 2, 2):
        a = edges[i]
        b = edges[i + 1]

        li[a].append(b)

    ids = [0] * (n + 1)
    low_ids = [0] * (n + 1)
    on_stack = [False] * (n + 1)
    stack = []
    sccs = []
    next_id = 1

    for i in range(1, n + 1):
        if ids[i] == 0:
            ids[i] = next_id
            low_ids[i] = next_id
            next_id += 1
            dfs(i)
            
    result = []
    vt = [False] * (n + 1)
    for scc_id in range(len(sccs)):
        value = sccs[scc_id].pop()
        sccs[scc_id].add(value)

        if dfs2(value, scc_id):
            result.extend(sccs[scc_id])

    result.sort()
    if not result:
        print()
    else:
        print(*result)
```
