### 배열 돌리기 4 17406

```python
from itertools import permutations as pm
from copy import deepcopy

n, m, k = map(int, input().split())
li = [list(map(int, input().split())) for _ in range(n)]
opers = [tuple(map(int, input().split())) for _ in range(k)]

def rotate_part(r, c, s):
    sx = r - s - 1
    sy = c - s - 1
    ex = r + s - 1
    ey = c + s - 1
    
    mx = r - 1
    d = 0
    while True:
        if sx + d == mx:
            break
        temp = li[sx + d][sy + d]
        for i in range(sx + d, ex - d):
            li[i][sy + d] = li[i + 1][sy + d]
        for j in range(sy + d, ey - d):
            li[ex - d][j] = li[ex - d][j + 1]
        for i in range(ex - d, sx + d, -1):
            li[i][ey - d] = li[i - 1][ey - d]
        for j in range(ey - d, sy + d, -1):
            li[sx + d][j] = li[sx + d][j - 1]
        li[sx + d][sy + d + 1] = temp
        
        d += 1

get_min = lambda: min(sum(v) for v in li)

ori = deepcopy(li)
cases = pm(opers, len(opers))
min_result = float('inf')
for case in cases:
    li = deepcopy(ori)
    for r, c, s in case:
        rotate_part(r, c, s)
    min_result = min(min_result, get_min())
    
print(min_result)
```

### 나무 재테크 16235 (신규)

```python
from collections import deque

n, m, k = map(int, input().split())
A = [list(map(int, input().split())) for _ in range(n)]

li = [[deque() for _ in range(n)] for _ in range(n)]
nt = [[5] * n for _ in range(n)] # nutrient
for _ in range(m):
    r, c, age = map(int, input().split())
    r -= 1
    c -= 1
    li[r][c].append(age)
    
di = [-1, -1, -1, 0, 0, 1, 1, 1]
dj = [-1, 0, 1, -1, 1, -1, 0, 1]

for _ in range(k):
    for i in range(n):
        for j in range(n):
            ti = 0
            for ti in range(len(li[i][j]) - 1, -1, -1):
                if li[i][j][ti] > nt[i][j]:
                    ti += 1
                    break
                nt[i][j] -= li[i][j][ti]
                li[i][j][ti] += 1
            while ti:
                nt[i][j] += li[i][j].popleft() // 2
                ti -= 1
                
    for i in range(n):
        for j in range(n):
            for v in li[i][j]:
                if v % 5 == 0:
                    for d in range(8):
                        ni = i + di[d]
                        nj = j + dj[d]
                        if ni < 0 or ni >= n or nj < 0 or nj >= n:
                            continue
                        li[ni][nj].append(1)
            
            nt[i][j] += A[i][j]
        
count = 0
for i in range(n):
    for j in range(n):
        count += len(li[i][j])
        
print(count)
```

### ⚾ 17281 (신규)

```python
# 맞았음, 값이 커질 때 비트마스킹 쓰면 손해일 수 있음, 덱보다 그냥 조건문이 나을 때도 있음,
# 반복문 안에서 조건문 많이 쓰게 될 때 밖에서 미리 처리해주고 조건문 없애는 편이 나을 때도 있음
from itertools import permutations as pm

n = int(input())
rs = [list(map(int, input().split())) for _ in range(n)] # results

cases = pm(range(1, 9), 8)
max_score = 0
for case in cases:
    case = case[:3] + (0,) + case[3:]
    p1 = p2 = p3 = 0
    cur = 0 # 현재 n번 타자 (0~8)
    score = 0
    for i in range(n):
        oc = 0 # out count
        while True:
            r = rs[i][case[cur]]
            if r == 1:
                score += p3
                p3, p2 = p2, p1
                p1 = 1
            elif r == 2:
                score += p3 + p2
                p3 = p1
                p1 = 0
                p2 = 1
            elif r == 3:
                score += p3 + p2 + p1
                p3 = p2 = p1 = 0
                p3 = 1
            elif r == 4:
                score += p3 + p2 + p1 + 1
                p3 = p2 = p1 = 0    
            else:
                oc += 1

            cur += 1
            if cur >= 9:
                cur = 0
                
            if oc == 3:
                break
                
        p3 = p2 = p1 = 0

    if score > max_score:
        max_score = score

print(max_score)
```

### 청소년 상어 19236

```python
from collections import deque
from copy import deepcopy

n = 4
dx = [-1, -1, 0, 1, 1, 1, 0, -1]
dy = [0, -1, -1, -1, 0, 1, 1, 1]

li = [] # li[x][y] = [물고기 번호, 방향], 상어 번호: 0, 빈 칸의 경우 [-1, -1]
for _ in range(n):
    temp = list(map(int, input().split()))
    row = []
    for i in range(0, 8, 2):
        row.append([temp[i], temp[i + 1] - 1])
    li.append(row)

# s = '''2 6 10 8 6 7 9 4
# 1 7 16 6 4 2 5 8
# 3 7 8 6 7 6 14 8
# 12 7 15 4 11 3 13 3'''
# s = s.split('\n')
# li = []
# for i in range(n):
#     temp = list(map(int, s[i].split()))
#     row = []
#     for i in range(0, 8, 2):
#         row.append([temp[i], temp[i + 1] - 1])
#     li.append(row)

base_sc = li[0][0][0] # base score
li[0][0][0] = 0

flocs = [[-1, -1] for _ in range(17)] # fish locations, flocs[물고기 번호] = [x, y], 해당 물고기가 없는 경우 [-1, -1]
for i in range(n):
    for j in range(n):
        num = li[i][j][0]
        flocs[num] = [i, j]

def mf():
    """move fishes"""
    
    for num in range(1, 17):
        if flocs[num][0] == -1:
            continue
        x, y = flocs[num]    
        for _ in range(8):
            nx, ny = x + dx[li[x][y][1]], y + dy[li[x][y][1]]
            if nx < 0 or nx >= n or ny < 0 or ny >= n or li[nx][ny][0] == 0:
                li[x][y][1] += 1
                if li[x][y][1] >= 8:
                    li[x][y][1] -= 8
                continue
                
            if li[nx][ny][0] == -1:
                flocs[num] = [nx, ny]
            else:          
                ofn = li[nx][ny][0] # other fish number
                flocs[num], flocs[ofn] = flocs[ofn], flocs[num]   
            li[x][y], li[nx][ny] = li[nx][ny], li[x][y]
            break
        else:
            li[x][y][1] -= 1
            
max_sc = 0
queue = deque()
queue.append((li, flocs, base_sc))
while queue:
    li, flocs, sc = queue.popleft()
    max_sc = max(max_sc, sc)
    
    mf()
    x, y = flocs[0]
    d = li[x][y][1]
    
    li[x][y] = [-1, -1]
    while True:
        x, y = x + dx[d], y + dy[d]
        if x < 0 or x >= n or y < 0 or y >= n:
            break
          
        if li[x][y][0] != -1:
            nli, nflocs, nsc = deepcopy(li), deepcopy(flocs), sc
            efn = nli[x][y][0] # eaten fish num
            nsc += nli[x][y][0]
            nflocs[efn] = [-1, -1]
            nli[x][y] = [0, nli[x][y][1]]
            nflocs[0] = [x, y]
            
            queue.append((nli, nflocs, nsc))
            
print(max_sc)
```

### 2048 (Easy) 12100

```python
from collections import deque
from copy import deepcopy

# 한 줄씩 해당하는 방향쪽으로 전부 당기고, 해당하는 방향쪽부터 합치고, 다시 해당하는 방향쪽으로 전부 당김
def move(li, d):
    # 위쪽 이동
    if d == 0:
        for j in range(n):
            while True:
                is_moved = False
                for i in range(n - 1, 0, -1):
                    if li[i][j] != 0 and li[i - 1][j] == 0:
                        li[i - 1][j], li[i][j] = li[i][j], li[i - 1][j]
                        is_moved = True
                if not is_moved:
                    break
            for i in range(n - 1):
                if li[i][j] and li[i + 1][j] == li[i][j]:
                    li[i][j] *= 2
                    li[i + 1][j] = 0
            while True:
                is_moved = False
                for i in range(n - 1, 0, -1):
                    if li[i][j] != 0 and li[i - 1][j] == 0:
                        li[i - 1][j], li[i][j] = li[i][j], li[i - 1][j]
                        is_moved = True
                if not is_moved:
                    break
    # 아래쪽 이동
    elif d == 1:
        for j in range(n):
            while True:
                is_moved = False
                for i in range(n - 1):
                    if li[i][j] != 0 and li[i + 1][j] == 0:
                        li[i + 1][j], li[i][j] = li[i][j], li[i + 1][j]
                        is_moved = True
                if not is_moved:
                    break
            for i in range(n - 1, 0, -1):
                if li[i][j] and li[i - 1][j] == li[i][j]:
                    li[i][j] *= 2
                    li[i - 1][j] = 0  
            while True:
                is_moved = False
                for i in range(n - 1):
                    if li[i][j] != 0 and li[i + 1][j] == 0:
                        li[i + 1][j], li[i][j] = li[i][j], li[i + 1][j]
                        is_moved = True
                if not is_moved:
                    break
    # 왼쪽 이동
    elif d == 2:
        for i in range(n):
            while True:
                is_moved = False
                for j in range(n - 1, 0, -1):
                    if li[i][j] != 0 and li[i][j - 1] == 0:
                        li[i][j - 1], li[i][j] = li[i][j], li[i][j - 1]
                        is_moved = True
                if not is_moved:
                    break
            for j in range(n - 1):
                if li[i][j] and li[i][j + 1] == li[i][j]:
                    li[i][j] *= 2
                    li[i][j + 1] = 0
            while True:
                is_moved = False
                for j in range(n - 1, 0, -1):
                    if li[i][j] != 0 and li[i][j - 1] == 0:
                        li[i][j - 1], li[i][j] = li[i][j], li[i][j - 1]
                        is_moved = True
                if not is_moved:
                    break
    # 오른쪽 이동
    else:
        for i in range(n):
            while True:
                is_moved = False
                for j in range(n - 1):
                    if li[i][j] != 0 and li[i][j + 1] == 0:
                        li[i][j + 1], li[i][j] = li[i][j], li[i][j + 1]
                        is_moved = True
                if not is_moved:
                    break
            for j in range(n - 1, 0, -1):
                if li[i][j] and li[i][j - 1] == li[i][j]:
                    li[i][j] *= 2
                    li[i][j - 1] = 0    
            while True:
                is_moved = False
                for j in range(n - 1):
                    if li[i][j] != 0 and li[i][j + 1] == 0:
                        li[i][j + 1], li[i][j] = li[i][j], li[i][j + 1]
                        is_moved = True
                if not is_moved:
                    break

n = int(input())
li = [list(map(int, input().split())) for _ in range(n)]

dx = [-1, 1, 0, 0]
dy = [0, 0, -1, 1]

queue = deque()
queue.append((li, 0)) # li, count
max_sc = 0 # max_score
while queue:
    li, count = queue.popleft()
    if count == 5:
        max_sc = max(max_sc, max(sum(li, [])))
        continue
        
    for d in range(4):
        nli = deepcopy(li)
        move(nli, d)
        queue.append((nli, count + 1))
        
print(max_sc)
```

### 소코반 4577 (신규)

```python
t = 1
while True:
    n, m = map(int, input().split())
    if n == 0:
        break
    li = [list(input()) for _ in range(n)]
    moves = input()

    mn = 'UDLR' # move names
    dx = [-1, 1, 0, 0]
    dy = [0, 0, -1, 1]

    botc = 0 # box on target count
    ts = set() # targets
    x, y = -1, -1 # 플레이어 위치
    for i in range(n):
        for j in range(m):
            if li[i][j] == 'W':
                li[i][j] = '.'
                ts.add((i, j))
                x, y = i, j
            elif li[i][j] == 'B':
                li[i][j] = 'b'
                ts.add((i, j))
                botc += 1
            elif li[i][j] == 'w':
                li[i][j] = '.'
                x, y = i, j
            elif li[i][j] == '+':
                li[i][j] = '.'
                ts.add((i, j))

    tc = len(ts) # target count

    for move in moves:
        d = mn.index(move)

        nx, ny = x + dx[d], y + dy[d]
        if li[nx][ny] == '.':
            x, y = nx, ny
        elif li[nx][ny] == 'b':
            tlx, tly = nx + dx[d], ny + dy[d] # target loc x
            if li[tlx][tly] != '.':
                continue
            if (nx, ny) in ts:
                botc -= 1
            li[nx][ny] = '.'
            x, y = nx, ny
            if (tlx, tly) in ts:
                botc += 1
            li[tlx][tly] = 'b'

        if botc == tc:
            print(f'Game {t}: complete')
            break
    else:
        print(f'Game {t}: incomplete')

    li[x][y] = 'w'
    for i in range(n):
        for j in range(m):
            if (i, j) in ts:
                if li[i][j] == 'b':
                    li[i][j] = 'B'
                elif li[i][j] == 'w':
                    li[i][j] = 'W'
                else:
                    li[i][j] = '+'

    print('\n'.join(''.join(v) for v in li))
    t += 1
```

### 모노미노도미노 2 20061

```python
import sys
# input = lambda: sys.stdin.readline().rstrip()

k = int(input())
bs = [list(map(int, input().split())) for _ in range(k)] # blocks

# k = 8
# s = '''1 1 1
# 2 3 0
# 3 2 2
# 3 2 3
# 3 1 3
# 2 0 0
# 3 2 0
# 3 1 2'''
# bs = [list(map(int, v.split())) for v in s.split('\n')]

n = 10
li = [[False] * n for _ in range(n)]

def gb(t, x, y):
    """get block"""
    
    if t == 1:
        return [[x, y]]
    elif t == 2:
        return [[x, y], [x, y + 1]]
    elif t == 3:
        return [[x, y], [x + 1, y]]    

def mob(t, x, y):
    """move one block"""
    
    dx = [1, 0]
    dy = [0, 1]
    
    for d in range(2): 
        b = gb(t, x, y)
        while True:
            is_end = False
            for bx, by in b:
                if bx + dx[d] >= 10 or by + dy[d] >= 10 or li[bx + dx[d]][by + dy[d]]:
                    is_end = True
                    break
            if is_end:
                break
            for i in range(len(b)):
                b[i][d] += 1
        for bx, by in b:
            li[bx][by] = True
            
def gs(sc):
    """ get score """
    
    i = 9
    while i > 5:
        if sum(li[i][:4]) == 4:
            sc += 1
            
            for ni in range(i - 1, 2, -1):
                li[ni + 1][:4] = li[ni][:4]
        else:
            i -= 1
            
    j = 9
    while j > 5:
        temp = 0
        for i in range(4):
            if li[i][j]:
                temp += 1
                
        if temp == 4:
            sc += 1
            
            for nj in range(j - 1, 2, -1):
                for i in range(4):
                    li[i][nj + 1] = li[i][nj]
        else:
            j -= 1
            
    return sc

def db():
    """delete outer blocks"""
    
    for _ in range(2):
        if sum(li[5][:4]) >= 1:
            for ni in range(n - 2, 2, -1):
                li[ni + 1][:4] = li[ni][:4]
   
        for i in range(4):
            if li[i][5]:
                for nj in range(n - 2, 2, -1):
                    for i in range(4):
                        li[i][nj + 1] = li[i][nj]
                break
                
sc = 0
for t, x, y in bs:
    mob(t, x, y)
    sc = gs(sc)
    db()
    
print(sc)
print(sum(sum(li, [])))
```

### 모노미노도미노 19235

```python
import sys
# input = lambda: sys.stdin.readline().rstrip()

k = int(input())
bs = [list(map(int, input().split())) for _ in range(k)]

# k = 9
# s = '''2 0 0
# 2 2 0
# 2 3 0
# 3 0 0
# 1 1 0
# 3 2 0'''
# bs = [list(map(int, v.split())) for v in s.split('\n')]

n = 10
li = [[0] * n for _ in range(n)]

def gb(t, x, y):
    '''get block'''
    
    if t == 1:
        return [(x, y)]
    elif t == 2:
        return [(x, y), (x, y + 1)]
    elif t == 3:
        return [(x, y), (x + 1, y)]

def mobd(b, bn):
    '''move one block to down'''
    
    xs = []
    ys = []
    for x, y in b:
        li[x][y] = 0
        xs.append(x)
        ys.append(y)

    max_x = max(xs)
    cands = []
    for y in ys:
        for nx in range(max_x, 10):
            if li[nx][y]:
                nx -= 1
                break
        cands.append(nx)
    max_x = min(cands)
        
    if len(b) == 2 and ys[0] == ys[1]:
        li[max_x][y] = li[max_x - 1][y] = bn
    else:
        for y in ys:
            li[max_x][y] = bn
            
def mobr(b, bn):
    '''move one block to right'''
    
    xs = []
    ys = []
    for x, y in b:
        li[x][y] = 0
        xs.append(x)
        ys.append(y)
    
    max_y = max(ys)
    cands = []
    for x in xs:
        for ny in range(max_y, 10):
            if li[x][ny]:
                ny -= 1
                break
        cands.append(ny)
    max_y = min(cands)
        
    if len(b) == 2 and xs[0] == xs[1]:
        li[x][max_y] = li[x][max_y - 1] = bn
    else:
        for x in xs:
            li[x][max_y] = bn    
    
def mob(b, bn):
    '''move one block'''
    
    mobd(b, bn)
    mobr(b, bn)
    
def mlb(ln, is_down=True):
    '''move line block, ln: line num'''
    
    if is_down:
        x = ln
        dx = [-1, 0]
        dy = [0, 1]
        grouped = set()
        for y in range(4):
            if li[x][y] == 0:
                continue
            if (x, y) in grouped:
                continue
            grouped.add((x, y))
            b = [(x, y)]
            for i in range(2):
                nx, ny = x + dx[i], y + dy[i]
                if ny >= 4:
                    continue
                if li[x][y] == li[nx][ny]:
                    grouped.add((nx, ny))
                    b.append((nx, ny))
            mobd(b, li[x][y])
    else:
        y = ln
        dx = [0, 1]
        dy = [-1, 0]
        grouped = set()
        for x in range(4):
            if li[x][y] == 0:
                continue
            if (x, y) in grouped:
                continue
            grouped.add((x, y))
            b = [(x, y)]
            for i in range(2):
                nx, ny = x + dx[i], y + dy[i]
                if nx >= 4:
                    continue
                if li[x][y] == li[nx][ny]:
                    grouped.add((nx, ny))
                    b.append((nx, ny))
            mobr(b, li[x][y])        
    
def gs(sc):
    '''get score'''
    
    # 아래쪽 점수 획득 및 라인 제거 및 이동
    while True:
        nsc = sc
        for x in range(6, 10):
            temp = 0
            for y in range(4):
                if li[x][y] != 0:
                    temp += 1
            if temp == 4:
                nsc += 1
                li[x][0] = li[x][1] = li[x][2] = li[x][3] = 0
        
        if nsc == sc:
            break
        sc = nsc
        for x in range(8, 3, -1):
            mlb(x, True)
            
    # 오른쪽 점수 획득 및 라인 제거 및 이동
    while True:
        nsc = sc
        for y in range(6, 10):
            temp = 0
            for x in range(4):
                if li[x][y] != 0:
                    temp += 1
            if temp == 4:
                nsc += 1
                li[0][y] = li[1][y] = li[2][y] = li[3][y] = 0
        
        if nsc == sc:
            break
        sc = nsc
        for y in range(8, 3, -1):
            mlb(y, False)
            
    return sc
    
def db():
    '''delete outer blocks'''
    
    # 아래쪽 연한 색깔 부분 확인 후 라인 제거 및 이동
    count = 0
    for x in [4, 5]:
        for y in range(4):
            if li[x][y] != 0:
                count += 1
                break
                
    tds = [] # to deletes
    if count >= 1:
        tds.append(9)
    if count == 2:
        tds.append(8)
        
    for x in tds:
        for y in range(4):
            li[x][y] = 0
    
    if count >= 1:
        for x in range(8, 3, -1):
            mlb(x, True)
            
    # 오른쪽 연한 색깔 부분 확인 후 라인 제거 및 이동
    count = 0
    for y in [4, 5]:
        for x in range(4):
            if li[x][y] != 0:
                count += 1
                break
                
    tds = [] # to deletes
    if count >= 1:
        tds.append(9)
    if count == 2:
        tds.append(8)
        
    for y in tds:
        for x in range(4):
            li[x][y] = 0
    
    if count >= 1:
        for y in range(8, 3, -1):
            mlb(y, False)    
        
sc = 0
bn = 1 # block num
for t, x, y in bs:
    b = gb(t, x, y)
    mob(b, bn)
    sc = gs(sc)
    db()
    sc = gs(sc)
    
    bn += 1
    
print(sc)
count = 0
for i in range(n):
    for j in range(n):
        if li[i][j]:
            count += 1
print(count)
```

### 개미 4307 (신규)

```python
import sys
input = lambda: sys.stdin.readline().rstrip()

for _ in range(int(input())):
    l, n = map(int, input().split())
    li = [int(input()) for _ in range(n)]
    li.sort()

    shortest = 0
    for v in li:
        shortest = max(shortest, min(v, l - v))

    longest = max(l - li[0], li[-1])

    print(shortest, longest)
```

### 역학 조사 19541 (신규)

```python
# 틀렸음
import sys
input = lambda: sys.stdin.readline().rstrip()

def check(mt, status, is_positive=0):
    """mt: meeting"""
    
    for i in range(1, len(mt)):
        if status[mt[i]] == is_positive:
            return True
    return False

n, m = map(int, input().split())

get_id = lambda x: int(x) - 1
li = [list(map(get_id, input().split())) for _ in range(m)]
rs = list(map(int, input().split())) # results

ss = rs[:] # starts

for i in range(m):
    if check(li[i], rs, 0):
        for j in range(1, len(li[i])):
            ss[li[i][j]] = 0     
            
to_check = ss[:]
for i in range(m):
    if check(li[i], to_check, 1):
        for j in range(1, len(li[i])):
            to_check[li[i][j]] = 1

if to_check == rs:
    print('YES')
    print(*ss)
else:
    print('NO')
```

```python
# 맞았음
import sys
input = lambda: sys.stdin.readline().rstrip()

def check(mt, status, is_positive=0):
    """mt: meeting"""
    
    for i in range(1, len(mt)):
        if status[mt[i]] == is_positive:
            return True
    return False

n, m = map(int, input().split())

get_id = lambda x: int(x) - 1
li = [list(map(get_id, input().split())) for _ in range(m)]
rs = list(map(int, input().split())) # results

ss = rs[:] # starts

for i in range(m - 1, -1, -1):
    if check(li[i], ss, 0):
        for j in range(1, len(li[i])):
            ss[li[i][j]] = 0
            
to_check = ss[:]
for i in range(m):
    if check(li[i], to_check, 1):
        for j in range(1, len(li[i])):
            to_check[li[i][j]] = 1

if to_check == rs:
    print('YES')
    print(*ss)
else:
    print('NO')
```
