### 단절점 11266

```python
import sys
sys.setrecursionlimit(int(1e5) + 100)
# input = lambda: sys.stdin.readline().rstrip()

def dfs(value, is_root):
    global next_num
    
    child_count = 0
    min_visited = vt[value]
    for v in li[value]:
        if vt[v] == 0:
            vt[v] = next_num
            next_num += 1
            temp = dfs(v, False)
         
            if not is_root and temp >= vt[value]:
                is_ans[value] = True
            
            min_visited = min(min_visited, temp)
            child_count += 1
        else:
            min_visited = min(min_visited, vt[v])
    
    if is_root and child_count >= 2:
        is_ans[value] = True
        
    return min_visited
        
n, m = map(int, input().split())

li = [[] for _ in range(n + 1)]
for _ in range(m):
    a, b = map(int, input().split())
    
    li[a].append(b)
    li[b].append(a)
    
vt = [0] * (n + 1)
next_num = 1
is_ans = [False] * (n + 1)
for i in range(1, n + 1):
    if vt[i] != 0:
        continue
    vt[i] = next_num
    next_num += 1
    
    dfs(i, True)

r = []
for i in range(1, n + 1):
    if is_ans[i]:
        r.append(i)
        
print(len(r))
print(*r)
```

### 단절선 11400

```python
import sys
input = lambda: sys.stdin.readline().rstrip()

## PYRIVAL BOOTSTRAP
# https://github.com/cheran-senthil/PyRival/blob/master/pyrival/misc/bootstrap.py
# This decorator allows for recursion without actually doing recursion
## @bootstrap, yield when getting and returning value in recursive functions, end of functions
from types import GeneratorType

def bootstrap(f, stack=[]):
    def wrappedfunc(*args, **kwargs):
        if stack:
            return f(*args, **kwargs)
        else:
            to = f(*args, **kwargs)
            while True:
                if type(to) is GeneratorType:
                    stack.append(to)
                    to = next(to)
                else:
                    stack.pop()
                    if not stack:
                        break
                    to = stack[-1].send(to)
            return to
    return wrappedfunc

@bootstrap
def dfs(value, to_evade):
    global next_num
    
    min_vt = vt[value]
    for v in li[value]:
        if vt[v] != 0:
            if v != to_evade:
                min_vt = min(min_vt, vt[v])
        else:
            vt[v] = next_num
            next_num += 1
            
            temp = yield dfs(v, value)
            
            if temp > vt[value]:
                a, b = value, v
                if a > b:
                    a, b = b, a
                r.append((a, b))
            
            min_vt = min(min_vt, temp)
            
    yield min_vt

n, m = map(int, input().split())
li = [[] for _ in range(n + 1)]
for _ in range(m):
    a, b = map(int, input().split())
    
    li[a].append(b)
    li[b].append(a)
    
next_num = 1
r = []
vt = [0] * (n + 1)
for i in range(1, n + 1):
    if vt[i]:
        continue
    vt[i] = next_num
    next_num += 1
    dfs(i, -1)

r.sort()

print(len(r))
for v in r:
    print(*v)
```
