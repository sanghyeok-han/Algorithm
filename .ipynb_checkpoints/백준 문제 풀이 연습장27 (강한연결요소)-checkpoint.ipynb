### 성격 진단 테스트 4305

```python
import sys
sys.setrecursionlimit(int(1e5) + 100)
# input = lambda: sys.stdin.readline().rstrip()

def dfs(value):
    vt[value] = True
    
    for v in li[value]:
        if vt[v]:
            continue
        dfs(v)
        
    stack.append(value)
    
def dfs2(value):
    vt[value] = True
    temp_r.append(value) 
    
    for v in rli[value]:
        if vt[v]:
            continue
        dfs2(v)
        
get_value = lambda x: ord(x) - 65

while True:
    n = int(input())
    if n == 0:
        break

    st = set()
    li = [[] for _ in range(26)]
    rli = [[] for _ in range(26)]
    for _ in range(n):
        temp = list(map(get_value, input().split()))
        a = temp.pop()
        st.add(a)
        for b in temp:
            if a == b:
                continue
            st.add(b)

            li[a].append(b)
            rli[b].append(a)

    stack = []
    vt = [False] * 26
    for i in range(26):
        if i not in st:
            continue
        if not vt[i]:
            dfs(i)

    vt = [False] * 26
    r = []
    while stack:
        value = stack.pop()
        if vt[value]:
            continue

        temp_r = []
        dfs2(value)
        r.append(sorted(temp_r))

    r.sort()

    for v in r:
        print(*[chr(w + 65) for w in v])

    print()
```

### 그래프의 싱크 6543

```python
import sys
sys.setrecursionlimit(int(1e5) + 100)
# input = lambda: sys.stdin.readline().rstrip()

def dfs(value):
    global next_id
    
    stack.append(value)
    on_stack[value] = True
    min_id = ids[value]
    
    for v in li[value]:
        if ids[v] != 0:
            if not on_stack[v]:
                continue
            min_id = min(min_id, low_ids[v])
        else:
            ids[v] = next_id
            low_ids[v] = next_id
            next_id += 1
            temp = dfs(v)
            min_id = min(min_id, temp)
        
    low_ids[value] = min_id
         
    if ids[value] == low_ids[value]:
        scc = set()
        while True:
            v = stack.pop()
            scc.add(v)
            on_stack[v] = False
            
            if value == v:
                break
        sccs.append(scc)
        
    return min_id

def dfs2(value, scc_id):
    vt[value] = True
    
    for v in li[value]:
        if v not in sccs[scc_id]:
            return False
        if not vt[v]:
            if not dfs2(v, scc_id):
                return False
        
    return True
    
while True:
    temp = list(map(int, input().split()))
    if temp[0] == 0:
        break
    n, m = temp
    
    li = [[] for _ in range(n + 1)]
    edges = list(map(int, input().split()))
    for i in range(0, m * 2, 2):
        a = edges[i]
        b = edges[i + 1]

        li[a].append(b)

    ids = [0] * (n + 1)
    low_ids = [0] * (n + 1)
    on_stack = [False] * (n + 1)
    stack = []
    sccs = []
    next_id = 1

    for i in range(1, n + 1):
        if ids[i] == 0:
            ids[i] = next_id
            low_ids[i] = next_id
            next_id += 1
            dfs(i)
            
    result = []
    vt = [False] * (n + 1)
    for scc_id in range(len(sccs)):
        value = sccs[scc_id].pop()
        sccs[scc_id].add(value)

        if dfs2(value, scc_id):
            result.extend(sccs[scc_id])

    result.sort()
    if not result:
        print()
    else:
        print(*result)
```

```python
import io, os, sys
input = io.BytesIO(os.read(0,os.fstat(0).st_size)).readline

## PYRIVAL BOOTSTRAP
# https://github.com/cheran-senthil/PyRival/blob/master/pyrival/misc/bootstrap.py
# This decorator allows for recursion without actually doing recursion
## @bootstrap, yield when getting and returning value in recursive functions, end of functions
from types import GeneratorType

def bootstrap(f, stack=[]):
    def wrappedfunc(*args, **kwargs):
        if stack:
            return f(*args, **kwargs)
        else:
            to = f(*args, **kwargs)
            while True:
                if type(to) is GeneratorType:
                    stack.append(to)
                    to = next(to)
                else:
                    stack.pop()
                    if not stack:
                        break
                    to = stack[-1].send(to)
            return to
    return wrappedfunc

def invert(num):
    if num % 2 == 0:
        num += 1
    else:
        num -= 1
    
    return num

@bootstrap
def dfs(value):
    for v in li[value]:
        if vt[v]:
            continue
        vt[v] = True
        yield dfs(v)
        
    stack.append(value)
    yield

@bootstrap 
def dfs2(value):
    for v in rli[value]:
        if vt[v]:
            continue
        vt[v] = True
        temp.append(v)
        yield dfs2(v)
    yield

n, m = map(int, input().split())

li = [[] for _ in range(n * 2)]
rli = [[] for _ in range(n * 2)]
for _ in range(m):
    a, b = map(int, input().split())
    
    if a < 0:
        a += 1
        na = (-a) * 2 + 1
    else:
        a -= 1
        na = a * 2
        
    if b < 0:
        b += 1
        nb = (-b) * 2 + 1
    else:
        b -= 1
        nb = b * 2    
    
    li[invert(na)].append(nb)
    li[invert(nb)].append(na)
    rli[nb].append(invert(na))
    rli[na].append(invert(nb))
    
nn = n * 2
    
stack = []
vt = [False] * nn
for i in range(nn):
    if vt[i]:
        continue
    vt[i] = True
    dfs(i)
    
vt = [False] * nn
scc = []
while stack:
    value = stack.pop()
    if vt[value]:
        continue
    vt[value] = True
        
    temp = [value]
    dfs2(value)
    scc.append(temp.copy())
    
ip = True
for v in scc:
    v_st = set(v)
    
    for w in v_st:
        if w % 2 == 0 and (w + 1) in v_st:
            ip = False
            break
            
if ip:
    print(1)
else:
    print(0)
```

```python
from collections import deque
import io, os, sys
input = io.BytesIO(os.read(0,os.fstat(0).st_size)).readline

## PYRIVAL BOOTSTRAP
# https://github.com/cheran-senthil/PyRival/blob/master/pyrival/misc/bootstrap.py
# This decorator allows for recursion without actually doing recursion
## @bootstrap, yield when getting and returning value in recursive functions, end of functions
from types import GeneratorType

def bootstrap(f, stack=[]):
    def wrappedfunc(*args, **kwargs):
        if stack:
            return f(*args, **kwargs)
        else:
            to = f(*args, **kwargs)
            while True:
                if type(to) is GeneratorType:
                    stack.append(to)
                    to = next(to)
                else:
                    stack.pop()
                    if not stack:
                        break
                    to = stack[-1].send(to)
            return to
    return wrappedfunc

def invert(num):
    if num % 2 == 0:
        num += 1
    else:
        num -= 1
    
    return num

@bootstrap
def dfs(value):
    for v in li[value]:
        if vt[v]:
            continue
        vt[v] = True
        yield dfs(v)
        
    stack.append(value)
    yield

@bootstrap 
def dfs2(value):
    for v in rli[value]:
        if vt[v]:
            continue
        vt[v] = True
        temp.append(v)
        yield dfs2(v)
    yield

while True:
    try:
        n, m = map(int, input().split())

        li = [[] for _ in range(n * 2)]
        rli = [[] for _ in range(n * 2)]
        for _ in range(m):
            a, b = map(int, input().split())

            if a < 0:
                a += 1
                na = (-a) * 2 + 1
            else:
                a -= 1
                na = a * 2

            if b < 0:
                b += 1
                nb = (-b) * 2 + 1
            else:
                b -= 1
                nb = b * 2    

            li[invert(na)].append(nb)
            li[invert(nb)].append(na)
            rli[nb].append(invert(na))
            rli[na].append(invert(nb))

        nn = n * 2

        stack = []
        vt = [False] * nn
        for i in range(nn):
            if vt[i]:
                continue
            vt[i] = True
            dfs(i)

        vt = [False] * nn
        scc = []
        while stack:
            value = stack.pop()
            if vt[value]:
                continue
            vt[value] = True

            temp = [value]
            dfs2(value)
            scc.append(temp.copy())

        ip = True
        for v in scc:
            v_st = set(v)

            for w in v_st:
                if w % 2 == 0 and (w + 1) in v_st:
                    ip = False
                    break

        if not ip:
            print('no')
        else:
            vt = [False] * nn
            vt[0] = True

            queue = deque()
            queue.append(0)
            while queue:
                value = queue.popleft()
                if value == 1:
                    ip = False
                    break

                for v in li[value]:
                    if not vt[v]:
                        vt[v] = True
                        queue.append(v)

            if ip:
                print('yes')
            else:
                print('no')
    except:
        break
```
