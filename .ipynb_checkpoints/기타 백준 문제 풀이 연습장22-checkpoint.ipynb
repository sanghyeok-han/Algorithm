```python
import io, os, sys
input = io.BytesIO(os.read(0, os.fstat(0).st_size)).readline
print = sys.stdout.write

n = int(input())

is_cows = [[False] * 1001 for _ in range(1001)]
is_comfs = [[False] * 1001 for _ in range(1001)]

dx = [0, 0, -1, 1]
dy = [-1, 1, 0, 0]

comf_c = 0
for _ in range(n):
    x, y = map(int, input().split())
    
    is_cows[x][y] = True
    
    c = 0
    to_check = []
    for d in range(4):
        nx = x + dx[d]
        ny = y + dy[d]
        
        if nx < 0 or nx > 1000 or ny < 0 or ny > 1000:
            continue
        if is_cows[nx][ny]:
            to_check.append((nx, ny))
            c += 1
        
    if c == 3:
        is_comfs[x][y] = True
        comf_c += 1
        
    for x, y in to_check:
        c = 0
        for d in range(4):
            nx = x + dx[d]
            ny = y + dy[d]

            if nx < 0 or nx > 1000 or ny < 0 or ny > 1000:
                continue
            if is_cows[nx][ny]:
                c += 1
        
        if c == 3 and not is_comfs[x][y]:
            comf_c += 1
            is_comfs[x][y] = True
        elif c != 3 and is_comfs[x][y]:
            comf_c -= 1
            is_comfs[x][y] = False
            
    print(str(comf_c) + '\n')
```

```python
import sys
input = lambda: sys.stdin.readline().rstrip()

mod = 720

def to_total_m(tm):
    h, m = map(int, tm.split(':'))
    
    total_m = h * 60 + m
    return total_m

n = int(input())
li = [input() for _ in range(n)]

min_r = n
for R in range(720):
    st = set()
    reduced = 0
    for i in range(n):
        total_m = to_total_m(li[i])

        init_total_m = total_m - (i * R)
        init_total_m %= mod
        
        if init_total_m in st:
            reduced += 1
        else:
            st.add(init_total_m)
            
    min_r = min(min_r, n - reduced)
    
print(min_r)
```

```python
import io, os, sys
input = io.BytesIO(os.read(0, os.fstat(0).st_size)).readline

inf = float('inf')

n = int(input())

vt = [False] * (n + 1)
d = dict()
li = [[] for _ in range(n + 1)]
result = 0
for _ in range(n):
    a, b, p = map(int, input().split())
    
    if a > b:
        a, b = b, a
    
    if a == b:
        vt[a] = True
        result += p
        continue
        
    tp = (a, b)
    if tp in d:
        result += min(d[tp], p)
        vt[a] = vt[b] = True
    else:
        li[a].append((b, p))
        li[b].append((a, p))
        
        d[tp] = p
        
for num in range(1, n + 1):
    if vt[num]:
        continue
    vt[num] = True
    
    start = num
    nnum = li[num][0][0]    
    ps = [li[num][0][1]]
    vt[nnum] = True
    
    num = nnum
    while True:
        for nnum, p in li[num]:
            if not vt[nnum]:
                vt[nnum] = True
                ps.append(p)
                num = nnum
                break
        else:
            for nnum, p in li[num]:
                if nnum == start:
                    ps.append(p)
                    break
            break
            
    ps_len = len(ps)
    
    dp = [[inf] * 2 for _ in range(ps_len)] # selected, not selected
    dp[0][0] = ps[0]
    
    for i in range(1, ps_len):
        dp[i][0] = min(dp[i - 1]) + ps[i]
        dp[i][1] = dp[i - 1][0]
    
    min_to_add = min(dp[ps_len - 1])
    
    dp = [[inf] * 2 for _ in range(ps_len)] # selected, not selected
    dp[0][1] = 0
    
    for i in range(1, ps_len):
        dp[i][0] = min(dp[i - 1]) + ps[i]
        dp[i][1] = dp[i - 1][0]
        
    min_to_add = min(min_to_add, dp[ps_len - 1][0])
    
    result += min_to_add
    
print(result)
```

```python
from collections import defaultdict as dd
import io, os, sys
input = io.BytesIO(os.read(0, os.fstat(0).st_size)).readline

inf = float('inf')

n = int(input())
x_to_ys = dd(list)
for _ in range(n):
    x, y = map(int, input().split())
    
    x_to_ys[x].append(y)

xs = sorted(x_to_ys)
xs_len = len(xs)

maxes_to_right = [0] * xs_len
mins_to_right = [0] * xs_len

cumul_max = -inf
cumul_min = inf
for i in range(xs_len):
    cumul_max = max(cumul_max, max(x_to_ys[xs[i]]))
    maxes_to_right[i] = cumul_max
    
    cumul_min = min(cumul_min, min(x_to_ys[xs[i]]))
    mins_to_right[i] = cumul_min

maxes_to_left = [0] * xs_len
mins_to_left = [0] * xs_len
    
cumul_max = -inf
cumul_min = inf
for i in range(xs_len - 1, -1, -1):
    cumul_max = max(cumul_max, max(x_to_ys[xs[i]]))
    maxes_to_left[i] = cumul_max
    
    cumul_min = min(cumul_min, min(x_to_ys[xs[i]]))
    mins_to_left[i] = cumul_min
    
max_x = xs[-1]
min_x = xs[0]

max_y = maxes_to_right[-1]
min_y = mins_to_right[-1]

surface = (max_x - min_x + 1) * 2 + (max_y - min_y + 1) * 2

maxes = [0] * xs_len
mins = [0] * xs_len
area = 0
for i in range(xs_len):
    maxes[i] = min(maxes_to_left[i], maxes_to_right[i])
    mins[i] = max(mins_to_left[i], mins_to_right[i])
    area += maxes[i] - mins[i] + 1
    
for i in range(1, xs_len):
    min_max = min(maxes[i - 1], maxes[i])
    max_min = max(mins[i - 1], mins[i])
    
    if min_max >= max_min:
        area += (min_max - max_min + 1) * max(0, (xs[i] - xs[i - 1] - 1))
    else:
        area += max(0, (xs[i] - xs[i - 1] - 1)) + (max_min - min_max)
    
print(surface, area)
```

```python
n = int(input())
li = [0] + list(map(int, input().split()))

nn = 2 ** n
dp = [[-1] * 2 for _ in range(nn)] # [[to_min, to_max], ..]
dp_to_left = [[-1] * 2 for _ in range(nn)]
dp_to_right = [[-1] * 2 for _ in range(nn)]

dp[1] = [li[1], li[1]]

for start in range(1, n):
    s = 2 ** start
    e = 2 ** (start + 1) - 1
    
    dp_to_left[s][0] = li[s]
    dp_to_left[s][0] += min(0, dp[s // 2][1])
    
    dp_to_left[s][1] = li[s]
    dp_to_left[s][1] += max(0, dp[s // 2][0])
    
    for i in range(s + 1, e + 1):
        dp_to_left[i][0] = li[i]
        dp_to_left[i][0] += min(0, dp[i // 2][1], dp_to_left[i - 1][1])
        
        dp_to_left[i][1] = li[i]
        dp_to_left[i][1] += max(0, dp[i // 2][0], dp_to_left[i - 1][0])
        
    dp_to_right[e][0] = li[e]
    dp_to_right[e][0] += min(0, dp[e // 2][1])
    
    dp_to_right[e][1] = li[e]
    dp_to_right[e][1] += max(0, dp[e // 2][0])
    
    for i in range(e - 1, s - 1, -1):
        dp_to_right[i][0] = li[i]
        dp_to_right[i][0] += min(0, dp[i // 2][1], dp_to_right[i + 1][1])
        
        dp_to_right[i][1] = li[i]
        dp_to_right[i][1] += max(0, dp[i // 2][0], dp_to_right[i + 1][0]) 
        
    for i in range(s, e + 1):
        dp[i][0] = min(dp_to_left[i][0], dp_to_right[i][0])
        dp[i][1] = max(dp_to_left[i][1], dp_to_right[i][1])

result = []
for i in range(1, nn):
    result.append(dp[i][0])
    
print(*result)
```

```python
mod = 1000000007

class SegmentTree:
    def __init__(self, data, default=10**15, func=lambda a, b: max(a,b)):
        """initialize the segment tree with data"""
        self._default = default
        self._func = func
        self._len = len(data)
        self._size = _size = 1 << (self._len - 1).bit_length()
 
        self.data = [default] * (2 * _size)
        self.data[_size:_size + self._len] = data
        for i in reversed(range(_size)):
            self.data[i] = func(self.data[i + i], self.data[i + i + 1])
 
    def __delitem__(self, idx):
        self[idx] = self._default
 
    def __getitem__(self, idx):
        return self.data[idx + self._size]
 
    def __setitem__(self, idx, value):
        idx += self._size
        self.data[idx] = value
        idx >>= 1
        while idx:
            self.data[idx] = self._func(self.data[2 * idx], self.data[2 * idx + 1])
            idx >>= 1
 
    def __len__(self):
        return self._len
 
    def query(self, start, stop):
        if start == stop:
            return self.__getitem__(start)
        stop += 1
        start += self._size
        stop += self._size
 
        res = self._default
        while start < stop:
            if start & 1:
                res = self._func(res, self.data[start])
                start += 1
            if stop & 1:
                stop -= 1
                res = self._func(res, self.data[stop])
            start >>= 1
            stop >>= 1
        return res
 
    def __repr__(self):
        return "SegmentTree({0})".format(self.data)
    
func = lambda a, b: a + b

n = int(input())
li = list(map(int, input().split()))

li = [(li[i], i) for i in range(n)]
li.sort(key=lambda x: x[0])

count_seg = SegmentTree([0] * (n + 1), 0, func=func)
sum_seg = SegmentTree([0] * (n + 1), 0, func=func)

result = 0
for value, index in li:
    sum_v = sum_seg.query(0, index)
    count_v = count_seg.query(0, index)
    
    result += count_v * value - sum_v
    result %= mod
    
    sum_seg[0] += value
    sum_seg[0] %= mod
    sum_seg[index + 1] -= value
    sum_seg[index + 1] %= mod
    
    count_seg[0] += 1
    count_seg[index + 1] -= 1
    
print(result)
```
