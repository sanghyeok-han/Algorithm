```python
alphas = 'RGYB'

A = input()
B = input()

A_cs = [0] * 4
B_cs = [0] * 4

for v in A:
    if v in alphas:
        A_cs[alphas.index(v)] += 1
        
for v in B:
    if v in alphas:
        B_cs[alphas.index(v)] += 1
        
diffs = [A_cs[i] - B_cs[i] for i in range(4)]
result = sum([v for v in diffs if v >= 1])

print(result)
```

```python
mod = 1000000007

n, e = map(int, input().split())

if n <= e:
    print(0)
else:
    dp = [[0] * (n + 2) for _ in range(n + 1)]
    sdp = [[0] * (n + 2) for _ in range(n + 1)]

    dp[1][0] = 1
    for length in range(2, n + 1):
        for sc in range(1, n + 1):
            dp[length][sc] += dp[length - 1][sc - 1]
            dp[length][sc] %= mod

        for sc in range(n + 1):
            for a_length in range(1, length):
                b_length = length - 1 - a_length

                t1 = dp[a_length][sc] * sdp[b_length][sc + 1]
                t2 = sdp[a_length][sc + 1] * dp[b_length][sc]
                t3 = dp[a_length][sc] * dp[b_length][sc]

                t1 %= mod
                t2 %= mod
                t3 %= mod

                dp[length][sc] += t1 + t2 + t3
                dp[length][sc] %= mod

        cumul = 0
        for sc in range(n, -1, -1):
            cumul += dp[length][sc]
            cumul %= mod
            sdp[length][sc] = cumul

    result = dp[n][e]

    print(result)
```

https://www.acmicpc.net/contest/problem/1124/5 F번 - 트리와 쿼리 10^9

```python
import io, os, sys
input = io.BytesIO(os.read(0, os.fstat(0).st_size)).readline

q = int(input())

pa = {}
for _ in range(q):
    tp, a, b = map(int, input().split())
    
    if tp == 1:
        pa[b] = a
    else:
        a_route = [a]
        cur = a
        while cur != 1:
            if cur in pa:
                cur = pa[cur]
            else:
                cur //= 2
                
            a_route.append(cur)
            
        b_route = [b]
        cur = b
        while cur != 1:
            if cur in pa:
                cur = pa[cur]
            else:
                cur //= 2
                
            b_route.append(cur)
            
        sames = []
        while a_route and b_route and a_route[-1] == b_route[-1]:
            sames.append(a_route[-1])
            a_route.pop()
            b_route.pop()
            
        result = sum(a_route) + sum(b_route) + sames[-1]
        
        print(result)
```

https://www.acmicpc.net/contest/problem/1124/5 E번 - 영업의 신

```python
import io, os, sys
input = io.BytesIO(os.read(0, os.fstat(0).st_size)).readline

# https://github.com/tatyam-prime/SortedSet/blob/main/SortedSet.py
import math
from bisect import bisect_left, bisect_right
from typing import Generic, Iterable, Iterator, TypeVar, Union, List
T = TypeVar('T')

class SortedSet(Generic[T]):
    BUCKET_RATIO = 50
    REBUILD_RATIO = 170

    def _build(self, a=None) -> None:
        "Evenly divide `a` into buckets."
        if a is None: a = list(self)
        size = self.size = len(a)
        bucket_size = int(math.ceil(math.sqrt(size / self.BUCKET_RATIO)))
        self.a = [a[size * i // bucket_size : size * (i + 1) // bucket_size] for i in range(bucket_size)]
    
    def __init__(self, a: Iterable[T] = []) -> None:
        "Make a new SortedSet from iterable. / O(N) if sorted and unique / O(N log N)"
        a = list(a)
        if not all(a[i] < a[i + 1] for i in range(len(a) - 1)):
            a = sorted(set(a))
        self._build(a)

    def __iter__(self) -> Iterator[T]:
        for i in self.a:
            for j in i: yield j

    def __reversed__(self) -> Iterator[T]:
        for i in reversed(self.a):
            for j in reversed(i): yield j
    
    def __len__(self) -> int:
        return self.size
    
    def __repr__(self) -> str:
        return "SortedSet" + str(self.a)
    
    def __str__(self) -> str:
        s = str(list(self))
        return "{" + s[1 : len(s) - 1] + "}"

    def _find_bucket(self, x: T) -> List[T]:
        "Find the bucket which should contain x. self must not be empty."
        for a in self.a:
            if x <= a[-1]: return a
        return a

    def __contains__(self, x: T) -> bool:
        if self.size == 0: return False
        a = self._find_bucket(x)
        i = bisect_left(a, x)
        return i != len(a) and a[i] == x

    def add(self, x: T) -> bool:
        "Add an element and return True if added. / O(√N)"
        if self.size == 0:
            self.a = [[x]]
            self.size = 1
            return True
        a = self._find_bucket(x)
        i = bisect_left(a, x)
        if i != len(a) and a[i] == x: return False
        a.insert(i, x)
        self.size += 1
        if len(a) > len(self.a) * self.REBUILD_RATIO:
            self._build()
        return True

    def discard(self, x: T) -> bool:
        "Remove an element and return True if removed. / O(√N)"
        if self.size == 0: return False
        a = self._find_bucket(x)
        i = bisect_left(a, x)
        if i == len(a) or a[i] != x: return False
        a.pop(i)
        self.size -= 1
        if len(a) == 0: self._build()
        return True
    
    def lt(self, x: T) -> Union[T, None]:
        "Find the largest element < x, or None if it doesn't exist."
        for a in reversed(self.a):
            if a[0] < x:
                return a[bisect_left(a, x) - 1]

    def le(self, x: T) -> Union[T, None]:
        "Find the largest element <= x, or None if it doesn't exist."
        for a in reversed(self.a):
            if a[0] <= x:
                return a[bisect_right(a, x) - 1]

    def gt(self, x: T) -> Union[T, None]:
        "Find the smallest element > x, or None if it doesn't exist."
        for a in self.a:
            if a[-1] > x:
                return a[bisect_right(a, x)]

    def ge(self, x: T) -> Union[T, None]:
        "Find the smallest element >= x, or None if it doesn't exist."
        for a in self.a:
            if a[-1] >= x:
                return a[bisect_left(a, x)]
    
    def __getitem__(self, x: int) -> T:
        "Return the x-th element, or IndexError if it doesn't exist."
        if x < 0: x += self.size
        if x < 0: raise IndexError
        for a in self.a:
            if x < len(a): return a[x]
            x -= len(a)
        raise IndexError
    
    def index(self, x: T) -> int:
        "Count the number of elements < x."
        ans = 0
        for a in self.a:
            if a[-1] >= x:
                return ans + bisect_left(a, x)
            ans += len(a)
        return ans

    def index_right(self, x: T) -> int:
        "Count the number of elements <= x."
        ans = 0
        for a in self.a:
            if a[-1] > x:
                return ans + bisect_right(a, x)
            ans += len(a)
        return
    
n, m, k = map(int, input().split())

wins = [0] * n
win_c = 0

ssts = [SortedSet() for _ in range(m)]
cumuls = [{} for _ in range(n)]

for pn in range(n):
    temp = list(map(int, input().split()))
    for i in range(0, 2 * k, 2):
        mn, t = temp[i], temp[i + 1]
        mn -= 1
        
        cumuls[pn][mn] = t
        ssts[mn].add((t, pn))
        
for mn in range(m):
    if ssts[mn]:
        winner = ssts[mn][-1][1]
        wins[winner] += 1
        
for pn in range(n):
    if wins[pn] == k:
        win_c += 1
        
q = int(input())
for _ in range(q):
    pn, mn, v = map(int, input().split())
    pn -= 1
    mn -= 1
    
    init = cumuls[pn][mn]
    
    cur_winner = ssts[mn][-1][1]
    
    sst_index = ssts[mn].index((init, -1))
    ssts[mn].discard(ssts[mn][sst_index])
    
    nxt = init + v
    cumuls[pn][mn] = nxt
    
    ssts[mn].add((nxt, pn))
    
    nxt_winner = ssts[mn][-1][1]
    
    if cur_winner != nxt_winner:
        if wins[cur_winner] == k:
            win_c -= 1
        wins[cur_winner] -= 1
        
        wins[nxt_winner] += 1
        if wins[nxt_winner] == k:
            win_c += 1
            
    print(win_c)
```

https://www.acmicpc.net/contest/problem/1134/7 G번 - 돌베어 법칙

```python
from collections import deque

n = int(input())
s = input()

if s == '.' * n:
    print(0)
else:
    min_c = 2000
    for term in range(1, n):
        deq = deque()
        c = 0
        for i in range(n):
            if s[i] == '#':
                while deq and deq[0] + term < i:
                    deq.popleft()
                    
                if deq and deq[0] + term == i:
                    deq.popleft()
                    deq.append(i)
                else:
                    c += 1
                    deq.append(i)
        min_c = min(min_c, c)
        
    print(min_c)
```

```python
inf = float('inf')

s, n = map(int, input().split())
E = list(map(int, input().split()))
k, l = map(int, input().split())

min_r = inf
min_r_i = inf
for i in range(n):
    e = E[i]
    
    diff = abs(s - e)
    
    if diff >= k * 2:
        diff -= k * 2
        needed = diff * l
    else:
        needed = k * 2 - diff
        
    if needed < min_r:
        min_r = needed
        min_r_i = i

print(min_r, min_r_i + 1)
```

https://www.acmicpc.net/contest/problem/1134/9 I번 - 공부 계획하기

```python
inf = float('inf')

n, t = map(int, input().split())
S = [list(map(int, input().split())) for _ in range(n)]
D = list(map(int, input().split()))

dp = [[0] * (t + 1) for _ in range(n)]
_from = [[0] * (t + 1) for _ in range(n)]

dp[0] = S[0]
for i in range(1, n):
    for tm in range(t + 1):
        max_sc = -1
        max_sc_ftm = -1
        for ftm in range(tm + 1):
            study_time = tm - ftm
            sc = dp[i - 1][ftm] + S[i][study_time]
            
            if sc > max_sc:
                max_sc = sc
                max_sc_ftm = ftm
                
        dp[i][tm] = max_sc
        _from[i][tm] = max_sc_ftm

max_r = -inf
max_r_tm = -1
for tm in range(t + 1):
    temp = dp[n - 1][tm] - D[tm]
    if temp > max_r:
        max_r = temp
        max_r_tm = tm

study_times = [0] * n
cur = max_r_tm
for i in range(n - 1, -1, -1):
    fcur = _from[i][cur]
    study_time = cur - fcur
    study_times[i] = study_time
    cur = fcur

print(max_r)
print(*study_times)
```

https://www.acmicpc.net/contest/problem/1134/8 H번 - 준영이의 사랑

```python
from itertools import permutations as perm

li = [1, 1, 1]

max_r = 0
max_r_cases = []
for case in perm(li):
    temp = 0
    for i in range(len(case)):
        temp += case[i] * case[i - 1]
        
        if temp == max_r:
            max_r = temp
            max_r_cases.append(case)
        elif temp > max_r:
            max_r = temp
            max_r_cases = [case]
            
print(max_r)
print(max_r_cases)
```

```python
import io, os, sys
# input = io.BytesIO(os.read(0, os.fstat(0).st_size)).readline

inf = float('inf')

## PYRIVAL BOOTSTRAP
# https://github.com/cheran-senthil/PyRival/blob/master/pyrival/misc/bootstrap.py
# This decorator allows for recursion without actually doing recursion
## @bootstrap, yield when getting and returning value in recursive functions, end of functions
from types import GeneratorType

def bootstrap(f, stack=[]):
    def wrappedfunc(*args, **kwargs):
        if stack:
            return f(*args, **kwargs)
        else:
            to = f(*args, **kwargs)
            while True:
                if type(to) is GeneratorType:
                    stack.append(to)
                    to = next(to)
                else:
                    stack.pop()
                    if not stack:
                        break
                    to = stack[-1].send(to)
            return to
    return wrappedfunc

@bootstrap
def dfs(value):
    vt_dfs[value] = True
        
    max_down_dist = 0
    for v, d in li[value]:
        if vt_dfs[v]:
            continue
        
        dist = d + (yield dfs(v))
        max_down_dist = max(max_down_dist, dist)
        
    max_down_dists[value] = max_down_dist
    yield max_down_dist
    
@bootstrap
def reroot(value, max_up_dist):
    vt_reroot[value] = True
    
    dp[value] = max(max_up_dist, max_down_dists[value])
    
    temp = []
    for v, d in li[value]:
        if vt_reroot[v]:
            continue
            
        temp.append((v, max_down_dists[v], d, max_down_dists[v] + d))
        
    temp.sort(key=lambda x: -x[3])
    
    if temp:
        max_down_dist = temp[0][3]
        for i in range(1, len(temp)):
            v, md, d, md_d = temp[i]
            
            yield reroot(v, max(max_up_dist, max_down_dist) + d)

        if len(temp) >= 2:
            yield reroot(temp[0][0], max(max_up_dist, temp[1][3]) + temp[0][2])
        else:
            yield reroot(temp[0][0], temp[0][2] + max_up_dist)
        
    yield
    
    
n = int(input())
li = [[] for _ in range(n)]
for _ in range(n - 1):
    u, v, d = map(int, input().split())
    u -= 1
    v -= 1

    li[u].append((v, d))
    li[v].append((u, d))

max_down_dists = [0] * n
vt_dfs = [False] * n
dp = [0] * n
dfs(0)

dp[0] = max_down_dists[0]

vt_reroot = [False] * n
reroot(0, 0)

for v in dp:
    print(v)
```

```python
inf = float('inf')

def S_k(num):
    sv = 0
    for digit in str(num):
        sv += int(digit) ** k
        
    return sv

a, b, k = map(int, input().split())

dp = [inf] * 4000000

for v in range(a, b + 1): 
    if dp[v] != inf:
        continue
    
    st = {v}
    seq = [v]

    cur = v
    while True:
        ncur = S_k(cur)

        if ncur in st:
            index = seq.index(ncur)
            min_v = min(seq[index:])
            
            for i in range(len(seq) - 1, index - 1, -1):
                dp[seq[i]] = min_v
                
            for i in range(index - 1, -1, -1):
                min_v = min(min_v, seq[i])
                dp[seq[i]] = min_v
            
            break
        elif dp[ncur] != inf:
            min_v = dp[ncur]
            
            for i in range(len(seq) - 1, -1, -1):
                min_v = min(min_v, seq[i])
                dp[seq[i]] = min_v
            
            break
            
        st.add(ncur)
        seq.append(ncur)

        cur = ncur
    
result = sum(dp[a:b + 1])

print(result)
```

https://www.acmicpc.net/problem/27560 Moo Route

```python
n = int(input())
li = list(map(int, input().split())) + [0]

cur = 0
result = []
while True:
    while True:
        if li[cur] >= 2:
            result.append('R')
            li[cur] -= 1
            cur += 1
        else:
            cur -= 1
            break
    
    is_all_zero = True
    while True:
        if li[cur] == 1:
            if is_all_zero:
                li[cur] -= 1
                result.append('L')
                cur -= 1
                if cur == -1:
                    cur += 1
                    break
            else:
                cur += 1
                break
        else:
            li[cur] -= 1
            result.append('L')
            if li[cur] != 0:
                is_all_zero = False
            cur -= 1
            if cur == -1:
                cur += 1
                break
                    
    if cur == 0 and is_all_zero:
        break
    
result = ''.join(result)

print(result)
```

https://www.acmicpc.net/problem/30108 교육적인 트리 문제

```python
from heapq import heappush, heappop

get_v = lambda x: int(x) - 1

n = int(input())
P = [-1] + list(map(get_v, input().split()))
A = list(map(int, input().split()))

li = [[] for _ in range(n)]
for i in range(1, n):
    p = P[i]
    cur = i
    
    li[p].append(cur)
    
heap = []
heappush(heap, (-A[0], 0))

result = []
cumul = 0
while heap:
    a, value = heappop(heap)
    a = -a
    
    cumul += a
    result.append(cumul)
    
    for v in li[value]:
        heappush(heap, (-A[v], v))
        
result = '\n'.join(map(str, result))

print(result)
```

https://www.acmicpc.net/problem/1848 동굴 탐험

```python
from heapq import heappush, heappop
import io, os, sys
input = io.BytesIO(os.read(0, os.fstat(0).st_size)).readline

inf = float('inf')

def dij(dists, heap):
    while heap:
        dist, value = heappop(heap)

        if dist > dists[value]:
            continue

        for v in li[value]:
            d = mat[value][v]

            ndist = dist + d
            if ndist < dists[v]:
                dists[v] = ndist

                heappush(heap, (ndist, v))
    
    return dists
    
n, m = map(int, input().split())

mat = [[inf] * n for _ in range(n)]
li = [[] for _ in range(n)]
adj_to_0s = set()
for _ in range(m):
    a, b, c, d = map(int, input().split())
    a -= 1
    b -= 1
    
    if a > b:
        a, b = b, a
        c, d = d, c
    
    li[a].append(b)
    li[b].append(a)
    
    mat[a][b] = min(mat[a][b], c)
    mat[b][a] = min(mat[b][a], d)
    
    if a == 0:
        adj_to_0s.add(b)
        
adj_to_0s = list(adj_to_0s)
adj_to_0s_len = len(adj_to_0s)

min_result = inf
w = 1
while True:
    lefts = []
    rights = []
    for i in range(adj_to_0s_len):
        if (i // w) % 2 == 0:
            lefts.append(adj_to_0s[i])
        else:
            rights.append(adj_to_0s[i])
            
    if not lefts or not rights:
        break

    dists = [inf] * n
    dists[0] = 0
    heap = []
    for value in lefts:
        dist = mat[0][value]

        heappush(heap, (dist, value))
        dists[value] = dist
        
    dists = dij(dists, heap)

    for v in rights:
        dist = mat[v][0]

        cand = dists[v] + dist
        min_result = min(min_result, cand)

    dists = [inf] * n
    dists[0] = 0
    heap = []
    for value in rights:
        dist = mat[0][value]

        heappush(heap, (dist, value))
        dists[value] = dist
    
    dists = dij(dists, heap)

    for v in lefts:
        dist = mat[v][0]

        cand = dists[v] + dist
        min_result = min(min_result, cand)
        
    w *= 2

print(min_result)
```

https://www.acmicpc.net/problem/1099 알 수 없는 문장

```python
from collections import Counter

inf = float('inf')

s = input()
s_len = len(s)

n = int(input())
li = [input() for _ in range(n)]

cts = [Counter(v) for v in li]

dp = [[inf] * s_len for _ in range(s_len)]
for i in range(s_len):
    for j in range(i, s_len):
        temp = s[i:j + 1]
        temp_ct = Counter(temp)
        
        for k in range(n):
            if temp_ct == cts[k]:
                cost = 0
                for l in range(i, j + 1):
                    if s[l] != li[k][l - i]:
                        cost += 1
                        
                dp[i][j] = min(dp[i][j], cost)
                
for term in range(1, s_len):
    for start in range(s_len - term):
        end = start + term
        
        for left_end in range(start, end):
            dp[start][end] = min(dp[start][end], dp[start][left_end] + dp[left_end + 1][end])
            
if dp[0][s_len - 1] == inf:
    print(-1)
else:
    print(dp[0][s_len - 1])
```

https://www.acmicpc.net/problem/17410 수열과 쿼리 1.5

```python
from bisect import bisect_right
import io, os, sys
input = io.BytesIO(os.read(0, os.fstat(0).st_size)).readline
print = sys.stdout.write

n = int(input())
li = list(map(int, input().split()))
m = int(input())

cri = int(n ** 0.5) + 1

parts = []
cur = 0
temp = []
for i in range(n):
    if cur < cri:
        temp.append(li[i])
        cur += 1
    else:
        parts.append(sorted(temp))
        cur = 1
        temp = [li[i]]
        
if temp:
    parts.append(sorted(temp))
    
parts_start_is = []
part_lens = []
cur = 0
for v in parts:
    parts_start_is.append(cur)
    
    v_len = len(v)
    cur += v_len
    part_lens.append(v_len)
    
result = []
for _ in range(m):
    tp, *qu = map(int, input().split())
    
    if tp == 1:
        i, v = qu
        i -= 1
        
        part_index = bisect_right(parts_start_is, i) - 1
        parts[part_index].remove(li[i])
        
        li[i] = v
        
        parts[part_index].append(li[i])
        parts[part_index].sort()
    
    elif tp == 2:
        i, j, k = qu
        
        i -= 1
        j -= 1
        
        left_part_index = bisect_right(parts_start_is, i) - 1
        right_part_index = bisect_right(parts_start_is, j) - 1
        
        c = 0
        if left_part_index == right_part_index:
            for index in range(i, j + 1):
                if li[index] > k:
                    c += 1
        else:
            for index in range(i, parts_start_is[left_part_index + 1]):
                if li[index] > k:
                    c += 1
            for index in range(parts_start_is[right_part_index], j + 1):
                 if li[index] > k:
                    c += 1               
            
            for part_index in range(left_part_index + 1, right_part_index):
                c += part_lens[part_index] - bisect_right(parts[part_index], k)
   
        result.append(c)
    
print('\n'.join(map(str, result)))
```

https://www.acmicpc.net/problem/1354 무한 수열 2

```python
def recur(num):
    if num <= 0:
        return 1
    if num < cri and dp[num] != -1:
        return dp[num]
            
    f_num1 = num // p - x
    f_num2 = num // q - y
    
    r = recur(f_num1) + recur(f_num2)
    
    if num < cri:
        dp[num] = r
    
    return r
    
n, p, q, x, y = map(int, input().split())

cri = 8000000
dp = [-1] * cri

result = recur(n)

print(result)
```

https://www.acmicpc.net/problem/1148 단어 만들기

```python
from collections import Counter
import sys
input = lambda: sys.stdin.readline().rstrip()

li = []
while True:
    s = input()
    
    if s == '-':
        break
        
    li.append(Counter(s))
    
while True:
    s = input()
    
    if s == '#':
        break
    
    ct = Counter(s)
    
    max_c = -1
    min_c = 100000000
    max_c_centers = set()
    min_c_centers = set()
    for center in s:
        c = 0
        for fct in li:
            if fct[center] == 0:
                continue
            
            ip = True
            for key in fct:
                if ct[key] < fct[key]:
                    ip = False
                    break
                    
            if ip:
                c += 1
                
        if c > max_c:
            max_c = c
            max_c_centers = {center}
        elif c == max_c:
            max_c_centers.add(center)
            
        if c < min_c:
            min_c = c
            min_c_centers = {center}
        elif c == min_c:
            min_c_centers.add(center)
            
    r1 = ''.join(sorted(min_c_centers))
    r2 = min_c
    r3 = ''.join(sorted(max_c_centers))
    r4 = max_c
    
    print(r1, r2, r3, r4)
```

https://www.acmicpc.net/problem/21610 마법사 상어와 비바라기

```python
dx = [0, -1, -1, -1, 0, 1, 1, 1]
dy = [-1, -1, 0, 1, 1, 1, 0, -1]

n, m = map(int, input().split())
li = [list(map(int, input().split())) for _ in range(n)]

clouds = [[n - 1, 0], [n - 1, 1], [n - 2, 0], [n - 2, 1]]
for _ in range(m):
    d, s = map(int, input().split())
    d -= 1
    
    for cloud in clouds:
        cloud[0] += dx[d] * s
        cloud[0] %= n
        
        cloud[1] += dy[d] * s
        cloud[1] %= n
               
    for x, y in clouds:
        li[x][y] += 1
        
    news = []
    for x, y in clouds:
        c = 0
        for nd in [1, 3, 5, 7]:
            nx = x + dx[nd]
            ny = y + dy[nd]
            
            if nx < 0 or nx >= n or ny < 0 or ny >= n:
                continue
            
            if li[nx][ny] >= 1:
                c += 1
                
        news.append(c)
        
    i = 0
    for x, y in clouds:
        li[x][y] += news[i]
        i += 1
        
    clouds = [tuple(cloud) for cloud in clouds]
    clouds = set(clouds)
    nclouds = []
    for i in range(n):
        for j in range(n):
            if li[i][j] >= 2 and (i, j) not in clouds:
                li[i][j] -= 2
                nclouds.append([i, j])
                
    clouds = nclouds.copy()
        
result = sum(sum(v) for v in li)

print(result)
```
