```python
import sys
sys.setrecursionlimit(110000)
input = lambda: sys.stdin.readline().rstrip()

inf = float('inf')

def dfs(value, depth):
    depths[value] = depth
    
    for v, d in li[value]:
        if depths[v] == -1:
            parents[v][0] = value
            min_dists[v][0] = d
            max_dists[v][0] = d
            dfs(v, depth + 1)

def lca(a, b, max_times):
    if depths[a] < depths[b]:
        a, b = b, a
        
    min_r = inf
    max_r = -inf
    for cur_tms in range(max_times, -1, -1):
        if depths[a] - 2 ** cur_tms >= depths[b]:
            min_r = min(min_r, min_dists[a][cur_tms])
            max_r = max(max_r, max_dists[a][cur_tms])
            a = parents[a][cur_tms]
            
    for cur_tms in range(max_times, -1, -1):
        if parents[a][cur_tms] != parents[b][cur_tms]:
            min_r = min(min_r, min_dists[a][cur_tms])
            min_r = min(min_r, min_dists[b][cur_tms])
            max_r = max(max_r, max_dists[a][cur_tms])
            max_r = max(max_r, max_dists[b][cur_tms])
            a = parents[a][cur_tms]
            b = parents[b][cur_tms]
            
    if a != b:
        min_r = min(min_r, min_dists[a][0])
        min_r = min(min_r, min_dists[b][0])
        max_r = max(max_r, max_dists[a][0])
        max_r = max(max_r, max_dists[b][0])
        a = parents[a][0]
        b = parents[b][0]
        
    return min_r, max_r
                 
n = int(input())

li = [[] for _ in range(n + 1)]
for _ in range(n - 1):
    u, v, d = map(int, input().split())
    
    li[u].append((v, d))
    li[v].append((u, d))
    
max_times = 16
    
parents = [[0] * (max_times + 1) for _ in range(n + 1)]
min_dists = [[inf] * (max_times + 1) for _ in range(n + 1)]
max_dists = [[-inf] * (max_times + 1) for _ in range(n + 1)]

depths = [-1] * (n + 1)

root = 1
dfs(root, 0)

for cur_times in range(1, max_times + 1):
    for i in range(1, n + 1):
        p1 = parents[i][cur_times - 1]
        p2 = parents[p1][cur_times - 1]
        
        min_d1 = min_dists[i][cur_times - 1]
        min_d2 = min_dists[p1][cur_times - 1]
        max_d1 = max_dists[i][cur_times - 1]
        max_d2 = max_dists[p1][cur_times - 1]
        
        parents[i][cur_times] = p2
        min_dists[i][cur_times] = min(min_d1, min_d2)
        max_dists[i][cur_times] = max(max_d1, max_d2)

for _ in range(int(input())):
    u, v = map(int, input().split())
    
    print(*lca(u, v, max_times))
```

https://www.acmicpc.net/problem/13511 트리와 쿼리 2

```python
import sys
sys.setrecursionlimit(110000)
input = lambda: sys.stdin.readline().rstrip()

def dfs(value, depth):
    depths[value] = depth
    
    for v, c in li[value]:
        if depths[v] == -1:
            parents[v][0] = value
            costs[v][0] = c
            dists[v][0] = 1
            dfs(v, depth + 1)
            
def lca(oa, ob, k, max_times):
    a, b = oa, ob
    
    a_dist = 0
    b_dist = 0
    cost = 0
    
    for cur_tms in range(max_times, -1, -1):
        if depths[a] - 2 ** cur_tms >= depths[b]:
            a_dist += dists[a][cur_tms]
            cost += costs[a][cur_tms]
            a = parents[a][cur_tms]
        if depths[b] - 2 ** cur_tms >= depths[a]:
            b_dist += dists[b][cur_tms]
            cost += costs[b][cur_tms]
            b = parents[b][cur_tms]

    for cur_tms in range(max_times, -1, -1):
        if parents[a][cur_tms] != parents[b][cur_tms]:
            a_dist += dists[a][cur_tms]
            cost += costs[a][cur_tms]
            a = parents[a][cur_tms]            
            b_dist += dists[b][cur_tms]
            cost += costs[b][cur_tms]
            b = parents[b][cur_tms]
            
    if a != b:
        a_dist += dists[a][0]
        cost += costs[a][0]
        a = parents[a][0]            
        b_dist += dists[b][0]
        cost += costs[b][0]
        b = parents[b][0]
    
    if k == -1:
        return cost
    else:
        a, b = oa, ob
        
        if k == 1:
            return a
        elif k <= a_dist + 1:
            to_up = k - 1
            for cur_tms in range(max_times, -1, -1):
                if 2 ** cur_tms <= to_up:
                    to_up -= 2 ** cur_tms
                    a = parents[a][cur_tms]
            return a
        else:
            to_up = (a_dist + b_dist + 1) - k
            for cur_tms in range(max_times, -1, -1):
                if 2 ** cur_tms <= to_up:
                    to_up -= 2 ** cur_tms
                    b = parents[b][cur_tms]
            return b

n = int(input())
li = [[] for _ in range(n + 1)]

for _ in range(n - 1):
    u, v, c = map(int, input().split())
    li[u].append((v, c))
    li[v].append((u, c))
    
max_times = 16
parents = [[0] * (max_times + 1) for _ in range(n + 1)]
costs = [[0] * (max_times + 1) for _ in range(n + 1)]
dists = [[0] * (max_times + 1) for _ in range(n + 1)]

depths = [-1] * (n + 1)
root = 1
dfs(root, 0)

for cur_tms in range(1, max_times + 1):
    for i in range(1, n + 1):
        p1 = parents[i][cur_tms - 1]
        p2 = parents[p1][cur_tms - 1]
        parents[i][cur_tms] = p2
        
        c1 = costs[i][cur_tms - 1]
        c2 = costs[p1][cur_tms - 1]
        costs[i][cur_tms] = c1 + c2
        
        d1 = dists[i][cur_tms - 1]
        d2 = dists[p1][cur_tms - 1]
        dists[i][cur_tms] = d1 + d2

for _ in range(int(input())):
    qu = list(map(int, input().split()))
    
    tp = qu[0]
    if tp == 1:
        r = lca(qu[1], qu[2], -1, max_times)
    else:
        r = lca(qu[1], qu[2], qu[3], max_times)
        
    print(r)
```

```python
# 처음 제출했을 때 틀리고 상하행선을 구분해야 하는 줄 알고 구현했다가 틀린 버전
import sys
sys.setrecursionlimit(230000)
# input = lambda: sys.stdin.readline().rstrip()

def dfs(value, depth):
    depths[value] = depth
    
    for v in li[value]:
        if depths[v] == -1:
            parents[v][0] = value
            dfs(v, depth + 1)

def lca(a, b):
    if depths[a] < depths[b]:
        a, b = b, a
    
    for cur_tms in range(max_tms, -1, -1):
        if depths[a] - 2 ** cur_tms >= depths[b]:
            a = parents[a][cur_tms]
    
    for cur_tms in range(max_tms, -1, -1):
        if parents[a][cur_tms] != parents[b][cur_tms]:
            a = parents[a][cur_tms]
            b = parents[b][cur_tms]
    
    if a != b:
        a = parents[a][0]
        b = parents[b][0]
    
    return a

def dfs_to_up(value):
    global max_r
    global max_r_case
    
    vt[value] = True
    
    cur = 0
    for v in li[value]:
        if not vt[v]:
            temp = dfs_to_up(v)
            if temp > max_r or temp == max_r and [v, value] < max_r_case:
                max_r = temp
                max_r_case = [v, value]
            cur += temp
            
    cur += deltas_to_up[value]
    
    return cur

def dfs_to_down(value, temp):
    global max_r
    global max_r_case
    
    vt[value] = True
    
    temp += deltas_to_down[value]
    for v in li[value]:
        if not vt[v]:
            if temp > max_r or temp == max_r and [value, v] < max_r_case:
                max_r = temp
                max_r_case = [value, v]
            dfs_to_down(v, temp)
            
n, q = map(int, input().split())
li = [[] for _ in range(n + 1)]

for _ in range(n - 1):
    a, b = map(int, input().split())
    
    li[a].append(b)
    li[b].append(a)
    
max_tms = 17
depths = [-1] * (n + 1)
parents = [[0] * (max_tms + 1) for _ in range(n + 1)]

root = 1
dfs(1, 0)

for cur_tms in range(1, max_tms + 1):
    for i in range(1, n + 1):
        p1 = parents[i][cur_tms - 1]
        p2 = parents[p1][cur_tms - 1]
        
        parents[i][cur_tms] = p2
        
deltas_to_up = [0] * (n + 1)
deltas_to_down = [0] * (n + 1)

for _ in range(q):
    f, t = map(int, input().split())
    
    temp = lca(f, t)
    
    if temp == t:
        deltas_to_up[f] += 1
        deltas_to_up[t] -= 1
    elif temp == f:
        deltas_to_down[f] += 1
        deltas_to_down[t] -= 1
    else:
        deltas_to_up[f] += 1
        deltas_to_up[temp] -= 1
        deltas_to_down[temp] += 1
        deltas_to_down[t] -= 1      
        
max_r = -1
max_r_case = []

vt = [False] * (n + 1)
dfs_to_up(1)

vt = [False] * (n + 1)
dfs_to_down(1, 0)

print(*max_r_case, max_r)
```

https://www.acmicpc.net/problem/10623 Breadth-First Search by Foxpower

```python
from collections import deque
import sys
input = lambda: sys.stdin.readline().rstrip()

## PYRIVAL BOOTSTRAP
# https://github.com/cheran-senthil/PyRival/blob/master/pyrival/misc/bootstrap.py
# This decorator allows for recursion without actually doing recursion
## @bootstrap, yield when getting and returning value in recursive functions, end of functions
from types import GeneratorType

def bootstrap(f, stack=[]):
    def wrappedfunc(*args, **kwargs):
        if stack:
            return f(*args, **kwargs)
        else:
            to = f(*args, **kwargs)
            while True:
                if type(to) is GeneratorType:
                    stack.append(to)
                    to = next(to)
                else:
                    stack.pop()
                    if not stack:
                        break
                    to = stack[-1].send(to)
            return to
    return wrappedfunc

@bootstrap
def dfs(value, depth):
    depths[value] = depth
    
    for v in li[value]:
        parents[v][0] = value
        yield dfs(v, depth + 1)
        
    yield

def lca(a, b):
    if depths[a] < depths[b]:
        a, b = b, a
    
    total = 0
    for cur_tms in range(max_tms, -1, -1):
        if depths[a] - 2 ** cur_tms >= depths[b]:
            total += 2 ** cur_tms
            a = parents[a][cur_tms] 
    
    for cur_tms in range(max_tms, -1, -1):
        if parents[a][cur_tms] != parents[b][cur_tms]:
            total += 2 ** (cur_tms + 1)
            a = parents[a][cur_tms]
            b = parents[b][cur_tms]
    
    if a != b:
        total += 2
        a = parents[a][0]
        b = parents[b][0]
    
    return total

n = int(input())
info = list(map(int, input().split()))

li = [[] for _ in range(n + 1)]
for i in range(n - 1):
    v = i + 2
    value = info[i]
    
    li[value].append(v)
    
for i in range(1, n + 1):
    li[i].sort()

max_tms = 16
depths = [-1] * (n + 1)
parents = [[0] * (max_tms + 1) for _ in range(n + 1)]

dfs(1, 0)

for cur_tms in range(1, max_tms + 1):
    for i in range(1, n + 1):
        p1 = parents[i][cur_tms - 1]
        p2 = parents[p1][cur_tms - 1]
        
        parents[i][cur_tms] = p2

queue = deque()
queue.append(1)
seq = []
while queue:
    value = queue.popleft()
    seq.append(value)
    
    for v in li[value]:
        queue.append(v)

total = 0
for i in range(n - 1):
    total += lca(seq[i], seq[i + 1])
    
print(total)
```

https://www.acmicpc.net/problem/24520 Meet In The Middle

```python
import sys
sys.setrecursionlimit(110000)
input = lambda: sys.stdin.readline().rstrip()

inf = float('inf')

def dfs(value, depth):
    depths[value] = depth
    
    for v, w in li[value]:
        if depths[v] == -1:
            parents[v][0] = value
            dists[v][0] = w
            dfs(v, depth + 1)
            
def lca(a, b):
    if depths[b] > depths[a]:
        a, b = b, a
        
    total = 0
    for cur_tms in range(max_tms, -1, -1):
        if depths[a] - 2 ** cur_tms >= depths[b]:
            total += dists[a][cur_tms]
            a = parents[a][cur_tms]
            
    for cur_tms in range(max_tms, -1, -1):
        if parents[a][cur_tms] != parents[b][cur_tms]:
            total += dists[a][cur_tms]
            total += dists[b][cur_tms]
            a = parents[a][cur_tms]
            b = parents[b][cur_tms]

    if a != b:
        total += dists[a][0]
        total += dists[b][0]
        a = parents[a][0]
        b = parents[b][0]
        
    return total, a

n, k = map(int, input().split())

li = [[] for _ in range(n + 1)]
for _ in range(n - 1):
    u, v, w = map(int, input().split())
    
    li[u].append((v, w))
    li[v].append((u, w))
    
root = 1
depths = [-1] * (n + 1)
max_tms = 16
parents = [[0] * (max_tms + 1) for _ in range((n + 1))]
dists = [[inf] * (max_tms + 1) for _ in range((n + 1))]
dfs(root, 0)

for cur_tms in range(1, max_tms + 1):
    for i in range(1, n + 1):
        p1 = parents[i][cur_tms - 1]
        p2 = parents[p1][cur_tms - 1]
        
        parents[i][cur_tms] = p2
        
        d1 = dists[i][cur_tms - 1]
        d2 = dists[p1][cur_tms - 1]
        
        dist = d1 + d2
        if p2 == 0:
            dist = inf
        dists[i][cur_tms] = dist
        
for _ in range(k):
    u, v = map(int, input().split())
    
    dist, lca_value = lca(u, v)
    if dist % 2 == 1:
        print(-1)
    else:
        half_d = dist // 2
        
        if lca_value == v:
            u, v = v, u
        
        if lca_value == u:
            total = 0
            for cur_tms in range(max_tms - 1, -1, -1):
                if total + dists[v][cur_tms] <= half_d:
                    total += dists[v][cur_tms]
                    v = parents[v][cur_tms]
                    
            if total == half_d:
                print(v)
            else:
                print(-1)
        else:
            r = -1
            nu, nv = u, v
            
            max_can_move = lca(u, lca_value)[0]
            if half_d <= max_can_move:
                total = 0
                for cur_tms in range(max_tms - 1, -1, -1):
                    if total + dists[nu][cur_tms] <= half_d:
                        total += dists[nu][cur_tms]
                        nu = parents[nu][cur_tms]

                if total == half_d:
                    r = nu
                    
            max_can_move = lca(v, lca_value)[0]
            if half_d <= max_can_move:
                total = 0
                for cur_tms in range(max_tms - 1, -1, -1):
                    if total + dists[nv][cur_tms] <= half_d:
                        total += dists[nv][cur_tms]
                        nv = parents[nv][cur_tms]

                if total == half_d:
                    r = nv
                    
            print(r)
```
