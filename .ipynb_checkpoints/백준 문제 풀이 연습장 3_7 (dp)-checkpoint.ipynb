```python
# 틀렸음
import sys
# input = lambda: sys.stdin.readline().rstrip()

inf = float('inf')

for _ in range(int(input())):
    C, D, d = map(int, input().split())
    a = list(map(int, input().split()))
    b = list(map(int, input().split()))

    a.pop()
    b.pop()

    a_len = len(a)
    b_len = len(b)

    dp = [[inf] * b_len for _ in range(a_len)]

    ai = 0
    for bi in range(b_len):
        diff = bi - ai
        base = C * diff
        base += D * int(diff >= 1) + d * diff

        if bi == b_len - 1:
            diff = a_len - 1 - ai

            base += C * diff
            base += D * int(diff >= 1) + d * diff   

        if a[ai] == b[bi]:
            dp[ai][bi] = base + C
        else:
            dp[ai][bi] = base + C * 2

    bi = 0
    for ai in range(1, a_len):
        diff = ai - bi
        base = C * diff
        base += D * int(diff >= 1) + d * diff

        if ai == a_len - 1:
            diff = b_len - 1 - bi

            base += C * diff
            base += D * int(diff >= 1) + d * diff   

        if a[ai] == b[bi]:
            dp[ai][bi] = base + C
        else:
            dp[ai][bi] = base + C * 2

    for ai in range(1, a_len):
        for bi in range(1, b_len):
            min_temp = inf

            fai = ai - 1
            for fbi in range(bi):
                base = dp[fai][fbi]

                diff = bi - 1 - fbi
                base += C * diff
                base += D * int(diff >= 1) + d * diff

                if a[ai] == b[bi]:
                    temp = base + C
                else:
                    temp = base + C * 2

                min_temp = min(min_temp, temp)

            fbi = bi - 1
            for fai in range(ai):
                base = dp[fai][fbi]

                diff = ai - 1 - fai
                base += C * diff
                base += D * int(diff >= 1) + d * diff

                if a[ai] == b[bi]:
                    temp = base + C
                else:
                    temp = base + C * 2

                min_temp = min(min_temp, temp)

            if ai == a_len - 1:
                diff = b_len - 1 - bi

                base = C * diff
                base += D * int(diff >= 1) + d * diff

                min_temp += base

            if bi == b_len - 1:
                diff = a_len - 1 - ai

                base = C * diff
                base += D * int(diff >= 1) + d * diff

                min_temp += base

            dp[ai][bi] = min_temp

    min_r = inf

    ai = a_len - 1
    for bi in range(b_len):
        min_r = min(min_r, dp[ai][bi])

    bi = b_len - 1
    for ai in range(a_len):
        min_r = min(min_r, dp[ai][bi])

    print(min_r)
```

```python
import sys
input = lambda: sys.stdin.readline().rstrip()

inf = float('inf')

n, t = map(int, input().split())

dp = [[-inf] * n for _ in range(t + 1)]

cs = []
es = []
for i in range(n):
    c, e = map(int, input().split())
    
    cs.append(c)
    es.append(e)
    
    if c == 0:
        dp[0][i] = 0

li = [list(map(int, input().split())) for _ in range(n)]

for cur_t in range(1, t + 1):
    for cur_n in range(n):
        max_temp = dp[cur_t - 1][cur_n] + es[cur_n]
        for f_n in range(n):
            f_t = cur_t - li[f_n][cur_n]
            if f_t >= 0 and dp[f_t][f_n] >= cs[cur_n]:
                max_temp = max(max_temp, dp[f_t][f_n])
                
        dp[cur_t][cur_n] = max_temp
        
max_r = 0
for cur_n in range(n):
    max_r = max(max_r, dp[t][cur_n])
    
print(max_r)
```

https://www.acmicpc.net/problem/22983 조각 체스판

```python
import sys
input = lambda: sys.stdin.readline().rstrip()

n, m = map(int, input().split())
li = [input() for _ in range(n)]

conts1 = [[0] * m for _ in range(n)]
conts2 = [[0] * m for _ in range(n)]

for i in range(n):
    conts1[i][0] = 1 
    for j in range(1, m):
        if li[i][j] != li[i][j - 1]:
            conts1[i][j] = conts1[i][j - 1] + 1
        else:
            conts1[i][j] = 1
            
for j in range(m):
    conts2[0][j] = 1
    for i in range(1, n):
        if li[i][j] != li[i - 1][j]:
            conts2[i][j] = conts2[i - 1][j] + 1
        else:
            conts2[i][j] = 1

dp = [[0] * m for _ in range(n)]

for i in range(n):
    dp[i][0] = 1
    
for j in range(m):
    dp[0][j] = 1
    
total = n + m - 1

for i in range(1, n):
    for j in range(1, m):
        if li[i][j] != li[i - 1][j - 1]:
            dp[i][j] = 1
            total += 1
        else:
            dp[i][j] = min(dp[i - 1][j - 1] + 1, conts1[i][j], conts2[i][j])
            total += dp[i][j]
            
print(total)
```

```python
mod = int(1e9) + 7
max_num = 20001

factos = [0] * (max_num + 1)
factos[0] = 1
invs = [0] * (max_num + 1)
cur = 1
for i in range(1, max_num + 1):
    cur *= i
    cur %= mod
    factos[i] = cur
    
invs[max_num] = pow(factos[max_num], mod - 2, mod)
for i in range(max_num - 1, -1, -1):
    invs[i] = (i + 1) * invs[i + 1] % mod

def nCk(n, k, mod):
    if n < k:
        return 0
    if k < 0:
        return 0
    return factos[n] * invs[k] * invs[n - k] % mod

n = int(input())

r = nCk(2 * n, n, mod) * pow(n + 1, mod - 2, mod)
r %= mod

print(r)
```

https://www.acmicpc.net/problem/1126 같은 탑

```python
# 시간 초과 확실
max_num = 250000

n = int(input())
li = list(map(int, input().split()))

dp = [set() for _ in range(max_num + 1)]
for i in range(n):
    for j in range(max_num, li[i] - 1, -1):
        if not dp[j]:
            nj = j - li[i]
            if dp[nj] or nj == 0:
                dp[j] = dp[nj].copy()
                dp[j].add(i)
                
result = -1
for num in range(max_num, 0, -1):
    if dp[num]:
        ndp = [False] * (num + 1)
        ndp[0] = True
        
        for index in range(n):
            if index not in dp[num]:
                for j in range(num, li[index] - 1, -1):
                    nj = j - li[index]
                    if ndp[nj]:
                        ndp[j] = True
                        
        if ndp[num]:
            result = num
            break
            
print(result)
```

https://www.acmicpc.net/problem/17192 Snakes

```python
inf = float('inf')

n, k = map(int, input().split())
li = list(map(int, input().split()))

maxes = [[0] * n for _ in range(n)]
sums = [[0] * n for _ in range(n)]

for i in range(n):
    for j in range(i, n):
        if i == j:        
            maxes[i][j] = li[i]
            sums[i][j] = li[i]
        else:
            maxes[i][j] = max(maxes[i][j - 1], li[j])
            sums[i][j] = sums[i][j - 1] + li[j]
        
dp = [[inf] * n for _ in range(k + 1)]

for i in range(n):
    dp[0][i] = maxes[0][i] * (i + 1) - sums[0][i]
    
for cur_k in range(1, k + 1):
    for cur_n in range(cur_k, n):
        min_temp = inf
        for f_n in range(cur_k - 1, cur_n):
            temp = dp[cur_k - 1][f_n] + maxes[f_n + 1][cur_n] * (cur_n - (f_n + 1) + 1) - sums[f_n + 1][cur_n]
            min_temp = min(min_temp, temp)
        dp[cur_k][cur_n] = min_temp
    
r = dp[k][n - 1]

print(r)
```

https://www.acmicpc.net/problem/18790 N의 배수 (1)

```python
# 접근 방식이 잘못되었음
from math import gcd

n = int(input())
li = list(map(int, input().split()))
nn = 2 * n - 1

if 0 in li:
    li.remove(0)
    print(0, *li[:n - 1])
else:
    dp_c = [inf] * (n + 1)
    dp_c[1] = 0
    dp = [set() for _ in range(n + 1)]
    
    for i in range(nn):
        g = gcd(li[i], n)
        for j in range(n, g - 1, -1):
            if j % g == 0:
                nj = j // g
                if nj == 1 or dp[nj]:
                    if dp_c[nj] + 1 < dp_c[j]:
                        dp_c[j] = dp_c[nj] + 1
                        dp[j] = dp[nj].copy()
                        dp[j].add(i)
```

https://www.acmicpc.net/problem/8895 막대 배치

```python
import sys
# input = lambda: sys.stdin.readline().rstrip()

def nCk(n, k):
    if n < k:
        return 0
    if k < 0:
        return 0
    
    if n - k < k:
        k = n - k
        
    num = 1
    for i in range(n, n - k, -1):
        num *= i
    
    denom = 1
    for i in range(2, k + 1):
        denom *= i
        
    return num // denom

def nPk(n, k):
    temp = 1
    for i in range(n, n - k, -1):
        temp *= i
        
    return temp

def recur(cur_n, cur_target):
    if dp[cur_n][cur_target] != -1:
        return dp[cur_n][cur_target]
    if cur_n < cur_target or cur_target == 0 and cur_n >= 1:
        dp[cur_n][cur_target] = 0
        return 0
    
    total = 0
    for left_c in range(cur_n):
        right_c = cur_n - 1 - left_c
        w = nPk(cur_n - 1, left_c)
        temp = recur(right_c, cur_target - 1)
        temp *= w
        total += temp
        
    dp[cur_n][cur_target] = total
    return total    

dp = [[-1] * 21 for _ in range(21)] # dp[n][target]
dp[0][0] = 1
dp[1][1] = 1

for cur_n in range(21):
    for cur_target in range(21):
        dp[cur_n][cur_target] = recur(cur_n, cur_target)

for _ in range(int(input())):
    n, l, r = map(int, input().split())

    total = 0
    for left_c in range(n):
        right_c = n - 1 - left_c

        w = nCk(n - 1, left_c)        
        temp = dp[left_c][l - 1] * dp[right_c][r - 1]
        temp *= w
        total += temp

    print(total)
```

https://www.acmicpc.net/problem/1328 고층 빌딩

```python
mod = 1000000007

def nCk(n, k):
    if n < k:
        return 0
    if k < 0:
        return 0
    
    if n - k < k:
        k = n - k
        
    num = 1
    for i in range(n, n - k, -1):
        num *= i
    
    denom = 1
    for i in range(2, k + 1):
        denom *= i
        
    return num // denom

def nPk(n, k):
    temp = 1
    for i in range(n, n - k, -1):
        temp *= i
        temp %= mod
        
    return temp

def recur(cur_n, cur_target):
    if dp[cur_n][cur_target] != -1:
        return dp[cur_n][cur_target]
    if cur_n < cur_target or cur_target == 0 and cur_n >= 1:
        dp[cur_n][cur_target] = 0
        return 0
    
    total = 0
    for left_c in range(cur_n):
        right_c = cur_n - 1 - left_c
        w = nPk(cur_n - 1, left_c)
        temp = recur(right_c, cur_target - 1) % mod
        temp *= w
        temp %= mod
        total += temp
        total %= mod
        
    dp[cur_n][cur_target] = total
    return total    

dp = [[-1] * 101 for _ in range(101)] # dp[n][target]
dp[0][0] = 1
dp[1][1] = 1

for cur_n in range(101):
    for cur_target in range(101):
        dp[cur_n][cur_target] = recur(cur_n, cur_target)

n, l, r = map(int, input().split())

total = 0
for left_c in range(n):
    right_c = n - 1 - left_c

    w = nCk(n - 1, left_c) % mod
    temp = dp[left_c][l - 1] * dp[right_c][r - 1] % mod
    temp *= w
    temp %= mod
    total += temp
    total %= mod

print(total)
```

https://www.acmicpc.net/problem/1311 할 일 정하기 1

```python
inf = float('inf')

def recur(ts, cur_f_to_match):
    if dp[ts] != -1:
        return dp[ts]

    min_r = inf
    for cur_t_to_match in range(n):
        tnum = 1 << cur_t_to_match
        if ts & tnum:
            nts = ts - tnum
            recur(nts, cur_f_to_match + 1)
            temp_r = dp[nts] + costs[cur_f_to_match][cur_t_to_match]
            min_r = min(min_r, temp_r)
            
    dp[ts] = min_r

n = int(input())
costs = [list(map(int, input().split())) for _ in range(n)]

init = 2 ** n - 1
dp = [-1] * (init + 1)
dp[0] = 0

for cur_t in range(n):
    dp[1 << cur_t] = costs[n - 1][cur_t]

recur(init, 0)

result = dp[init]

print(result)
```

```python
mod = 1000000009

n, k = map(int, input().split())

dp_a = [0] * (n + 1)
dp_ab = [0] * (n + 1)
dp_aba = [0] * (n + 1)
dp_abab = [0] * (n + 1)
dp_abc = [0] * (n + 1)
dp_abcb = [0] * (n + 1)
dp = [0] * (n + 1)

dp_a[1] = 1
dp[1] = k

for i in range(2, n + 1):
    dp_a[i] = dp[i - 1] - dp_ab[i - 1] - dp_abab[i - 1]
    dp_ab[i] = dp_a[i - 1]
    dp_aba[i] = dp_ab[i - 1] + dp_abab[i - 1]
    dp_abab[i] = dp_aba[i - 1]
    dp_abc[i] = dp_ab[i - 1]
    dp_abcb[i] = dp_abc[i - 1]
    
    dp_a[i] %= mod
    dp_ab[i] %= mod
    dp_aba[i] %= mod
    dp_abab[i] %= mod
    dp_abc[i] %= mod
    dp_abcb[i] %= mod

    dp[i] = dp[i - 1] * k - dp_abab[i - 1] - dp_abcb[i - 1]
    dp[i] %= mod
    
r = dp[n]

print(r)
```

```python
mod = 1000000007

n, m, p = map(int, input().split())

dp = [[0] * (n + 1) for _ in range(p + 1)]
dp[1][1] = n

for cur_p in range(2, p + 1):
    for cur_n in range(1, min(cur_p, n) + 1):
        t1 = dp[cur_p - 1][cur_n - 1] * (n - (cur_n - 1))
        t2 = dp[cur_p - 1][cur_n] * max(0, cur_n - m)
        
        dp[cur_p][cur_n] = (t1 + t2) % mod
        
print(dp[p][n])
```

https://www.acmicpc.net/problem/8984 막대기

```python
import sys
input = lambda: sys.stdin.readline().rstrip()

n, l = map(int, input().split())

xs = set()
ys = set()
li = []
for i in range(n):
    x, y = map(int, input().split())
    
    xs.add(x)
    ys.add(y)
    li.append((x, y))
    
xs = sorted(xs)
ys = sorted(ys)

x_to_i = {}
y_to_i = {}

x_len = len(xs)
y_len = len(ys)
for i in range(x_len):
    x_to_i[xs[i]] = i
for i in range(y_len):
    y_to_i[ys[i]] = i

xi_to_yi = [[] for _ in range(x_len)]
for x, y in li:
    xi = x_to_i[x]
    yi = y_to_i[y]
    
    xi_to_yi[xi].append(yi)
    
for xi in range(x_len):
    xi_to_yi[xi].sort()

dp_x = [0] * x_len
dp_y = [0] * y_len
for xi in range(x_len):
    max_temps = []
    max_temp = 0
    for yi in xi_to_yi[xi]:
        dist = abs(xs[xi] - ys[yi]) + l
        temp = dp_y[yi] + dist
        max_temp = max(max_temp, temp)
        max_temps.append(max_temp)
    
    dp_x[xi] = max_temp
    yi = xi_to_yi[xi][0]
    dist = abs(xs[xi] - ys[yi]) + l
    dp_y[yi] = max(dp_y[yi], dist)
    
    for i in range(1, len(xi_to_yi[xi])):
        yi = xi_to_yi[xi][i]
        dist = abs(xs[xi] - ys[yi]) + l
        dp_y[yi] = max(dp_y[yi], max_temps[i - 1] + dist)
        
max_r = max(max(dp_x), max(dp_y))

print(max_r)
```

https://www.acmicpc.net/problem/5626 제단

```python
mod = 1000000007

n = int(input())
li = list(map(int, input().split()))

if n == 1:
    if li[0] in [-1, 0]:
        print(1)
    else:
        print(0)
else:
    max_h = n // 2
    dp = [[0] * (max_h + 1) for _ in range(n)]
    dp[0][0] = 1

    for i in range(1, n):
        if li[i] == -1:
            for cur_h in range(max_h + 1):
                for f_h in [cur_h - 1, cur_h, cur_h + 1]:
                    if 0 <= f_h <= max_h:
                        dp[i][cur_h] += dp[i - 1][f_h]
                dp[i][cur_h] %= mod
        else:
            cur_h = li[i]
            if cur_h > max_h:
                break
            for f_h in [cur_h - 1, cur_h, cur_h + 1]:
                if 0 <= f_h <= max_h:
                    dp[i][cur_h] += dp[i - 1][f_h]
            dp[i][cur_h] %= mod

    print(dp[n - 1][0])
```

https://www.acmicpc.net/problem/3056 007

```python
n = int(input())
li = [list(map(int, input().split())) for _ in range(n)]

dp = [0] * (1 << n)

cur_j = 0
d = dict()
for cur_w in range(n):
    d[1 << cur_w] = li[cur_j][cur_w]

for cur_j in range(1, n):
    nd = dict()
    
    for key in d:
        cur_w_bit = 1
        for cur_w in range(n):
            if key & cur_w_bit == 0:
                nkey = key | cur_w_bit
                temp = d[key] * li[cur_j][cur_w]
                if nkey not in nd or temp > nd[nkey]:
                    nd[nkey] = temp
            cur_w_bit *= 2
                
    d = nd.copy()
    
result = d[(1 << n) - 1]
result /= 100 ** (n - 1)

print(result)
```

https://www.acmicpc.net/problem/17528 Two Machines

```python
# 시간 초과
inf = float('inf')

n = int(input())
li = [list(map(int, input().split())) for _ in range(n)]

max_a = 62500

dp = [[inf] * (max_a + 1) for _ in range(n)]

a, b = li[0]
dp[0][a] = 0
dp[0][0] = b

for i in range(1, n):
    a_added, b_added = li[i]
    
    for fav in range(max_a + 1):
        fbv = dp[i - 1][fav]
        
        av = fav + a_added
        if av <= max_a:
            dp[i][av] = min(dp[i][av], fbv)
            
        bv = fbv + b_added
        dp[i][fav] = min(dp[i][fav], bv)
    
min_r = inf
for a, b in enumerate(dp[n - 1]):
    min_r = min(min_r, max(a, b))

print(min_r)
```

```python
# 시간 초과
inf = float('inf')

def reduce(i, av):
    return i * 70000 + av

n = int(input())
li = [list(map(int, input().split())) for _ in range(n)]

dp = [inf] * (250 * 70000 + 1)
max_a = 250 * 250

a, b = li[0]
dp[reduce(0, a)] = 0
dp[reduce(0, 0)] = b

for i in range(1, n):
    a_added, b_added = li[i]
    
    for fav in range(max_a + 1):
        fbv = dp[reduce(i - 1, fav)]
        
        av = fav + a_added
        if av <= max_a:
            dp[reduce(i, av)] = min(dp[reduce(i, av)], fbv)
            
        bv = fbv + b_added
        dp[reduce(i, fav)] = min(dp[reduce(i, fav)], bv)
    
min_r = inf
for a in range(max_a + 1):
    b = dp[reduce(n - 1, a)]
    min_r = min(min_r, max(a, b))

print(min_r)
```

https://www.acmicpc.net/problem/10982 행운쿠키 제작소

```python
#include <cstdio>

int INF = 99999999;

int t;
int n;
int arr[1002][2] = {0,};
int max_a = 100000;
int dp[100001] = {0,};
int ndp[100001] = {0,};
int a, b, a_added, b_added, fav, fbv, av, bv, max_v, min_r;

int main()
{
    scanf("%d", &t);
    
    for (int tn=0;tn<t;tn++){
        scanf("%d", &n);
        
        for (int i=0;i<n;i++){
            scanf("%d", &arr[i][0]);
            scanf("%d", &arr[i][1]);
        }
        
        for (int a=0;a<max_a + 1;a++){
            dp[a] = INF;
            ndp[a] = INF;
        }
        
        a = arr[0][0];
        b = arr[0][1];
        
        dp[a] = 0;
        dp[0] = b;
        for (int i=1;i<n;i++){
            a_added = arr[i][0];
            b_added = arr[i][1];
            
            for (int fav=0;fav<max_a + 1;fav++){
                fbv = dp[fav];
                
                av = fav + a_added;
                if (av <= max_a){
                    if (fbv < ndp[av]){
                        ndp[av] = fbv;
                    }
                }
                
                bv = fbv + b_added;
                if (bv < ndp[fav]){
                    ndp[fav] = bv;
                }
            }
            
            for (int av=0;av<max_a + 1;av++){
                dp[av] = ndp[av];
                ndp[av] = INF;
            }
        }
        
        min_r = INF;
        for (int a=0;a<max_a + 1;a++){
            b = dp[a];
            
            max_v = a;
            if (b > max_v){
                max_v = b;
            }
            
            if (max_v < min_r){
                min_r = max_v;
            }
        }
        
        printf("%d\n", min_r);
    }
}
```

https://www.acmicpc.net/problem/7812 중앙 트리

```python
import io, os, sys
sys.setrecursionlimit(11000)
input = io.BytesIO(os.read(0,os.fstat(0).st_size)).readline

def dfs(value, cumul_w):
    vt[value] = True
    
    costs[value] = cumul_w
    
    sub_c = 1
    for v, w in li[value]:
        if not vt[v]:  
            sub_c += dfs(v, cumul_w + w)
        
    sub_size[value] = sub_c
    
    return sub_c
    
def reroot(value, former_value, former_w):
    vt[value] = True
    
    if value != 0:
        cur_sub_c = sub_size[value]
        from_top = n - cur_sub_c

        diff = from_top - cur_sub_c
        dp[value] = dp[former_value] + diff * former_w
    
    for v, w in li[value]:
        if not vt[v]:
            reroot(v, value, w)

while True:
    n = int(input())
    if n == 0:
        break
    
    li = [[] for _ in range(n)]
    for _ in range(n - 1):
        a, b, w = map(int, input().split())
        li[a].append((b, w))
        li[b].append((a, w))

    sub_size = [0] * n
    costs = [0] * n
    vt = [False] * n
    dfs(0, 0)

    dp = [0] * n
    dp[0] = sum(costs)

    vt = [False] * n
    reroot(0, -1, 0)

    result = min(dp)

    print(result)
```

https://www.acmicpc.net/problem/20188 등산 마니아

```python
import io, os, sys
input = io.BytesIO(os.read(0,os.fstat(0).st_size)).readline

## PYRIVAL BOOTSTRAP
# https://github.com/cheran-senthil/PyRival/blob/master/pyrival/misc/bootstrap.py
# This decorator allows for recursion without actually doing recursion
## @bootstrap, yield when getting and returning value in recursive functions, end of functions
from types import GeneratorType

def bootstrap(f, stack=[]):
    def wrappedfunc(*args, **kwargs):
        if stack:
            return f(*args, **kwargs)
        else:
            to = f(*args, **kwargs)
            while True:
                if type(to) is GeneratorType:
                    stack.append(to)
                    to = next(to)
                else:
                    stack.pop()
                    if not stack:
                        break
                    to = stack[-1].send(to)
            return to
    return wrappedfunc

@bootstrap
def dfs(value, cumul):
    vt[value] = True
    
    depths[value] = cumul
    
    sub_c = 0
    dubbed_c = 0
    for v in li[value]:
        if not vt[v]:
            temp_sub_c = yield dfs(v, cumul + 1)
            
            dubbed_c += sub_c * temp_sub_c
            sub_c += temp_sub_c
    
    dubbed_c += sub_c
    dubbeds[value] = dubbed_c * cumul        
    
    sub_c += 1
    sub_sizes[value] = sub_c
    
    yield sub_c
    
n = int(input())

li = [[] for _ in range(n + 1)]
for _ in range(n - 1):
    a, b = map(int, input().split())
    li[a].append(b)
    li[b].append(a)
    
sub_sizes = [0] * (n + 1)
depths = [0] * (n + 1)
dubbeds = [0] * (n + 1)
vt = [False] * (n + 1)
dfs(1, 0)

total = sum(depths) * (n - 1)
dubbed = sum(dubbeds)

result = total - dubbed

print(result)
```
