### 무지의 먹방 라이브

```python
def solution(food_times, k):
    food_times = [[v, i] for i, v in enumerate(food_times)]
    food_times.sort()
    
    accul = 0
    indices_to_delete = set()
    for i in range(len(food_times)):
        food_times[i][0] -= accul
        if food_times[i][0] == 0:
            indices_to_delete.add(food_times[i][1])
            continue
        remained = k - food_times[i][0] * (len(food_times) - len(indices_to_delete))
        if remained >= 0:
            indices_to_delete.add(food_times[i][1])
            accul += food_times[i][0]
            k = remained 
        else:
            break
    
    left = [i for i in range(len(food_times)) if i not in indices_to_delete]
    if not left:
        return -1
    return left[k % len(left)] + 1
```

### 자물쇠와 열쇠

```python
from copy import deepcopy

def rotate(key, m):
    new_key = deepcopy(key)
    for i in range(m):
        for j in range(m):
            new_key[i][j] = key[j][m - 1 - i]      
    return new_key

def get_zero_type_count(new_lock, x, y, m):
    zero_type_count = 0
    for i in range(x, x + m):
        for j in range(y, y + m):
            if new_lock[i][j] == 0:
                zero_type_count += 1
    return zero_type_count

def compare(new_lock, key, x, y, m):
    for i in range(x, x + m):
        for j in range(y, y + m):
            if new_lock[i][j] == -1:
                continue
            if new_lock[i][j] != 1 ^ key[i - x][j - y]:
                return False
    return True
                
def solution(key, lock):
    n, m = len(lock), len(key)
    new_lock_len = n + m * 2 - 2
    new_lock = [[-1] * new_lock_len for _ in range(new_lock_len)]
    # -1: 빈 공간, 0: 홈, 1: 돌기
    for i in range(m - 1, m + n - 1):
        for j in range(m - 1, m + n - 1):
            new_lock[i][j] = lock[i - (m - 1)][j - (m - 1)]
            
    # 채워야하는 홈의 개수
    zero_type_count = 0
    for i in range(n):
        for j in range(n):
            if lock[i][j] == 0:
                zero_type_count += 1
    
    for turn_count in range(4):
        for i in range(new_lock_len - m + 1):
            for j in range(new_lock_len - m + 1):
                if get_zero_type_count(new_lock, i, j, m) < zero_type_count:
                    continue
                if compare(new_lock, key, i, j, m):
                    return True
        if turn_count <= 2:
            key = rotate(key, m)
    return False
```

### 기둥과 보 설치

```python
def solution(n, build_frame):
    # 구조물의 좌표는 기둥은 아래쪽, 보는 왼쪽을 기준으로 저장
    columns = set()
    rows = set()
    for x, y, t, c in build_frame:
        # 기둥을 설치하는 경우ㅁ
        if t == 0 and c == 1:
            # 바닥이 아닌 경우
            if y > 0:
                # 밑에 다른 기둥이나 보가 없는 경우
                if (x, y - 1) not in columns and (x - 1, y) not in rows and (x, y) not in rows:
                    continue
            columns.add((x, y))
        # 기둥을 삭제하는 경우
        elif t == 0 and c == 0:
            # 위에 다른 기둥이 있는 경우
            if (x, y + 1) in columns:
                # 위 기둥을 받칠 보가 없는 경우
                if not ((x - 1, y + 1) in rows or (x, y + 1) in rows):
                    continue
            # 기둥 위 왼쪽으로 보가 있는 경우
            if (x - 1, y + 1) in rows:
                # 기둥 왼쪽에 다른 기둥이 없는 경우
                if (x - 1, y) not in columns:
                    # 해당 보가 좌우로 연결되어있지 않은 경우
                    if (x - 2, y + 1) not in rows or (x, y + 1) not in rows:
                        continue
            # 기둥 위 오른쪽으로 보가 있는 경우
            if (x, y + 1) in rows:
                # 기둥 오른쪽에 다른 기둥이 없는 경우
                if (x + 1, y) not in columns:
                    # 해당 보가 좌우로 연결되어있지 않은 경우
                    if (x - 1, y + 1) not in rows or (x + 1, y + 1) not in rows:
                        continue
            columns.remove((x, y))
        # 보를 설치하는 경우
        elif t == 1 and c == 1:
            # 밑에 기둥이 없는 경우
            if not ((x, y - 1) in columns or (x + 1, y - 1) in columns):
                # 양 옆으로 보가 연결되지 않은 경우
                if (x - 1, y) not in rows or (x + 1, y) not in rows:
                    continue
            rows.add((x, y))
        # 보를 삭제하는 경우
        else:
            # 왼쪽 기둥이 없는 경우
            if (x, y - 1) not in columns:
                # 왼쪽으로 이어진 보가 있는 경우
                if (x - 1, y) in rows:
                    # 왼쪽 보를 지탱하는 기둥이 없는 경우
                    if (x - 1, y - 1) not in columns:
                        continue
                # 왼쪽으로 이어진 보가 없는 경우
                else:
                    # 보의 왼쪽 위에 기둥이 있는 경우
                    if (x, y) in columns:
                        continue
            # 오른족 기둥이 없는 경우
            if (x + 1, y - 1) not in columns:
                # 오른쪽으로 이어진 보가 있는 경우
                if (x + 1, y) in rows:
                    # 오른쪽 보를 지탱하는 기둥이 없는 경우
                    if (x + 2, y - 1) not in columns:
                        continue
                # 오른쪽으로 이어진 보가 없는 경우
                else:
                    # 보의 오른쪽 위에 기둥이 있는 경우
                    if (x + 1, y) in columns:
                        continue                     
            rows.remove((x, y))
            
    rows = [[v[0], v[1], 1] for v in rows]
    columns = [[v[0], v[1], 0] for v in columns]
    total = []
    total.extend(rows)
    total.extend(columns)
    total.sort()
    return total
```

### 치킨 배달

```python
from itertools import combinations as comb

def get_distance(a, b):
    return abs(a[0] - b[0]) + abs(a[1] - b[1])

n, m = map(int, input().split())
li = [list(map(int, input().split())) for _ in range(n)]

chickens = []
houses = []
for i in range(n):
    for j in range(n):
        if li[i][j] == 1:
            houses.append((i, j))
        elif li[i][j] == 2:
            chickens.append((i, j))
chicken_cases = comb(chickens, m)

min_total = float('inf')
for chicken_case in chicken_cases:
    total = 0
    for house in houses:
        min_value = float('inf')
        for chicken in chicken_case:
            min_value = min(min_value, get_distance(chicken, house))
        total += min_value
    min_total = min(min_total, total)
    
print(min_total)
```

### 외벽 점검

```python
# 시간 초과
from collections import deque

def bfs(weaks, dists, n):
    queue = deque()
    queue.append((0, weaks, dists))
    while queue:
        index, weaks, dists = queue.popleft()
        if index == len(dists):
            continue
        for start in range(n):
            new_weaks = weaks.copy()
            for i in range(start, start + dists[index] + 1):
                if i % n in new_weaks:
                    new_weaks.remove(i % n)
                if not new_weaks:
                    return index + 1
            queue.append((index + 1, new_weaks, dists))
    return -1
        
def solution(n, weaks, dists):
    dists.sort(reverse=True)
    return bfs(weaks, dists, n)
```

```python
# 시간 초과, 개선되었음
from collections import deque

def bfs(weaks, dists, n):
    queue = deque()
    queue.append((0, weaks))
    while queue:
        dist_index, weaks = queue.popleft()
        if dist_index == len(dists):
            continue
        for start in weaks:
            end = start + dists[dist_index]
            if end < n:
                new_weaks = [v for v in weaks if v < start or v > end]
            else:
                new_weaks = [v for v in weaks if end % n < v < start]
            if not new_weaks:
                return dist_index + 1
            queue.append((dist_index + 1, new_weaks))
    return -1
        
def solution(n, weaks, dists):
    dists.sort(reverse=True)
    return bfs(weaks, dists, n)
```

```python
# 맞았음, 가장 큰 거리로도 하나밖에 커버 안되는 케이스는 조기 종료하는 로직을 추가하였음
from collections import deque

def bfs(weaks, dists, n):
    min_result = float('inf')
    
    queue = deque()
    queue.append((0, weaks))
    while queue:
        dist_index, weaks = queue.popleft()
        is_appended = False
        for start in weaks:
            end = start + dists[dist_index]
            if end < n:
                new_weaks = [v for v in weaks if v < start or v > end]
            else:
                new_weaks = [v for v in weaks if end % n < v < start]
            if not new_weaks:
                min_result = dist_index + 1
                return min_result
            if len(new_weaks) == len(weaks) - 1:
                continue
            if dist_index + 1 < len(dists):
                is_appended = True
                queue.append((dist_index + 1, new_weaks))
        if not is_appended:
            if len(weaks) <= len(dists) - dist_index:
                min_result = min(min_result, dist_index + len(weaks))
    return min_result
        
def solution(n, weaks, dists):
    dists.sort(reverse=True)
    min_result = bfs(weaks, dists, n)
    if min_result == float('inf'):
        return -1
    else:
        return min_result
```

### 특정 거리의 도시 찾기

```python
# 기존 코드
import heapq
import sys
input = sys.stdin.readline

n, m, k, x = map(int, input().split())
li = [[] for _ in range(n + 1)]
for _ in range(m):
    s, e = map(int, input().split())
    li[s].append(e)
    
def dij():
    distances = [float('inf')] * (n + 1)
    distances[x] = 0
    
    heap = []
    heapq.heappush(heap, (x, 0))
    
    while heap:
        value, distance = heapq.heappop(heap)
        
        for v in li[value]:
            d = distance + 1
            
            if d < distances[v]:
                distances[v] = d
                heapq.heappush(heap, (v, d))
                
    return distances

distances = dij()
distances = [i for i, v in enumerate(distances) if v == k]
distances.sort()

if distances:
    for d in distances:
        print(d)
else:
    print(-1)
```

```python
# 기존 코드
from collections import deque
from itertools import combinations
import copy


n, m = map(int, input().split())
o_coords = [] # original coords
for _ in range(n):
    o_coords.append(list(map(int, input().split())))

safes = []
viruses = []


for i in range(n):
    for j in range(m):
        if o_coords[i][j] == 0:
            safes.append((i, j))
        elif o_coords[i][j] == 2:
            viruses.append((i, j))
            
cases = list(combinations(safes, 3))


def bfs(coords):
    queue = deque()
    queue.extend(viruses)
    
    while queue:
        x, y = queue.popleft()
        
        for a, b in ((x-1, y), (x+1, y), (x, y-1), (x, y+1)):
            if 0 <= a < n and 0 <= b < m:
                if coords[a][b] == 0:
                    coords[a][b] = 2
                    queue.append((a, b))
        
    count = 0
    for i in range(n):
        for j in range(m):
            if coords[i][j] == 0:
                count += 1
    return count


max_safe_count = 0

for case in cases:
    coords = copy.deepcopy(o_coords)
    for wall in case:
        x, y = wall
        coords[x][y] = 1
    safe_count = bfs(coords)
    if safe_count > max_safe_count:
        max_safe_count = safe_count
        
print(max_safe_count)
```

### 경쟁적 전염

```python
# 기존 코드
from collections import deque
import sys
input = sys.stdin.readline

n, k = map(int, input().split())

li = []
for _ in range(n):
    li.append(list(map(int, input().split())))
    
s, target_x, target_y = map(int, input().split())
target_x -= 1
target_y -= 1

starts = []
for i in range(n):
    for j in range(n):
        if li[i][j] > 0:
            starts.append((i, j, 0, li[i][j]))
            
starts.sort(key=lambda x: x[-1])
starts = [v[:3] for v in starts]
queue = deque()
queue.extend(starts)

dx = [0, 0, -1, 1]
dy = [-1, 1, 0, 0]

while queue:
    x, y, second = queue.popleft()
    
    if second == s:
        continue
    
    for i in range(4):
        nx = x + dx[i]
        ny = y + dy[i]
        
        if 0 <= nx < n and 0 <= ny < n:
            if li[nx][ny] == 0:
                li[nx][ny] = li[x][y]
                queue.append((nx, ny, second + 1))
            
print(li[target_x][target_y])
```

```python
# 맞았음, python으로는 시간 초과
from collections import deque
from copy import deepcopy
count = 0

n, k = map(int, input().split())
li = [list(map(int, input().split())) for _ in range(n)]
s, target_x, target_y = map(int, input().split())
target_x -= 1
target_y -= 1

dx = [0, 0, -1, 1]
dy = [-1, 1, 0, 0]

viruses = [[] for _ in range(k + 1)]
for i in range(n):
    for j in range(n):
        if li[i][j] >= 1:
            viruses[li[i][j]].append((i, j))

for _ in range(s):
    new_viruses = [[] for _ in range(k + 1)]
    for virus_num in range(1, k + 1):
        queue = deque()
        queue.extend(viruses[virus_num])
        while queue:
            x, y = queue.popleft()
            
            for i in range(4):
                count += 1
                nx = x + dx[i]
                ny = y + dy[i]
                if nx < 0 or nx >= n or ny < 0 or ny >= n:
                    continue
                if li[nx][ny] == 0:
                    li[nx][ny] = virus_num
                    new_viruses[virus_num].append((nx, ny))
    viruses = deepcopy(new_viruses)
            
print(li[target_x][target_y])
print('c', count)
```

```python
from collections import deque
count = 0

n, k = map(int, input().split())
li = [list(map(int, input().split())) for _ in range(n)]
s, target_x, target_y = map(int, input().split())
target_x -= 1
target_y -= 1

dx = [0, 0, -1, 1]
dy = [-1, 1, 0, 0]

viruses = []
for i in range(n):
    for j in range(n):
        if li[i][j] >= 1:
            viruses.append((li[i][j], i, j, 0))
viruses.sort()

if s >= 1:
    queue = deque()
    queue.extend(viruses)
    while queue:
        virus_num, x, y, second = queue.popleft()

        for i in range(4):
            count += 1
            nx = x + dx[i]
            ny = y + dy[i]
            if nx < 0 or nx >= n or ny < 0 or ny >= n:
                continue

            if li[nx][ny] == 0:
                li[nx][ny] = virus_num
                if second + 1 < s:
                    queue.append((virus_num, nx, ny, second + 1))
            
print(li[target_x][target_y])
print('c', count)
```
