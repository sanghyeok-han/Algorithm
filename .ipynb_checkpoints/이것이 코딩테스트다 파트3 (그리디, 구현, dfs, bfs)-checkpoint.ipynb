### 곱하기 혹은 더하기

```python
li = list(map(int, input()))
result = li[0]
for i in range(1, len(li)):
    if result <= 1 or li[i] <= 1:
        result += li[i]
    else:
        result *= li[i]
        
print(result)
```

### 볼링공 고르기

```python
n, m = map(int, input().split())
li = list(map(int, input().split()))
stack = sorted(li, reverse=True)

current_value = 0
current_count = 0
result = 0
while stack:
    value = stack.pop()
    if value == current_value:
        current_count += 1
    else:
        result += current_count * (len(stack) + 1)
        current_value = value
        current_count = 1
        
print(result)
```

### 무지의 먹방 라이브

```python
def solution(food_times, k):
    food_times = [[v, i] for i, v in enumerate(food_times)]
    food_times.sort()
    
    accul = 0
    indices_to_delete = set()
    for i in range(len(food_times)):
        food_times[i][0] -= accul
        if food_times[i][0] == 0:
            indices_to_delete.add(food_times[i][1])
            continue
        remained = k - food_times[i][0] * (len(food_times) - len(indices_to_delete))
        if remained >= 0:
            indices_to_delete.add(food_times[i][1])
            accul += food_times[i][0]
            k = remained 
        else:
            break
    
    left = [i for i in range(len(food_times)) if i not in indices_to_delete]
    if not left:
        return -1
    return left[k % len(left)] + 1
```

```python
# 기존 코드
def solution(s):
    from collections import deque

    if len(s) == 1:
        return 1

    min_result = float('inf')
    for compress in range(1, len(s) // 2 + 1):
        print(compress)
        temp_result = []
        to_compare = []
        temp = []
        same_count = 1
        for i in range(0, len(s), compress):
            if i + compress < len(s) + 1:
                if not to_compare:
                    to_compare = s[i:i + compress]
                else:
                    temp = s[i:i + compress]
                    if to_compare == temp:
                        same_count += 1
                    else:
                        if same_count >= 2:
                            temp_result.append(same_count)
                            same_count = 1

                        temp_result.extend(to_compare)
                        to_compare = temp[:]
            else:
                break

        if same_count >= 2:
            temp_result.append(same_count)
        temp_result.extend(to_compare)

        if len(s[i + compress:]) >= 1:
            temp_result.extend(s[i:i + compress])
        elif len(s[i:i + compress]) < compress:
            temp_result.extend(s[i:i + compress])

        temp_result = "".join(map(str, temp_result))
        if len(temp_result) < min_result:
            min_result = len(temp_result)

    return min_result
```

```python
def add_to_temp_result(count, temp_result, stack, current):
    if stack:
        if count >= 2:
            temp_result += str(count) + stack[-1]
            count = 1
        else:
            temp_result += stack[-1]
            stack.pop()
    stack.append(current)
    return count, temp_result    

def solution(s): 
    min_length = float('inf')
    for unit in range(1, len(s) + 1):
        stack = []
        temp_result = ''
        count = 1
        for i in range(0, len(s), unit):
            current = s[i:i + unit]
            if stack and stack[-1] == current:
                count += 1
                continue
            count, temp_result = add_to_temp_result(count, temp_result, stack, current)
        count, temp_result = add_to_temp_result(count, temp_result, stack, current)
        rest = s[i + unit:]
        if rest:
            temp_result += rest
        min_length = min(min_length, len(temp_result))
    return min_length
```

### 자물쇠와 열쇠

```python
from copy import deepcopy

def rotate(key, m):
    new_key = deepcopy(key)
    for i in range(m):
        for j in range(m):
            new_key[i][j] = key[j][m - 1 - i]      
    return new_key

def get_zero_type_count(new_lock, x, y, m):
    zero_type_count = 0
    for i in range(x, x + m):
        for j in range(y, y + m):
            if new_lock[i][j] == 0:
                zero_type_count += 1
    return zero_type_count

def compare(new_lock, key, x, y, m):
    for i in range(x, x + m):
        for j in range(y, y + m):
            if new_lock[i][j] == -1:
                continue
            if new_lock[i][j] != 1 ^ key[i - x][j - y]:
                return False
    return True
                
def solution(key, lock):
    n, m = len(lock), len(key)
    new_lock_len = n + m * 2 - 2
    new_lock = [[-1] * new_lock_len for _ in range(new_lock_len)]
    # -1: 빈 공간, 0: 홈, 1: 돌기
    for i in range(m - 1, m + n - 1):
        for j in range(m - 1, m + n - 1):
            new_lock[i][j] = lock[i - (m - 1)][j - (m - 1)]
            
    # 채워야하는 홈의 개수
    zero_type_count = 0
    for i in range(n):
        for j in range(n):
            if lock[i][j] == 0:
                zero_type_count += 1
    
    for turn_count in range(4):
        for i in range(new_lock_len - m + 1):
            for j in range(new_lock_len - m + 1):
                if get_zero_type_count(new_lock, i, j, m) < zero_type_count:
                    continue
                if compare(new_lock, key, i, j, m):
                    return True
        if turn_count <= 2:
            key = rotate(key, m)
    return False
```

```python
# 기존 코드
from collections import deque

n = int(input())
li = [[0] * n for _ in range(n)] # 0: 빈칸, 1: 뱀이 차지하는 칸, 2: 사과 

k = int(input())
for _ in range(k):
    x, y = map(int, input().split())
    x, y = x - 1, y - 1
    li[x][y] = 2
    
l = int(input())
snake_infos = deque()
for _ in range(l):
    sec, turn = input().split()
    sec = int(sec)
    snake_infos.append((sec, turn))

li[0][0] = 1
snake_body = deque()
snake_body.append((0, 0))

# 0: 상, 1: 하, 2: 좌, 3: 우
dx = [-1, 1, 0, 0]
dy = [0, 0, -1, 1]

def change_dir(current_dir, turn):
    if current_dir == 0:
        if turn == 'L':
            current_dir = 2
        else:
            current_dir = 3
    elif current_dir == 1:
        if turn == 'L':
            current_dir = 3
        else:
            current_dir = 2                
    elif current_dir == 2:
        if turn == 'L':
            current_dir = 1
        else:
            current_dir = 0            
    else:
        if turn == 'L':
            current_dir = 0
        else:
            current_dir = 1
    return current_dir

current_snake_info = snake_infos.popleft()

current_dir = 3
current_sec = 1
x, y = 0, 0
while True:
    x += dx[current_dir]
    y += dy[current_dir]
    
    if x < 0 or x >= n or y < 0 or y >= n:
        break
    if li[x][y] == 1:
        break
        
    if li[x][y] == 2:
        snake_body.append((x, y))
        li[x][y] = 1
    else:
        snake_body.append((x, y))
        li[x][y] = 1
        tail_x, tail_y = snake_body.popleft()
        li[tail_x][tail_y] = 0
        
    if current_snake_info:
        if current_sec == current_snake_info[0]:
            current_dir = change_dir(current_dir, current_snake_info[1])
                    
            if snake_infos:
                current_snake_info = snake_infos.popleft()
            else:
                current_snake_info = ""
            
    current_sec += 1
            
print(current_sec)
```

```python
from collections import deque

# 상, 우, 하, 좌
dx = [-1, 0, 1, 0]
dy = [0, 1, 0, -1]

snakes = deque()
snakes.append((0, 0))

n = int(input())
k = int(input())
apples = set()
for _ in range(k):
    apples.add(tuple(map(lambda x: int(x) - 1, input().split())))
l = int(input())
turns = deque()
for _ in range(l):
    s, d = input().split()
    s = int(s)
    d = 1 if d == 'D' else -1
    turns.append((s, d))
    
move_count = 0
current_d = 1
current_second = 0
x, y = 0, 0
while True:
    print(snakes)
    nx, ny = x + dx[current_d % 4], y + dy[current_d % 4]
    current_second += 1
    
    if nx < 0 or nx >= n or ny < 0 or ny >= n:
        break
    if (nx, ny) in snakes:
        break
    snakes.append((nx, ny))
    if (nx, ny) not in apples:
        snakes.popleft()
    else:
        apples.remove((nx, ny))
    if turns and turns[0][0] == current_second:
        current_d += turns[0][1]
        turns.popleft()
    
    x, y = nx, ny
    
print(current_second)
```

### 기둥과 보 설치

```python
def solution(n, build_frame):
    # 구조물의 좌표는 기둥은 아래쪽, 보는 왼쪽을 기준으로 저장
    columns = set()
    rows = set()
    for x, y, t, c in build_frame:
        # 기둥을 설치하는 경우ㅁ
        if t == 0 and c == 1:
            # 바닥이 아닌 경우
            if y > 0:
                # 밑에 다른 기둥이나 보가 없는 경우
                if (x, y - 1) not in columns and (x - 1, y) not in rows and (x, y) not in rows:
                    continue
            columns.add((x, y))
        # 기둥을 삭제하는 경우
        elif t == 0 and c == 0:
            # 위에 다른 기둥이 있는 경우
            if (x, y + 1) in columns:
                # 위 기둥을 받칠 보가 없는 경우
                if not ((x - 1, y + 1) in rows or (x, y + 1) in rows):
                    continue
            # 기둥 위 왼쪽으로 보가 있는 경우
            if (x - 1, y + 1) in rows:
                # 기둥 왼쪽에 다른 기둥이 없는 경우
                if (x - 1, y) not in columns:
                    # 해당 보가 좌우로 연결되어있지 않은 경우
                    if (x - 2, y + 1) not in rows or (x, y + 1) not in rows:
                        continue
            # 기둥 위 오른쪽으로 보가 있는 경우
            if (x, y + 1) in rows:
                # 기둥 오른쪽에 다른 기둥이 없는 경우
                if (x + 1, y) not in columns:
                    # 해당 보가 좌우로 연결되어있지 않은 경우
                    if (x - 1, y + 1) not in rows or (x + 1, y + 1) not in rows:
                        continue
            columns.remove((x, y))
        # 보를 설치하는 경우
        elif t == 1 and c == 1:
            # 밑에 기둥이 없는 경우
            if not ((x, y - 1) in columns or (x + 1, y - 1) in columns):
                # 양 옆으로 보가 연결되지 않은 경우
                if (x - 1, y) not in rows or (x + 1, y) not in rows:
                    continue
            rows.add((x, y))
        # 보를 삭제하는 경우
        else:
            # 왼쪽 기둥이 없는 경우
            if (x, y - 1) not in columns:
                # 왼쪽으로 이어진 보가 있는 경우
                if (x - 1, y) in rows:
                    # 왼쪽 보를 지탱하는 기둥이 없는 경우
                    if (x - 1, y - 1) not in columns:
                        continue
                # 왼쪽으로 이어진 보가 없는 경우
                else:
                    # 보의 왼쪽 위에 기둥이 있는 경우
                    if (x, y) in columns:
                        continue
            # 오른족 기둥이 없는 경우
            if (x + 1, y - 1) not in columns:
                # 오른쪽으로 이어진 보가 있는 경우
                if (x + 1, y) in rows:
                    # 오른쪽 보를 지탱하는 기둥이 없는 경우
                    if (x + 2, y - 1) not in columns:
                        continue
                # 오른쪽으로 이어진 보가 없는 경우
                else:
                    # 보의 오른쪽 위에 기둥이 있는 경우
                    if (x + 1, y) in columns:
                        continue                     
            rows.remove((x, y))
            
    rows = [[v[0], v[1], 1] for v in rows]
    columns = [[v[0], v[1], 0] for v in columns]
    total = []
    total.extend(rows)
    total.extend(columns)
    total.sort()
    return total
```

```python
def solution(n, build_frame):
    
    pillar = set()
    bo = set()
    
    for i in build_frame:
        if i[0] < 0 or i[0] > n or i[1] < 0 or i[1] > n:
            continue
            
        
        
        if i[3] == 1 and i[2] == 0:  # 설치 & 기둥인 경우
            if i[1] == 0:
                pillar.add((i[0], i[1], 0))
                
            elif (i[0], i[1]-1, 0) in pillar:
                pillar.add((i[0], i[1], 0))
                
            elif (i[0]-1, i[1], 1) in bo or (i[0], i[1], 1) in bo:
                pillar.add((i[0], i[1], 0))
                
        elif i[3] == 0 and i[2] == 0:   # 삭제 & 기둥일 때,
            if (i[0], i[1]+1, 0) not in pillar and ((i[0], i[1]+1, 1) not in bo and (i[0]-1, i[1]+1, 1) not in bo):
                pillar.remove((i[0], i[1], 0))
                
            elif (i[0], i[1]+1, 0) in pillar: # 만약에 삭제 대상의 위에 기둥이 있는 경우
                if (i[0]-1, i[1]+1, 1) in bo or (i[0], i[1]+1, 1) in bo: # 양 옆에 보가 하나라도 있는 경우엔
                    pillar.remove((i[0], i[1], 0))
            
            elif (i[0]-1, i[1]+1, 1) in bo and (i[0], i[1]+1, 1) in bo: # 만약에 삭제 대상 기둥의 양 위에 보가 있을 때
                if (((i[0]-2, i[1]+1, 1) in bo or (i[0]-1, i[1], 0) in pillar) and 
                    ((i[0]+1, i[1]+1, 1) in bo or (i[0]+1, i[1], 0) in pillar)):
                    pillar.remove((i[0], i[1], 0))
                    
            elif (i[0]-1, i[1]+1, 1) in bo and (i[0], i[1]+1, 1) not in bo:
                if (i[0]-1, i[1], 0) in pillar:
                    pillar.remove((i[0], i[1], 0))
                    
            elif (i[0]-1, i[1]+1, 1) not in bo and (i[0], i[1]+1, 1) in bo:
                if (i[0]+1, i[1], 0) in pillar:
                    pillar.remove((i[0], i[1], 0))
            
            
                    
        
        elif i[3] == 1 and i[2] == 1: # 설치 & 보인 경우
            if (i[0], i[1]-1, 0) in pillar or (i[0]+1, i[1]-1, 0) in pillar: # 보의 끝에 하나라도 기둥이 깔려있는 경우
                bo.add((i[0], i[1], 1))
            elif (i[0]-1, i[1], 1) in bo and (i[0]+1, i[1], 1) in bo:
                bo.add((i[0], i[1], 1))
                
        elif i[3] == 0 and i[2] == 1: # 삭제 & 보일 때,
            if (((i[0], i[1], 0) not in pillar and (i[0]+1, i[1], 0) not in pillar) and 
                  ((i[0]-1, i[1], 1) not in bo and (i[0]+1, i[1], 1) not in bo)):
                bo.remove((i[0], i[1], 1))
                
            elif (i[0], i[1], 0) in pillar:  # 만약에 삭제 대상의 바로 위에 기둥이 있는 경우
                if (i[0]-1, i[1], 1) in bo or (i[0], i[1]-1, 0) in pillar: # 그 기둥의 왼쪽 아래에 보가 있거나 밑에 또다른 기둥.
                    bo.remove((i[0], i[1], 1))
            
            elif (i[0]+1, i[1], 0) in pillar:  # 만약에 삭제 대상의 바로 위가 아닌 반대쪽 끝에 기둥이 있는 경우
                if (i[0]+1, i[1], 1) in bo or (i[0]+1, i[1]-1, 0) in pillar:
                    bo.remove((i[0], i[1], 1))
                            
            elif (i[0]-1, i[1], 1) in bo and (i[0]+1, i[1], 1) in bo: # 제거 대상 양쪽에 보가 있는 경우
                if (((i[0]-1, i[1]-1, 0) in pillar or (i[0], i[1]-1, 0) in pillar) and 
                    ((i[0]+1, i[1]-1, 0) in pillar or (i[0]+2, i[1]-1, 0) in pillar)):
                    bo.remove((i[0], i[1], 1))
                    
            elif (i[0]-1, i[1], 1) not in bo and (i[0]+1, i[1], 1) in bo: # 제거 대상 오른쪽에만 보가 있는 경우
                if (i[0]+1, i[1]-1, 0) in pillar or (i[0]+2, i[1]-1, 0) in pillar:
                    bo.remove((i[0], i[1], 1))
                    
            elif (i[0]-1, i[1], 1) in bo and (i[0]+1, i[1], 1) not in bo: # 제거 대상 왼쪽에만 보가 있는 경우
                if (i[0]-1, i[1]-1, 0) in pillar or (i[0], i[1]-1, 0) in pillar:
                    bo.remove((i[0], i[1], 1))
                    
            
        
    answer = list(pillar | bo)
        
    answer = sorted(answer, key = lambda x : (x[0], x[1], x[2]))
    answer = [list(v) for v in answer] 
                
        
    return answer
```

### 치킨 배달

```python
from itertools import combinations as comb

def get_distance(a, b):
    return abs(a[0] - b[0]) + abs(a[1] - b[1])

n, m = map(int, input().split())
li = [list(map(int, input().split())) for _ in range(n)]

chickens = []
houses = []
for i in range(n):
    for j in range(n):
        if li[i][j] == 1:
            houses.append((i, j))
        elif li[i][j] == 2:
            chickens.append((i, j))
chicken_cases = comb(chickens, m)

min_total = float('inf')
for chicken_case in chicken_cases:
    total = 0
    for house in houses:
        min_value = float('inf')
        for chicken in chicken_case:
            min_value = min(min_value, get_distance(chicken, house))
        total += min_value
    min_total = min(min_total, total)
    
print(min_total)
```

### 외벽 점검

```python
# 시간 초과
from collections import deque

def bfs(weaks, dists, n):
    queue = deque()
    queue.append((0, weaks, dists))
    while queue:
        index, weaks, dists = queue.popleft()
        if index == len(dists):
            continue
        for start in range(n):
            new_weaks = weaks.copy()
            for i in range(start, start + dists[index] + 1):
                if i % n in new_weaks:
                    new_weaks.remove(i % n)
                if not new_weaks:
                    return index + 1
            queue.append((index + 1, new_weaks, dists))
    return -1
        
def solution(n, weaks, dists):
    dists.sort(reverse=True)
    return bfs(weaks, dists, n)
```

### 특정 거리의 도시 찾기

```python
# 기존 코드
import heapq
import sys
input = sys.stdin.readline

n, m, k, x = map(int, input().split())
li = [[] for _ in range(n + 1)]
for _ in range(m):
    s, e = map(int, input().split())
    li[s].append(e)
    
def dij():
    distances = [float('inf')] * (n + 1)
    distances[x] = 0
    
    heap = []
    heapq.heappush(heap, (x, 0))
    
    while heap:
        value, distance = heapq.heappop(heap)
        
        for v in li[value]:
            d = distance + 1
            
            if d < distances[v]:
                distances[v] = d
                heapq.heappush(heap, (v, d))
                
    return distances

distances = dij()
distances = [i for i, v in enumerate(distances) if v == k]
distances.sort()

if distances:
    for d in distances:
        print(d)
else:
    print(-1)
```

### 괄호 변환

```python
# 기존 코드
def reverse(s):
    new = []
    for v in s:
        if v == '(':
            new.append(')')
        else:
            new.append('(')
    return "".join(new)

def disjoint(s, index):
    return s[:index + 1], s[index + 1:]
    
def check_first_valanced_index(s):
    count_open = 0
    count_close = 0
    for i in range(len(s)):
        if s[i] == '(':
            count_open += 1
        else:
            count_close += 1
    
        if count_open and count_open == count_close:
            return i
    
def check_rightness(s):
    stack = [s[0]]
    for i in range(1, len(s)):
        if s[i] == '(':
            stack.append(s[i])
        else:
            if stack and stack[-1] == '(':
                stack.pop()
            else:
                return False
    if stack:
        return False
    
    return True

def recur(p): 
    if not p:
        return ''
    
    first_valanced_index = check_first_valanced_index(p)
    u, v = disjoint(p, first_valanced_index)
    
    if check_rightness(u):
        return u + recur(v)
    else:
        return '(' + recur(v) + ')' + reverse(u[1:-1])

def solution(p):
    return recur(p)
```

```python
def disjoint(p):
    counts = [0, 0] # open_count, close_count
    for i in range(len(p)):
        if p[i] == '(':
            counts[0] += 1
        else:
            counts[1] += 1
        if counts[0] >= 1 and counts[0] == counts[1]:
            return p[:i + 1], p[i + 1:]

def check(s):
    stack = []
    for v in s:
        if v == '(':
            stack.append('(')
        else:
            if stack:
                stack.pop()
            else:
                return False
    return True
        
def reverse(s):
    new_s = []
    for v in s:
        if v == '(':
            new_s.append(')')
        else:
            new_s.append('(')
    return ''.join(new_s)
    
def recur(p):
    if not p:
        return p
    
    u, v = disjoint(p)
    if check(u):
        return u + recur(v)
    else:
        return '(' + recur(v) + ')' + reverse(u[1:-1])

def solution(p):
    return recur(p)
```

### 연산자 끼워 넣기

```python
from itertools import permutations as p

n = int(input())
nums = list(map(int, input().split()))
operators = list(map(int, input().split()))
operators = '+' * operators[0] + '-' * operators[1] + '*' * operators[2] + '/' * operators[3]

cases = p(operators, n - 1)
max_value = -float('inf')
min_value = float('inf')

for case in cases:
    temp = nums[0]
    for i in range(n - 1):
        if case[i] == '+':
            temp += nums[i + 1]
        elif case[i] == '-':
            temp -= nums[i + 1]
        elif case[i] == '*':
            temp *= nums[i + 1]
        elif case[i] == '/':
            if temp < 0:
                temp = -temp
                temp //= nums[i + 1]
                temp = -temp
            else:
                temp //= nums[i + 1]
            
    max_value = max(max_value, temp)
    min_value = min(min_value, temp)
    
print(max_value)
print(min_value)
```

```python
# 효율성 개선
n = int(input())
nums = list(map(int, input().split()))
operators = list(map(int, input().split()))

max_value = -float('inf')
min_value = float('inf')

# +-*/
def dfs(index, temp, add, minus, mul, div):
    global max_value
    global min_value
    
    if index == n - 1:
        max_value = max(max_value, temp)
        min_value = min(min_value, temp)
        return
        
    if add:
        dfs(index + 1, temp + nums[index + 1], add - 1, minus, mul, div)   
    if minus:
        dfs(index + 1, temp - nums[index + 1], add, minus - 1, mul, div)      
    if mul:
        dfs(index + 1, temp * nums[index + 1], add, minus, mul - 1, div) 
    if div:
        temp = -(-temp // nums[index + 1]) if temp < 0 else temp // nums[index + 1]
        dfs(index + 1, temp, add, minus, mul, div - 1)

add, minus, mul, div = operators
dfs(0, nums[0], add, minus, mul, div)
    
print(max_value)
print(min_value)
```

### 감시 피하기

```python
from itertools import combinations as comb

n = int(input())
li = [input().split() for _ in range(n)]

teachers = []
blanks = []
for i in range(n):
    for j in range(n):
        if li[i][j] == 'X':
            blanks.append((i, j))
        elif li[i][j] == 'T':
            teachers.append((i, j))

dx = [-1, 1, 0, 0]
dy = [0, 0, -1, 1]
            
def solve():
    cases = comb(blanks, 3)
    for case in cases:
        case = set(case)
        found_students = set()
        
        for x, y in teachers:
            for i in range(4):
                nx, ny = x, y
                while True:
                    nx += dx[i]
                    ny += dy[i]
                    if nx < 0 or nx >= n or ny < 0 or ny >= n:
                        break
                    if (nx, ny) in case:
                        break
                    if li[nx][ny] == 'S':
                        found_students.add((nx, ny))
        if len(found_students) == 0:
            return 'YES'
    return 'NO'

print(solve())
```

### 인구 이동

```python
# 기존 코드
from collections import deque

n, l, r = map(int, input().split())

li = []
for _ in range(n):
    li.append(list(map(int, input().split())))

dx = [0, 0, -1, 1]
dy = [-1, 1, 0, 0]

count = 0
while True:
    is_changed = False
    visited = [[False] * n for _ in range(n)]
    for i in range(n):
        for j in range(n):
            if visited[i][j]:
                continue
                
            to_change = []
            to_change.append((i, j))
            
            cumul_popul = []
            cumul_popul.append(li[i][j])
            
            visited[i][j] = True
            queue = deque()
            queue.append((i, j))
            
            while queue:
                x, y = queue.popleft()
                
                for k in range(4):
                    nx = x + dx[k]
                    ny = y + dy[k]
                    
                    if 0 <= nx < n and 0 <= ny < n:
                        if l <= abs(li[x][y] - li[nx][ny]) <= r:
                            if not visited[nx][ny]:
                                visited[nx][ny] = True
                                to_change.append((nx, ny))
                                cumul_popul.append(li[nx][ny])
                                queue.append((nx, ny))
                            
            if len(to_change) >= 2:
                is_changed = True
                distributed_popul = sum(cumul_popul) // len(to_change)
                
                for x, y in to_change:
                    li[x][y] = distributed_popul

    if not is_changed:
        break
        
    count += 1

print(count)
```

```python
# 각 반복마다 탐색을 시작할 필요가 있는 곳만 추려서 효율성 개선함
from collections import deque
from itertools import product
from copy import deepcopy

n, l, r = map(int, input().split())
li = [list(map(int, input().split())) for _ in range(n)]

dx = [-1, 1, 0, 0]
dy = [0, 0, -1, 1]

to_search = product(range(n), range(n))
count = 0
while True:
    is_moved = False
    visited = [[False] * n for _ in range(n)]
    new_to_search = []
    for i, j in to_search:
        if visited[i][j]:
            continue
        visited[i][j] = True    
        team = [(i, j)]
        popul = li[i][j]

        queue = deque()
        queue.append((i, j))
        while queue:
            x, y = queue.popleft()
            for k in range(4):
                nx = x + dx[k]
                ny = y + dy[k]

                if nx < 0 or nx >= n or ny < 0 or ny >= n:
                    continue    
                if l <= abs(li[nx][ny] - li[x][y]) <= r:
                    if visited[nx][ny]:
                        continue
                    visited[nx][ny] = True
                    queue.append((nx, ny))
                    team.append((nx, ny))
                    popul += li[nx][ny]
        if len(team) >= 2:
            is_moved = True
            avg = popul // len(team)
            for x, y in team:
                li[x][y] = avg
            new_to_search.extend(team)
    if is_moved:
        count += 1
        to_search = deepcopy(new_to_search)
    else:
        break
        
print(count)
```

### 블록 이동하기

```python
# 기존 코드
def solution(board):
    from collections import deque

    n = len(board)
    answer = 0

    # 로봇 현재 위치 기준: 오른쪽, 밑
    queue = deque()
    queue.append((0, 1, 0, 0)) # (x, y, status, count), status의 경우 0: 가로, 1: 세로

    visited = [[[float('inf')] * n for _ in range(n)] for _ in range(2)] # visited[status][x][y]
    visited[0][0][1] = 0
    while queue:
        x, y, status, count = queue.popleft()

        if x == n - 1 and y == n - 1:
            print(visited)
            return count

        # 현재 가로인 경우
        if status == 0:
            if y - 2 >= 0:
                if board[x][y - 2] == 0:
                    if visited[status][x][y] + 1 < visited[0][x][y - 1]:
                        visited[0][x][y - 1] = visited[status][x][y] + 1
                        queue.append((x, y - 1, status, count + 1))

            if y + 1 < n:
                if board[x][y + 1] == 0:
                    if visited[status][x][y] + 1 < visited[0][x][y + 1]:
                        visited[0][x][y + 1] = visited[status][x][y] + 1
                        queue.append((x, y + 1, status, count + 1))

            if x + 1 < n:
                if board[x + 1][y - 1] == 0 and board[x + 1][y] == 0:
                    if visited[status][x][y] + 1 < visited[0][x + 1][y]:
                        visited[0][x + 1][y] = visited[status][x][y] + 1
                        queue.append((x + 1, y, status, count + 1))

                    if visited[status][x][y] + 1 < visited[1][x + 1][y - 1]:
                        visited[1][x + 1][y - 1] = visited[status][x][y] + 1
                        queue.append((x + 1, y - 1, 1, count + 1))

                    if visited[status][x][y] + 1 < visited[1][x + 1][y]:
                        visited[1][x + 1][y] = visited[status][x][y] + 1
                        queue.append((x + 1, y, 1, count + 1))                        

            if x - 1 >= 0:
                if board[x - 1][y - 1] == 0 and board[x - 1][y] == 0:
                    if visited[status][x][y] + 1 < visited[0][x - 1][y]:
                        visited[0][x - 1][y] = visited[status][x][y] + 1
                        queue.append((x - 1, y, status, count + 1))

                    if visited[status][x][y] + 1 < visited[1][x][y - 1]:
                        visited[1][x][y - 1] = visited[status][x][y] + 1
                        queue.append((x, y - 1, 1, count + 1))

                    if visited[status][x][y] + 1 < visited[1][x][y]:
                        visited[1][x][y] = visited[status][x][y] + 1
                        queue.append((x, y, 1, count + 1))

        # 현재 세로인 경우
        else:
            if x - 2 >= 0:
                if board[x - 2][y] == 0:
                    if visited[status][x][y] + 1 < visited[1][x - 1][y]:
                        visited[1][x - 1][y] = visited[status][x][y] + 1
                        queue.append((x - 1, y, status, count + 1))   

            if x + 1 < n:
                if board[x + 1][y] == 0:
                    if visited[status][x][y] + 1 < visited[1][x + 1][y]:
                        visited[1][x + 1][y] = visited[status][x][y] + 1
                        queue.append((x + 1, y, status, count + 1))

            if y - 1 >= 0:
                if board[x - 1][y - 1] == 0 and board[x][y - 1] == 0:
                    if visited[status][x][y] + 1 < visited[1][x][y - 1]:
                        visited[1][x][y - 1] = visited[status][x][y] + 1
                        queue.append((x, y - 1, status, count + 1))

                    if visited[status][x][y] + 1 < visited[0][x - 1][y]:
                        visited[0][x - 1][y] = visited[status][x][y] + 1
                        queue.append((x - 1, y, 0, count + 1))

                    if visited[status][x][y] + 1 < visited[0][x][y]:
                        visited[0][x][y] = visited[status][x][y] + 1
                        queue.append((x, y, 0, count + 1))

            if y + 1 < n:
                if board[x - 1][y + 1] == 0 and board[x][y + 1] == 0:
                    if visited[status][x][y] + 1 < visited[1][x][y + 1]:
                        visited[1][x][y + 1] = visited[status][x][y] + 1
                        queue.append((x, y + 1, status, count + 1))

                    if visited[status][x][y] + 1 < visited[0][x - 1][y + 1]:
                        visited[0][x - 1][y + 1] = visited[status][x][y] + 1
                        queue.append((x - 1, y + 1, 0, count + 1))

                    if visited[status][x][y] + 1 < visited[0][x][y + 1]:
                        visited[0][x][y + 1] = visited[status][x][y] + 1
                        queue.append((x, y + 1, 0, count + 1))
```

```python
from collections import deque

def solution(board):
    n = len(board)
    m = len(board[0])
    INF = float('inf')
    
    # 상, 하, 좌, 우
    dx = [-1, 1, 0, 0]
    dy = [0, 0, -1, 1]
    
    d = [[[INF] * m for _ in range(n)] for _ in range(2)] # d[state][x][y]
    d[0][0][0] = d[0][0][1] = 0
    
    queue = deque()
    queue.append((0, 0, 0, 1, 0, 0)) # sx, sy, ex, ey, state (1: 가로, 2: 세로), count  * 가로의 경우 왼쪽, 세로의 경우 위쪽이 sx, sy
    while queue:
        sx, sy, ex, ey, state, count = queue.popleft()
        if ex == n - 1 and ey == m - 1:
            return min(d[0][ex][ey], d[1][ex][ey])
        
        for i in range(4):
            nsx, nsy, nex, ney = sx + dx[i], sy + dy[i], ex + dx[i], ey + dy[i]
            if nsx < 0 or nsx >= n or nsy < 0 or nsy >= m or nex < 0 or nex >= m or nex < 0 or ney >= m:
                continue
            if board[nsx][nsy] or board[nex][ney]:
                continue
                
            if d[state][nsx][nsy] == INF or d[state][nex][ney] == INF:
                d[state][nsx][nsy] = min(d[state][nsx][nsy], count + 1)
                d[state][nex][ney] = min(d[state][nex][ney], count + 1)
                queue.append((nsx, nsy, nex, ney, state, count + 1))

            nstate = state ^ 1
            if (state == 0 and (i == 0 or i == 1)) or (state == 1 and (i == 2 or i == 3)):
                for c in [(sx, sy, nsx, nsy), (ex, ey, nex, ney)]:
                    if d[nstate][c[2]][c[3]] == INF or d[nstate][c[0]][c[1]] == INF:
                        d[nstate][c[2]][c[3]] = min(d[nstate][c[2]][c[3]], count + 1)
                        d[nstate][c[0]][c[1]] = min(d[nstate][c[0]][c[1]], count + 1)
                        queue.append((min(c[0], c[2]), min(c[1], c[3]), max(c[0], c[2]), max(c[1], c[3]), nstate, count + 1))
```
