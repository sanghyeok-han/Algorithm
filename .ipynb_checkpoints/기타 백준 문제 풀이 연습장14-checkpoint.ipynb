```python
from itertools import combinations as comb
from math import gcd

def recur(bit):
    if vt[bit]:
        return
    vt[bit] = True
    
    for i in range(n):
        if bit & (1 << i):
            f_bit = bit ^ (1 << i)
            added_bit = 1 << i
            
            f_num_len = lens[f_bit]
            added_num_len = lens[added_bit]
            
            recur(f_bit)
            
            for f_re in range(k):
                cur_re = f_re * mods_10[added_num_len] + nums[i]
                cur_re %= k
                
                cs[bit][cur_re] += cs[f_bit][f_re]
                
n = int(input())
li = [input() for _ in range(n)]
k = int(input())

nums = [int(v) % k for v in li]

lens = [0] * (1 << n)
for i in range(n):
    lens[1 << i] = len(li[i])

for c in range(2, n + 1):
    for case in comb(range(n), c):
        sv = 0
        for ci in range(1, c):
            sv += 1 << case[ci]
            
        lens[(1 << case[0]) + sv] = lens[1 << case[0]] + lens[sv]

mods_10 = [0] * 751

cur = 1 % k
mods_10[0] = cur
for i in range(1, 751):
    cur *= 10
    cur %= k
    mods_10[i] = cur

        
cs = [[0] * k for _ in range(1 << n)]
vt = [False] * (1 << n)
for i in range(n):
    cs[1 << i][nums[i]] = 1
    vt[1 << i] = True
    
recur((1 << n) - 1)

num = cs[(1 << n) - 1][0]
denom = sum(cs[(1 << n) - 1])

g = gcd(num, denom)

num //= g
denom //= g

print(f'{num}/{denom}')
```

```python
import io, os, sys
input = io.BytesIO(os.read(0,os.fstat(0).st_size)).readline

# https://github.com/tatyam-prime/SortedSet/blob/main/SortedSet.py
import math
from bisect import bisect_left, bisect_right
from typing import Generic, Iterable, Iterator, TypeVar, Union, List
T = TypeVar('T')

class SortedSet(Generic[T]):
    BUCKET_RATIO = 50
    REBUILD_RATIO = 170

    def _build(self, a=None) -> None:
        "Evenly divide `a` into buckets."
        if a is None: a = list(self)
        size = self.size = len(a)
        bucket_size = int(math.ceil(math.sqrt(size / self.BUCKET_RATIO)))
        self.a = [a[size * i // bucket_size : size * (i + 1) // bucket_size] for i in range(bucket_size)]
    
    def __init__(self, a: Iterable[T] = []) -> None:
        "Make a new SortedSet from iterable. / O(N) if sorted and unique / O(N log N)"
        a = list(a)
        if not all(a[i] < a[i + 1] for i in range(len(a) - 1)):
            a = sorted(set(a))
        self._build(a)

    def __iter__(self) -> Iterator[T]:
        for i in self.a:
            for j in i: yield j

    def __reversed__(self) -> Iterator[T]:
        for i in reversed(self.a):
            for j in reversed(i): yield j
    
    def __len__(self) -> int:
        return self.size
    
    def __repr__(self) -> str:
        return "SortedSet" + str(self.a)
    
    def __str__(self) -> str:
        s = str(list(self))
        return "{" + s[1 : len(s) - 1] + "}"

    def _find_bucket(self, x: T) -> List[T]:
        "Find the bucket which should contain x. self must not be empty."
        for a in self.a:
            if x <= a[-1]: return a
        return a

    def __contains__(self, x: T) -> bool:
        if self.size == 0: return False
        a = self._find_bucket(x)
        i = bisect_left(a, x)
        return i != len(a) and a[i] == x

    def add(self, x: T) -> bool:
        "Add an element and return True if added. / O(√N)"
        if self.size == 0:
            self.a = [[x]]
            self.size = 1
            return True
        a = self._find_bucket(x)
        i = bisect_left(a, x)
        if i != len(a) and a[i] == x: return False
        a.insert(i, x)
        self.size += 1
        if len(a) > len(self.a) * self.REBUILD_RATIO:
            self._build()
        return True

    def discard(self, x: T) -> bool:
        "Remove an element and return True if removed. / O(√N)"
        if self.size == 0: return False
        a = self._find_bucket(x)
        i = bisect_left(a, x)
        if i == len(a) or a[i] != x: return False
        a.pop(i)
        self.size -= 1
        if len(a) == 0: self._build()
        return True
    
    def lt(self, x: T) -> Union[T, None]:
        "Find the largest element < x, or None if it doesn't exist."
        for a in reversed(self.a):
            if a[0] < x:
                return a[bisect_left(a, x) - 1]

    def le(self, x: T) -> Union[T, None]:
        "Find the largest element <= x, or None if it doesn't exist."
        for a in reversed(self.a):
            if a[0] <= x:
                return a[bisect_right(a, x) - 1]

    def gt(self, x: T) -> Union[T, None]:
        "Find the smallest element > x, or None if it doesn't exist."
        for a in self.a:
            if a[-1] > x:
                return a[bisect_right(a, x)]

    def ge(self, x: T) -> Union[T, None]:
        "Find the smallest element >= x, or None if it doesn't exist."
        for a in self.a:
            if a[-1] >= x:
                return a[bisect_left(a, x)]
    
    def __getitem__(self, x: int) -> T:
        "Return the x-th element, or IndexError if it doesn't exist."
        if x < 0: x += self.size
        if x < 0: raise IndexError
        for a in self.a:
            if x < len(a): return a[x]
            x -= len(a)
        raise IndexError
    
    def index(self, x: T) -> int:
        "Count the number of elements < x."
        ans = 0
        for a in self.a:
            if a[-1] >= x:
                return ans + bisect_left(a, x)
            ans += len(a)
        return ans

    def index_right(self, x: T) -> int:
        "Count the number of elements <= x."
        ans = 0
        for a in self.a:
            if a[-1] > x:
                return ans + bisect_right(a, x)
            ans += len(a)
        return 
    
n = int(input())
li = [int(input()) for _ in range(n)]

sst = SortedSet()
sst.add((li[0], 0))
print(0)

cumul = 0
for i in range(1, n):
    index = sst.index((li[i], 0))
    
    max_depth = 0
    if index - 1 >= 0:
        depth = sst[index - 1][1]
        max_depth = max(max_depth, depth)       
    if index < len(sst):
        depth = sst[index][1]
        max_depth = max(max_depth, depth) 
        
    sst.add((li[i], max_depth + 1))
    cumul += max_depth + 1
    
    print(cumul)
```

```python
import io, os, sys
# input = io.BytesIO(os.read(0,os.fstat(0).st_size)).readline

n, k = map(int, input().split())
li = [list(map(int, input().split())) for _ in range(n)]

deltas = [0] * 1000001
for i in range(n):
    l, r = li[i]
    deltas[l] += 1
    deltas[r] -= 1
    
ws = [0] * 1000001
cur = 0
for i in range(1000001):
    cur += deltas[i]
    ws[i] = cur
    
start = 0
end = 0
cur = ws[0]
ra, rb = -1, -1
while True:
    if cur < k:
        end += 1
        if end > 1000000:
            break
        cur += ws[end]
    elif cur > k:
        cur -= ws[start]
        start += 1
        if start > end:
            end = start
            if end > 1000000:
                break
            cur += ws[end]
    else:
        ra = start
        rb = end
        break
        
if ra == -1:
    print(0, 0)
else:
    print(ra, rb + 1)
```

https://www.acmicpc.net/problem/9345 디지털 비디오 디스크(DVDs)

```python
import io, os, sys
input = io.BytesIO(os.read(0,os.fstat(0).st_size)).readline

inf = float('inf')

class SegmentTree:
    def __init__(self, data, default=10**15, func=lambda a, b: max(a,b)):
        """initialize the segment tree with data"""
        self._default = default
        self._func = func
        self._len = len(data)
        self._size = _size = 1 << (self._len - 1).bit_length()
 
        self.data = [default] * (2 * _size)
        self.data[_size:_size + self._len] = data
        for i in reversed(range(_size)):
            self.data[i] = func(self.data[i + i], self.data[i + i + 1])
 
    def __delitem__(self, idx):
        self[idx] = self._default
 
    def __getitem__(self, idx):
        return self.data[idx + self._size]
 
    def __setitem__(self, idx, value):
        idx += self._size
        self.data[idx] = value
        idx >>= 1
        while idx:
            self.data[idx] = self._func(self.data[2 * idx], self.data[2 * idx + 1])
            idx >>= 1
 
    def __len__(self):
        return self._len
 
    def query(self, start, stop):
        if start == stop:
            return self.__getitem__(start)
        stop += 1
        start += self._size
        stop += self._size
 
        res = self._default
        while start < stop:
            if start & 1:
                res = self._func(res, self.data[start])
                start += 1
            if stop & 1:
                stop -= 1
                res = self._func(res, self.data[stop])
            start >>= 1
            stop >>= 1
        return res
 
    def __repr__(self):
        return "SegmentTree({0})".format(self.data)
    
for _ in range(int(input())):
    n, k = map(int, input().split())

    max_seg = SegmentTree(range(n), -inf, func=max)
    min_seg = SegmentTree(range(n), inf, func=min)

    for _ in range(k):
        q, a, b = map(int, input().split())

        if q == 0:
            max_seg[a], max_seg[b] = max_seg[b], max_seg[a]
            min_seg[a], min_seg[b] = min_seg[b], min_seg[a]
        else:
            max_v = max_seg.query(a, b)
            min_v = min_seg.query(a, b)

            ip = True
            if min_v != a or max_v != b:
                ip = False

            if ip:
                print('YES')
            else:
                print('NO')
```

https://www.acmicpc.net/problem/5817 고통받는 난쟁이들

```python
import io, os, sys
input = io.BytesIO(os.read(0,os.fstat(0).st_size)).readline

inf = float('inf')

class SegmentTree:
    def __init__(self, data, default=10**15, func=lambda a, b: max(a,b)):
        """initialize the segment tree with data"""
        self._default = default
        self._func = func
        self._len = len(data)
        self._size = _size = 1 << (self._len - 1).bit_length()
 
        self.data = [default] * (2 * _size)
        self.data[_size:_size + self._len] = data
        for i in reversed(range(_size)):
            self.data[i] = func(self.data[i + i], self.data[i + i + 1])
 
    def __delitem__(self, idx):
        self[idx] = self._default
 
    def __getitem__(self, idx):
        return self.data[idx + self._size]
 
    def __setitem__(self, idx, value):
        idx += self._size
        self.data[idx] = value
        idx >>= 1
        while idx:
            self.data[idx] = self._func(self.data[2 * idx], self.data[2 * idx + 1])
            idx >>= 1
 
    def __len__(self):
        return self._len
 
    def query(self, start, stop):
        if start == stop:
            return self.__getitem__(start)
        stop += 1
        start += self._size
        stop += self._size
 
        res = self._default
        while start < stop:
            if start & 1:
                res = self._func(res, self.data[start])
                start += 1
            if stop & 1:
                stop -= 1
                res = self._func(res, self.data[stop])
            start >>= 1
            stop >>= 1
        return res
 
    def __repr__(self):
        return "SegmentTree({0})".format(self.data)

n, k = map(int, input().split())
li = [0] + list(map(int, input().split()))

locs = [0] * (n + 1)
for i in range(n + 1):
    v = li[i]
    locs[v] = i

max_seg = SegmentTree(locs, -inf, func=max)
min_seg = SegmentTree(locs, inf, func=min)

for _ in range(k):
    q, x, y = map(int, input().split())

    if q == 1:
        a_loc, b_loc = x, y
        a, b = li[a_loc], li[b_loc]
        
        max_seg[a] = b_loc
        max_seg[b] = a_loc
        
        min_seg[a] = b_loc
        min_seg[b] = a_loc
        
        li[a_loc], li[b_loc] = li[b_loc], li[a_loc]
    else:
        a, b = x, y
        
        max_v = max_seg.query(a, b)
        min_v = min_seg.query(a, b)

        ip = True
        length = b - a + 1
        if max_v - min_v + 1 > length:
            ip = False

        if ip:
            print('YES')
        else:
            print('NO')
```

```python
import io, os, sys
input = io.BytesIO(os.read(0,os.fstat(0).st_size)).readline

inf = float('inf')

def reduce(cur_r, cur_c, cur_d, cur_turn_count):
    return cur_turn_count + cur_d * (max_turn_c + 1) + cur_c * (max_turn_c + 1) * 2 + cur_r * (max_turn_c + 1) * 2 * c

# def expand(reduced):
#     cur_turn_count = reduced  % (max_turn_c + 1)
#     reduced //= (max_turn_c + 1)
    
#     cur_d = reduced % 2
#     reduced //= 2
    
#     cur_c = reduced % c
#     reduced //= c
    
#     cur_r = reduced
    
#     return cur_r, cur_c, cur_d, cur_turn_count

for _ in range(int(input())):
    r, c, l, g = map(int, input().split())

    move_right = [list(map(int, input().split())) for _ in range(r)]
    move_down = [list(map(int, input().split())) for _ in range(r - 1)]

    max_turn_c = max(r, c) * 2 + 1

#     dp = [[[[inf] * (max_turn_c + 1) for _ in range(2)] for _ in range(c)] for _ in range(r)] # dp[cur_r][cur_c][cur_d][cur_turn_count] = needed fuel amount
#     dp[0][0][0][0] = dp[0][0][1][0] = 0
#     direction seq: [to_down, to_right]
    
    dp = [inf] * (max_turn_c + 1) * 2 * c * r
    dp[reduce(0, 0, 0, 0)] = 0
    dp[reduce(0, 0, 1, 0)] = 0
    
    j = 0
    for i in range(1, r):
        dp[reduce(i, j, 0, 0)] = dp[reduce(i - 1, j, 0, 0)] + move_down[i - 1][j]

    i = 0
    for j in range(1, c):
        dp[reduce(i, j, 1, 0)] = dp[reduce(i, j - 1, 1, 0)] + move_right[i][j - 1]

    for i in range(1, r):
        for j in range(1, c):
            for cur_turn_count in range(max_turn_c + 1):
                
                to_down = dp[reduce(i - 1, j, 0, cur_turn_count)] + move_down[i - 1][j]
                to_right = dp[reduce(i, j - 1, 1, cur_turn_count)] + move_right[i][j - 1]

                if cur_turn_count >= 1:
                    to_down = min(to_down, dp[reduce(i - 1, j, 1, cur_turn_count - 1)] + move_down[i - 1][j])
                    to_right = min(to_right, dp[reduce(i, j - 1, 0, cur_turn_count - 1)] + move_right[i][j - 1])
                    
                dp[reduce(i, j, 0, cur_turn_count)] = to_down
                dp[reduce(i, j, 1, cur_turn_count)] = to_right

    i = r - 1
    j = c - 1

    min_turn_used = inf
    for cur_turn_count in range(max_turn_c + 1):
        if dp[reduce(i, j, 0, cur_turn_count)] <= g or dp[reduce(i, j, 1, cur_turn_count)] <= g:
            min_turn_used = cur_turn_count
            break

    if min_turn_used == inf:
        print(-1)
    else:
        base_tm_used = (r + c - 2) * l
        result = base_tm_used + min_turn_used

        print(result)
```

https://www.acmicpc.net/problem/14517 팰린드롬 개수 구하기 (Large)

```python
mod = 10007

s = input()
n = len(s)

dp = [[0] * n for _ in range(n)]

for i in range(n):
    dp[i][i] = 1
    
for i in range(n - 1):
    dp[i][i + 1] = 2
    if s[i] == s[i + 1]:
        dp[i][i + 1] += 1
        
for length in range(3, n + 1):
    for start in range(n - length + 1):
        end = start + length - 1
        
        dp[start][end] = dp[start][end - 1] + dp[start + 1][end] - dp[start + 1][end - 1]
        if s[start] == s[end]:
            dp[start][end] += dp[start + 1][end - 1] + 1
            
        dp[start][end] %= mod
        
result = dp[0][n - 1]

print(result)
```

```python
import io, os, sys
input = io.BytesIO(os.read(0,os.fstat(0).st_size)).readline

# https://github.com/tatyam-prime/SortedSet/blob/main/SortedMultiset.py
import math
from bisect import bisect_left, bisect_right, insort
from typing import Generic, Iterable, Iterator, TypeVar, Union, List
T = TypeVar('T')

class SortedMultiset(Generic[T]):
    BUCKET_RATIO = 50
    REBUILD_RATIO = 170

    def _build(self, a=None) -> None:
        "Evenly divide `a` into buckets."
        if a is None: a = list(self)
        size = self.size = len(a)
        bucket_size = int(math.ceil(math.sqrt(size / self.BUCKET_RATIO)))
        self.a = [a[size * i // bucket_size : size * (i + 1) // bucket_size] for i in range(bucket_size)]
    
    def __init__(self, a: Iterable[T] = []) -> None:
        "Make a new SortedMultiset from iterable. / O(N) if sorted / O(N log N)"
        a = list(a)
        if not all(a[i] <= a[i + 1] for i in range(len(a) - 1)):
            a = sorted(a)
        self._build(a)

    def __iter__(self) -> Iterator[T]:
        for i in self.a:
            for j in i: yield j

    def __reversed__(self) -> Iterator[T]:
        for i in reversed(self.a):
            for j in reversed(i): yield j
    
    def __len__(self) -> int:
        return self.size
    
    def __repr__(self) -> str:
        return "SortedMultiset" + str(self.a)
    
    def __str__(self) -> str:
        s = str(list(self))
        return "{" + s[1 : len(s) - 1] + "}"

    def _find_bucket(self, x: T) -> List[T]:
        "Find the bucket which should contain x. self must not be empty."
        for a in self.a:
            if x <= a[-1]: return a
        return a

    def __contains__(self, x: T) -> bool:
        if self.size == 0: return False
        a = self._find_bucket(x)
        i = bisect_left(a, x)
        return i != len(a) and a[i] == x

    def count(self, x: T) -> int:
        "Count the number of x."
        return self.index_right(x) - self.index(x)

    def add(self, x: T) -> None:
        "Add an element. / O(√N)"
        if self.size == 0:
            self.a = [[x]]
            self.size = 1
            return
        a = self._find_bucket(x)
        insort(a, x)
        self.size += 1
        if len(a) > len(self.a) * self.REBUILD_RATIO:
            self._build()

    def discard(self, x: T) -> bool:
        "Remove an element and return True if removed. / O(√N)"
        if self.size == 0: return False
        a = self._find_bucket(x)
        i = bisect_left(a, x)
        if i == len(a) or a[i] != x: return False
        a.pop(i)
        self.size -= 1
        if len(a) == 0: self._build()
        return True

    def lt(self, x: T) -> Union[T, None]:
        "Find the largest element < x, or None if it doesn't exist."
        for a in reversed(self.a):
            if a[0] < x:
                return a[bisect_left(a, x) - 1]

    def le(self, x: T) -> Union[T, None]:
        "Find the largest element <= x, or None if it doesn't exist."
        for a in reversed(self.a):
            if a[0] <= x:
                return a[bisect_right(a, x) - 1]

    def gt(self, x: T) -> Union[T, None]:
        "Find the smallest element > x, or None if it doesn't exist."
        for a in self.a:
            if a[-1] > x:
                return a[bisect_right(a, x)]

    def ge(self, x: T) -> Union[T, None]:
        "Find the smallest element >= x, or None if it doesn't exist."
        for a in self.a:
            if a[-1] >= x:
                return a[bisect_left(a, x)]
    
    def __getitem__(self, x: int) -> T:
        "Return the x-th element, or IndexError if it doesn't exist."
        if x < 0: x += self.size
        if x < 0: raise IndexError
        for a in self.a:
            if x < len(a): return a[x]
            x -= len(a)
        raise IndexError

    def index(self, x: T) -> int:
        "Count the number of elements < x."
        ans = 0
        for a in self.a:
            if a[-1] >= x:
                return ans + bisect_left(a, x)
            ans += len(a)
        return ans

    def index_right(self, x: T) -> int:
        "Count the number of elements <= x."
        ans = 0
        for a in self.a:
            if a[-1] > x:
                return ans + bisect_right(a, x)
            ans += len(a)
        return ans

n, s = map(int, input().split())
q = int(input())
qs = [int(input()) for _ in range(q)]

ms = SortedMultiset()
for _ in range(n):
    ms.add((0, 0))

li = [[0] * 2 for _ in range(n + 1)]
for qi in range(q):
    num = qs[qi]
    
    f_info = tuple(li[num])
    ms.discard(f_info)
    
    li[num][0] -= 1
    li[num][1] = qi
    
    info = tuple(li[num])
    ms.add(info)
    
    cur_info = tuple(li[1])
    
    cur_rank = -1
    if cur_info[0] == 0:
        cur_rank = n
    else:
        cur_rank = ms.index(cur_info) + 1
    
    target_info = tuple(ms[s - 1])
    
    needed = -1
    if cur_rank <= s:
        needed = 0
    else:
        diff = (-target_info[0]) - (-cur_info[0])
        needed = diff + 1
    
    print(cur_rank, needed)
```

```python
import io, os, sys
# input = io.BytesIO(os.read(0,os.fstat(0).st_size)).readline

get_value = lambda x: int(x) - 1

n, m = map(int, input().split())
hs = list(map(int, input().split()))

li = [[] for _ in range(n)]
for _ in range(m):
    u, v = map(get_value, input().split())
    
    if hs[u] < hs[v]:
        li[u].append(v)
    elif hs[u] > hs[v]:
        li[v].append(u)
    
nhs = [(hs[i], i) for i in range(n)]
nhs.sort(key=lambda x: -x[0])

dp = [1] * n

for i in range(n):
    value = nhs[i][1]
    
    temp = 0
    for v in li[value]:
        temp = max(temp, dp[v])
        
    dp[value] += temp
        
print('\n'.join(map(str, dp)))
```

```python
def rotate(li, n):
    '''type 1 -> 2 -> 4 -> 3 -> 1'''
    
    nli = [[0] * n for _ in range(n)]
    
    for i in range(n):
        for j in range(n):
            nli[j][n - 1 - i] = li[i][j]
    return nli

tiles1 = [[[1]], [[1, 1], [1, 0]], [[1, 1, 2, 2], [1, 5, 5, 2], [3, 5, 0, 0], [3, 3, 0, 0]]] # d, r empty
tiles2 = [[[1]], [[2, 2], [0, 2]]] # d, l empty
tiles3 = [[[1]], [[3, 0], [3, 3]]] # u, r empty
tiles4 = [[[1]], [[0, 4], [4, 4]]] # u, l empty

ntiles = rotate(tiles1[-1], 2 ** len(tiles3))
tiles2.append(ntiles)

ntiles = rotate(tiles2[-1], 2 ** len(tiles3))
tiles4.append(ntiles)

ntiles = rotate(tiles4[-1], 2 ** len(tiles3))
tiles3.append(ntiles)

for cur_n in range(3, 8):
    size = 2 ** cur_n

    half = size // 2
    qart = size // 4
    temp = [[0] * size for _ in range(size)]

    for i in range(half):
        for j in range(half):
            temp[i][j] = tiles1[cur_n - 1][i][j]

    for i in range(half):
        for j in range(half, size):
            temp[i][j] = tiles2[cur_n - 1][i][j - half]

    for i in range(half, size):
        for j in range(half):
            temp[i][j] = tiles3[cur_n - 1][i - half][j]

    for i in range(qart, half + qart):
        for j in range(qart, half + qart):
            temp[i][j] = tiles1[cur_n - 1][i - qart][j - qart]

    tiles1.append(temp)

    ntiles = rotate(tiles1[-1], size)
    tiles2.append(ntiles)

    ntiles = rotate(tiles2[-1], size)
    tiles4.append(ntiles)

    ntiles = rotate(tiles4[-1], size)
    tiles3.append(ntiles)

k = int(input())
x, y = map(int, input().split())

size = 2 ** k

ox = size - y
oy = x - 1

result = [[0] * size for _ in range(size)]

def recur(cur_n, sx, sy, ox, oy):
    if cur_n == 0:
        return
    
    size = 2 ** cur_n
    half = size // 2
    
    if ox < half and oy < half:
        for i in range(size):
            for j in range(size):
                result[sx + i][sy + j] = tiles4[cur_n][i][j]
        recur(cur_n - 1, sx, sy, ox, oy)
    elif ox < half and oy >= half:
        for i in range(size):
            for j in range(size):
                result[sx + i][sy + j] = tiles3[cur_n][i][j]
        nsx = sx
        nsy = sy + half
        nox = ox
        noy = oy - half
        recur(cur_n - 1, nsx, nsy, nox, noy)
    elif ox >= half and oy < half:
        for i in range(size):
            for j in range(size):
                result[sx + i][sy + j] = tiles2[cur_n][i][j]
        nsx = sx + half
        nsy = sy
        nox = ox - half
        noy = oy
        recur(cur_n - 1, nsx, nsy, nox, noy)
    else:
        for i in range(size):
            for j in range(size):
                result[sx + i][sy + j] = tiles1[cur_n][i][j]
        nsx = sx + half
        nsy = sy + half
        nox = ox - half
        noy = oy - half
        recur(cur_n - 1, nsx, nsy, nox, noy)
        
recur(k, 0, 0, ox, oy)
    
nxt_num = 1
nresult = [[0] * size for _ in range(size)]
nresult[ox][oy] = -1
for i in range(size - 1):
    for j in range(size):
        if nresult[i][j] != 0:
            continue
        
        cur = result[i][j]
        
        nresult[i][j] = nxt_num
        if j - 1 >= 0 and result[i + 1][j - 1] == result[i + 1][j] == cur and nresult[i + 1][j - 1] == nresult[i + 1][j] == 0:
            nresult[i + 1][j - 1] = nresult[i + 1][j] = nxt_num
        elif j + 1 < size and result[i][j + 1] == result[i + 1][j] == cur and nresult[i][j + 1] == nresult[i + 1][j] == 0:
            nresult[i][j + 1] = nresult[i + 1][j] = nxt_num
        elif j + 1 < size and result[i + 1][j] == result[i + 1][j + 1] == cur and nresult[i + 1][j] == nresult[i + 1][j + 1] == 0:
            nresult[i + 1][j] = nresult[i + 1][j + 1] = nxt_num
        elif j + 1 < size and result[i][j + 1] == result[i + 1][j + 1] == cur and nresult[i][j + 1] == nresult[i + 1][j + 1] == 0:
            nresult[i][j + 1] = nresult[i + 1][j + 1] = nxt_num
            
        nxt_num += 1
        
for v in nresult:
    print(*v)
```
