```python
from collections import deque
import sys
# input = lambda: sys.stdin.readline().rstrip()

lowers = 'abcdefghijklmnopqrstuvwxyz'
uppers = lowers.upper()
d = {v: 1 << lowers.index(v) for v in lowers}
    
for _ in range(int(input())):
    n, m = map(int, input().split())
    li = [list(input()) for _ in range(n)]
    given = set(input())
    
    keys = 0
    if given != {'0'}:
        for key in given:
            keys |= d[key]

    found_door_locs = {v: set() for v in uppers}
    queue = deque()
    docu_num = '1'
    collected = set()
    for i in range(n):
        for j in range(m):
            if li[i][j] == '$':
                li[i][j] = docu_num
                docu_num = int(docu_num)
                docu_num += 1
                docu_num = str(docu_num)
            
            if i == 0 or i == n - 1 or j == 0 or j == m - 1:
                if li[i][j] == '.' or li[i][j].isdigit() or li[i][j].islower():
                    queue.append((i, j))
                    if li[i][j].isdigit():
                        collected.add(li[i][j])
                    elif li[i][j].islower():
                        keys |= d[li[i][j]]
                elif li[i][j].isupper():
                    found_door_locs[li[i][j]].add((i, j))
                    if li[i][j].lower() in given:
                        queue.append((i, j))

    visited = [[set() for _ in range(m)] for _ in range(n)]
    for x, y in queue:
        visited[x][y].add(keys)

    dx = [-1, 1, 0, 0]
    dy = [0, 0, -1, 1]

    while queue:
        x, y = queue.popleft()
        for i in range(4):
            nx = x + dx[i]
            ny = y + dy[i]

            if nx < 0 or nx >= n or ny < 0 or ny >= m:
                continue
            if li[nx][ny] == '*':
                continue
            if keys in visited[nx][ny]:
                continue        
            visited[nx][ny].add(keys)

            if li[nx][ny].isdigit():
                collected.add(li[nx][ny])
            elif li[nx][ny].isupper():
                found_door_locs[li[nx][ny]].add((nx, ny))
                if not (d[li[nx][ny].lower()] & keys):
                    continue
            elif li[nx][ny].islower():
                keys |= d[li[nx][ny]]
                queue.extend(found_door_locs[li[nx][ny].upper()])
            queue.append((nx, ny))

    print(len(collected))
```

### 그대, 그머가 되어 14496

```python
from collections import deque
import sys
# input = lambda: sys.stdin.readline().rstrip()

s, e = map(int, input().split())
n, m = map(int, input().split())
li = [[] for _ in range(n + 1)]
for _ in range(m):
    a, b = map(int, input().split())
    li[a].append(b)
    li[b].append(a)

def bfs():    
    dists = [float('inf')] * (n + 1)
    dists[s] = 0
    queue = deque()
    queue.append(s)
    while queue:
        value = queue.popleft()
        if value == e:
            return dists[value]
        
        for v in li[value]:
            if dists[v] == float('inf'):
                dists[v] = dists[value] + 1
                queue.append(v)
    return -1

print(bfs())
```

### 점프 게임 15558

```python
from collections import deque

n, k = map(int, input().split())
li = [list(map(int, input())), list(map(int, input()))]

def bfs():
    visited = [[False] * n for _ in range(2)]
    visited[0][0] = True
    queue = deque()
    queue.append((0, 0))
    t = 0
    while queue:
        for _ in range(len(queue)):
            is_on_right, x = queue.popleft()

            dr = [0, 0, 1]
            dx = [-1, 1, k]
            for i in range(3):
                nr = is_on_right ^ dr[i]
                nx = x + dx[i]

                if nx <= t:
                    continue
                if nx >= n:
                    return 1

                if li[nr][nx] == 0:
                    continue
                if visited[nr][nx]:
                    continue
                visited[nr][nx] = True
                queue.append((nr, nx))
        t += 1
    return 0
          
print(bfs())
```

### 두 로봇 15971

```python
from collections import deque
import sys
input = lambda: sys.stdin.readline().rstrip()

n, s, e = map(int, input().split())
li = [[] for _ in range(n + 1)]

for _ in range(n - 1):
    a, b, d = map(int, input().split())
    li[a].append((b, d))
    li[b].append((a, d))
    
def bfs(start, end):    
    dists = [-1] * (n + 1)
    dists[start] = 0
    queue = deque()
    queue.append((start, 0)) # node, max dist edge 
    while queue:
        value, mde = queue.popleft()
        if value == e:
            return dists[value] - mde

        for v, d in li[value]:
            if dists[v] == -1:
                nmde = max(mde, d)
                dists[v] = dists[value] + d
                queue.append((v, nmde))
        
print(bfs(s, e))
```

### 움직이는 미로 탈출 16954

```python
from collections import deque

li = [input() for _ in range(8)]

n = 8
dx = [-1, -1, -1, 0, 0, 0, 1, 1, 1]
dy = [-1, 0, 1, -1, 0, 1, -1, 0, 1]

def move_walls():
    for i in range(7, 0, -1):
        li[i] = li[i - 1]
    li[0] = '.' * 8

def bfs():
    queue = deque()
    queue.append((n - 1, 0))
    while True:
        visited = [[False] * n for _ in range(n)]
        for _ in range(len(queue)):
            x, y = queue.popleft()
            if li[x][y] == '#':
                continue
                
            for i in range(9):
                nx = x + dx[i]
                ny = y + dy[i]    
                if nx < 0 or nx >= n or ny < 0 or ny >= n:
                    continue
                if li[nx][ny] == '#':
                    continue
                if nx == 0 and ny == n - 1:
                    return 1
                if not visited[nx][ny]:
                    visited[nx][ny] = True
                    queue.append((nx, ny))
                    
        if not queue:
            return 0
        move_walls()
        
print(bfs())
```

### 직사각형 탈출 16973

```python
# 맞았음, 누적합 이용도 가능함
from collections import deque
import sys
# input = lambda: sys.stdin.readline().rstrip()

n, m = map(int, input().split())
li = [list(map(int, input().split())) for _ in range(n)]
h, w, sr, sc, fr, fc = map(int, input().split())

sr, sc, fr, fc = sr - 1, sc - 1, fr - 1, fc - 1

def check(x, y, d):
    if d == 0:
        for j in range(y, y + w):
            if li[x][j] == 1:
                return False
    elif d == 1:
        for j in range(y, y + w):
            if li[x + h - 1][j] == 1:
                return False
    elif d == 2:
        for i in range(x, x + h):
            if li[i][y] == 1:
                return False
    else:
        for i in range(x, x + h):
            if li[i][y + w - 1] == 1:
                return False
            
    return True

dx = [-1, 1, 0, 0]
dy = [0, 0, -1, 1]

dists = [[-1] * m for _ in range(n)]
dists[sr][sc] = 0
queue = deque()
queue.append((sr, sc))

result = -1
while queue:
    x, y = queue.popleft()
    if x == fr and y == fc:
        result = dists[x][y]
        break
        
    for d in range(4):
        nx, ny = x + dx[d], y + dy[d]
        if nx < 0 or nx + h - 1 >= n or ny < 0 or ny + w - 1 >= m:
            continue
        if dists[nx][ny] != -1:
            continue
        if check(nx, ny, d):
            dists[nx][ny] = dists[x][y] + 1
            queue.append((nx, ny))
            
print(result)
```

### 배달 1175 (다시 풀음)

```python
from collections import deque

n, m = map(int, input().split())
li = [input() for _ in range(n)]

sx = sy = 0
ts = [] # targets
for i in range(n):
    for j in range(m):
        if li[i][j] == 'S':
            sx = i
            sy = j
        elif li[i][j] == 'C':
            ts.append((i, j))
            
dx = [-1, 1, 0, 0]
dy = [0, 0, -1, 1]

dists = [[[[float('inf')] * 3 for _ in range(4)] for _ in range(m)] for _ in range(n)] # dists[x][y][들어온 방향][두 곳 전부 미방문, 1번째 장소 방문, 2번째 장소 방문]
for i in range(4):
    dists[sx][sy][i][0] = 0

queue = deque()
for d in range(4):
    queue.append((sx, sy, d, 0)) # x, y, 들어온 방향, 선물 방문할 곳 방문 ststus

min_dist = float('inf')
while queue:
    x, y, d, s = queue.popleft()
    
    for nd in range(4):
        if nd == d:
            continue
        nx = x + dx[nd]
        ny = y + dy[nd]
        
        if nx < 0 or nx >= n or ny < 0 or ny >= m:
            continue
        if li[nx][ny] == '#':
            continue
        
        ns = s
        if nx == ts[0][0] and ny == ts[0][1]:
            if s == 0:
                ns = 1
            elif s == 2:
                ns = 3
        elif nx == ts[1][0] and ny == ts[1][1]:
            if s == 0:
                ns = 2
            elif s == 1:
                ns = 3
                
        ndist = dists[x][y][d][s] + 1
        if ns == 3:
            min_dist = min(min_dist, ndist)
            continue

        if ndist < dists[nx][ny][nd][ns]:
            dists[nx][ny][nd][ns] = ndist
            queue.append((nx, ny, nd, ns))
            
if min_dist == float('inf'):
    print(-1)
else:
    print(min_dist)
```

### 탈옥 9376

```python
from collections import deque

# n, m = map(int, input().split())
# li = [input() for _ in range(n)]

n, m = 9, 9
s = '''*#**#**#*
*#**#**#*
*#**#**#*
*#**.**#*
*#*#.#*#*
*$##*##$*
*#*****#*
*.#.#.#.*
*********'''
li = s.split('\n')

starts = []
ps = [] # prisoners
for i in range(n):
    for j in range(m):
        if i == 0 or j == 0 or i == n - 1 or j == m - 1:
            if li[i][j] != '*':
                starts.append((i, j))
        if li[i][j] == '$':
            ps.append((i, j))

dx = [-1, 1, 0, 0]
dy = [0, 0, -1, 1]
        
p1x, p1y = ps[0]
p2x, p2y = ps[1]

mr = float('inf') # min result
for sx, sy in starts:
    bc = 0 # broke count
    s = 0 # 구출 상태
    
    if li[sx][sy] == '$':
        if sx == ps[0][0] and sy == ps[0][1]:
            s = 1
        else:
            s = 2
    elif li[sx][sy] == '#':
        bc = 1
    
    # bcs[n][m][0: 구출 못함, 1: 1번 구출, 2, 2번 구출, 3: 둘 다 구출] = 문 부순 최소 횟수
    bcs = [[[float('inf')] * 4 for _ in range(m)] for _ in range(n)] # broken door counts
    bcs[sx][sy][s] = bc
    queue = deque()  
    queue.append((sx, sy, s))

    while queue:
        x, y, s = queue.popleft()

        for i in range(4):
            nx = x + dx[i]
            ny = y + dy[i]

            if nx < 0 or nx >= n or ny < 0 or ny >= m:
                continue
            if li[nx][ny] == '*':
                continue

            bc = bcs[x][y][s] # broken door count
            nbc = bc
            ns = s 
            if li[nx][ny] == '#':
                    nbc += 1
            elif li[nx][ny] == '$':        
                if nx == ps[0][0] and ny == ps[0][1]:
                    if s == 0:
                        ns = 1
                    elif s == 2:
                        ns = 3
                else:
                    if s == 0:
                        ns = 2
                    elif s == 1:
                        ns = 3

            if nbc < bcs[nx][ny][ns]:
                bcs[nx][ny][ns] = nbc

                if ns != 3:
                    queue.append((nx, ny, ns))

    
            if li[nx][ny] != '*' and li[nx][ny] != '$':
                if bcs[nx][ny][1] != float('inf') and bcs[nx][ny][2] != float('inf') and bcs[nx][ny][0] != float('inf'):
                    mr = min(mr, bcs[p1x][p1y][1] + bcs[p2x][p2y][2] - bcs[nx][ny][0])
                    queue.pop()
                    
    r = min(bcs[p1x][p1y][3], bcs[p2x][p2y][3], bcs[p1x][p1y][1] + bcs[p2x][p2y][2])
    mr = min(mr, r)

print(mr)
```
