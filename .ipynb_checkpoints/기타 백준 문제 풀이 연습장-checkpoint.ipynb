그룹 단어 체커 1316번

```python
from collections import defaultdict

n = int(input())
input_li = []

for _ in range(n):
    input_li.append(input())
    
result_num = 0
for s in input_li: 
    li = list(s)
    d = defaultdict(int)
    

    indices_to_delete = []
    for i in range(len(li) - 1):
        if li[i] == li[i + 1]:
            indices_to_delete.append(i + 1)

    for i in reversed(indices_to_delete):
        del li[i]


    for v in li:
        d[v] += 1

    sum_values = sum(d.values())

    if sum_values == len(set(li)):
        result_num += 1

print(result_num)
```

Birds Rituals 20680번

```python
n, m = map(int, input().split(" "))
state = input().split(" ")

for _ in range(m):
    c = input().split(" ")
    
    if c[0] == "insert":
        state.insert(int(c[2]), c[1])
    elif c[0] == "relocate":
        current_position = state.index(c[1])
        state.pop(current_position)
        state.insert(current_position + int(c[2]), c[1])
    elif c[0] == "depart":
        state.remove(c[1])
        
print(" ".join(state))
```

줄 세우기 2252 - 그래프 이론, 위상 정렬

```python
# 맞았음
from collections import deque
import sys

n, m = map(int, sys.stdin.readline().split())

indegree_li = [0 for _ in range(n + 1)]
array = [[] for _ in range(n + 1)]

for _ in range(m):
    a, b = map(int, sys.stdin.readline().split())
    indegree_li[b] += 1
    array[a].append(b)

queue = deque()
for i in range(1, n + 1):
    if indegree_li[i] == 0:
        queue.append(i)
        
result = []
while queue:
    value = queue.popleft()
    result.append(value)
    
    for b in array[value]:
        indegree_li[b] -= 1
        
        if indegree_li[b] == 0:
            queue.append(b)
            
print(' '.join(map(str, result)))
```

### ACM Craft 1005

```python
# 맞았음
from collections import deque
import sys

case_num = int(sys.stdin.readline())
for _ in range(case_num):
    n, k = map(int, sys.stdin.readline().split())

    time_to_spend_li = [0] + list(map(int, sys.stdin.readline().split()))
    indegree_li = [0] * (n + 1)
    post_array = [[] for _ in range(n + 1)] # 각 값은 후행함.
    pre_array = [[] for _ in range(n + 1)] # 각 값은 선행함.

    total_time_spent_dict = {i: 0 for i in range(1, n + 1)}

    for _ in range(k):
        a, b = map(int, sys.stdin.readline().split())
        indegree_li[b] += 1
        post_array[a].append(b)
        pre_array[b].append(a)

    queue = deque()

    for i in range(1, n + 1):
        if indegree_li[i] == 0:
            queue.append(i)

    purpose = int(sys.stdin.readline())
    while queue:
        value = queue.popleft()

        total_time_spent_dict[value] += time_to_spend_li[value]

        # 각 건물 설치가 완료되었을 때 이전 설치 루트 중 가장 긴 시간을 추가해줌
        max_time_spent = 0
        for pre in pre_array[value]:
            if total_time_spent_dict[pre] > max_time_spent:
                max_time_spent = total_time_spent_dict[pre]

        total_time_spent_dict[value] += max_time_spent

        if value == purpose:
            print(total_time_spent_dict[value])
            break

        for b in post_array[value]:
            indegree_li[b] -= 1
            if indegree_li[b] == 0:
                queue.append(b)
```

```python
# 시간초과 예상
from itertools import permutations
from itertools import chain

k = int(input())
inequality_li = list(input()) + ["x"]

permutation_iter = permutations(range(10), k + 1)

min_value = float('inf')
max_value = 0

while True:
    try:
        case = next(permutation_iter)
        if eval("".join(map(str, chain(*list(zip(case, inequality_li)))))[:-1]):
            value = int("".join(map(str, case)))
            if value < min_value:
                min_value = value
            if value > max_value:
                max_value = value
    except StopIteration:
        break
        
print(min_value)
print(max_value)
```

```python
# 맞았음
k = int(input())
inequalities = input().replace(" ", "")

remain_li = [str(i) for i in range(10)]
possible_li = []
is_value_added = False

def recur(s, remain_li, remain_inequalities):
    global is_value_added
    s += remain_inequalities[0]
    remain_inequalities = remain_inequalities[1:]
    for v in remain_li:
        if is_value_added:
            return
        
        new_remain_li = remain_li[:]
        new_remain_li.remove(v)
        s += v
        
        if eval(s):
            if not remain_inequalities:
                new_s = s
                new_s = new_s.replace("<", "")
                new_s = new_s.replace(">", "")
                possible_li.append(new_s)
                is_value_added = True
            else:
                recur(s, new_remain_li, remain_inequalities)
        s = s[:-1]
    

s = ''
for v in remain_li:
    s += v
    new_remain_li = remain_li[:]
    new_remain_li.remove(v)
    recur(s, new_remain_li, inequalities)
    if is_value_added:
        is_value_added = False
        break
    s = s[:-1]

s = ''
for v in reversed(remain_li):
    s += v
    new_remain_li = list(reversed(remain_li))
    new_remain_li.remove(v)
    recur(s, new_remain_li, inequalities)
    if is_value_added:
        break
    s = s[:-1]
print(f'{possible_li[-1]}\n{possible_li[0]}')
```

```python
n = 890
count = -1
current = -1
while count < n:
    current += 1
    if checker(current):
        count += 1
print(current)
```

```python
def checker(current):
    current = str(current)
    
    while len(current) >= 2:
        if current[-2] > current[-1]:
            current = current[:-1]
        else:
            return False
    return True
```

```python
# 예시 코드 바탕 구현
n = int(input())
li = list(map(int, input().split()))
dp = [1] * n

for i in range(1, n):
    for j in range(i):
        if li[j] < li[i]:
            dp[i] = max(dp[i], dp[j] + 1)
            
print(max(dp))
```

### 조합 2407

```python
from decimal import getcontext, Decimal

getcontext().prec = 1000

n, m = map(int, input().split())

nom = 1
for i in range(n, n - m, -1):
    nom *= i

denom = 1
for i in range(1, m + 1):
    denom *= i
    
print(Decimal(nom) // Decimal(denom))
```
