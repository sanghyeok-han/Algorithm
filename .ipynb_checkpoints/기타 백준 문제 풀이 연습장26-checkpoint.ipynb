https://www.acmicpc.net/problem/17408 수열과 쿼리 24

```python
import io, os, sys
input = io.BytesIO(os.read(0, os.fstat(0).st_size)).readline

inf = float('inf')

class SegmentTree:
    def __init__(self, data, default=10**15, func=lambda a, b: max(a,b)):
        """initialize the segment tree with data"""
        self._default = default
        self._func = func
        self._len = len(data)
        self._size = _size = 1 << (self._len - 1).bit_length()
 
        self.data = [default] * (2 * _size)
        self.data[_size:_size + self._len] = data
        for i in reversed(range(_size)):
            self.data[i] = func(self.data[i + i], self.data[i + i + 1])
 
    def __delitem__(self, idx):
        self[idx] = self._default
 
    def __getitem__(self, idx):
        return self.data[idx + self._size]
 
    def __setitem__(self, idx, value):
        idx += self._size
        self.data[idx] = value
        idx >>= 1
        while idx:
            self.data[idx] = self._func(self.data[2 * idx], self.data[2 * idx + 1])
            idx >>= 1
 
    def __len__(self):
        return self._len
 
    def query(self, start, stop):
        if start == stop:
            return self.__getitem__(start)
        stop += 1
        start += self._size
        stop += self._size
 
        res = self._default
        while start < stop:
            if start & 1:
                res = self._func(res, self.data[start])
                start += 1
            if stop & 1:
                stop -= 1
                res = self._func(res, self.data[stop])
            start >>= 1
            stop >>= 1
        return res
 
    def __repr__(self):
        return "SegmentTree({0})".format(self.data)
    

n = int(input())
li = [0] + list(map(int, input().split()))
m = int(input())

def func(a, b):
    if type(a) == float or type(a) == int:
        a = [a]
    if type(b) == float or type(b) == int:
        b = [b]
    
    return sorted(a + b, reverse=True)[:2]

seg = SegmentTree(li, -inf, func=func)

for _ in range(m):
    tp, *qu = map(int, input().split())
    
    if tp == 1:
        i, v = qu
        
        li[i] = v
        seg[i] = v
    elif tp == 2:
        l, r = qu
        
        result = sum(seg.query(l, r))
        
        print(result)
```

https://www.acmicpc.net/problem/11961 High Card Low Card (Platinum)

```python
import io, os, sys
input = io.BytesIO(os.read(0, os.fstat(0).st_size)).readline

# https://github.com/tatyam-prime/SortedSet/blob/main/SortedSet.py
import math
from bisect import bisect_left, bisect_right
from typing import Generic, Iterable, Iterator, TypeVar, Union, List
T = TypeVar('T')

class SortedSet(Generic[T]):
    BUCKET_RATIO = 50
    REBUILD_RATIO = 170

    def _build(self, a=None) -> None:
        "Evenly divide `a` into buckets."
        if a is None: a = list(self)
        size = self.size = len(a)
        bucket_size = int(math.ceil(math.sqrt(size / self.BUCKET_RATIO)))
        self.a = [a[size * i // bucket_size : size * (i + 1) // bucket_size] for i in range(bucket_size)]
    
    def __init__(self, a: Iterable[T] = []) -> None:
        "Make a new SortedSet from iterable. / O(N) if sorted and unique / O(N log N)"
        a = list(a)
        if not all(a[i] < a[i + 1] for i in range(len(a) - 1)):
            a = sorted(set(a))
        self._build(a)

    def __iter__(self) -> Iterator[T]:
        for i in self.a:
            for j in i: yield j

    def __reversed__(self) -> Iterator[T]:
        for i in reversed(self.a):
            for j in reversed(i): yield j
    
    def __len__(self) -> int:
        return self.size
    
    def __repr__(self) -> str:
        return "SortedSet" + str(self.a)
    
    def __str__(self) -> str:
        s = str(list(self))
        return "{" + s[1 : len(s) - 1] + "}"

    def _find_bucket(self, x: T) -> List[T]:
        "Find the bucket which should contain x. self must not be empty."
        for a in self.a:
            if x <= a[-1]: return a
        return a

    def __contains__(self, x: T) -> bool:
        if self.size == 0: return False
        a = self._find_bucket(x)
        i = bisect_left(a, x)
        return i != len(a) and a[i] == x

    def add(self, x: T) -> bool:
        "Add an element and return True if added. / O(√N)"
        if self.size == 0:
            self.a = [[x]]
            self.size = 1
            return True
        a = self._find_bucket(x)
        i = bisect_left(a, x)
        if i != len(a) and a[i] == x: return False
        a.insert(i, x)
        self.size += 1
        if len(a) > len(self.a) * self.REBUILD_RATIO:
            self._build()
        return True

    def discard(self, x: T) -> bool:
        "Remove an element and return True if removed. / O(√N)"
        if self.size == 0: return False
        a = self._find_bucket(x)
        i = bisect_left(a, x)
        if i == len(a) or a[i] != x: return False
        a.pop(i)
        self.size -= 1
        if len(a) == 0: self._build()
        return True
    
    def lt(self, x: T) -> Union[T, None]:
        "Find the largest element < x, or None if it doesn't exist."
        for a in reversed(self.a):
            if a[0] < x:
                return a[bisect_left(a, x) - 1]

    def le(self, x: T) -> Union[T, None]:
        "Find the largest element <= x, or None if it doesn't exist."
        for a in reversed(self.a):
            if a[0] <= x:
                return a[bisect_right(a, x) - 1]

    def gt(self, x: T) -> Union[T, None]:
        "Find the smallest element > x, or None if it doesn't exist."
        for a in self.a:
            if a[-1] > x:
                return a[bisect_right(a, x)]

    def ge(self, x: T) -> Union[T, None]:
        "Find the smallest element >= x, or None if it doesn't exist."
        for a in self.a:
            if a[-1] >= x:
                return a[bisect_left(a, x)]
    
    def __getitem__(self, x: int) -> T:
        "Return the x-th element, or IndexError if it doesn't exist."
        if x < 0: x += self.size
        if x < 0: raise IndexError
        for a in self.a:
            if x < len(a): return a[x]
            x -= len(a)
        raise IndexError
    
    def index(self, x: T) -> int:
        "Count the number of elements < x."
        ans = 0
        for a in self.a:
            if a[-1] >= x:
                return ans + bisect_left(a, x)
            ans += len(a)
        return ans

    def index_right(self, x: T) -> int:
        "Count the number of elements <= x."
        ans = 0
        for a in self.a:
            if a[-1] > x:
                return ans + bisect_right(a, x)
            ans += len(a)
        return 

n = int(input())
A = [int(input()) for _ in range(n)]

B = list(set(range(1, n * 2 + 1)) - set(A))

sst = SortedSet(B)
L = [0] * n
for i in range(n):
    if i >= 1:
        L[i] = L[i - 1]
        
        
    index = sst.index(A[i])
    if index != len(sst):
        L[i] += 1
        sst.discard(sst[index])
        
sst = SortedSet(B)
R = [0] * n
for i in range(n - 1, -1, -1):
    if i < n - 1:
        R[i] = R[i + 1]
        
    index = sst.index(A[i]) - 1
    if index != -1:
        R[i] += 1
        sst.discard(sst[index])
        
result = max(L[n - 1], R[0])
for i in range(1, n):
    result = max(result, L[i - 1] + R[i])
    
print(result)
```

```python
from collections import deque
import io, os, sys
input = io.BytesIO(os.read(0, os.fstat(0).st_size)).readline

# https://github.com/tatyam-prime/SortedSet/blob/main/SortedMultiset.py
import math
from bisect import bisect_left, bisect_right, insort
from typing import Generic, Iterable, Iterator, TypeVar, Union, List
T = TypeVar('T')

class SortedMultiset(Generic[T]):
    BUCKET_RATIO = 50
    REBUILD_RATIO = 170

    def _build(self, a=None) -> None:
        "Evenly divide `a` into buckets."
        if a is None: a = list(self)
        size = self.size = len(a)
        bucket_size = int(math.ceil(math.sqrt(size / self.BUCKET_RATIO)))
        self.a = [a[size * i // bucket_size : size * (i + 1) // bucket_size] for i in range(bucket_size)]
    
    def __init__(self, a: Iterable[T] = []) -> None:
        "Make a new SortedMultiset from iterable. / O(N) if sorted / O(N log N)"
        a = list(a)
        if not all(a[i] <= a[i + 1] for i in range(len(a) - 1)):
            a = sorted(a)
        self._build(a)

    def __iter__(self) -> Iterator[T]:
        for i in self.a:
            for j in i: yield j

    def __reversed__(self) -> Iterator[T]:
        for i in reversed(self.a):
            for j in reversed(i): yield j
    
    def __len__(self) -> int:
        return self.size
    
    def __repr__(self) -> str:
        return "SortedMultiset" + str(self.a)
    
    def __str__(self) -> str:
        s = str(list(self))
        return "{" + s[1 : len(s) - 1] + "}"

    def _find_bucket(self, x: T) -> List[T]:
        "Find the bucket which should contain x. self must not be empty."
        for a in self.a:
            if x <= a[-1]: return a
        return a

    def __contains__(self, x: T) -> bool:
        if self.size == 0: return False
        a = self._find_bucket(x)
        i = bisect_left(a, x)
        return i != len(a) and a[i] == x

    def count(self, x: T) -> int:
        "Count the number of x."
        return self.index_right(x) - self.index(x)

    def add(self, x: T) -> None:
        "Add an element. / O(√N)"
        if self.size == 0:
            self.a = [[x]]
            self.size = 1
            return
        a = self._find_bucket(x)
        insort(a, x)
        self.size += 1
        if len(a) > len(self.a) * self.REBUILD_RATIO:
            self._build()

    def discard(self, x: T) -> bool:
        "Remove an element and return True if removed. / O(√N)"
        if self.size == 0: return False
        a = self._find_bucket(x)
        i = bisect_left(a, x)
        if i == len(a) or a[i] != x: return False
        a.pop(i)
        self.size -= 1
        if len(a) == 0: self._build()
        return True

    def lt(self, x: T) -> Union[T, None]:
        "Find the largest element < x, or None if it doesn't exist."
        for a in reversed(self.a):
            if a[0] < x:
                return a[bisect_left(a, x) - 1]

    def le(self, x: T) -> Union[T, None]:
        "Find the largest element <= x, or None if it doesn't exist."
        for a in reversed(self.a):
            if a[0] <= x:
                return a[bisect_right(a, x) - 1]

    def gt(self, x: T) -> Union[T, None]:
        "Find the smallest element > x, or None if it doesn't exist."
        for a in self.a:
            if a[-1] > x:
                return a[bisect_right(a, x)]

    def ge(self, x: T) -> Union[T, None]:
        "Find the smallest element >= x, or None if it doesn't exist."
        for a in self.a:
            if a[-1] >= x:
                return a[bisect_left(a, x)]
    
    def __getitem__(self, x: int) -> T:
        "Return the x-th element, or IndexError if it doesn't exist."
        if x < 0: x += self.size
        if x < 0: raise IndexError
        for a in self.a:
            if x < len(a): return a[x]
            x -= len(a)
        raise IndexError

    def index(self, x: T) -> int:
        "Count the number of elements < x."
        ans = 0
        for a in self.a:
            if a[-1] >= x:
                return ans + bisect_left(a, x)
            ans += len(a)
        return ans

    def index_right(self, x: T) -> int:
        "Count the number of elements <= x."
        ans = 0
        for a in self.a:
            if a[-1] > x:
                return ans + bisect_right(a, x)
            ans += len(a)
        return ans

n, k = map(int, input().split())
li = [list(map(int, input().split())) for _ in range(n)]

li.sort(key=lambda x: x[1])

r = 0
cur_end = -1
ms = SortedMultiset()
for s, e in li:
    if s <= cur_end:
        continue

    index = ms.index(s) - 1
    if index >= 0:
        ms.discard(ms[index])
        ms.add(e)
    else:
        if len(ms) < k - 1:
            ms.add(e)
        else:
            r += 1
            cur_end = e
            
print(r)
```

```python
n = int(input())
A = input()
B = input()

same_as_firsts = []
first = B[0]
for i in range(n):
    if A[i] == first:
        same_as_firsts.append(i)
        
start = 0
end = len(same_as_firsts) - 1
possible_rightest = -1
while start <= end:
    mid = (start + end) // 2
    
    ip = True
    stack = []
    i = same_as_firsts[mid]
    for ai in range(i):
        stack.append(A[ai])
    
    ai = i + 1
    bi = 1
    if bi == n:
        possible_rightest = i
        start = mid + 1
    else:
        while True:
            if ai >= n:
                if stack[-1] != B[bi]:
                    ip = False
                    break
                else:
                    stack.pop()
                    bi += 1
                    if bi == n:
                        break
            else:
                if stack and stack[-1] == B[bi]:
                    stack.pop()
                    bi += 1
                    if bi == n:
                        break
                elif B[bi] == A[ai]:
                    ai += 1
                    bi += 1
                    if bi == n:
                        break
                else:
                    stack.append(A[ai])
                    ai += 1

        if ip:
            possible_rightest = mid
            start = mid + 1
        else:
            end = mid - 1
            
if possible_rightest == -1:
    print(0)
else:
    result = 0
    for i in range(possible_rightest + 1):
        result += same_as_firsts[i] + 1
        
    print(result)
```

```python
from collections import deque
import io, os, sys
input = io.BytesIO(os.read(0, os.fstat(0).st_size)).readline

def rorate(li):
    new_li = [deque([0] * n) for _ in range(n)]
    
    for i in range(n):
        for j in range(n):
            new_li[i][j] = li[n - 1 - j][i]
            
    return new_li 

n = int(input())
li = [deque(map(int, input().split())) for _ in range(n)]
q = int(input())

for _ in range(q):
    tp, *qu = map(int, input().split())
    
    if tp == 1:
        i = qu[0] - 1
        
        li[i].rotate()
    else:
        li = rorate(li)
        
for v in li:
    print(*v)
```

https://www.acmicpc.net/contest/problem/985/5 E번 - Every? Only One's Marble

```python
import sys
input = lambda: sys.stdin.readline().rstrip()

def reflect_move(move_c):
    global bought_c, island_remained, cur_use, cur_loc, cur_g_i, money, social_money, ip

    cur_loc += move_c
    while cur_loc >= len(board):
        cur_loc -= len(board)
        money += to_earn

    if board[cur_loc] == 0:
        pass
    elif type(board[cur_loc]) == int:
        if board[cur_loc] <= money:
            money -= board[cur_loc]
            board[cur_loc] = 0
            bought_c += 1
    elif board[cur_loc] == 's':
        pass
    elif board[cur_loc] == 'i':
        island_remained = 3
        pass
    elif board[cur_loc] == 'm':
        money += social_money
        social_money = 0
    elif board[cur_loc] == 't':
        cur_loc = 0
        money += to_earn
    elif board[cur_loc] == 'g':
        tp, x = golds[cur_g_i]
        cur_g_i += 1
        if cur_g_i == len(golds):
            cur_g_i = 0

        if tp == 1:
            money += x
        elif tp == 2:
            money -= x
            if money < 0:
                ip = False
                pass
        elif tp == 3:
            money -= x
            if money < 0:
                ip = False
                pass
            social_money += x
        elif tp == 4:
            reflect_move(x)

n, money, to_earn, g = map(int, input().split())
golds = [list(map(int, input().split())) for _ in range(g)]
normals = [input().split() for _ in range(4 * n - 8)]

board = [0] * (4 * n - 4)
cur_c = 0

# 's': 시작 칸, 'i': 무인도 칸, 'm': 사회복지기금 칸, 't': 우주여행 칸, 'g': 황금열쇠 칸
term = n - 2
board[0] = 's'
board[0 + term + 1] = 'i'
board[0 + (term + 1) * 2] = 'm'
board[0 + (term + 1) * 3] = 't'

cur_nornal_i = 0
to_buy_c = 0
for i in range(1, 4 * n - 4):
    if board[i] == 0:
        tp, *qu = normals[cur_nornal_i]
        
        if tp == 'G':
            board[i] = 'g'
        else:
            to_buy_c += 1
            cost = int(qu[0])
            board[i] = cost
            
        cur_nornal_i += 1
    
total_use = int(input())

bought_c = 0
island_remained = 0
cur_use = 0
cur_loc = 0
cur_g_i = 0
social_money = 0
ip = True
while True:
    a, b = map(int, input().split())
    cur_use += 1
    
    if island_remained:
        if a != b:
            island_remained -= 1
        else:
            island_remained = 0
    else:
        sv = a + b
        reflect_move(sv)
    
    if cur_use == total_use:
        break
        
if not ip:
    print('LOSE')
else:
    if bought_c == to_buy_c:
        print('WIN')
    else:
        print('LOSE')
```

https://www.acmicpc.net/problem/8987 수족관 3

```python
from math import log2, ceil
import io, os, sys
input = io.BytesIO(os.read(0, os.fstat(0).st_size)).readline

inf = float('inf')

## PYRIVAL BOOTSTRAP
# https://github.com/cheran-senthil/PyRival/blob/master/pyrival/misc/bootstrap.py
# This decorator allows for recursion without actually doing recursion
## @bootstrap, yield when getting and returning value in recursive functions, end of functions
from types import GeneratorType

def bootstrap(f, stack=[]):
    def wrappedfunc(*args, **kwargs):
        if stack:
            return f(*args, **kwargs)
        else:
            to = f(*args, **kwargs)
            while True:
                if type(to) is GeneratorType:
                    stack.append(to)
                    to = next(to)
                else:
                    stack.pop()
                    if not stack:
                        break
                    to = stack[-1].send(to)
            return to
    return wrappedfunc

class SegmentTree:
    def __init__(self, data, default=10**15, func=lambda a, b: max(a,b)):
        """initialize the segment tree with data"""
        self._default = default
        self._func = func
        self._len = len(data)
        self._size = _size = 1 << (self._len - 1).bit_length()
 
        self.data = [default] * (2 * _size)
        self.data[_size:_size + self._len] = data
        for i in reversed(range(_size)):
            self.data[i] = func(self.data[i + i], self.data[i + i + 1])
 
    def __delitem__(self, idx):
        self[idx] = self._default
 
    def __getitem__(self, idx):
        return self.data[idx + self._size]
 
    def __setitem__(self, idx, value):
        idx += self._size
        self.data[idx] = value
        idx >>= 1
        while idx:
            self.data[idx] = self._func(self.data[2 * idx], self.data[2 * idx + 1])
            idx >>= 1
 
    def __len__(self):
        return self._len
 
    def query(self, start, stop):
        if start == stop:
            return self.__getitem__(start)
        stop += 1
        start += self._size
        stop += self._size
 
        res = self._default
        while start < stop:
            if start & 1:
                res = self._func(res, self.data[start])
                start += 1
            if stop & 1:
                stop -= 1
                res = self._func(res, self.data[stop])
            start >>= 1
            stop >>= 1
        return res
 
    def __repr__(self):
        return "SegmentTree({0})".format(self.data)
    
def min_func(a, b):
    av, ai = a
    bv, bi = b
    
    if av <= bv:
        return [av, ai]
    else:
        return [bv, bi]
    
def max_func(a, b):
    av, ai = a
    bv, bi = b
    
    if av >= bv:
        return [av, ai]
    else:
        return [bv, bi]
    
@bootstrap
def recur(start, end, value):
    if start == end:
        yield
    
    left_r = value - 1
    if left_r >= start:
        left_v = seg_min.query(start, left_r)[1]
        c_tree[value][0] = left_v
        c_tree_p[left_v] = value
        yield recur(start, left_r, left_v)
    
    right_l = value + 1
    
    if right_l <= end:
        right_v = seg_min.query(right_l, end)[1]
        c_tree[value][1] = right_v
        c_tree_p[right_v] = value
        
        yield recur(right_l, end, right_v)
        
    yield
      
@bootstrap
def dfs(value, x_left, x_right, parent_h, sum_water):
    global vt_num
    
    cur_h = ys[value][0]
    cur_x_left, cur_x_right = xs[value]
    
    n_to_s[value][0] = vt_num
    s_to_n[vt_num] = value
    water = (x_right - x_left) * (cur_h - parent_h)
    sum_water += water
    
    waters[value] = water
    sum_waters[vt_num][0] = sum_water
    sum_waters[vt_num][1] = value
    
    left_v = c_tree[value][0]
    right_v = c_tree[value][1]
    
    if left_v != -1:
        vt_num += 1
        yield dfs(left_v, x_left, cur_x_left, cur_h, sum_water)
        
    if right_v != -1:
        vt_num += 1
        yield dfs(right_v, cur_x_right, x_right, cur_h, sum_water)
        
    n_to_s[value][1] = vt_num
    
    yield
    
def construct_max(node, s, e):
    if s == e:
        seg_max[node] = sum_waters[s]
        return seg_max[node]
    mid = (s + e) // 2
    
    t1 = construct_max(node * 2, s, mid)
    t2 = construct_max(node * 2 + 1, mid + 1, e)
    
    seg_max[node] = max_func(t1, t2)
    
    return seg_max[node]

def query_max(node, s, e, pending, qs, qe):
    seg_max[node][0] += pending
    pendings_max[node] += pending
    
    if e < qs or s > qe:
        return [-inf, -inf]
    if qs <= s <= e <= qe:
        return seg_max[node]
    else:
        pending = pendings_max[node]
        pendings_max[node] = 0
        
        mid = (s + e) // 2
        
        t1 = query_max(node * 2, s, mid, pending, qs, qe)
        t2 = query_max(node * 2 + 1, mid + 1, e, pending, qs, qe)
        
        return max_func(t1, t2)
    
def add_range_max(node, s, e, pending, ts, te, to_add):
    seg_max[node][0] += pending
    pendings_max[node] += pending
    
    if ts <= s <= e <= te:
        pendings_max[node] += to_add
        seg_max[node][0] += to_add
        
        return seg_max[node]
    elif e < ts or s > te:
        return seg_max[node]
    else:
        pending = pendings_max[node]
        pendings_max[node] = 0
        
        mid = (s + e) // 2
          
        t1 = add_range_max(node * 2, s, mid, pending, ts, te, to_add)
        t2 = add_range_max(node * 2 + 1, mid + 1, e, pending, ts, te, to_add)
        
        seg_max[node] = max_func(t1, t2)

        return seg_max[node]
    
n = int(input())

ys = []
xs = []

input()
for i in range(n // 2 - 1):
    x1, y = map(int, input().split())
    x2, y = map(int, input().split())
    
    xs.append((x1, x2))
    ys.append([y, i])
input()

k = int(input())
    
y_len = len(ys)
seg_min = SegmentTree(ys, [inf, inf], func=min_func)
c_tree = [[-1, -1] for _ in range(y_len)] # cartesian_tree
c_tree_p = [-1] * y_len # cartesian tree parent

root = seg_min.query(0, y_len - 1)[1]
recur(0, y_len - 1, root)

n_to_s = [[-1] * 2 for _ in range(y_len)] # node_num to seq_num
s_to_n = [-1] * y_len
vt_num = 0
waters = [-1] * y_len # node_num 기준
sum_waters = [[inf, inf] for _ in range(y_len)] # sum_waters[seq_num] = [sum_water, node_num]

dfs(root, 0, xs[-1][-1], 0, 0)

size = 2 ** (ceil(log2(y_len)) + 1) + 1

seg_max = [[-inf, -inf] for _ in range(size)]
pendings_max = [0] * size
construct_max(1, 0, y_len - 1)

result = 0
for _ in range(k):
    amount, index = query_max(1, 0, y_len - 1, 0, 0, y_len - 1)
    result += amount
    
    cur = index
    if waters[cur] >= 0:
        while True:
            start, end = n_to_s[cur]

            add_range_max(1, 0, y_len - 1, 0, start, end, -waters[cur])
            waters[cur] = -1

            cur = c_tree_p[cur]
            if cur == -1:
                break
            if waters[cur] == -1:
                break
            
print(result)
```

```python
#include <bits/stdc++.h>
using namespace std;

const int inf = 50000000;

int n_node_num;
vector<int> seg_node_num_to_num;
vector<vector<int>> nli;
vector<int> segtree;
vector<int> li;

int construct(int node, int start, int end) {
    n_node_num += 1;
    
    int cur_n_node_num = n_node_num;
    seg_node_num_to_num[node] = cur_n_node_num;
    li.push_back(0);
    
    for (int i = start; i <= end; i++) {
        nli[i].push_back(cur_n_node_num);
    }
    
    if (start == end) {
        segtree[node] = li[start];
        
        li[cur_n_node_num] = segtree[node];
        return segtree[node];
    }
    int mid = (start + end) / 2;
    
    int t1 = construct(node * 2, start, mid);
    int t2 = construct(node * 2 + 1, mid + 1, end);
    
    segtree[node] = max(t1, t2);
    
    li[cur_n_node_num] = segtree[node];
    return segtree[node];
}

void query(int node, int s, int e, int qs, int qe, int idx) {
    if (e < qs || s > qe) {
        return;
    }
    if (qs <= s && e <= qe) {
        int cur_node_num = seg_node_num_to_num[node];
        nli[cur_node_num].push_back(idx);
        return;
    } else {
        int mid = (s + e) / 2;
        
        query(node * 2, s, mid, qs, qe, idx);
        query(node * 2 + 1, mid + 1, e, qs, qe, idx);
    }

}

vector<int> ids, low_ids, sk;
vector<bool> on_stack;
vector<vector<int>> sccs;
int next_id;

int dfs(int value) {
    sk.push_back(value);
    on_stack[value] = true;
    int min_id = ids[value];
    
    for (int v : nli[value]) {
        if (ids[v] != 0) {
            if (!on_stack[v]){
                continue;
            }
            min_id = min(min_id, low_ids[v]);
        } else {
            ids[v] = next_id;
            low_ids[v] = next_id;
            next_id += 1;
            int temp = dfs(v);
            min_id = min(min_id, temp);
        }
    }
    low_ids[value] = min_id;
    
    if (ids[value] == low_ids[value]) {
        vector<int> scc;
        while (true) {
            int v = sk.back();
            sk.pop_back();
            scc.push_back(v);
            on_stack[v] = false;
            
            if (value == v) break;
        }
        sccs.push_back(scc);
    }
    
    return min_id;
}

int main() {
    ios_base::sync_with_stdio(false);
    cin.tie(nullptr);

    int n;
    cin >> n;
    li.resize(n);

    for (int i = 0; i < n; i++) {
        cin >> li[i];
    }
    vector<pair<int, int>> intervals(n);

    for (auto &interval : intervals) {
        int x, y;
        cin >> x >> y;
        interval = {x - 1, y - 1};
    }
    
    int size = pow(2, ceil(log2(n)) + 1) + 1;
    segtree.assign(size, -inf);
    n_node_num = n - 1;
    seg_node_num_to_num.assign(size, -1);
    nli.resize(n + size);
    construct(1, 0, n - 1);

    for (int i = 0; i < n; i++) {
        int intv_s, intv_e;
        tie(intv_s, intv_e) = intervals[i];
        query(1, 0, n - 1, intv_s, intv_e, i);
    }
    
    int nn = li.size();
    ids.assign(nn, 0);
    low_ids.assign(nn, 0);
    on_stack.assign(nn, false);
    next_id = 1;

    for (int i = 0; i < nn; i++) {
        if (ids[i] == 0) {
            ids[i] = next_id;
            low_ids[i] = next_id;
            next_id += 1;
            dfs(i);
        }
    }
    
    vector<int> scc_nums(nn, 0);
    for (int scc_i = 0; scc_i < sccs.size(); scc_i++) {
        for (int v : sccs[scc_i]) {
            scc_nums[v] = scc_i;
        }
    }
        
    vector<set<int>> affecting_sccs(sccs.size());
    for (int u = 0; u < nn; u++) {
        for (int v : nli[u]) {
            affecting_sccs[scc_nums[v]].insert(scc_nums[u]);
        }
    }

    vector<int> scc_result(sccs.size(), 0);
    vector<int> result(n, 0);
    for (int scc_i = sccs.size() - 1; scc_i >= 0; scc_i--) {
        for (int affecting_scc_i : affecting_sccs[scc_i]) {
            scc_result[scc_i] = max(scc_result[scc_i], scc_result[affecting_scc_i]);
        }

        if (sccs[scc_i].size() >= 2){
            for (int v : sccs[scc_i]) {
                scc_result[scc_i] = max(scc_result[scc_i], li[v]);
            }
        }

        int max_v = scc_result[scc_i];
        for (int v : sccs[scc_i]) {
            if (v < n) result[v] = max_v;
        }
    }
    
    for (int res : result) {
        cout << res << " ";
    }
    cout << endl;

    return 0;
}
```

https://www.acmicpc.net/contest/problem/1093/4 D번 - 마라탕 재료 고르기

```python
from itertools import combinations as comb

inf = float('inf')

n, k = map(int, input().split())
li = [list(map(int, input().split())) for _ in range(n)]

max_r = -inf
for case in comb(range(n), k):
    temp_r = 0
    for a in case:
        for b in case:
            if a == b:
                continue
            
            temp_r += li[a][b]
    
    temp_r //= 2
    max_r = max(max_r, temp_r)
    
print(max_r)
```

https://www.acmicpc.net/contest/problem/1093/5 E번 - 광기의 PS

```python
import io, os, sys
input = io.BytesIO(os.read(0, os.fstat(0).st_size)).readline

n, l = map(int, input().split())

result = 0
info = []
for _ in range(n):
    k, t = map(int, input().split())
    
    if t <= 5:
        result += t
    else:
        result += t
        
        info.append((k * t, 5 * k))
        
cur = 0
info.sort(key=lambda x: -x[1])
for needed, subtracted in info:
    if cur + needed >= l:
        result += cur + needed - l
        cur = l
    else:
        cur += needed
    
    cur -= subtracted
    
print(result)
```

https://www.acmicpc.net/contest/problem/1093/6 F번 - 누가 이길까

```python
from collections import deque

n, l = map(int, input().split())
a, b = map(int, input().split())
ds = list(map(int, input().split()))
hs = list(map(int, input().split()))

deq = deque([0] * (l + 1))
for i in range(n):
    needed = hs[i] // b
    if hs[i] % b:
        needed += 1
        
    deq[ds[i]] = max(deq[ds[i]], needed)
    
ip = True
c = 0
while True:
    can_move = True
    for i in range(1, a + 2):
        if i >= len(deq):
            break
        
        remained = i - deq[i]
        if deq[i] >= 1 and remained < 2:
            can_move = False
            break
    
    if can_move:
        c += 1
        deq.popleft()
        deq.popleft()
        deq.append(0)
        
        if len(deq) == 1:
            break
    else:
        for i in range(1, a + 1):
            if i >= len(deq):
                break
            remained = i - deq[i]
            if remained < 0:
                ip = False
                break
        
            if deq[i] >= 1:
                deq[i] -= 1
                
        c += 1
        deq.popleft()
        deq.append(0)
           
    if not ip:
        break
    
if not ip:
    print(-1)
else:
    print(c)
```

https://www.acmicpc.net/problem/10848 팔렘방의 다리

```python
# https://github.com/tatyam-prime/SortedSet/blob/main/SortedMultiset.py
import math
from bisect import bisect_left, bisect_right, insort
from typing import Generic, Iterable, Iterator, TypeVar, Union, List
T = TypeVar('T')

class SortedMultiset(Generic[T]):
    BUCKET_RATIO = 50
    REBUILD_RATIO = 170

    def _build(self, a=None) -> None:
        "Evenly divide `a` into buckets."
        if a is None: a = list(self)
        size = self.size = len(a)
        bucket_size = int(math.ceil(math.sqrt(size / self.BUCKET_RATIO)))
        self.a = [a[size * i // bucket_size : size * (i + 1) // bucket_size] for i in range(bucket_size)]
    
    def __init__(self, a: Iterable[T] = []) -> None:
        "Make a new SortedMultiset from iterable. / O(N) if sorted / O(N log N)"
        a = list(a)
        if not all(a[i] <= a[i + 1] for i in range(len(a) - 1)):
            a = sorted(a)
        self._build(a)

    def __iter__(self) -> Iterator[T]:
        for i in self.a:
            for j in i: yield j

    def __reversed__(self) -> Iterator[T]:
        for i in reversed(self.a):
            for j in reversed(i): yield j
    
    def __len__(self) -> int:
        return self.size
    
    def __repr__(self) -> str:
        return "SortedMultiset" + str(self.a)
    
    def __str__(self) -> str:
        s = str(list(self))
        return "{" + s[1 : len(s) - 1] + "}"

    def _find_bucket(self, x: T) -> List[T]:
        "Find the bucket which should contain x. self must not be empty."
        for a in self.a:
            if x <= a[-1]: return a
        return a

    def __contains__(self, x: T) -> bool:
        if self.size == 0: return False
        a = self._find_bucket(x)
        i = bisect_left(a, x)
        return i != len(a) and a[i] == x

    def count(self, x: T) -> int:
        "Count the number of x."
        return self.index_right(x) - self.index(x)

    def add(self, x: T) -> None:
        "Add an element. / O(√N)"
        if self.size == 0:
            self.a = [[x]]
            self.size = 1
            return
        a = self._find_bucket(x)
        insort(a, x)
        self.size += 1
        if len(a) > len(self.a) * self.REBUILD_RATIO:
            self._build()

    def discard(self, x: T) -> bool:
        "Remove an element and return True if removed. / O(√N)"
        if self.size == 0: return False
        a = self._find_bucket(x)
        i = bisect_left(a, x)
        if i == len(a) or a[i] != x: return False
        a.pop(i)
        self.size -= 1
        if len(a) == 0: self._build()
        return True

    def lt(self, x: T) -> Union[T, None]:
        "Find the largest element < x, or None if it doesn't exist."
        for a in reversed(self.a):
            if a[0] < x:
                return a[bisect_left(a, x) - 1]

    def le(self, x: T) -> Union[T, None]:
        "Find the largest element <= x, or None if it doesn't exist."
        for a in reversed(self.a):
            if a[0] <= x:
                return a[bisect_right(a, x) - 1]

    def gt(self, x: T) -> Union[T, None]:
        "Find the smallest element > x, or None if it doesn't exist."
        for a in self.a:
            if a[-1] > x:
                return a[bisect_right(a, x)]

    def ge(self, x: T) -> Union[T, None]:
        "Find the smallest element >= x, or None if it doesn't exist."
        for a in self.a:
            if a[-1] >= x:
                return a[bisect_left(a, x)]
    
    def __getitem__(self, x: int) -> T:
        "Return the x-th element, or IndexError if it doesn't exist."
        if x < 0: x += self.size
        if x < 0: raise IndexError
        for a in self.a:
            if x < len(a): return a[x]
            x -= len(a)
        raise IndexError

    def index(self, x: T) -> int:
        "Count the number of elements < x."
        ans = 0
        for a in self.a:
            if a[-1] >= x:
                return ans + bisect_left(a, x)
            ans += len(a)
        return ans

    def index_right(self, x: T) -> int:
        "Count the number of elements <= x."
        ans = 0
        for a in self.a:
            if a[-1] > x:
                return ans + bisect_right(a, x)
            ans += len(a)
        return ans

k, n = map(int, input().split())

li = []
base = 0
for _ in range(n):
    p, s, q, t = input().split()
    s = int(s)
    t = int(t)
    
    if p == q:
        base += abs(s - t)
    else:
        li.append((s, t))
    
nn = len(li)
if k == 1:
    nli = []
    for v in li:
        nli.extend(v)
    
    nli.sort()
    result = base + nli[nn // 2]
else:
    li.sort(key=lambda x: (x[0] + x[1]))
    min_to_add = 0
        
    left_max_ms = SortedMultiset()
    left_min_ms = SortedMultiset()
    right_max_ms = SortedMultiset()
    right_min_ms = SortedMultiset()
    
    left_max_sum = 0
    left_min_sum = 0
    right_max_sum = 0
    right_min_sum = 0
    
    for i in range(nn):
        for v in li[i]:
            if not left_max_ms:
                left_max_ms.add(-v)
                left_max_sum += v
            else:  
                if v <= -left_max_ms[0]:
                    left_max_ms.add(-v)
                    left_max_sum += v
                else:
                    left_min_ms.add(v)
                    left_min_sum += v

                if len(left_max_ms) >= len(left_min_ms) + 2:
                    w = -left_max_ms[0]
                    left_max_ms.discard(left_max_ms[0])
                    left_max_sum -= w

                    left_min_ms.add(w)
                    left_min_sum += w
                elif len(left_min_ms) >= len(left_max_ms) + 2:
                    w = left_min_ms[0]
                    left_min_ms.discard(left_min_ms[0])
                    left_min_sum -= w

                    left_max_ms.add(-w)
                    left_max_sum += w
                
    if len(left_max_ms) == len(left_min_ms):
        cri = -left_max_ms[0]
    elif len(left_max_ms) >= len(left_min_ms):
        cri = -left_max_ms[0]
    else:
        cri = left_min_ms[0]    
    
    to_add_left = (cri * len(left_max_ms) - left_max_sum) + (left_min_sum -  cri * len(left_min_ms))
    to_add_right = 0
    min_to_add = to_add_left + to_add_right
    
    for i in range(nn - 1, -1, -1):
        for v in li[i]:
            if v in left_min_ms:
                left_min_ms.discard(v)
                left_min_sum -= v
            else:
                left_max_ms.discard(-v)
                left_max_sum -= v
                
            if len(left_max_ms) >= len(left_min_ms) + 2:
                w = -left_max_ms[0]
                left_max_ms.discard(left_max_ms[0])
                left_max_sum -= w
                
                left_min_ms.add(w)
                left_min_sum += w
            elif len(left_min_ms) >= len(left_max_ms) + 2:
                w = left_min_ms[0]
                left_min_ms.discard(left_min_ms[0])
                left_min_sum -= w
                
                left_max_ms.add(-w)
                left_max_sum += w            

            if not right_max_ms:
                right_max_ms.add(-v)
                right_max_sum += v
            else:  
                if v <= -right_max_ms[0]:
                    right_max_ms.add(-v)
                    right_max_sum += v
                else:
                    right_min_ms.add(v)
                    right_min_sum += v

                if len(right_max_ms) >= len(right_min_ms) + 2:
                    w = -right_max_ms[0]
                    right_max_ms.discard(right_max_ms[0])
                    right_max_sum -= w

                    right_min_ms.add(w)
                    right_min_sum += w
                elif len(right_min_ms) >= len(right_max_ms) + 2:
                    w = right_min_ms[0]
                    right_min_ms.discard(right_min_ms[0])
                    right_min_sum -= w

                    right_max_ms.add(-w)
                    right_max_sum += w 
        
        if left_max_ms or left_min_ms:
            if len(left_max_ms) == len(left_min_ms):
                cri = -left_max_ms[0]
            elif len(left_max_ms) >= len(left_min_ms):
                cri = -left_max_ms[0]
            else:
                cri = left_min_ms[0]    

            to_add_left = (cri * len(left_max_ms) - left_max_sum) + (left_min_sum -  cri * len(left_min_ms)) 
        else:
            to_add_left = 0
        
        if len(right_max_ms) == len(right_min_ms):
            cri = -right_max_ms[0]
        elif len(right_max_ms) >= len(right_min_ms):
            cri = -right_max_ms[0]
        else:
            cri = right_min_ms[0]    

        to_add_right = (cri * len(right_max_ms) - right_max_sum) + (right_min_sum -  cri * len(right_min_ms)) 
        
        min_to_add = min(min_to_add, to_add_left + to_add_right)
        
    result = base + min_to_add
    
print(result)
```
