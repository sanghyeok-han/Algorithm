### 탈옥 9376 (신규)

```python
# 방향이 잘못되었음

from collections import deque

# h, w = map(int, input().split())
# li = [list(input()) for _ in range(h)]

n, m = 5, 9
s = '''****#****
*..#.#..*
****.****
*$#.#.#$*
*********'''
li = [['.'] + list(v) + ['.'] for v in s.split('\n')]

li.insert(0, ['.'] * (w + 2))
li.append(['.'] * (w + 2))

num = 1
pc = [] # prisoners coord
for i in range(n + 2):
    for j in range(m + 2):
        if li[i][j] == '$':
            pc.append((i, j))
            li[i][j] = 'p' + str(num)
            num += 1
        elif li[i][j] == '.':
            li[i][j] = 0
        elif li[i][j] == '#':
            li[i][j] = 1
            
# broken walls, bw[x][y][is_p1_visited][is_p2_visited] = 부순 벽의 수
bw = [[[[float('inf')] * 2 for _ in range(2)] for _ in range(m + 2)] for _ in range(n + 2)]
bw[0][0][0][0] = 0

dx = [-1, 1, 0, 0]
dy = [0, 0, -1, 1]

queue = deque()
queue.append((0, 0, 0, 0, set())) # x, y, is_p1_visited, is_p2_visited, broken_walls
while queue:
    x, y, ip1v, ip2v = queue.popleft()
    bwc = bw[x][y][ip1v][ip2v] # broken wall count
    
    for i in range(4):
        nx, ny = x + dx[i], y + dy[i]
        
        if nx < 0 or nx >= n or ny < 0 or ny >= m:
            continue
        if li[nx][ny] == '*':
            continue
            
        nip1v, nip2v = ip1v, ip2v
        nbwc = bwc
        
        if li[nx][ny] == 'p1':
            nip1v = 1
        elif li[nx][ny] == 'p2':
            nip2v = 1
        else:
            nbwc += li[nx][ny]
        
        if nbwc < bw[nx][ny][nip1v][nip2v]:
            bw[nx][ny][nip1v][nip2v] = nbwc
            queue.append((nx, ny, nip1v, nip2v))
            
mr = float('inf') # min_result
for x, y in pc:
    mr = min(mr, bw[x][y][1][1])
    
print(mr)
```

### 벽 부수고 이동하기 4 16946

```python
from collections import deque
import sys
# input = lambda: sys.stdin.readline().rstrip()

n, m = map(int, input().split())
li = [list(map(int, input())) for _ in range(n)]

dx = [-1, 1, 0, 0]
dy = [0, 0, -1 ,1]

vt = [[False] * m for _ in range(n)]

for i in range(n):
    for j in range(m):
        if li[i][j] >= 1:
            continue
        if vt[i][j]:
            continue
        vt[i][j] = True
        
        ws = set() # walls
        c = 1 # count
        
        queue = deque()
        queue.append((i, j))
        
        while queue:
            x, y = queue.popleft()
            
            for k in range(4):
                nx, ny = x + dx[k], y + dy[k]
                if nx < 0 or nx >= n or ny < 0 or ny >= m:
                    continue
                
                if li[nx][ny]:
                    ws.add((nx, ny))
                else:
                    if vt[nx][ny]:
                        continue
                    vt[nx][ny] = True
                    
                    c += 1
                    queue.append((nx, ny))

        for wx, wy in ws:
            li[wx][wy] += c
            
get_value = lambda x: str(x % 10)

print('\n'.join(''.join(map(get_value, v)) for v in li))
```

### 가장 많은 글자 1371

```python
from collections import Counter

s = ''
while True:
    try:
        s += input()
    except:
        break

c = Counter(s)
del c[' ']
c = c.most_common()
max_count = c[0][1]
result = []
for v in c:
    if v[1] == max_count:
        result.append(v[0])
        
print(''.join(sorted(result)))
```

### 공통 순열 1622

```python
from collections import Counter

while True:
    try:
        a = input()
        b = input()
    except:
        break

    ca = Counter(a)
    cb = Counter(b)
    c = ca & cb

    alphas = sorted(c)
    result = ''
    for v in alphas:
        result += v * c[v]

    print(result)
```
