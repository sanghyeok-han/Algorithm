```python
import io, os, sys
input = io.BytesIO(os.read(0, os.fstat(0).st_size)).readline

for _ in range(int(input())):
    n = int(input())

    c = 0
    for i in range(n):
        a, b = map(int, input().split())

        if a > b:
            c += 1

    print(c)
```

```python
import sys
input = lambda: sys.stdin.readline().rstrip()

for _ in range(int(input())):
    li = [input() for _ in range(3)]

    win = 'DRAW'
    for i in range(3):
        if len(set(li[i])) == 1:
            if li[i][0] != '.':
                win = li[i][0]

    for j in range(3):
        st = set()
        for i in range(3):
            st.add(li[i][j])

        if len(st) == 1:
            v = st.pop()
            if v != '.':
                win = v

    if li[0][0] == li[1][1] == li[2][2]:
        if li[0][0] != '.':
            win = li[0][0]

    if li[0][2] == li[1][1] == li[2][0]:
        if li[0][2] != '.':
            win = li[1][1]

    print(win)
```

```python
import io, os, sys
input = io.BytesIO(os.read(0, os.fstat(0).st_size)).readline
print = sys.stdout.write


for _ in range(int(input())):
    n, m, h = map(int, input().split())
    li = [sorted(map(int, input().split())) for _ in range(n)]

    rs = [] # (point, penalty)
    for i in range(n):
        cumul_t = 0
        penalty = 0
        point = 0
        for v in li[i]:
            cumul_t += v
            if cumul_t > h:
                break

            penalty += cumul_t
            point += 1

        rs.append((point, penalty))

    cri = rs[0]
    rs.sort(key=lambda x: (-x[0], x[1]))

    result = 0
    for i in range(n):
        if rs[i] == cri:
            result = i + 1
            break

    print(str(result) + '\n')
```

```python
import io, os, sys
input = io.BytesIO(os.read(0, os.fstat(0).st_size)).readline

for _ in range(int(input())):
    n, d, h = map(int, input().split())
    li = list(map(int, input().split()))

    total = n * d * h / 2
    for i in range(1, n):
        diff = li[i] - li[i - 1]

        if diff < h:
            trimmed = h - diff

            total -= d * (trimmed / h) * trimmed / 2

    print(total)
```

```python
import io, os, sys
input = io.BytesIO(os.read(0, os.fstat(0).st_size)).readline
print = sys.stdout.write

st = set()
cri = int(1e6)

for k in range(2, 100000):
    base = 1 + k + k ** 2
    if base > cri:
        break
        
    st.add(base)
    
    cur = base
    w = k ** 3
    while True:
        cur += w
        w *= k
        
        if cur > cri:
            break
        st.add(cur)
        
for _ in range(int(input())):
    n = int(input())
    
    if n in st:
        print('YES\n')
    else:
        print('NO\n')
```

```python
import sys
input = lambda: sys.stdin.readline().rstrip()

get_value = lambda x: int(x) - 1

for _ in range(int(input())):
    n = int(input())

    ct1 = [0] * 9
    li = list(map(get_value, input().split()))
    for v in li:
        ct1[v] += 1

    print('- 0')
    sys.stdout.flush()

    ct2 = [0] * 9
    li = list(map(get_value, input().split()))
    for v in li:
        ct2[v] += 1

    inc_num = -1
    for num in range(9):
        if ct1[num] < ct2[num]:
            inc_num = num
            break

    if inc_num != -1:
        to_dels = []
        for i in range(len(li)):
            if li[i] != inc_num:
                to_dels.append(i + 1)

        print('-', len(to_dels), *to_dels)
        sys.stdout.flush()

        while True:
            is_end = False
            li = list(map(get_value, input().split()))
            print('- 0')
            sys.stdout.flush()
        
            for i in range(len(li)):
                if li[i] != inc_num:
                    assert len(set(li)) == 1
                    is_end = True
                    result = i + 1
                    break

            if is_end:
                print('!', result)
                sys.stdout.flush()
                break
    else:
        print('- 0')
        sys.stdout.flush()

        ct3 = [0] * 9
        li = list(map(get_value, input().split()))
        for v in li:
            ct3[v] += 1    

        inc_num = -1
        for num in range(9):
            if ct2[num] < ct3[num]:
                inc_num = num
                break    

        to_dels = []
        for i in range(len(li)):
            if li[i] != inc_num:
                to_dels.append(i + 1)

        print('-', len(to_dels), *to_dels)
        sys.stdout.flush()

        while True:
            is_end = False
            li = list(map(get_value, input().split()))
            print('- 0')
            sys.stdout.flush()

            for i in range(len(li)):
                if li[i] != inc_num:
                    assert len(set(li)) == 1
                    is_end = True
                    result = i + 1
                    break

            if is_end:
                print('!', result)
                sys.stdout.flush()
                break
```
