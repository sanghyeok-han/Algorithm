```python
import sys
# input = lambda: sys.stdin.readline().rstrip()

for _ in range(int(input())):
    n = int(input())

    rc = 0
    result = []
    if n % 2 == 0:
        rc = n // 2
        start = 1
        for i in range(rc):
            result.append((start, start + 3 * n // 2 + 2))
            start += 3
    else:
        rc = n // 2 + 1
        start = 1
        for i in range(rc - 1):
            result.append((start, start + 3 * (n - 1) // 2 + 2))
            start += 3
        result.append((2, n * 3))

    print(rc)
    for v in result:
        print(*v)
```

```python
def recur(li):
    can_win = False
    
    if li[0] == 0:
        return False
    
    for i in range(1, n):
        li[0] -= 1
        li[0], li[i] = li[i], li[0]
        temp = recur(li)
        li[0], li[i] = li[i], li[0]
        li[0] += 1
        
        if not temp:
            can_win = True
            break
            
    return can_win
            
n = 4
for a1 in range(1, 6):
    for a2 in range(1, 6):
        for a3 in range(1, 6):
            for a4 in range(1, 6):
                print(recur([a1, a2, a3, a4]), a1, a2, a3, a4)
        
        
```

```python
홀수 길이 범위 경우는 모든 xor 결과가 0이 되어야 됨

짝수 길이 범위 경우는 모든 xor 결과가 0이 되어야 하고
xor 누적합 기준 홀수 자리에 0이 하나 더 있어야 함
```

```python
xor 누적합 만든 다음

홀수의 경우에는 해당 구간 끝과 시작 - 1과 같은 값이 있어야 함
또한 짝수 번째 인덱스 값 중에 시작 - 1과 같은 값이 있어야 함

홀수고 조건 충족하고 누적값 다 같으면 답은 0
아니면 답은 1

짝수고 조건 충족하고 누적값 다 같으면 답은 0
아니고 진짜 배열에서 맨 앞이나 맨 뒤가 0이면 답은 1
아니면 답은 2
```

```python
# 시간 초과
from collections import defaultdict as dd
from bisect import bisect_left
import sys
input = lambda: sys.stdin.readline().rstrip()

n, q = map(int, input().split())
li = list(map(int, input().split()))

sli = []
cur = 0
for v in li:
    cur ^= v
    sli.append(cur)
    
all_d = dd(list)
even_d = dd(list)
odd_d = dd(list)
for i in range(n):
    v = sli[i]
    
    all_d[str(v)].append(i)
    
    if i % 2 == 0:
        even_d[str(v)].append(i)
    else:
        odd_d[str(v)].append(i)

for _ in range(q):
    l, r = map(int, input().split())
    l -= 1
    r -= 1

    # odd case
    if (r - l) % 2 == 0:
        if l == 0:
            to_find = 0
        else:
            to_find = sli[l - 1]

        if sli[r] != to_find:
            print(-1)
        else:
            a = bisect_left(all_d[str(to_find)], l)
            b = bisect_left(all_d[str(to_find)], r + 0.1) - 1

            c = b - a + 1
            if c == r - l + 1:
                print(0)
            else:
                print(1)
    else:
        if l == 0:
            to_find = 0
        else:
            to_find = sli[l - 1]

        if sli[r] != to_find:
            print(-1)
        else:
            if l % 2 == 0:
                a = bisect_left(even_d[str(to_find)], l)
                b = bisect_left(even_d[str(to_find)], r + 0.1) - 1
            else:
                a = bisect_left(odd_d[str(to_find)], l)
                b = bisect_left(odd_d[str(to_find)], r + 0.1) - 1 

            c = b - a + 1
            if c <= 0:
                print(-1)
            else:
                a = bisect_left(all_d[str(to_find)], l)
                b = bisect_left(all_d[str(to_find)], r + 0.1) - 1

                c = b - a + 1
                if c == r - l + 1:
                    print(0)
                else:
                    if li[l] == 0 or li[r] == 0:
                        print(1)
                    else:
                        print(2)
```
