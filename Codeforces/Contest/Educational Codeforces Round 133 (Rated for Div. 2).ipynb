```python
mod = 998244353

n, ok = map(int, input().split())

total = [0] * (n + 1)
deltas = [0] * (n + 1)
deltas[0] = 1
former = [0] * (n + 1)

start = 0
nxt = ok
k = ok
while True:
    if start > n:
        break

    for x in range(start, n + 1):
        total[x] += deltas[x]
        total[x] %= mod
        if x + k <= n:
            deltas[x + k] += former[x] + deltas[x]
            deltas[x + k] %= mod

        former[x] = deltas[x]
        deltas[x] = 0
        
    start += nxt
    nxt += 1
    k += 1
    
print(*total[1:])
```
