```python
import sys, os, io
input = io.BytesIO(os.read(0,os.fstat(0).st_size)).readline

## PYRIVAL BOOTSTRAP
# https://github.com/cheran-senthil/PyRival/blob/master/pyrival/misc/bootstrap.py
# This decorator allows for recursion without actually doing recursion
## @bootstrap, yield when getting and returning value in recursive functions, end of functions
from types import GeneratorType

def bootstrap(f, stack=[]):
    def wrappedfunc(*args, **kwargs):
        if stack:
            return f(*args, **kwargs)
        else:
            to = f(*args, **kwargs)
            while True:
                if type(to) is GeneratorType:
                    stack.append(to)
                    to = next(to)
                else:
                    stack.pop()
                    if not stack:
                        break
                    to = stack[-1].send(to)
            return to
    return wrappedfunc

@bootstrap
def dfs(value, cs, cl):
    children_c = len(li[value])
    
    ncs = 0
    ncl = 0
    
    if children_c:
        ncs = cs // children_c
        ncl = cl // children_c
        if cl % children_c:
            ncl += 1
    
    svs = []
    for v in li[value]:
        yield dfs(v, ncs, ncl)
        svs.append(dp[v])
        
    svs.sort(key=lambda x: -(x[1] - x[0]))
    
    s_case = cs * scs[value]
    l_case = cl * scs[value]
    if children_c >= 1:
        can_use_upper = cs % children_c
        for i in range(can_use_upper):
            s_case += svs[i][1]
        for i in range(can_use_upper, children_c):
            s_case += svs[i][0]

        can_use_upper = cl % children_c
        if can_use_upper == 0:
            can_use_upper = children_c
        for i in range(can_use_upper):
            l_case += svs[i][1]
        for i in range(can_use_upper, children_c):
            l_case += svs[i][0]
    
    dp[value] = [s_case, l_case]
    
    yield

for _ in range(int(input())):
    n, k = map(int, input().split())
    ps = list(map(int, input().split()))
    scs = [0] + list(map(int, input().split()))

    dp = [[0] * 2 for _ in range(n + 1)]

    li = [[] for _ in range(n + 1)]
    for i in range(n - 1):
        p = ps[i]
        c = i + 2

        li[p].append(c)

    dfs(1, k, k)

    result = max(dp[1])

    print(result)
```

```python
import io, os, sys
input = io.BytesIO(os.read(0,os.fstat(0).st_size)).readline

def get_sum(num):
    return num * (num + 1) // 2

for _ in range(int(input())):
    n = int(input())
    li = list(map(int, input().split()))

    locs = [0] * n
    for i, v in enumerate(li):
        locs[v] = i

    total = 0
    last_num = 0
    cur_l = cur_r = locs[0]
    vt = {0}
    while last_num < n:
        loc = locs[last_num]

        if loc < cur_l:
            for i in range(cur_l - 1, loc - 1, -1):
                vt.add(li[i])
            cur_l = loc
        elif loc > cur_r:
            for i in range(cur_r + 1, loc + 1):
                vt.add(li[i])
            cur_r = loc

        while last_num + 1 in vt:
            last_num += 1

        rp = (last_num + 1) * 2 # range_possible

        cr = cur_r - cur_l + 1 # cur_range

        if cr > rp:
            last_num += 1
            continue

        remained_length = rp - cr
        remained_left = remained_right = 0
        if last_num == n - 1:
            remained_left = cur_l
            remained_right = n - 1 - cur_r
        else:
            nxt_loc = locs[last_num + 1]

            if nxt_loc < cur_l:
                remained_left = min(cur_l - nxt_loc - 1, remained_length)
                remained_right = min(n - 1 - cur_r, remained_length)
            else:
                remained_left = min(cur_l, remained_length)
                remained_right = min(nxt_loc - cur_r - 1, remained_length)

        cur_to_right = cr + remained_right
        all_covered_left = rp - cur_to_right
        all_covered_left = min(all_covered_left, remained_left)

        total += (all_covered_left + 1) * (remained_right + 1)

        part_covered_left = remained_left - all_covered_left

        part_covering_start_v = remained_right

        part_sum = get_sum(remained_right)
        to_minus = get_sum(part_covering_start_v - part_covered_left)
        part_sum -= to_minus

        total += part_sum

    #     print(last_num, remained_left, remained_right, (all_covered_left + 1) * (remained_right + 1), part_sum)

        last_num += 1

    print(total)
```

```python
import io, os, sys
input = io.BytesIO(os.read(0,os.fstat(0).st_size)).readline

inf = float('inf')

for _ in range(int(input())):
    l, r, x = map(int, input().split())
    a, b = map(int, input().split())

    if a == b:
        print(0)
    else:
        r1 = inf # start inc
        if a + x <= r:
            if a + x <= b:
                r1 = 1
            else:
                cur = r
                if b <= cur - x:
                    r1 = 2
                else:
                    cur = l
                    if cur + x <= b:
                        r1 = 3

        r2 = inf # start dec
        if l <= a - x:
            if b <= a - x:
                r2 = 1
            else:
                cur = l
                if cur + x <= b:
                    r2 = 2
                else:
                    cur = r
                    if b <= cur - x:
                        r2 = 3

        result = min(r1, r2)
        if result == inf:
            result = -1

        print(result)
```

https://codeforces.com/problemset/problem/1761/C C. Set Construction

```python
from collections import deque
import sys
input = lambda: sys.stdin.readline().rstrip()

for _ in range(int(input())):
    n = int(input())
    li = [input() for _ in range(n)]

    nli = [[] for _ in range(n)]
    pres = [[] for _ in range(n)]
    indegree = [0] * n
    for i in range(n):
        for j in range(n):
            if li[i][j] == '1':
                nli[i].append(j)
                pres[j].append(i)
                indegree[j] += 1

    result = [{i + 1} for i in range(n)]
    queue = deque()
    for i in range(n):
        if indegree[i] == 0:
            queue.append(i)

    while queue:
        value = queue.popleft()

        for pre in pres[value]:
            result[value] |= result[pre]

        for v in nli[value]:
            indegree[v] -= 1
            if indegree[v] == 0:
                queue.append(v)

    for i in range(n):
        print(len(result[i]), *result[i])
```

```python
import io, os, sys
input = io.BytesIO(os.read(0,os.fstat(0).st_size)).readline

def get_divisors(num):
    divisors = set()
    for mod in range(1, int(num ** 0.5) + 1):
        if num % mod == 0:
            divisors.add(mod)
            divisors.add(num // mod)
            
    return divisors

for _ in range(int(input())):
    a, b, c ,d = map(int, input().split())

    ab = a * b
    a_divs = get_divisors(a)
    b_divs = get_divisors(b)

    divs = set()
    for av in a_divs:
        for bv in b_divs:
            divs.add(av * bv)

    # gcd(x, ab)

    rx = ry = -1
    for g in divs:
        nab = ab // g

        needed = (b + 1) // nab
        if (b + 1) % nab:
            needed += 1

        y = needed * nab
        if y > d:
            continue

        needed2 = (a + 1) // g
        if (a + 1) % g:
            needed2 += 1

        x = needed2 * g
        if x > c:
            continue

        rx, ry = x, y
        break
    
    print(rx, ry)
```

```python
import io, os, sys
# input = io.BytesIO(os.read(0,os.fstat(0).st_size)).readline

for _ in range(int(input())):
    n = int(input())
    li = list(map(int, input().split()))

    if n == 2:
        t1 = sum(li)
        t2 = abs(li[0] - li[1]) * 2

        result = max(t1, t2)
    elif n == 3:
        max_v = max(li)

        if li[0] == max_v or li[2] == max_v:
            result = max_v * n
        else:
            min_v = min(li)
            
            result = sum(li)
            result = max(result, max(li[0], li[2]) * 3)
            result = max(result, (max_v - min_v) * n)
            result = max(result, (max_v - min_v) * 2 + max(li[0], li[2]))   
    else:
        result = max(li) * n
        
    print(result)
```

```python
from heapq import heappush, heappop
import io, os, sys
input = io.BytesIO(os.read(0,os.fstat(0).st_size)).readline

for _ in range(int(input())):
    n = int(input())
    li = list(map(int, input().split()))

    half = n // 2
    indices = [-1] * (n + 1)
    for i in range(half):
        indices[li[i]] = i

    st = set(li)

    li_ex = []
    for num in range(n, 0, -1):
        if num not in st:
            li_ex.append(num)

    heap = []
    result = [0] * (n + 1)

    ip = True
    cur_to_check = n
    for ex in li_ex:
        for num in range(cur_to_check, ex, -1):
            heappush(heap, -indices[num])

        if not heap:
            ip = False
            break

        index = -heappop(heap)

        result[li[index]] = ex

        cur_to_check = ex - 1

    if not ip:
        print(-1)
    else:
        nresult = []
        for v in li:
            nresult.append(result[v])
            nresult.append(v)

        print(*nresult)
```
