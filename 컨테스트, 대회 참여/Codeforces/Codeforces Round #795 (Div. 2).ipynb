```python
from collections import deque
import sys
# input = lambda: sys.stdin.readline().rstrip()

for _ in range(int(input())):
    n = int(input())
    li = list(map(int, input().split()))

    ip = True
    r = []
    fv = li[0]
    deq = deque([1])
    for i in range(1, n):
        if li[i] == fv:
            deq.append(i + 1)
        else:
            if len(deq) == 1:
                ip = False
                break
            else:
                deq.rotate()
                r.extend(deq)
                deq = deque()
                deq.append(i + 1)
                fv = li[i]

    if len(deq) == 1:
        ip = False
    deq.rotate()
    r.extend(deq)

    if not ip:
        print(-1)
    else:
        print(*r)
```

```python
import sys
# input = lambda: sys.stdin.readline().rstrip()

n = int(input())
li = list(map(int, input().split()))

ip = True
cur = -1
for i in range(n):
    if li[i] > 0:
        if cur == 1:
            ip = False
            break
        cur = 1
    elif li[i] <= -1:
        cur = -1

if ip:
    s = 0
    e = 0
    cur = li[e]
    cur_check = li[e]
    while s <= e <= n - 1:
        if cur <= 0:
            s = e + 1
            e += 1
            if e >= n:
                break
            cur_check = li[e]
            cur = li[e]
        else:
            e += 1
            if e >= n:
                break
            if cur > cur_check:
                ip = False
                break  

if ip:
    print('YES')
else:
    print('NO')
```
