```python
import sys
input = lambda: sys.stdin.readline().rstrip()

for _ in range(int(input())):
    n = int(input())
    li = list(map(int, input().split()))

    ip = False
    sv = sum(li)
    for i in range(n):
        if (sv - li[i]) % (n - 1) == 0:
            if li[i] == (sv - li[i]) // (n - 1):
                ip = True
                break

    if ip:
        print('YES')
    else:
        print('NO')
```

```python
import sys
input = lambda: sys.stdin.readline().rstrip()

for _ in range(int(input())):
    n = int(input())
    li = list(map(int, input().split()))

    used = [False] * n
    c = 0
    for i in range(1, n):
        if li[i - 1] > li[i] and not used[i - 1]:
            used[i - 1] = True
            used[i] = True
            c += 1

    print(c)
```

```python
from itertools import permutations as perm

li = [1, 2, 3, 4, 5]
n = len(li)

for case in perm(li):
    case = list(case)
    case.append(case[0])
    
    ip = True
    for i in range(1, n):
        if case[i - 1] < case[i] and case[i] > case[i + 1]:
            continue
        if case[i - 1] > case[i] and case[i] < case[i + 1]:
            continue
        ip = False
        break
        
    if ip:
        print(case[:n])
```

### 업솔빙

```python
# 컨테스트 때 해당 확인용 브루트포스 코드로
# 2~n번째 원소는 검증 가능하지만 첫번째 원소는 검증이 안된다는 것을 간과했었음
from itertools import permutations as perm

li = [1, 2, 3, 4, 5]
n = len(li)

for case in perm(li):
    case = list(case)
    case.append(case[0])
    case.append(case[1])

    ip = True
    for i in range(1, n + 1):
        if case[i - 1] < case[i] and case[i] > case[i + 1]:
            continue
        if case[i - 1] > case[i] and case[i] < case[i + 1]:
            continue
        ip = False
        break
        
    if ip:
        print(case[:n])
```

```python
from random import randint
import sys
# input = lambda: sys.stdin.readline().rstrip()
 
rd = randint(1, 1e9)
 
for _ in range(int(input())):
    n = int(input())
    li = sorted(map(int, input().split()))

    if n % 2 == 1:
        print('NO')
        continue
    
    d = dict()
    max_c = 1
    for v in li:
        w = v ^ rd
        if w in d:
            d[w] += 1
            max_c = max(max_c, d[w])
        else:
            d[w] = 1

    if max_c > n // 2:
        print('NO')
    else:
        r = []
        for i in range(n // 2):
            r.append(li[i + n // 2])
            r.append(li[i])

        r.append(r[0])
        ip = True
        for i in range(1, n):
            if r[i - 1] < r[i] and r[i] > r[i + 1]:
                continue
            if r[i - 1] > r[i] and r[i] < r[i + 1]:
                continue
            ip = False
            break

        if ip:
            print('YES')
            r.pop()
            print(*r)
            continue

        r = []
        for i in range(n // 2):
            r.append(li[i])
            r.append(li[i + n // 2])

        r.append(r[0])
        ip = True
        for i in range(1, n):
            if r[i - 1] < r[i] and r[i] > r[i + 1]:
                continue
            if r[i - 1] > r[i] and r[i] < r[i + 1]:
                continue
            ip = False
            break

        if ip:
            print('YES')
            r.pop()
            print(*r)
        else:
            print('NO')
```
