```python
a, b = map(int, input().split())

if a == b:
    print(a // 2)
elif a > b:
    total = 0
    
    start = 0
    end = b
    
    c = 0
    while start <= end:
        mid = (start + end) // 2
        
        na = a - mid * 3
        nb = b - mid
        
        if na >= nb:
            c = mid
            start = mid + 1
        else:
            end = mid - 1
            
    total += c
    a -= c * 3
    b -= c
    
    total += min(a // 2, b // 2)
    
    print(total)
elif a < b:
    total = 0

    start = 0
    end = a

    c = 0
    while start <= end:
        mid = (start + end) // 2

        na = a - mid
        nb = b - mid * 3

        if na <= nb:
            c = mid
            start = mid + 1
        else:
            end = mid - 1

    total += c
    a -= c
    b -= c * 3

    total += min(a // 2, b // 2)

    print(total)    
```

```python
from collections import deque

get_value = lambda x: int(x) - 1

# n = int(input())
# b = list(map(get_value, input().split()))
# p = deque(map(get_value, input().split()))

n = 5
b = [2, 0, 2, 2, 0]
p = deque([2, 0, 1, 4, 3])

li = [set() for _ in range(n)]
root = 0
for child_node in range(n):
    if b[child_node] == child_node:
        root = child_node
    else:
        li[b[child_node]].add(child_node)

ws = [0] * n
dists = [0] * n
dist_to_make = 1
available = set()

ip = True # is possible 

def dfs(value):
    global ip
    global available
    global dist_to_make
    
    if not ip:
        return
    
    available |= li[value]
    while available:
        if p[0] not in available:
            ip = False
            return
        else:
            v = p[0]
            p.popleft()
            new_dist = max(dist_to_make, dists[b[v]])
            dists[v] = new_dist
            dist_to_make = new_dist + 1
            ws[v] = new_dist - dists[b[v]]
            available.discard(v)
            dfs(v)
            
if root != p[0]:
    print(-1)
else:
    p.popleft()
    dfs(root)
    
    if not ip:
        print(-1)
    else:
        print(*ws)
```

```python
import os
import sys
from io import BytesIO, IOBase
import threading
threading.stack_size(2**26)
 

def main():
    global ip
    global available
    global dist_to_make
    global p
    
    from collections import deque

    get_value = lambda x: int(x) - 1

    def dfs(value):
        global ip
        global available
        global dist_to_make
        global p

        if not ip:
            return

        available |= li[value]
        while True:
            if not available:
                break

            if not ip:
                return
            if not p:
                ip = False
                return
            if p[0] not in available:
                ip = False
                return
            else:
                v = p[0]
                p.popleft()
                new_dist = max(dist_to_make, dists[b[v]])
                dists[v] = new_dist
                dist_to_make = new_dist + 1
                ws[v] = new_dist - dists[b[v]]
                available.discard(v)
                dfs(v)

    for _ in range(int(input())):
        n = int(input())
        b = list(map(get_value, input().split()))
        p = deque(map(get_value, input().split()))

        if n == 1:
            print(0)
        else:
            li = [set() for _ in range(n)]
            root = -1

            ip = True # is possible 
            for child_node in range(n):
                if b[child_node] == child_node:
                    if root == -1:
                        root = child_node
                    else:
                        ip = False
                else:
                    li[b[child_node]].add(child_node)

            if root == -1:
                ip = False

            if not ip:
                print(-1)
            else:
                ws = [0] * n
                dists = [0] * n
                dist_to_make = 1
                available = set()


                if root != p[0]:
                    print(-1)
                else:
                    p.popleft()
                    dfs(root)

                    if not ip:
                        print(-1)
                    else:
                        print(*ws)

    
    
    
    
    
    
    

# region fastio
 
BUFSIZE = 8192
 

class FastIO(IOBase):
    newlines = 0
 
    def __init__(self, file):
        self._fd = file.fileno()
        self.buffer = BytesIO()
        self.writable = "x" in file.mode or "r" not in file.mode
        self.write = self.buffer.write if self.writable else None
 
    def read(self):
        while True:
            b = os.read(self._fd, max(os.fstat(self._fd).st_size, BUFSIZE))
            if not b:
                break
            ptr = self.buffer.tell()
            self.buffer.seek(0, 2), self.buffer.write(b), self.buffer.seek(ptr)
        self.newlines = 0
        return self.buffer.read()
 
    def readline(self):
        while self.newlines == 0:
            b = os.read(self._fd, max(os.fstat(self._fd).st_size, BUFSIZE))
            self.newlines = b.count(b"\n") + (not b)
            ptr = self.buffer.tell()
            self.buffer.seek(0, 2), self.buffer.write(b), self.buffer.seek(ptr)
        self.newlines -= 1
        return self.buffer.readline()
 
    def flush(self):
        if self.writable:
            os.write(self._fd, self.buffer.getvalue())
            self.buffer.truncate(0), self.buffer.seek(0)
 
 
class IOWrapper(IOBase):
    def __init__(self, file):
        self.buffer = FastIO(file)
        self.flush = self.buffer.flush
        self.writable = self.buffer.writable
        self.write = lambda s: self.buffer.write(s.encode("ascii"))
        self.read = lambda: self.buffer.read().decode("ascii")
        self.readline = lambda: self.buffer.readline().decode("ascii")
 
 
sys.stdin, sys.stdout = IOWrapper(sys.stdin), IOWrapper(sys.stdout)
input = lambda: sys.stdin.readline().rstrip("\r\n")
 
# endregion
 
if __name__ == "__main__":
    ip = -1
    available = -1
    dist_to_make = -1
    p = -1
    
    threading.Thread(target=main).start()
```
