```python
def get_divisors(num):
    divisors = set()
    for mod in range(1, int(num ** 0.5) + 1):
        if num % mod == 0:
            divisors.add(mod)
            divisors.add(num // mod)
            
    return sorted(divisors)

n, k = map(int, input().split())
li = list(input())

for r in get_divisors(n):
    needed = 0
    for start in range(r):
        ct = dict()
        max_num = 1
        total_num = 0
        for i in range(start, n, r):
            total_num += 1
            if li[i] in ct:
                ct[li[i]] += 1
                max_num = max(max_num, ct[li[i]])
            else:
                ct[li[i]] = 1

        needed += total_num - max_num

    if needed <= k:
        print(r)
        break
```

```python
from heapq import heappush, heappop

n = int(input())
s = input()

ac = 0
is_r_found = False
cc = 0
li = []
for v in s:
    if v == 'A':
        if not is_r_found:
            ac += 1
        else:
            li.append(min(ac, cc))
            is_r_found = False
            cc = 0
            ac = 1
    elif v == 'R':
        if not is_r_found:
            is_r_found = True
        else:
            li.append(min(ac, cc))
            is_r_found = False
            ac = cc = 0
    else:
        if not is_r_found:
            ac = 0
        else:
            cc += 1
            
if cc:
    li.append(min(ac, cc))
    
c1 = 0
heap = []
for v in li:
    if v == 1:
        c1 += 1
    elif v > 1:
        heappush(heap, -v)
        
r = 0
while True:
    if not heap:
        if c1 == 0:
            break
        else:
            c1 -= 1
            r += 1
    else:
        v = -heappop(heap)
        v -= 1
        r += 1
        if v == 1:
            c1 += 1
        else:
            heappush(heap, -v)
            
    if c1 > 0:
        c1 -= 1
        r += 1
    else:
        if heap:
            heappop(heap)
            r += 1
        r += len(heap) * 2
        break
        
print(r)
```

```python
from itertools import permutations as perm
from bisect import bisect_left

n, x = 8, 7

max_lis = 0
rs = []
for case in perm(range(1, n + 1)):
    if case[0] != x:
        continue
    
    li = []
    for i in range(n - 1):
        li.append(abs(case[i + 1] - case[i]))
        
    nli = [li[0]]
    for i in range(1, len(li)):
        index = bisect_left(nli, li[i])

        if index == len(nli):
            nli.append(li[i])
        else:
            nli[index] = li[i]
    
    lis = len(nli)
    max_lis = max(max_lis, lis)
    rs.append([lis, case])
    
rs = [case for lis, case in rs if lis == max_lis]
for v in rs:
    temp = []
    for i in range(len(v) - 1):
        temp.append(v[i + 1] - v[i])
        
#     print(temp)

    print(v)
    
print(max_lis)
```
