```python
def check_area(tm):
    cri1 = 2 * 60
    if tm < 2 * 60:
        return 1
    if tm < 4 * 60:
        return 2
    if tm < 6 * 60:
        return 3
    if tm < 8 * 60:
        return 4
    if tm < 10 * 60:
        return 5
    return 6

s = input()
ws = [0] + list(map(int, input().split()))
areas = [False] * 7

h, m = map(int, s.split(':'))
tm = h * 60 + m
for _ in range(int(input())):
    q = input().split()
    
    status = q[1]
    
    if status == '^':
        areas[check_area(tm)] = True
    elif status == '10MIN':
        tm += 10
    elif status == '30MIN':
        tm += 30
    elif status == '50MIN':
        tm += 50
    elif status == '2HOUR':
        tm += 2 * 60
    elif status == '4HOUR':
        tm += 4 * 60
    else:
        tm += 9 * 60
        
    tm %= 12 * 60
    
total = 0
for i in range(1, 7):
    if not areas[i]:
        total += ws[i]
        
total = min(100, total)

print(total)
```

```python
from collections import defaultdict as dd
import sys
# input = lambda: sys.stdin.readline().rstrip()
       
atp = [set(), {'X'}, {'W'}, {'X', 'W'}, {'R'}, {'R', 'X'}, {'R', 'W'}, {'R', 'X', 'W'}]

u,f = map(int, input().split())

igs = dd(set) # included groups
for _ in range(u):
    temp = input().split()
    
    igs[temp[0]].add(temp[0])
    if len(temp) >= 2:
        groups = temp[1].split(',')
        
        for group in groups:
            igs[temp[0]].add(group)
    
fos = dd(set) # file owners
fogs = dd(set) # file owning groups
aus = dd(list)
    
for _ in range(f):
    fname, au, user, group = input().split()
    
    fos[fname].add(user)
    fogs[fname].add(group)
    
    au = list(map(int, au))
    au = [atp[au[0]], atp[au[1]], atp[au[2]]]
    
    if 'W' in au[2]:
        au[2].add('R')
    ### 만약 au[0] 에 'W'가 새로 있으면???
    
    for i in range(1, -1, -1):
        au[i] |= au[i + 1]
        
    aus[fname] = au.copy()
        
for _ in range(int(input())):
    user, fname, op = input().split()
    
    ip = False
    
    if op in aus[fname][2]:
        ip = True
    else:   
        if op in aus[fname][0]:
            if user in fos[fname]:
                ip = True
                
        if op in aus[fname][1]:
            for group in igs[user]:
                if group in fogs[fname]:
                    ip = True
                    break
                    
    if ip:
        print(1)
    else:
        print(0)
```

```python
# 열차 번호 끝자리가 홀수면 하행이고 오른쪽으로 이동하고
# 열차 번호 끝자리가 짝수면 상행이고 왼쪽으로 이동한다.
```

```python
import sys
input = lambda: sys.stdin.readline().rstrip()

# 0, 7, 14, 22, 32는 역
# 1~6, 8~13, 15~21, 23~31는 출발 전 확인할 구간
# 7, 14, 22, 32에 도착하면 1분 쉬어야 함

cc, hh = map(int, input().split())

starts = dict()
min_start = float('inf')
for _ in range(cc + hh):
    num, start = input().split()
    num = int(num)
    
    h, m = map(int, start.split(':'))
    
    start = h * 60 + m
    starts[num] = start
    
    min_start = min(min_start, start)
    
tm = min_start
li = [[] for _ in range(33)]
arrived = dict() # num: 도착 시간, 만약 arrived의 길이가 c + h와 같다면 종료함
ivs = [[0, 7], [7, 14], [14, 22], [22, 32]]
to_stop = {7, 14, 22}
while True:
    stopped = [[] for _ in range(33)]

    for i in range(33):
        for j in range(len(li[i]) - 1, -1, -1):
            if li[i][j][1] == 0:
                stopped[i].append(li[i][j][0])
                del li[i][j]
    
    for s, e in ivs:
        moved = False
        for i in range(s + 1, e):
            if li[i]:
                if li[i][0][0] % 2 == 1:
                    li[i + 1].append(li[i][0].copy())
                    if i + 1 in to_stop:
                        li[i + 1][-1][1] = 0
                    if i + 1 == 32:
                        arrived[li[i + 1][-1][0]] = tm
                else:
                    li[i - 1].append(li[i][0].copy())
                    if i - 1 in to_stop:
                        li[i - 1][-1][1] = 0
                    if i - 1 == 0:
                        arrived[li[i - 1][-1][0]] = tm
                moved = True
                del li[i][0]
                break
        if moved:
            continue
        
        cands = []
    
        c = 0
        for num, stat in li[s]:
            if num % 2 == 1 and stat == 1:
                cands.append([num, starts[num]])
                c += 1
   
        for num, stat in li[e]:
            if num % 2 == 0 and stat == 1:
                cands.append([num, starts[num]])
                c += 1
                
        if not cands:
            continue
                
        if c == 1 and cands[0][0] % 2 == 1:
            num = cands[0][0]
            for i in range(len(li[s])):
                if li[s][i][0] == num:
                    del li[s][i]
                    break
            li[s + 1].append([num, 1])
            moved = True
        if moved:
            continue
            
        if c == 1 and cands[0][0] % 2 == 0:
            num = cands[0][0]
            for i in range(len(li[e])):
                if li[e][i][0] == num:
                    del li[e][i]
                    break
            li[e - 1].append([num, 1])
            moved = True
        if moved:
            continue
            
        cands.sort(key=lambda x: (x[1], x[0]))
        target = cands[0][0]
        if target % 2 == 1:
            for i in range(len(li[s])):
                if li[s][i][0] == target:
                    del li[s][i]
                    break
            li[s + 1].append([target, 1])
        else:
            for i in range(len(li[e])):
                if li[e][i][0] == target:
                    del li[e][i]
                    break
            li[e - 1].append([target, 1])
            
            
    for i in range(33):
        for j in range(len(stopped[i])):
            li[i].append([stopped[i][j], 1])
            
    for num in starts:
        if starts[num] == tm:
            if num % 2 == 1:
                li[0].append([num, 1]) # 1: 다음 턴 움직이고자 함, 0: 다음 턴 그대로
            else:
                li[-1].append([num, 1])

    if len(arrived) == cc + hh:
        break
    
    tm += 1
    
arrived = list(arrived.items())
arrived.sort(key=lambda x: x[0])

for _, tm in arrived:
    tm %= (24 * 60)
    
    h = tm // 60
    m = tm % 60
    
    h = str(h).rjust(2, '0')
    m = str(m).rjust(2, '0')
    
    print(f'{h}:{m}')
```

```python
0
0   1
nkn mrk

노선별 역별로 환승역 여부 검사
환승역이 여러 개면 서로의 거리 계산

환승역끼리의 거리를 플로이드 워셜로 구해줌

쿼리 받고 만약 두 역이 공통으로 속한 노선이 있으면 해당 노선에서의 거리를 계산해줌

그렇지 않으면 해당 노선에 속한 환승역들까지의 거리를 계산함 [[환승역, 거리], [환승역, 거리]]
두 역의 위의 모든 조합을 보면서 최소 거리를 구함


역별 포함 노선 정보
노선별 역별 해당 노선에서의 번호 정보

노선별 포함 환승역 정보
환승역끼리의 거리 정보
```

```python
from collections import defaultdict as dd
import sys
# input = lambda: sys.stdin.readline().rstrip()

inf = float('inf')

n, q = map(int, input().split())

included_lines = dd(set)
lines = [dd(int) for _ in range(n)]

for line_num in range(n):
    li = input().split()
    c = int(li[0])
    for s_num in range(1, len(li)):
        s_name = li[s_num]
        included_lines[s_name].add(line_num)
        lines[line_num][s_name] = s_num
        
trans_by_line = [[] for _ in range(n)]
trans_name_to_num = dd(int)
cur_trans_num = 0
dist_info = [] # [from, to, dist]
for l_num in range(n):
    for s_name in lines[l_num]:
        if len(included_lines[s_name]) >= 2:
            trans_by_line[l_num].append(s_name)
            if s_name not in trans_name_to_num:
                trans_name_to_num[s_name] = cur_trans_num
                cur_trans_num += 1
    for i in range(len(trans_by_line[l_num])):
        for j in range(i + 1, len(trans_by_line[l_num])):
            a = trans_by_line[l_num][i]
            b = trans_by_line[l_num][j]
            dist = abs(lines[l_num][a] - lines[l_num][b])
            dist_info.append([a, b, dist])

dists = [[inf] * cur_trans_num for _ in range(cur_trans_num)]
for i in range(cur_trans_num):
    dists[i][i] = 0
    
for f, t, dist in dist_info:
    ff = trans_name_to_num[f]
    tt = trans_name_to_num[t]
    
    dists[ff][tt] = dist
    dists[tt][ff] = dist

for mid in range(cur_trans_num):
    for i in range(cur_trans_num):
        for j in range(cur_trans_num):
            dists[i][j] = min(dists[i][j], dists[i][mid] + dists[mid][j])
            
for _ in range(q):
    result = inf
    
    a, b = input().split()
    
    if a == b:
        result = 0
    
    st1 = included_lines[a]
    st2 = included_lines[b]
    
    st = st1 & st2
    while st:
        l_num = st.pop()
        dist = abs(lines[l_num][a] - lines[l_num][b])
        result = min(result, dist)
        
    a_trans = set()
    for l_num in st1:
        for trans_name in trans_by_line[l_num]:
            dist = abs(lines[l_num][a] - lines[l_num][trans_name])
            a_trans.add((trans_name_to_num[trans_name], dist))

    b_trans = set()
    for l_num in st2:
        for trans_name in trans_by_line[l_num]:
            dist = abs(lines[l_num][b] - lines[l_num][trans_name])
            b_trans.add((trans_name_to_num[trans_name], dist))

    a_trans = list(a_trans)
    b_trans = list(b_trans)

    for ati in range(len(a_trans)):
        for bti in range(len(b_trans)):
            dist = a_trans[ati][1] + b_trans[bti][1]
            dist += dists[a_trans[ati][0]][b_trans[bti][0]]
            result = min(result, dist)
                
    if result == inf:
        print(-1)
    else:
        print(result * 2)
```
