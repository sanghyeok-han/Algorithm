### RPG Extreme 17081

```python
전반적으로 기록해야 할 것:
    현재 턴수

초반에 기록해야 할 것:
    주인공 첫 시작 위치
    
전투 중 기록해야 할 것:
    첫 번째 맞교환인지 (Courage 착용시 공력력 2배, Hunter 착용시 보스 첫 공격에 데미지 0)
    
    
주인공 사망시: 몬스터 체력 초기화, RE 장신구 착용시 해당 장신구 사라지고
                                   주인공 첫 시작 위치로 복귀
```

```python
참고사항
전투가 시작되고, 캐릭터 혹은 몬스터가 죽을 때 까지의
모든 과정은 한 턴 내에 이루어진다.

공격력 상승 효과는 주인공의 현재 공격력 + 무기의 공격력이 최우선

Courage와 Dexterity 장신구 함께 착용시 공격력 2배 대신 3배 적용

보스 몬스터를 처치해서 게임이 끝날 경우에도 바로 게임이 끝나는 대신
적용할건 다 적용됨

주인공이 있는 칸은 그 칸에 무엇이 있든 @로 출력함
주인공이 사망하면 주인공 위치를 출력 안함

몬스터의 체력이 회복되는 경우가 있으므로 몬스터 정보의 사본으로 전투해야함

몬스터 퇴치 -> 경험치 반영 (Experience 장신구 고려)
```

```python
n, m = map(int, input().split())
li = [list(input()) for _ in range(n)]
comms = input()

item_c = 0
monster_c = 0
sx, sy = 0, 0
for i in range(n):
    for j in range(m):
        if li[i][j] == '&' or li[i][j] == 'M':
            monster_c += 1
        elif li[i][j] == 'B':
            item_c += 1
        elif li[i][j] == '@':
            sx, sy = i, j
            li[i][j] = '.'
x, y = sx, sy

d = dict()
for _ in range(monster_c):
    mx, my, ms, mw, ma, mh, me = input().split()
    mx = int(mx) - 1
    my = int(my) - 1
    mw = int(mw)
    ma = int(ma)
    mh = int(mh)
    me = int(me)
    
    d[(mx, my)] = [ms, mw, ma, mh, me]
    
for _ in range(item_c):
    ix, iy, itp, iinfo = input().split()
    ix = int(ix) - 1
    iy = int(iy) - 1
    if itp == 'W' or itp == 'A':
        iinfo = int(iinfo)
        
    d[(ix, iy)] = [itp, iinfo]

def move(x, y, com):
    nx, ny = x, y
    if com == 'L':
        ny -= 1
    elif com == 'R':
        ny += 1
    elif com == 'U':
        nx -= 1
    elif com == 'D':
        nx += 1
        
    if nx < 0 or nx >= n or ny < 0 or ny >= m or li[nx][ny] == '#':
        return x, y
    return nx, ny
    
def fight(x, y):
    global h
    
    if li[x][y] == 'M' and 'HU' in accs:
        h = max_h
    
    ms, mw, ma, mh, _ = d[(x, y)]
    first_ak_weight = 1
    if 'CO' in accs:
        first_ak_weight = 2
        if 'DX' in accs:
            first_ak_weight = 3
    
    turn = 1
    while True:
        if turn == 1:
            mh -= max(1, (w + weapon_w) * first_ak_weight - ma)
        else:
            mh -= max(1, (w + weapon_w) - ma)
            
        if mh <= 0:
            return True
        
        if turn == 1 and li[x][y] == 'M' and 'HU' in accs:
            pass
        else:
            h -= max(1, mw - (a + shield_a))
            
            if h <= 0:
                h = 0
                return False
        
        turn += 1
    
    
max_h, w, a = 20, 2, 2 # 최대 체력, 공격력 (무기 제외), 방어력
h = max_h
level = 1
exp = 0
ntnl = 5 * 1 # need to nxt level

accs = set() # accessories, 최대 4개, 중복 X 착용 가능
weapon_w = 0
shield_a = 0

result_words = ''
for i in range(len(comms)):
    x, y = move(x, y, comms[i])
#     print(i, comms[i], x, y, accs, h)
    
    if li[x][y] == '.':
        pass
    elif li[x][y] == 'B':
        li[x][y] = '.'
        itp, iinfo = d[(x, y)]
        if itp == 'W':
            weapon_w = iinfo
        elif itp == 'A':
            shield_a = iinfo
        elif itp == 'O':
            if len(accs) == 4 or iinfo in accs:
                continue
            else:
                accs.add(iinfo)
    elif li[x][y] == '^':
        if 'DX' in accs:
            h -= 1
        else:
            h -= 5
            
        if h <= 0:
            h = 0
            
            if 'RE' in accs:
                accs.remove('RE')
                
                h = max_h
                x, y = sx, sy
            else:
                result_words = 'YOU HAVE BEEN KILLED BY SPIKE TRAP..'
                break
            
    elif li[x][y] == '&' or li[x][y] == 'M':
        is_won = fight(x, y)
        
        if not is_won:
            if 'RE' in accs:
                accs.remove('RE')
                
                h = max_h
                x, y = sx, sy
            else:
                result_words = f'YOU HAVE BEEN KILLED BY {d[(x, y)][0]}..'
                break
        else:
            if 'HR' in accs:
                h += 3
                h = min(h, max_h)
            
            if 'EX' in accs:
                exp += int(1.2 * d[(x, y)][4])
            else:
                exp += d[(x, y)][4]
                
            if exp >= ntnl:
                level += 1
                max_h += 5
                w += 2
                a += 2
                h = max_h
                exp = 0
                ntnl = 5 * level
            if li[x][y] == 'M':
                result_words = 'YOU WIN!'
                break
                
            li[x][y] = '.'
else:
    result_words = 'Press any key to continue.'
    
last_turn = i + 1
if h <= 0:
    h = 0
else:
    li[x][y] = '@'

for v in li:
    print(''.join(v))
print(f'Passed Turns : {last_turn}')
print(f'LV : {level}')
print(f'HP : {h}/{max_h}')
print(f'ATT : {w}+{weapon_w}')
print(f'DEF : {a}+{shield_a}')
print(f'EXP : {exp}/{ntnl}')
print(result_words)
```

### 별 찍기 - 18 10993

```python
from collections import deque

def add_margin(temp):
    margin_len = (nxt - len(temp)) // 2
    temp = ' ' * margin_len + temp + ' ' * margin_len
    
    return temp

n = int(input())

if n == 1:
    print('*')
else:
    cur = 1
    deq = deque()
    deq.append('*')
    for cur_n in range(2, n + 1):
        ndeq = deque()
        nxt = cur * 2 + 3
        if cur_n % 2 == 0:
            for to_make in range(nxt, 0, -2):
                if to_make == nxt:
                    ndeq.appendleft('*' * to_make)
                    continue
                if deq:
                    temp = deq.popleft()
                    mid_side_len = (to_make - len(temp)) // 2
                    temp = '*' + ' ' * (mid_side_len - 1) + temp + ' ' * (mid_side_len - 1) + '*'
                    temp = add_margin(temp)
                    ndeq.append(temp)
                else:
                    mid_len = to_make - 2
                    if mid_len <= 0:
                        temp = '*'
                        temp = add_margin(temp)
                        ndeq.append(temp)
                    else:
                        temp = '*' + ' ' * mid_len + '*'
                        temp = add_margin(temp)
                        ndeq.append(temp)
        else:
            for to_make in range(nxt, 0, -2):
                if to_make == nxt:
                    ndeq.append('*' * to_make)
                    continue
                if deq:
                    temp = deq.pop()
                    mid_side_len = (to_make - len(temp)) // 2
                    temp = '*' + ' ' * (mid_side_len - 1) + temp + ' ' * (mid_side_len - 1) + '*'
                    temp = add_margin(temp)
                    ndeq.appendleft(temp)
                else:
                    mid_len = to_make - 2
                    if mid_len <= 0:
                        temp = '*'
                        temp = add_margin(temp)
                        ndeq.appendleft(temp)
                    else:
                        temp = '*' + ' ' * mid_len + '*'
                        temp = add_margin(temp)
                        ndeq.appendleft(temp)            

        deq = ndeq.copy()
        cur = nxt
        
    for v in deq:
        print(v.rstrip())
```
