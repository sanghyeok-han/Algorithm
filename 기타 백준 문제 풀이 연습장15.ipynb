https://www.acmicpc.net/problem/14310 Evaluation (Small) (다시 풀음)

```python
for tn in range(int(input())):
    n = int(input())

    d = {}

    to_dels = []
    for _ in range(n):
        a, b = input().split('=')

        s = b.index('(')
        e = b.index(')')

        temp = b[s + 1:e]
        if temp:
            d[a] = set(temp.split(','))
        else:
            to_dels.append(a)

    ip = True
    while True:
        n_to_dels = []
        for key in d:
            for v in to_dels:
                d[key].discard(v)

            if not d[key]:
                n_to_dels.append(key)

        if d and not n_to_dels:
            ip = False
            break

        for v in n_to_dels:
            del d[v]
        to_dels = n_to_dels.copy()

        if not d:
            break

    if ip:
        r = 'GOOD'
    else:
        r = 'BAD'
        
    print(f'Case #{tn + 1}: {r}')
```

```python
from heapq import heappush, heappop
import io, os, sys
input = io.BytesIO(os.read(0,os.fstat(0).st_size)).readline

inf = float('inf')

n, m, k = map(int, input().split())

li = [[] for _ in range(n + 1)]
for _ in range(m):
    a, b, t = map(int, input().split())
    
    li[a].append((b, t))
    li[b].append((a, t))

max_can_earn = [0] * (n + 1)
for _ in range(k):
    num, v = map(int, input().split())
    
    max_can_earn[num] = max(max_can_earn[num], v)
    
heap = []
    
dists = [[inf, inf] for _ in range(n + 1)]
dists[n][0] = 0
heappush(heap, (0, 0, n))
if max_can_earn[n] >= 1:
    dists[n - 1][1] = 0
    heappush(heap, (0, 1, n))
    
while heap:
    dist, ate, value = heappop(heap)
    if dist > dists[value][ate]:
        continue
    
    for v, d in li[value]:
        if not ate:
            ndist = dist + d
            if ndist < dists[v][0]:
                dists[v][0] = ndist
                heappush(heap, (ndist, 0, v))
            
            if max_can_earn[v] >= 1:
                ndist_a = ndist - max_can_earn[v]
                if ndist_a < dists[v][1]:
                    dists[v][1] = ndist_a
                    heappush(heap, (ndist_a, 1, v))
        else:
            ndist_a = dist + d
            if ndist_a < dists[v][1]:
                dists[v][1] = ndist_a
                heappush(heap, (ndist_a, 1, v))
                
for num in range(1, n):
    if dists[num][0] >= dists[num][1]:
        print(1)
    else:
        print(0)
```

https://www.acmicpc.net/problem/25512 트리를 간단하게 색칠하는 최소 비용

```python
from collections import deque
import io, os, sys
input = io.BytesIO(os.read(0,os.fstat(0).st_size)).readline

n = int(input())

li = [[] for _ in range(n)]
for _ in range(n - 1):
    p, c = map(int, input().split())
    li[p].append(c)
    
costs = [list(map(int, input().split())) for _ in range(n)]

vt = [False] * n
vt[0] = True

queue = deque()
queue.append((0, 0))

r1 = 0
while queue:
    value, color = queue.popleft()
    
    r1 += costs[value][color]
    for v in li[value]:
        if not vt[v]:
            vt[v] = True
            queue.append((v, color ^ 1))

vt = [False] * n
vt[0] = True

queue = deque()
queue.append((0, 1))
            
r2 = 0
while queue:
    value, color = queue.popleft()
    
    r2 += costs[value][color]
    for v in li[value]:
        if not vt[v]:
            vt[v] = True
            queue.append((v, color ^ 1))
            
r = min(r1, r2)

print(r)
```

```python
from collections import defaultdict as dd
import io, os, sys
input = io.BytesIO(os.read(0,os.fstat(0).st_size)).readline

while True:
    n, w = map(int, input().split())
    if n == w == 0:
        break
    
    li = [int(input()) for _ in range(n)]

    temp = dd(int)
    for v in li:
        temp[v // w] += 1

    max_h = max(temp.values())
    max_num = max(temp.keys())

    part = 1 / max_num
    cur = 0
    total = 0
    for i in range(max_num, -1, -1):
        h = temp[i]

        total += (h / max_h) * cur
        cur += part

    total += 0.01

    print(total)
```

```python
import io, os, sys
sys.setrecursionlimit(2000)
input = io.BytesIO(os.read(0,os.fstat(0).st_size)).readline

def find(v):
    if pa[v] != v:
        pa[v] = find(pa[v])
        
    return pa[v]

def union(a, b):
    if a > b:
        a, b = b, a
    pa[b] = a
    
n, m, q = map(int, input().split())

pa = list(range(n + 1))
for _ in range(m):
    a, b = map(int, input().split())
    
    ra = find(a)
    rb = find(b)
    union(ra, rb)
    
for _ in range(q):
    a, b = map(int, input().split())
    
    if find(a) != find(b):
        print('N')
    else:
        print('Y')
```

```python
from heapq import heappush, heappop
from collections import deque
import io, os, sys
input = io.BytesIO(os.read(0,os.fstat(0).st_size)).readline

inf = float('inf')

dx = [-1, 1, 0, 0]
dy = [0, 0, -1, 1]

n, m = map(int, input().split())
k = int(input())

heap = []
li = [[0] * m for _ in range(n)]
for _ in range(k):
    r, c, d = map(int, input().split())
    d += 1
    r -= 1
    c -= 1
    li[r][c] = d
    heappush(heap, (-d, r * m + c))
    
while heap:
    influ, rdd = heappop(heap)
    influ = -influ
    
    ninflu = influ - 1
    
    y = rdd % m
    x = rdd // m
    
    for d in range(4):
        nx = x + dx[d]
        ny = y + dy[d]
        
        if nx < 0 or nx >= n or ny < 0 or ny >= m:
            continue
        
        if li[nx][ny] < ninflu:
            li[nx][ny] = ninflu
            heappush(heap, (-ninflu, nx * m + ny))
            
dists = [[inf] * m for _ in range(n)]
dists[0][0] = 0

queue = deque()
queue.append((0, 0))
while queue:
    x, y = queue.popleft()
    
    for d in range(4):
        nx = x + dx[d]
        ny = y + dy[d]
        
        if nx < 0 or nx >= n or ny < 0 or ny >= m:
            continue
        if li[nx][ny] >= 1:
            continue
            
        if dists[nx][ny] == inf:
            dists[nx][ny] = dists[x][y] + 1
            queue.append((nx, ny))
            
if dists[n - 1][m - 1] == inf:
    print('NO')
else:
    print('YES')
    print(dists[n - 1][m - 1])
```

```python
from collections import deque
import io, os, sys
input = io.BytesIO(os.read(0,os.fstat(0).st_size)).readline

# https://github.com/tatyam-prime/SortedSet/blob/main/SortedSet.py
import math
from bisect import bisect_left, bisect_right
from typing import Generic, Iterable, Iterator, TypeVar, Union, List
T = TypeVar('T')

class SortedSet(Generic[T]):
    BUCKET_RATIO = 50
    REBUILD_RATIO = 170

    def _build(self, a=None) -> None:
        "Evenly divide `a` into buckets."
        if a is None: a = list(self)
        size = self.size = len(a)
        bucket_size = int(math.ceil(math.sqrt(size / self.BUCKET_RATIO)))
        self.a = [a[size * i // bucket_size : size * (i + 1) // bucket_size] for i in range(bucket_size)]
    
    def __init__(self, a: Iterable[T] = []) -> None:
        "Make a new SortedSet from iterable. / O(N) if sorted and unique / O(N log N)"
        a = list(a)
        if not all(a[i] < a[i + 1] for i in range(len(a) - 1)):
            a = sorted(set(a))
        self._build(a)

    def __iter__(self) -> Iterator[T]:
        for i in self.a:
            for j in i: yield j

    def __reversed__(self) -> Iterator[T]:
        for i in reversed(self.a):
            for j in reversed(i): yield j
    
    def __len__(self) -> int:
        return self.size
    
    def __repr__(self) -> str:
        return "SortedSet" + str(self.a)
    
    def __str__(self) -> str:
        s = str(list(self))
        return "{" + s[1 : len(s) - 1] + "}"

    def _find_bucket(self, x: T) -> List[T]:
        "Find the bucket which should contain x. self must not be empty."
        for a in self.a:
            if x <= a[-1]: return a
        return a

    def __contains__(self, x: T) -> bool:
        if self.size == 0: return False
        a = self._find_bucket(x)
        i = bisect_left(a, x)
        return i != len(a) and a[i] == x

    def add(self, x: T) -> bool:
        "Add an element and return True if added. / O(√N)"
        if self.size == 0:
            self.a = [[x]]
            self.size = 1
            return True
        a = self._find_bucket(x)
        i = bisect_left(a, x)
        if i != len(a) and a[i] == x: return False
        a.insert(i, x)
        self.size += 1
        if len(a) > len(self.a) * self.REBUILD_RATIO:
            self._build()
        return True

    def discard(self, x: T) -> bool:
        "Remove an element and return True if removed. / O(√N)"
        if self.size == 0: return False
        a = self._find_bucket(x)
        i = bisect_left(a, x)
        if i == len(a) or a[i] != x: return False
        a.pop(i)
        self.size -= 1
        if len(a) == 0: self._build()
        return True
    
    def lt(self, x: T) -> Union[T, None]:
        "Find the largest element < x, or None if it doesn't exist."
        for a in reversed(self.a):
            if a[0] < x:
                return a[bisect_left(a, x) - 1]

    def le(self, x: T) -> Union[T, None]:
        "Find the largest element <= x, or None if it doesn't exist."
        for a in reversed(self.a):
            if a[0] <= x:
                return a[bisect_right(a, x) - 1]

    def gt(self, x: T) -> Union[T, None]:
        "Find the smallest element > x, or None if it doesn't exist."
        for a in self.a:
            if a[-1] > x:
                return a[bisect_right(a, x)]

    def ge(self, x: T) -> Union[T, None]:
        "Find the smallest element >= x, or None if it doesn't exist."
        for a in self.a:
            if a[-1] >= x:
                return a[bisect_left(a, x)]
    
    def __getitem__(self, x: int) -> T:
        "Return the x-th element, or IndexError if it doesn't exist."
        if x < 0: x += self.size
        if x < 0: raise IndexError
        for a in self.a:
            if x < len(a): return a[x]
            x -= len(a)
        raise IndexError
    
    def index(self, x: T) -> int:
        "Count the number of elements < x."
        ans = 0
        for a in self.a:
            if a[-1] >= x:
                return ans + bisect_left(a, x)
            ans += len(a)
        return ans

    def index_right(self, x: T) -> int:
        "Count the number of elements <= x."
        ans = 0
        for a in self.a:
            if a[-1] > x:
                return ans + bisect_right(a, x)
            ans += len(a)
        return 

n, q, k = map(int, input().split())

# left, sorted, right
# <-               ->

is_rev = 0
deqs = [deque(), deque()]
cur_ldi = 0 # cur left deq index
sst = SortedSet()

for _ in range(q):
    qu = list(map(int, input().split()))
    
    tp = qu[0]
    if tp == 0:
        p = qu[1]
        
        deqs[cur_ldi].appendleft(p)
    elif tp == 1:
        is_rev = 0
        for i in range(2):
            while deqs[i]:
                v = deqs[i].popleft()
                sst.add(v)
        cur_ldi = 0
    else:
        is_rev ^= 1
        cur_ldi ^= 1

if k <= len(deqs[cur_ldi]):
    result = deqs[cur_ldi][k - 1]
else:
    k -= len(deqs[cur_ldi])
    if k <= len(sst):
        k -= 1
        if is_rev:
            k = len(sst) - 1 - k
        result = sst[k]
    else:
        k -= len(sst)
        k -= 1
        k = len(deqs[cur_ldi ^ 1]) - 1 - k
        result = deqs[cur_ldi ^ 1][k]
        
print(result)
```

```python
from collections import deque
import io, os, sys
input = io.BytesIO(os.read(0,os.fstat(0).st_size)).readline

# https://github.com/tatyam-prime/SortedSet/blob/main/SortedSet.py
import math
from bisect import bisect_left, bisect_right
from typing import Generic, Iterable, Iterator, TypeVar, Union, List
T = TypeVar('T')

class SortedSet(Generic[T]):
    BUCKET_RATIO = 50
    REBUILD_RATIO = 170

    def _build(self, a=None) -> None:
        "Evenly divide `a` into buckets."
        if a is None: a = list(self)
        size = self.size = len(a)
        bucket_size = int(math.ceil(math.sqrt(size / self.BUCKET_RATIO)))
        self.a = [a[size * i // bucket_size : size * (i + 1) // bucket_size] for i in range(bucket_size)]
    
    def __init__(self, a: Iterable[T] = []) -> None:
        "Make a new SortedSet from iterable. / O(N) if sorted and unique / O(N log N)"
        a = list(a)
        if not all(a[i] < a[i + 1] for i in range(len(a) - 1)):
            a = sorted(set(a))
        self._build(a)

    def __iter__(self) -> Iterator[T]:
        for i in self.a:
            for j in i: yield j

    def __reversed__(self) -> Iterator[T]:
        for i in reversed(self.a):
            for j in reversed(i): yield j
    
    def __len__(self) -> int:
        return self.size
    
    def __repr__(self) -> str:
        return "SortedSet" + str(self.a)
    
    def __str__(self) -> str:
        s = str(list(self))
        return "{" + s[1 : len(s) - 1] + "}"

    def _find_bucket(self, x: T) -> List[T]:
        "Find the bucket which should contain x. self must not be empty."
        for a in self.a:
            if x <= a[-1]: return a
        return a

    def __contains__(self, x: T) -> bool:
        if self.size == 0: return False
        a = self._find_bucket(x)
        i = bisect_left(a, x)
        return i != len(a) and a[i] == x

    def add(self, x: T) -> bool:
        "Add an element and return True if added. / O(√N)"
        if self.size == 0:
            self.a = [[x]]
            self.size = 1
            return True
        a = self._find_bucket(x)
        i = bisect_left(a, x)
        if i != len(a) and a[i] == x: return False
        a.insert(i, x)
        self.size += 1
        if len(a) > len(self.a) * self.REBUILD_RATIO:
            self._build()
        return True

    def discard(self, x: T) -> bool:
        "Remove an element and return True if removed. / O(√N)"
        if self.size == 0: return False
        a = self._find_bucket(x)
        i = bisect_left(a, x)
        if i == len(a) or a[i] != x: return False
        a.pop(i)
        self.size -= 1
        if len(a) == 0: self._build()
        return True
    
    def lt(self, x: T) -> Union[T, None]:
        "Find the largest element < x, or None if it doesn't exist."
        for a in reversed(self.a):
            if a[0] < x:
                return a[bisect_left(a, x) - 1]

    def le(self, x: T) -> Union[T, None]:
        "Find the largest element <= x, or None if it doesn't exist."
        for a in reversed(self.a):
            if a[0] <= x:
                return a[bisect_right(a, x) - 1]

    def gt(self, x: T) -> Union[T, None]:
        "Find the smallest element > x, or None if it doesn't exist."
        for a in self.a:
            if a[-1] > x:
                return a[bisect_right(a, x)]

    def ge(self, x: T) -> Union[T, None]:
        "Find the smallest element >= x, or None if it doesn't exist."
        for a in self.a:
            if a[-1] >= x:
                return a[bisect_left(a, x)]
    
    def __getitem__(self, x: int) -> T:
        "Return the x-th element, or IndexError if it doesn't exist."
        if x < 0: x += self.size
        if x < 0: raise IndexError
        for a in self.a:
            if x < len(a): return a[x]
            x -= len(a)
        raise IndexError
    
    def index(self, x: T) -> int:
        "Count the number of elements < x."
        ans = 0
        for a in self.a:
            if a[-1] >= x:
                return ans + bisect_left(a, x)
            ans += len(a)
        return ans

    def index_right(self, x: T) -> int:
        "Count the number of elements <= x."
        ans = 0
        for a in self.a:
            if a[-1] > x:
                return ans + bisect_right(a, x)
            ans += len(a)
        return 

n, q = map(int, input().split())

# left, sorted, right
# <-               ->

is_rev = 0
deqs = [deque(), deque()]
cur_ldi = 0 # cur left deq index
sst = SortedSet()

for _ in range(q):
    qu = list(map(int, input().split()))
    
    tp = qu[0]
    if tp == 0:
        p = qu[1]
        
        deqs[cur_ldi].appendleft(p)
    elif tp == 1:
        is_rev = 0
        for i in range(2):
            while deqs[i]:
                v = deqs[i].popleft()
                sst.add(v)
        cur_ldi = 0
    elif tp == 2:
        is_rev ^= 1
        cur_ldi ^= 1
    else:
        result = -1
        if deqs[cur_ldi]:
            result = deqs[cur_ldi].popleft()
        else:
            if sst:
                if is_rev == 0:
                    result = sst[0]
                else:
                    result = sst[-1]
                sst.discard(result)
            else:
                result = deqs[cur_ldi ^ 1].pop()

        print(result)
```
