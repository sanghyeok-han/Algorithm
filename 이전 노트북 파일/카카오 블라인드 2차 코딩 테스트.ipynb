1.  start api를 통해서 게임 현황을 초기화함

2.  Match API를 호출해서 현재 게임 진행 시간을 1초로 만들어줌

3.  WaitingLine API를 호출해서 현재 대기하고 있는 인원 수를 파악함
    - 대기하고 있는 인원들을 cur_wait 리스트 (cur_wait = [[유저 id, 추정 실력, 대기 시작 시간]])에 추가함
    - cur_wait를 (추정 실력 오름차순, 대기 시작 시간 오름차순)으로 정렬함.

4.  Match API를 통해 앞에서부터 차례대로 매칭을 시킴. 
    - 대기하고 있는 인원들 중 추정 실력을 모르는 사람들끼리, 혹은 한 명만 모르는 경우 바로 매칭을 시킴
    - 그러한 인원이 없다면 최소 기다린 시간이 5가 넘기 전까지 전부 대기시킴
    - 두 사람의 추정 실력의 차가 50000이 넘지 않는 경우 매칭을 시킴
    - 매칭시키면서 매칭시킨 인덱스를 저장함
    - 매칭시킨 인덱스를 cur_wait에서 제거함

5.  GameResult API를 호출해서 받은 게임 결과를 통해서 아이디별 실력 순서를 조정함
    - 고유 실력은 1000 이상 100000 이하의 정수
    - 고유 실력의 분포는 평균이 40000 표준편차가 20000인 정규분포
    - 추정 실력이 둘 다 0, 0이라면 진 쪽의 추정 실력을 20000으로 이긴 쪽의 추정 실력을 60000으로 만듦
        - 게임 플레이 시간을 기준으로 진 쪽에 예시1의 값을 더하고 이긴 쪽에 값을 빼줌
    - 한 명의 추정 실력만 0 인데 졌다면 이긴 사람의 추정 실력에 예시2의 값을 빼준 값을 할당함
    - 한 명의 추정 실력만 0 인데 이겼다면 진 사람의 추정 실력에 예시2의 값을 더해준 값을 할당함
    - 두 명의 추정 실력이 모두 0이 아니라면 둘의 추정 실력 +- 에시3

### 서비스 구현

```python
import requests
import time

cur_time = 0
ar = [0] * 31 # assumed ratings
cw = [] # current waiting
epts = 30.858585858585858 # expected play time when both players have same ratings

# 1번 과정
init_headers = {'X-Auth-Token': '7f20d29353db0c1a63dc26f7ab614593',
           'Content-Type': 'application/json'}
params = {'problem': 1}
res = requests.post('https://huqeyhi95c.execute-api.ap-northeast-2.amazonaws.com/prod/start', headers=init_headers, params=params)
auth_key = res.json()
auth_key = auth_key['auth_key']

headers = {'Authorization': auth_key,
           'Content-Type': 'application/json'}


# 2번 과정
dummy_pairs = [[1, 1]]

data = '{"pairs":' + str(dummy_pairs) + '}'
res = requests.put('https://huqeyhi95c.execute-api.ap-northeast-2.amazonaws.com/prod/match',
                    headers=headers, data=data)

matched_result = res.json()
cur_time = matched_result['time']

while cur_time < 594:
    if cur_time % 30 == 0:
        print(cur_time, ar)
        
    time.sleep(0.11)
    
    # 3번 과정
    res = requests.get('https://huqeyhi95c.execute-api.ap-northeast-2.amazonaws.com/prod/waiting_line', headers=headers)
    waiting_line = res.json()
    waiting_line = waiting_line['waiting_line']
    waiting_line = [[v['id'], ar[v['id']], v['from']] for v in waiting_line]
    cw.extend(waiting_line)
    cw.sort(key=lambda x: (x[1], x[2]))

    # 4번 과정
    cands = [v for v in cw if v[1] == 0 or cur_time - v[2] >= 1]
    pairs = []
    matched_ids = set()
    for i in range(len(cands) - 1):
        is_match_needed = False
        if cands[i][0] in matched_ids:
            continue

        if cands[i][1] == 0:
            is_match_needed = True
        else:
            if cands[i + 1][1] - cands[i][1] <= 80000:
                is_match_needed = True

        if is_match_needed:
            matched_ids |= {cands[i][0], cands[i + 1][0]}
            pairs.append([cands[i][0], cands[i + 1][0]])

    cw = [v for v in cw if v[0] not in matched_ids]

    data = '{"pairs":' + str(pairs) + '}'
    res = requests.put('https://huqeyhi95c.execute-api.ap-northeast-2.amazonaws.com/prod/match',
                        headers=headers, data=data)

    matched_result = res.json()
    cur_time = matched_result['time']

    # 5번 과정
    def cuar(winner, loser, taken):
        """change user's assumed ratings"""

        if ar[winner] == 0 and ar[loser] == 0:
            additional = ((taken - epts) * 500)
            ar[winner] = 60000 - additional
            ar[loser] = 20000 + additional
        elif ar[winner] == 0:
            additional = abs((taken - epts)* 1000)
            ar[winner] = ar[loser] + additional
        elif ar[loser] == 0:
            additional = abs((taken - epts)* 1000)
            ar[loser] = ar[winner] - additional  
        else:
            additional = (100000 - abs(winner - loser)) * 0.1 + abs(taken - epts) * 50
            ar[winner] += additional
            ar[loser] -= additional
            
#         if ar[winner] > 100000:
#             ar[winner] = 100000
#         if ar[loser] < 1000:
#             ar[loser] = 1000

    res = requests.get('https://huqeyhi95c.execute-api.ap-northeast-2.amazonaws.com/prod/game_result', headers=headers)
    game_result = res.json()
    game_result = game_result['game_result']
    for v in game_result:
        cuar(v['win'], v['lose'], v['taken'])

# 7번 과정
ar2 = [(i, v) for i, v in enumerate(ar) if i > 0]
ar2.sort()

changes = []
cur_grade = 1
cur_rating = -1
for i, (user_id, rating) in enumerate(ar2):
    if cur_rating == rating:
        changes.append({"id": user_id, "grade": cur_grade})
    else:
        cur_rating = rating
        cur_grade = i + 1
        changes.append({"id": user_id, "grade": cur_grade})

data = {"commands": changes}
data = str(data).replace("'", '"')

res = requests.put('https://huqeyhi95c.execute-api.ap-northeast-2.amazonaws.com/prod/change_grade',
                    headers=headers, data=data)

gss = res.json() # game server status
print(gss)

while cur_time < 600:
    dummy_pairs = [[1, 1]]
    data = '{"pairs":' + str(dummy_pairs) + '}'
    res = requests.put('https://huqeyhi95c.execute-api.ap-northeast-2.amazonaws.com/prod/match',
                        headers=headers, data=data)

    matched_result = res.json()
    if matched_result['status'] == 'finished':
        break
    
    print(matched_result)
```
