https://www.acmicpc.net/problem/25418 정수 a를 k로 만들기

```python
from collections import deque

inf = float('inf')

a, k = map(int, input().split())

dists = [inf] * (k + 1)
dists[a] = 0

queue = deque()
queue.append(a)
while queue:
    value = queue.popleft()
    
    v = value + 1
    if v <= k and dists[v] == inf:
        dists[v] = dists[value] + 1
        queue.append(v)
        
    v = value * 2
    if v <= k and dists[v] == inf:
        dists[v] = dists[value] + 1
        queue.append(v)
        
r = dists[k]

print(r)
```

```python
from itertools import product as prod

n = int(input())
li = sorted(map(int, input().split()))
t = int(input())

met_cs = 0
for case in prod(*[range(2)] * n):
    dirs = [-1] * n
    for i in range(n):
        if case[i] == 1:
            dirs[i] = 1
            
    met_c = 0

    for i in range(n):
        if dirs[i] == -1:
            continue
        cur = li[i]
        for j in range(i + 1, n):
            if li[j] - cur > t * 2:
                break
            if dirs[j] == -1:
                met_c += 1
                
    met_cs += met_c
    
result = met_cs / (2 ** n)

print(result)
```

```python
import io, os, sys
input = io.BytesIO(os.read(0, os.fstat(0).st_size)).readline

## PYRIVAL BOOTSTRAP
# https://github.com/cheran-senthil/PyRival/blob/master/pyrival/misc/bootstrap.py
# This decorator allows for recursion without actually doing recursion
## @bootstrap, yield when getting and returning value in recursive functions, end of functions
from types import GeneratorType

def bootstrap(f, stack=[]):
    def wrappedfunc(*args, **kwargs):
        if stack:
            return f(*args, **kwargs)
        else:
            to = f(*args, **kwargs)
            while True:
                if type(to) is GeneratorType:
                    stack.append(to)
                    to = next(to)
                else:
                    stack.pop()
                    if not stack:
                        break
                    to = stack[-1].send(to)
            return to
    return wrappedfunc

@bootstrap
def dfs(value):
    for v in li[value]:
        if vt[v]:
            continue
        vt[v] = True
        yield dfs(v)
        
    stack.append(value)
    
    yield

@bootstrap
def dfs2(value):
    for v in rli[value]:
        if vt[v]:
            continue
        vt[v] = True
        temp.append(v)
        yield dfs2(v)
        
    yield
    
@bootstrap
def dfs3(value, money):
    global result
    
    vt[value] = True
    
    money += moneies[value]
    
    if value in rests:
        result = max(result, money)
    
    for v in li[value]:
        if vt[v]:
            continue
            
        yield dfs3(v, money)
    
    yield
    
n, m = map(int, input().split())
li = [set() for _ in range(n + 1)]
rli = [set() for _ in range(n + 1)]
for _ in range(m):
    a, b = map(int, input().split())
    
    li[a].add(b)
    rli[b].add(a)
    
stack = []
vt = [False] * (n + 1)
for i in range(1, n + 1):
    if vt[i]:
        continue
    vt[i] = True
    dfs(i)
    
vt = [False] * (n + 1)
scc = []
while stack:
    value = stack.pop()
    if vt[value]:
        continue
    vt[value] = True
        
    temp = [value]
    dfs2(value)
    scc.append(temp.copy())
    
moneies = [0] + [int(input()) for _ in range(n)]
s, p = map(int, input().split())
rests = set(map(int, input().split()))

to_dels = set()
for group in scc:
    is_rest_in = False
    ins = set()
    outs = set()
    sum_v = 0
    for v in group:
        sum_v += moneies[v]
        
        if v in rests:
            is_rest_in = True
        
        for vv in rli[v]:
            ins.add(vv)
        for vv in li[v]:
            outs.add(vv)
            
    st = set(group)
            
    if 1 in st:
        rep = 1
    else:
        rep = group[0]
    
    for out in outs:
        if out not in st: 
            li[rep].add(out)
        
    for to_in in ins:
        if to_in not in st:
            li[to_in].add(rep)
        
    to_dels |= (st - {rep})
    
    moneies[rep] = sum_v
    if is_rest_in:
        rests.add(rep)
        
for value in range(1, n + 1):
    temp_to_dels = set()
    for v in li[value]:
        if v in to_dels:
            temp_to_dels.add(v)
            
    li[value] -= temp_to_dels
    
result = 0
vt = [False] * (n + 1)
dfs3(1, 0)

print(result)
```

```python
from collections import defaultdict as dd
from math import log2, ceil

def find(v):
    while v != pa[v]:
        pa[v] = pa[pa[v]]
        v = pa[v]
    return v

def update(node, l, r):
    if ul <= l and r <= ur:
        seg[node].append((a, b))
    elif r < ul or ur < l:
        return
    else:
        mid = (l + r) // 2
        
        update(node * 2, l, mid)
        update(node * 2 + 1, mid + 1, r)
        
def query(node, l, r):
    unioned_c = 0
    unioned_info = []
    for a, b in seg[node]:
        ra = find(a)
        rb = find(b)
     
        if ra != rb:
            if rank[ra] < rank[rb]:
                ra, rb = rb, ra

            pa[rb] = ra
            is_rank_added = False
            if rank[ra] == rank[rb]:
                rank[ra] += 1
                is_rank_added = True
                
            unioned_c += 1
            unioned_info.append((ra, rb, is_rank_added))
            print(unioned_info)
            
    if l == r:
        a, b = check_queries[l]
        ra = find(a)
        rb = find(b)
        
        if ra == rb:
            result[l] = 1
    else:
        mid = (l + r) // 2
        
        query(node * 2, l, mid)
        query(node * 2 + 1, mid + 1, r)
        
    for _ in range(unioned_c):
        ra, rb, is_rank_added = unioned_info.pop()
        
        pa[rb] = rb
        if is_rank_added:
            rank[ra] -= 1

n, m = map(int, input().split())

pendings = dd(int)
cur_check_used = -1
check_queries = []
edge_history = []
for qi in range(m):
    tp, a, b = map(int, input().split())
    
    if a > b:
        a, b = b, a
        
    if tp == 3:
        cur_check_used += 1
        check_queries.append((a, b))
    else:
        if (a, b) not in pendings:
            pendings[(a, b)] = cur_check_used
        else:
            f_check_used = pendings[(a, b)]
    
            if f_check_used < cur_check_used:
                edge_history.append((a, b, f_check_used + 1, cur_check_used))
                
            del pendings[(a, b)]
            
for key in pendings:
    f_check_used = pendings[key]
    edge_history.append((key[0], key[1], f_check_used + 1, cur_check_used))
    
size = 2 ** (ceil(log2(cur_check_used + 2)) + 1) + 1
seg = [[] for _ in range(size)]

for a, b, ul, ur in edge_history:
    update(1, 0, cur_check_used)
    
result = [0] * (cur_check_used + 1)
rank = [0] * (n + 1)
pa = list(range(n + 1))
query(1, 0, cur_check_used)

print(rank)
print(pa)

print('\n'.join(map(str, result)))
```
