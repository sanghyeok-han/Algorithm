https://codeforces.com/contest/1762/problem/D D. GCD Queries

```python
from collections import deque
import sys
input = lambda: sys.stdin.readline().rstrip()

for _ in range(int(input())):
    n = int(input())
    
    deq = deque(range(1, n + 1))
    
    while len(deq) >= 3:
        x = deq.popleft()
        y = deq.popleft()
        z = deq.popleft()

        print('?', x, y)
        sys.stdout.flush()
        r1 = int(input())   
        
        print('?', x, z)
        sys.stdout.flush()
        r2 = int(input())          
        
        if r1 == r2:
            deq.appendleft(y)
            deq.appendleft(z)
        elif r1 > r2:
            deq.appendleft(x)
            deq.appendleft(y)
        else:
            deq.appendleft(x)
            deq.appendleft(z)

    print('!', deq[0], deq[1])
    sys.stdout.flush()
    result = int(input())
    
    if result == -1:
        break
```

```python
from heapq import heappush, heappop
import sys, os, io
# input = io.BytesIO(os.read(0,os.fstat(0).st_size)).readline

for _ in range(int(input())):
    n, m = map(int, input().split())
    m -= 1
    
    li = list(map(int, input().split()))

    used_c = 0

    prefix = [0] * n
    cumul = 0
    for i in range(n):
        cumul += li[i]
        prefix[i] = cumul

    heap = [-li[m]]
    for i in range(m - 1, -1, -1):
        if prefix[i] < prefix[m]:
            used_c += 1
            
            fv = -heappop(heap)
            prefix[m] -= 2 * fv
        heappush(heap, -li[i])

    cumul = 0
    heap = []
    for i in range(m + 1, n):
        heappush(heap, li[i])
        cumul += li[i]

        if cumul < 0:
            fv = heappop(heap)

            used_c += 1
            cumul -= 2 * fv

    print(used_c)
```
