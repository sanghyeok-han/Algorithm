```python
# 틀렸음
from collections import deque, defaultdict as dd
import sys
input = lambda: sys.stdin.readline().rstrip()

n = int(input())
li = [input() for _ in range(n)]

nli = []
fronts = set()
graph = dd(list)
graph_r = dd(list)
for i in range(n):
    nli.append(li[i][0] + li[i][-1])
    
    f = li[i][0]
    e = li[i][-1]
    
    graph[f].append(e)
    graph_r[e].append(f)
    fronts.add(f)
    
vt_ends_w = set()
vt_ends_l = set()
vt_ends_inf = set()
queue = deque()
for i in range(n):
    if nli[i][1] not in fronts:
        vt_ends_w.add(nli[i][1])
        queue.append(nli[i][1])
    
    if nli[i][0] == nli[i][1]:
        vt_ends_inf.add(nli[i][1])
        
while True:
    for _ in range(len(queue)):
        cur_end = queue.popleft()

        for n_end in graph_r[cur_end]:
            if n_end not in vt_ends_w and n_end not in vt_ends_l and n_end not in vt_ends_inf:
                can_win = True
                is_not_fixed = False
                
                for nn_end in graph[n_end]:
                    if nn_end in vt_ends_w:
                        can_win = False
                        break
                    elif nn_end in vt_ends_inf:
                        can_inf = True
                    elif nn_end not in vt_ends_l:
                        is_not_fixed = True
                    
                if not can_win:
                    vt_ends_l.add(n_end)
                elif is_not_fixed:
                    continue
                elif can_inf:
                    vt_ends_inf.add(n_end)
                else:
                    vt_ends_w.add(n_end)
                    
                queue.append(n_end)
                
    if not queue:
        break
    
result = []
for v in li:
    if v[-1] in vt_ends_w:
        result.append(v)
        
result.sort()

print(len(result))
for v in result:
    print(v)
```

```python
from math import log2, ceil
import sys
input = lambda: sys.stdin.readline().rstrip()

## PYRIVAL BOOTSTRAP
# https://github.com/cheran-senthil/PyRival/blob/master/pyrival/misc/bootstrap.py
# This decorator allows for recursion without actually doing recursion
## @bootstrap, yield when getting and returning value in recursive functions, end of functions
from types import GeneratorType

def bootstrap(f, stack=[]):
    def wrappedfunc(*args, **kwargs):
        if stack:
            return f(*args, **kwargs)
        else:
            to = f(*args, **kwargs)
            while True:
                if type(to) is GeneratorType:
                    stack.append(to)
                    to = next(to)
                else:
                    stack.pop()
                    if not stack:
                        break
                    to = stack[-1].send(to)
            return to
    return wrappedfunc

def construct(node, s, e):
    if s == e:
        segtree[node] = earns[seq_to_node[s]]
        return segtree[node]
    mid = (s + e) // 2
    
    t1 = construct(node * 2, s, mid)
    t2 = construct(node * 2 + 1, mid + 1, e)
    
    segtree[node] = t1 + t2
    
    return segtree[node]

def update_range(node, s, e, pending):
    segtree[node] += pending * (e - s + 1)
    pendings[node] += pending
    
    if ts <= s <= e <= te:
        pendings[node] += to_add
        segtree[node] += to_add * (e - s + 1)
        
        return segtree[node]
    elif e < ts or s > te:
        return segtree[node]
    else:
        pending = pendings[node]
        pendings[node] = 0
        
        mid = (s + e) // 2
        
        t1 = update_range(node * 2, s, mid, pending)
        t2 = update_range(node * 2 + 1, mid + 1, e, pending)
        
        segtree[node] = t1 + t2
        
        return segtree[node]

def query(node, s, e, pending):
    segtree[node] += pending * (e - s + 1)
    pendings[node] += pending
    
    if e < qs or s > qe:
        return 0
    if qs <= s <= e <= qe:
        return segtree[node]
    else:
        pending = pendings[node]
        pendings[node] = 0
        
        mid = (s + e) // 2
        
        t1 = query(node * 2, s, mid, pending)
        t2 = query(node * 2 + 1, mid + 1, e, pending)
        
        return t1 + t2

@bootstrap
def dfs(value):
    global vt_num
    
    seq_to_node[vt_num] = value
    history[value][0] = vt_num
    
    for v in li[value]:
        vt_num += 1
        yield dfs(v)
    
    history[value][1] = vt_num
    
    yield
    
n, m = map(int, input().split())

size = 2 ** (ceil(log2(n)) + 1) + 1
segtree = [0] * size
pendings = [0] * size

li = [[] for _ in range(n)]
earns = [0] * n

earns[0] = int(input())

for c_num in range(1, n):
    earn, p_num = map(int, input().split())
    p_num -= 1
    
    li[p_num].append(c_num)
    earns[c_num] = earn

history = [[-1] * 2 for _ in range(n)]
seq_to_node = [-1] * n

vt_num = 0
dfs(0)

construct(1, 0, n - 1)

for _ in range(m):
    tp, *qu = input().split()
    
    if tp == 'p':
        a, x = map(int, qu)
        a -= 1
        
        ts, te = history[a]
        ts += 1
        to_add = x
        
        if ts <= te:
            update_range(1, 0, n - 1, 0)
    else:
        a = int(qu[0])
        a -= 1
        
        qs = qe = history[a][0]
        result = query(1, 0, n - 1, 0)
        
        print(result)
```

```python
from collections import deque
import io, os, sys
input = io.BytesIO(os.read(0,os.fstat(0).st_size)).readline

inf = float('inf')

def spfa():
    min_costs = [inf] * ran
    queue = deque()
    in_q = [False] * ran
    from_ = [-1] * ran
    
    min_costs[sc] = 0
    queue.append(sc)
    in_q[sc] = True
    while queue:
        value = queue.popleft()
        in_q[value] = False
        
        for v in G[value]:
            if F[value][v] < C[value][v]:     
                ncost = min_costs[value] + costs[value][v]
                if ncost < min_costs[v]:
                    min_costs[v] = ncost
                    from_[v] = value
                    
                    if not in_q[v]:
                        queue.append(v)
                        in_q[v] = True
                        
    return from_, min_costs[sk]    

n = int(input())
temp_costs = [list(map(int, input().split())) for _ in range(n)]

ran = 2 * n + 2
sc = 2 * n
sk = 2 * n + 1

G = [[] for _ in range(ran)]
C = [[0] * ran for _ in range(ran)]
F = [[0] * ran for _ in range(ran)]
costs = [[inf] * ran for _ in range(ran)] 

for i in range(n):
    for j in range(n):
        s_num = i
        p_num = n + j
        
        C[s_num][p_num] = inf
        G[s_num].append(p_num)
        G[p_num].append(s_num)
        costs[s_num][p_num] = temp_costs[i][j]
        costs[p_num][s_num] = -temp_costs[i][j]
        
for s_num in range(n):
    G[sc].append(s_num)
    C[sc][s_num] = 1
    costs[sc][s_num] = 0
    
for p_num in range(n, 2 * n):
    G[p_num].append(sk)
    C[p_num][sk] = 1
    costs[p_num][sk] = 0
    
total_flow = 0
total_cost = 0
while True:
    from_, cost = spfa()
    
    if from_[sk] == -1:
        break
        
    cur = sk
    can_flow = inf
    while True:
        fcur = from_[cur]
        
        can_flow = min(can_flow, C[fcur][cur] - F[fcur][cur])
        
        cur = fcur
        if cur == sc:
            break
    
    cur = sk
    while True:
        fcur = from_[cur]
        
        F[fcur][cur] += can_flow
        F[cur][fcur] -= can_flow
        
        cur = fcur
        if cur == sc:
            break
        
    total_flow += can_flow
    total_cost += can_flow * cost
    
print(total_cost)
```
