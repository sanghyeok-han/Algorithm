```python
import sys
input = lambda: sys.stdin.readline().rstrip()

class SegmentTree:
    def __init__(self, data, default=10**15, func=lambda a, b: max(a,b)):
        """initialize the segment tree with data"""
        self._default = default
        self._func = func
        self._len = len(data)
        self._size = _size = 1 << (self._len - 1).bit_length()
 
        self.data = [default] * (2 * _size)
        self.data[_size:_size + self._len] = data
        for i in reversed(range(_size)):
            self.data[i] = func(self.data[i + i], self.data[i + i + 1])
 
    def __delitem__(self, idx):
        self[idx] = self._default
 
    def __getitem__(self, idx):
        return self.data[idx + self._size]
 
    def __setitem__(self, idx, value):
        idx += self._size
        self.data[idx] = value
        idx >>= 1
        while idx:
            self.data[idx] = self._func(self.data[2 * idx], self.data[2 * idx + 1])
            idx >>= 1
 
    def __len__(self):
        return self._len
 
    def query(self, start, stop):
        if start == stop:
            return self.__getitem__(start)
        stop += 1
        start += self._size
        stop += self._size
 
        res = self._default
        while start < stop:
            if start & 1:
                res = self._func(res, self.data[start])
                start += 1
            if stop & 1:
                stop -= 1
                res = self._func(res, self.data[stop])
            start >>= 1
            stop >>= 1
        return res
 
    def __repr__(self):
        return "SegmentTree({0})".format(self.data)
    
n, q = map(int, input().split())
li = [0] * (n + 1)

func = lambda a, b: a + b
segtree = SegmentTree(li, 0, func = func)

for _ in range(q):
    tp, p, x = map(int, input().split())
    
    if tp == 1:
        segtree[p] += x
    else:
        q = x
        r = segtree.query(p, q)
        
        print(r)
```

```python
from collections import deque

def solve1(n, li):
    if n == 1:
        return 0
    else:
        is_max = False
        is_min = False

        cur_max = li[0]
        cur_min = li[0]
        cur_start = li[0]
        i = 0
        total = 0
        vt = set()

        deq = deque(range(1, n + 1))
        while True:
            v = li[i]

            while deq and deq[0] in vt:
                deq.popleft()
            while deq and deq[-1] in vt:
                deq.pop()

            if is_max:
                while li[i] != deq[0]:
                    vt.add(li[i])
                    i += 1

                is_max = False
                is_min = True
                total += 1
                if i == n - 1:
                    break
            elif is_min:
                while li[i] != deq[-1]:
                    vt.add(li[i])
                    i += 1

                is_max = True
                is_min = False
                total += 1
                if i == n - 1:
                    break
            else:
                cur_max = max(cur_max, v)
                cur_min = min(cur_min, v)

                if cur_start != cur_max and cur_start != cur_min or v != cur_max and v != cur_min:
                    total += 1
                    i -= 1
                    cur_max = li[i]
                    cur_min = li[i]
                    cur_start = li[i]
                    continue

                if deq[0] == v:
                    is_min = True
                    if cur_min != v or cur_max != v:
                        total += 1
                if deq[-1] == v:
                    is_max = True
                    if cur_min != v or cur_max != v:
                        total += 1

                vt.add(v)
                i += 1


        return total
    
import sys
# input =lambda: sys.stdin.readline()

def solve2(n, a):
    if n==1:
        return 0
    if n==2:
        return 1
    min1 = [-1 for i in range(n)]
    min2 = [-1 for i in range(n)]
    max1 = [-1 for i in range(n)]
    max2 = [-1 for i in range(n)]
    t = n-1
    for i in range(n-2,-1,-1):
        min1[i]=t
        if a[i]<a[t]:
            t = i
    t = n-1
    for i in range(n-2,-1,-1):
        max1[i]=t
        if a[i]>a[t]:
            t = i
    t = 0
    for i in range(1,n):
        min2[i]=t
        if a[i]<a[t]:
            t = i
    t = 0
    for i in range(1,n):
        max2[i]=t
        if a[i]>a[t]:
            t = i
    x = -1
    for i in range(n):
        if a[i]==1:
            x=i
    count1=0
    i=x
    t=1
    while i<n-1:
        if t==1:
            i=max1[i]
            t=0
            count1+=1
        else:
            i=min1[i]
            t=1
            count1+=1
    count2=0
    i=x
    t=1
    while i>0:
        if t==1:
            i=max2[i]
            t=0
            count2+=1
        else:
            i=min2[i]
            t=1
            count2+=1
    return count1 + count2
```

```python
from contextlib import contextmanager
import threading
import _thread
import time

class TimeoutException(Exception):
    def __init__(self, msg=''):
        self.msg = msg

@contextmanager
def time_limit(seconds, msg=''):
    timer = threading.Timer(seconds, lambda: _thread.interrupt_main())
    timer.start()
    try:
        yield
    except KeyboardInterrupt:
        raise TimeoutException("Timed out for operation {}".format(msg))
    finally:
        # if the action ends in specified time, timer is canceled
        timer.cancel()
        

def sol(lst):
    while lst[-1] == 9:
        continue
        
    n = len(lst)
    lst.sort()
    return lst[(n-1) // 2]

time_limit_sec = 1 # 시간 제한(초 단위)
test_cases = [[[1, 5, 5, 7]], [[1, 8, 4, 9]], [[3, 4, 5, 6]], [['a', 3]], [[8]]]
ans = [5, 0, 4, 0, 9]

for test_case_id in range(len(test_cases)):
    test_case = test_cases[test_case_id]
    
    time_start = time.time()
    is_time_over = False
    error = ''
    
    try:
        result = time_result = -1
        with time_limit(time_limit_sec, 'sleep'):
            result = sol(*test_case)
        time_result = round((time.time() - time_start) * 1000)
    except BaseException as error_m:
        error = str(error_m)
        if error == 'Timed out for operation sleep':
            is_time_over = True

    if error:
        if is_time_over:
            error = '시간 초과!'
        to_print = f'{test_case_id}: Failed:'.ljust(20, ' ')
        to_print += f'{error}'
    elif result != ans[test_case_id]:
        to_print = f'{test_case_id}: Failed:'.ljust(20, ' ')
        to_print += f'잘못된 답: {result}, 원래 정답: {ans[test_case_id]}'
    else:
        time_result = str(time_result) + 'ms'
        to_print = f'{test_case_id}: {result}'.ljust(20, ' ')
        to_print += f'Time elapsed: {time_result}'

    print(to_print)
```
