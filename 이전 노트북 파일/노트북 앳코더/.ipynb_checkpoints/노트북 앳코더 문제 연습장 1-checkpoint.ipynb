```python
ori = 'atcoder'

s = input()
li = []
for v in s:
    li.append(ori.index(v))
    
def merge_sort(li):
    global count
    
    if len(li) == 1:
        return li
    
    left = merge_sort(li[:len(li) // 2])
    right = merge_sort(li[len(li) // 2:])
    
    nl = len(left)
    nr = len(right)
    
    nli = []
    l_i = r_i = 0
    while True:
        if left[l_i] <= right[r_i]:
            nli.append(left[l_i])
            count += r_i
            l_i += 1
            
            if l_i == nl:
                while r_i < nr:
                    nli.append(right[r_i])
                    r_i += 1
                break    
        else:
            nli.append(right[r_i])
            r_i += 1
            
            if r_i == nr:
                while l_i < nl:
                    nli.append(left[l_i])
                    count += r_i
                    l_i += 1
                break
    return nli 
    
count = 0
merge_sort(li)

print(count)
```

```python
import sys
input = lambda: sys.stdin.readline().rstrip()

n, m = map(int, input().split())
xs = list(map(int, input().split()))
bonuses = [0] * (n + 1)
for _ in range(m):
    x, y = map(int, input().split())
    bonuses[x] = y
    
dp = [[0] * (n + 1) for _ in range(n)]
maxes = [0] * n

dp[0][0] = 0
dp[0][1] = xs[0] + bonuses[1]
maxes[0] = dp[0][1]

for i in range(1, n):
    dp[i][0] = maxes[i - 1]
    cur_max = dp[i][0]
    for cur_s in range(1, i + 2):
        temp = dp[i - 1][cur_s - 1] + xs[i] + bonuses[cur_s]
        cur_max = max(cur_max, temp)
        
        dp[i][cur_s] = temp
        
    maxes[i] = cur_max
    
r = maxes[n - 1]
    
print(r)
```

https://atcoder.jp/contests/abc262/tasks/abc262_e E - Red and Blue Graph

```python
import sys
input = lambda: sys.stdin.readline().rstrip()

mod = 998244353
max_num = 200000

factos = [0] * (max_num + 1)
factos[0] = 1
invs = [0] * (max_num + 1)
cur = 1
for i in range(1, max_num + 1):
    cur *= i
    cur %= mod
    factos[i] = cur
    
invs[max_num] = pow(factos[max_num], mod - 2, mod)
for i in range(max_num - 1, -1, -1):
    invs[i] = (i + 1) * invs[i + 1] % mod

def nCk(n, k):
    if n < k:
        return 0
    return factos[n] * invs[k] * invs[n - k] % mod

n, m, k = map(int, input().split())

ct = [0] * (n + 1)
for _ in range(m):
    u, v = map(int, input().split())
    
    ct[u] += 1
    ct[v] += 1
    
c0 = c1 = 0
for i in range(1, n + 1):
    if ct[i] % 2 == 0:
        c0 += 1
    else:
        c1 += 1
        
total = 0
for used_c1 in range(0, c1 + 1, 2):
    if used_c1 > k:
        break
        
    used_c0 = k - used_c1
    if used_c0 > c0:
        continue
        
    total += nCk(c1, used_c1) * nCk(c0, used_c0)
    total %= mod
    
print(total)
```

```python
import sys
input = lambda: sys.stdin.readline().rstrip()

def add(num):
    for v in included[num]:
        if v not in ct:
            ct[v] = 1
        else:
            ct[v] += 1
            
def sub(num):
    for v in included[num]:
        ct[v] -= 1
        if ct[v] == 0:
            del ct[v]

n, m = map(int, input().split())

included = [[] for _ in range(m + 1)]
for i in range(n):
    a, b = map(int, input().split())
    included[a].append(i)
    included[b].append(i)
    
start = 1
end = 1
ct = {}
add(1)
deltas = [0] * (m + 2)
while True:
    if len(ct) == n:
        min_len = end - start + 1
        max_len = m - start + 1
        deltas[min_len] += 1
        deltas[max_len + 1] -= 1
        
        sub(start)
        start += 1
        if start > end:
            end = start
            if end > m:
                break
            add(end)
    else:
        end += 1
        if end > m:
            break
        add(end)
        
result = [0] * (m + 1)
cur = 0
for i in range(m + 1):
    cur += deltas[i]
    result[i] = cur
    
print(*result[1:])
```

https://atcoder.jp/contests/abc259/tasks/abc259_d Circumferences

```python
from collections import deque
import sys
input = lambda: sys.stdin.readline().rstrip()

n = int(input())
sx, sy, tx, ty = map(int, input().split())
li = [list(map(int, input().split())) for _ in range(n)]

queue = deque()
vt = [False] * n
purpose = set()
for i in range(n):
    x, y, r = li[i]
    
    dist_s = (x - sx) ** 2 + (y - sy) ** 2
    if dist_s == r ** 2:
        vt[i] = True
        queue.append(i)    
     
    dist_s = (x - tx) ** 2 + (y - ty) ** 2
    if dist_s == r ** 2:
        purpose.add(i)
        
graph = [[] for _ in range(n)]
for i in range(n):
    x1, y1, r1 = li[i]
    for j in range(i + 1, n):
        x2, y2, r2 = li[j]
        
        dist_s = (x1 - x2) ** 2 + (y1 - y2) ** 2
        
        if dist_s > (r1 + r2) ** 2:
            continue
            
        nr1, nr2 = r1, r2
        if nr1 > nr2:
            nr1, nr2 = nr2, nr1
            
        if dist_s < (nr2 - nr1) ** 2:
            continue
            
        graph[i].append(j)
        graph[j].append(i)
        
ip = False
while queue:
    value = queue.popleft()
    if value in purpose:
        ip = True
        break
        
    for v in graph[value]:
        if not vt[v]:
            vt[v] = True
            queue.append(v)
        
if ip:
    print('Yes')
else:
    print('No')
```

https://atcoder.jp/contests/abc257/tasks/abc257_d D - Jumping Takahashi 2

```python
from collections import deque
import sys
# input = lambda: sys.stdin.readline().rstrip()

inf = float('inf')

def check(mid):
    vt = [False] * n    
    for i in range(n):
        if vt[i]:
            continue
        nvt = [False] * n
        nvt[i] = True
        queue = deque()
        queue.append(i)
        while queue:
            value = queue.popleft()
            vt[value] = True
            
            for v, dist in graph[value]:
                if not nvt[v] and li[value][2] * mid >= dist:
                    nvt[v] = True
                    queue.append(v)
                    
        if sum(nvt) == n:
            return True
        
    return False
        
n = int(input())
li = [list(map(int, input().split())) for _ in range(n)]
        
graph = [[] for _ in range(n)]
for i in range(n):
    x1, y1, p1 = li[i]
    for j in range(i + 1, n):
        x2, y2, p2 = li[j]
        
        dist = abs(x1 - x2) + abs(y1 - y2)
        graph[i].append((j, dist))
        graph[j].append((i, dist))
    
start = 1
end = 5 * int(1e9)
result = inf
while start <= end:
    mid = (start + end) // 2
    
    if check(mid):
        result = mid
        end = mid - 1
    else:
        start = mid + 1
        
print(result)
```
