```python
import sys
# input = lambda: sys.stdin.readline().rstrip()

mod = 1000000007
    
h, w = map(int, input().split())
li = [input() for _ in range(h)]

cont1 = [[0] * w for _ in range(h)]
total_tidy = 0
for i in range(h):
    temp = 0
    for j in range(w):
        if li[i][j] == '#':
            nj = j - 1
            while nj >= 0 and li[i][nj] != '#':
                cont1[i][nj] = temp
                nj -= 1
            temp = 0
        else:
            total_tidy += 1
            temp += 1
            
    if temp:
        nj = j
        while nj >= 0 and li[i][nj] != '#':
            cont1[i][nj] = temp
            nj -= 1
        temp = 0
            
cont2 = [[0] * w for _ in range(h)]
for j in range(w):
    temp = 0
    for i in range(h):
        if li[i][j] == '#':
            ni = i - 1
            while ni >= 0 and li[ni][j] != '#':
                cont2[ni][j] = temp
                ni -= 1
            temp = 0
        else:
            temp += 1
            
    if temp:
        ni = i
        while ni >= 0 and li[ni][j] != '#':
            cont2[ni][j] = temp
            ni -= 1
        temp = 0
        
amounts = 2000 * 2000 + 5
twos = [1] * amounts
cur = 1
for i in range(1, amounts):
    cur *= 2
    cur %= mod
    twos[i] = cur
        
total = 0
for i in range(h):
    for j in range(w):
        if li[i][j] == '#':
            continue
        included = cont1[i][j] + cont2[i][j] - 1
        excluded = total_tidy - included
        
        total += (twos[included] - 1) * twos[excluded]
        total %= mod
        
print(total)
```
