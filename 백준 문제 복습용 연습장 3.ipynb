```python
from collections import deque
import sys
input = lambda: sys.stdin.readline().rstrip()

n, m, k = map(int, input().split())
li = [input() for _ in range(n)]
get_value = lambda x: int(x) - 1
sx, sy, ex, ey = map(get_value, input().split())

# 하우상좌
dx = [1, 0, -1, 0]
dy = [0, 1, 0, -1]

def bfs():
    dists = [[-1] * m for _ in range(n)]
    dists[sx][sy] = 0

    queue = deque()
    queue.append((sx, sy, -1, False))
    while queue:
        x, y, d, is_ended = queue.popleft()

        for nd in range(4):
            if d != -1 and (d + 2) % 4 == nd:
                continue
            if not is_ended and d == nd:
                continue

            n_is_ended = False
            nx = x + dx[nd]
            ny = y + dy[nd]
            for added in range(k):
                if added == k - 1:
                    n_is_ended = True
                if nx < 0 or nx >= n or ny < 0 or ny >= m:
                    break
                if li[nx][ny] == '#':
                    break
                if dists[nx][ny] != -1:
                    if dists[nx][ny] <= dists[x][y]:
                        break
                else:
                    dists[nx][ny] = dists[x][y] + 1
                    if nx == ex and ny == ey:
                        return dists[nx][ny]
                    
                    queue.append((nx, ny, nd, n_is_ended))
                    
                nx += dx[nd]
                ny += dy[nd]
                    
    return -1

print(bfs())
```

### 숨바꼭질 1697

```python
from collections import deque

n, k = map(int, input().split())

def bfs():
    md = 100000 # max dist
    dists = [-1] * (md + 1)
    dists[n] = 0

    queue = deque([n])
    while queue:
        x = queue.popleft()
        if x == k:
            return dists[x]
        
        dx = [lambda x: x - 1, lambda x: x + 1, lambda x: 2 * x]
        for i in range(3):
            nx = dx[i](x)
            if nx < 0 or nx > md:
                continue
            if dists[nx] == -1:
                dists[nx] = dists[x] + 1
                queue.append(nx)
                
print(bfs())
```

### 연산자 끼워넣기 14888

```python
n = int(input())
li = list(map(int, input().split()))
ops = list(map(int, input().split())) # operators

max_value = -float('inf')
min_value = float('inf')

def dfs(cur_index, cumul, ops):
    global max_value
    global min_value
    
    if cur_index == n:
        max_value = max(max_value, cumul)
        min_value = min(min_value, cumul)
        return
    
    dx = [lambda c, nxt: c + nxt, lambda c, nxt: c - nxt, lambda c, nxt: c * nxt, lambda c, nxt: c // nxt if c >= 0 else -(abs(c) // nxt)]
    for i in range(4):  
        if ops[i]:
            ncumul = dx[i](cumul, li[cur_index])
            ops[i] -= 1
            dfs(cur_index + 1, ncumul, ops)
            ops[i] += 1
            
dfs(1, li[0], ops)

print(max_value)
print(min_value)
```

### 연구소 14502

```python
from itertools import combinations as comb
from copy import deepcopy
from collections import deque

n, m = map(int, input().split())
li = [list(map(int, input().split())) for _ in range(n)]

blanks = []
vs = [] # viruses
for i in range(n):
    for j in range(m):
        if li[i][j] == 0:
            blanks.append((i, j))
        elif li[i][j] == 2:
            vs.append((i, j))

dx = [-1, 1, 0, 0]
dy = [0, 0, -1, 1]
            
obc = len(blanks) # original blank count
cases = comb(blanks, 3)

msa = 0 # max safe area
for case in cases:
    bc = obc - 3
    nli = deepcopy(li)
    for x, y in case:
        nli[x][y] = 1
        
    queue = deque()
    queue.extend(vs)
    while queue:
        x, y = queue.popleft()
        
        for i in range(4):
            nx = x + dx[i]
            ny = y + dy[i]
            
            if nx < 0 or nx >= n or ny < 0 or ny >= m:
                continue
            if nli[nx][ny] == 0:
                bc -= 1
                nli[nx][ny] = 2
                queue.append((nx, ny))
                
    msa = max(msa, bc)
    
print(msa)
```

### 인구 이동 16234

```python
from collections import deque

n, l, r = map(int, input().split())
li = [list(map(int, input().split())) for _ in range(n)]

dx = [-1, 1, 0, 0]
dy = [0, 0, -1, 1]

count = 0
while True:
    is_moved = False
    vt = [[False] * n for _ in range(n)]
    for i in range(n):
        for j in range(n):
            if vt[i][j]:
                continue
            vt[i][j] = True
            
            populs = [li[i][j]]
            coords = [(i, j)]
            
            queue = deque()
            queue.append((i, j))
            while queue:
                x, y = queue.popleft()
                
                for k in range(4):
                    nx = x + dx[k]
                    ny = y + dy[k]
                    if nx < 0 or nx >= n or ny < 0 or ny >= n:
                        continue
                    diff = abs(li[x][y] - li[nx][ny])
                    if vt[nx][ny]:
                        continue
                    if diff < l or diff > r:
                        continue
                    vt[nx][ny] = True
                    populs.append(li[nx][ny])
                    coords.append((nx, ny))
                    queue.append((nx, ny))
                
            if len(populs) >= 2:
                is_moved = True
                sp = sum(populs) // len(populs) # shared population
                for x, y in coords:
                    li[x][y] = sp
    if is_moved:
        count += 1
    else:
        break
        
print(count)
```

### 유닛 이동시키기 2194

```python
from collections import deque
import sys
input = lambda: sys.stdin.readline().rstrip()

n, m, a, b, k = map(int, input().split())
li = [[0] * m for _ in range(n)]
get_value = lambda x: int(x) - 1
for _ in range(k):
    x, y = map(get_value, input().split())
    li[x][y] = 1
sx, sy = map(get_value, input().split())
ex, ey = map(get_value, input().split())

sum_li = [[0] * (m + 1) for _ in range(n + 1)]
for i in range(1, n + 1):
    temp = 0
    for j in range(1, m + 1):
        temp += li[i - 1][j - 1]
        sum_li[i][j] = temp + sum_li[i - 1][j]
        
def bfs():   
    dx = [-1, 1, 0, 0]
    dy = [0, 0, -1, 1]

    dists = [[-1] * m for _ in range(n)]
    dists[sx][sy] = 0

    queue = deque()
    queue.append((sx, sy))
    while queue:
        x, y = queue.popleft()

        for i in range(4):
            nx_s = x + dx[i]
            ny_s = y + dy[i]
            if nx_s < 0 or nx_s >= n or ny_s < 0 or ny_s >= m:
                continue
            nx_e = nx_s + a - 1
            ny_e = ny_s + b - 1
            if nx_e < 0 or nx_e >= n or ny_e < 0 or ny_e >= m:
                continue        
            if dists[nx_s][ny_s] != -1:
                continue
            if sum_li[nx_e + 1][ny_e + 1] - sum_li[nx_s][ny_e + 1] - sum_li[nx_e + 1][ny_s] + sum_li[nx_s][ny_s] >= 1:
                continue
            dists[nx_s][ny_s] = dists[x][y] + 1
            if nx_s == ex and ny_s == ey:
                return dists[nx_s][ny_s]
            queue.append((nx_s, ny_s))
    return -1

print(bfs())
```

```python
from collections import deque
import sys
# input = lambda: sys.stdin.readline().rstrip()

n, m = map(int, input().split())
li = [list(map(int, input())) for _ in range(n)]

def bfs():
    if n == 1 and m == 1:
        return 1
    
    dx = [-1, 1, 0, 0]
    dy = [0, 0, -1, 1]
    
    dists = [[[float('inf')] * 2 for _ in range(m)] for _ in range(n)]
    dists[0][0][1] = 1

    queue = deque()
    queue.append((0, 0, 1)) # x, y, k
    while queue:
        x, y, k = queue.popleft()
        for i in range(4):
            nx, ny = x + dx[i], y + dy[i]
            if nx < 0 or nx >= n or ny < 0 or ny >= m:
                continue

            value = dists[x][y][k] + 1
            if not li[nx][ny]:
                if value < dists[nx][ny][k]:
                    dists[nx][ny][k] = value
                    if nx == n - 1 and ny == m - 1:
                        return value
                    queue.append((nx, ny, k))
            else:
                if k and value < dists[nx][ny][0]:
                    dists[nx][ny][0] = value
                    if nx == n - 1 and ny == m - 1:
                        return value
                    queue.append((nx, ny, 0))
    return -1

print(bfs())
```

### 벽 부수고 이동하기 3 16933

```python
# 다른 사람 코드 (https://www.acmicpc.net/source/21694427)
# pypy3
import sys
from collections import deque
input = sys.stdin.readline
n, m, k = map(int, input().split())
load = [input().rstrip() for _ in range(n)]
visited = [[k+1 for _ in range(m)] for _ in range(n)]
nm = [(-1, 0), (0, -1), (1, 0), (0, 1)]
visited[0][0] = 0
q = deque([[0, 0, 0, 0]])
res = 1
while q:
    qlen = len(q)
    for _ in range(qlen):
        x, y, s, c = q.popleft()
        if x == n-1 and y == m-1:
            print(res)
            exit()
        for _x, _y in nm:
            nx = x+_x
            ny = y+_y
            if 0 <= nx < n and 0 <= ny < m:
                if load[nx][ny] == '0' and visited[nx][ny] > c:
                    visited[nx][ny] = c
                    q.append([nx, ny, 1 ^ s, c])
                elif load[nx][ny] == '1':
                    if visited[nx][ny] > c+1:
                        if s == 0:
                            visited[nx][ny] = c+1
                            q.append([nx, ny, 1 ^ s, c+1])
                        else:
                            q.append([x, y, 1 ^ s, c])
    res += 1
print(-1)
```
