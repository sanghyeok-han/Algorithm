### Boggle 9202

```python
import sys
input = lambda: sys.stdin.readline().rstrip()

class Node():
    def __init__(self, value, is_on=False):
        self.value = value
        self.children = {}
        self.is_on = is_on
        
    def append_child_node(self, value):
        if value in self.children:
            return self.children[value]
        else:
            self.children[value] = Node(value)
            return self.children[value]

dx = [-1, -1, -1, 0, 0, 1, 1, 1]
dy = [-1, 0, 1, -1, 1, -1, 0, 1]
        
def dfs(i, j, p_node):
    if li[i][j] in p_node.children:
        node = p_node.children[li[i][j]]
    else:
        return

    stack.append(li[i][j])
    on_stack[i][j] = True
    if node.is_on:
        st.add(''.join(stack))
        
    if len(stack) == 8:
        stack.pop()
        on_stack[i][j] = False
        return
        
    for d in range(8):
        ni = i + dx[d]
        nj = j + dy[d]
        
        if ni < 0 or ni >= 4 or nj < 0 or nj >= 4:
            continue
        if not on_stack[ni][nj]:
            dfs(ni, nj, node)

    stack.pop()
    on_stack[i][j] = False
    return
  
len_to_sc = [0, 0, 0, 1, 1, 2, 3, 5, 11]
    
n = int(input())
words = [input() for _ in range(n)]

root = Node('')
for word in words:
    cur = root
    for alpha in word:
        cur = cur.append_child_node(alpha)
        
    cur.is_on = True
    
input()

t = int(input())
for ti in range(t):
    li = [input() for _ in range(4)]

    st = set()
    on_stack = [[False] * 4 for _ in range(4)]
    stack = []
    for i in range(4):
        for j in range(4):
            dfs(i, j, root)

    r = list(st)
    r.sort(key=lambda x: (-len(x), x))

    max_sc = sum(len_to_sc[len(v)] for v in r)
    max_len_word = r[0]
    count = len(r)

    print(max_sc, max_len_word, count)
    
    if ti + 1 < t:
        input()
```

```python
import sys
input = lambda: sys.stdin.readline().rstrip()

class Node():
    def __init__(self, value):
        self.value = value
        self.children = {}

def dfs(node, depth):
    if depth >= 1:
        print('-' * ((depth - 1) * 2) + node.value)
    
    for key in sorted(node.children):
        dfs(node.children[key], depth + 1)
        
root = Node('')

n = int(input())
for _ in range(n):
    c, *li = input().split()
    c = int(c)
    
    cur = root
    for v in li:
        if v in cur.children:
            cur = cur.children[v]
        else:
            cur.children[v] = Node(v)
            cur = cur.children[v]
            
dfs(root, 0)
```

https://www.acmicpc.net/problem/5670 휴대폰 자판

```python
import sys
input = lambda: sys.stdin.readline().rstrip()

class Node():
    def __init__(self, value):
        self.value = value
        self.children = {}

while True:
    try:
        n = int(input())
        li = [input() for _ in range(n)]        

        root = Node('')

        for s in li:
            c = len(s)

            cur = root
            for v in s:
                if v in cur.children:
                    cur = cur.children[v]
                else:
                    cur.children[v] = Node(v)
                    cur = cur.children[v]

            cur.children[''] = Node('')

        total = 0
        for s in li:
            cur = root.children[s[0]]
            for i in range(1, len(s)):
                v = s[i]
                if len(cur.children) >= 2:
                    total += 1

                cur = cur.children[v]

        total += n
        result = '{:.2f}'.format(total / n)

        print(result)
    except:
        break
```

https://www.acmicpc.net/problem/19585 전설

```python
import io, os, sys
from io import BytesIO, IOBase
 
BUFSIZE = 8192

class FastIO(IOBase):
    newlines = 0
 
    def __init__(self, file):
        self._fd = file.fileno()
        self.buffer = BytesIO()
        self.writable = "x" in file.mode or "r" not in file.mode
        self.write = self.buffer.write if self.writable else None
 
    def read(self):
        while True:
            b = os.read(self._fd, max(os.fstat(self._fd).st_size, BUFSIZE))
            if not b:
                break
            ptr = self.buffer.tell()
            self.buffer.seek(0, 2), self.buffer.write(b), self.buffer.seek(ptr)
        self.newlines = 0
        return self.buffer.read()
 
    def readline(self):
        while self.newlines == 0:
            b = os.read(self._fd, max(os.fstat(self._fd).st_size, BUFSIZE))
            self.newlines = b.count(b"\n") + (not b)
            ptr = self.buffer.tell()
            self.buffer.seek(0, 2), self.buffer.write(b), self.buffer.seek(ptr)
        self.newlines -= 1
        return self.buffer.readline()
 
    def flush(self):
        if self.writable:
            os.write(self._fd, self.buffer.getvalue())
            self.buffer.truncate(0), self.buffer.seek(0)

class IOWrapper(IOBase):
    def __init__(self, file):
        self.buffer = FastIO(file)
        self.flush = self.buffer.flush
        self.writable = self.buffer.writable
        self.write = lambda s: self.buffer.write(s.encode("ascii"))
        self.read = lambda: self.buffer.read().decode("ascii")
        self.readline = lambda: self.buffer.readline().decode("ascii")

sys.stdin, sys.stdout = IOWrapper(sys.stdin), IOWrapper(sys.stdout)
input = lambda: sys.stdin.readline().rstrip()
print = sys.stdout.write

class Node():
    __slots__ = ['value', 'children', 'is_end']
    
    def __init__(self, value):
        self.value = value
        self.children = {}
        self.is_end = False

c, n = map(int, input().split())

colors = [input() for _ in range(c)]
nicks = {input() for _ in range(n)}

root_c = Node('')
for color in colors:
    cur = root_c
    for v in color:
        if v in cur.children:
            cur = cur.children[v]
        else:
            node = Node(v)
            cur.children[v] = node
            cur = node
    cur.is_end = True
                
q = int(input())
for _ in range(q):
    team = input()
    
    ip = False
    
    team_len = len(team)
    cur = root_c
    for i, v in enumerate(team):
        if v in cur.children:
            cur = cur.children[v]
            if cur.is_end:
                if team[i + 1:] in nicks:
                    ip = True
                    break
        else:
            break
            
    if ip:
        print('Yes\n')
    else:
        print('No\n')
```
