### https://atcoder.jp/contests/abc235/tasks/abc235_e E - MST + 1

```python
mst 경로 상에 있는 엣지가 더 적은 값으로 바뀌면 Yes

->

쿼리에 해당하는 것과 기본 구성 엣지를 구분해서 추가하고 정렬함
해당 엣지들로 mst 구성할 때 기본 구성 엣지를 연결할 필요가 있으면 연결하고 쿼리에 해당하는 엣지는 필요 여부만 파악함
```

```python
import sys
input = lambda: sys.stdin.readline().rstrip()

def find(v):
    if v == parent[v]:
        return v
    parent[v] = find(parent[v])
    return parent[v]

def union(a, b):
    if rank[a] >= rank[b]:
        parent[b] = a
        
        if rank[a] == rank[b]:
            rank[a] += 1
    else:
        parent[a] = b
        

n, m, q = map(int, input().split())

parent = list(range(n + 1))
rank = [0] * (n + 1)

edges = []
for _ in range(m):
    a, b, c = map(int, input().split())
    edges.append((a, b, c, -1))
    
for i in range(q):
    a, b, c = map(int, input().split())
    edges.append((a, b, c, i))
    
edges.sort(key=lambda x: x[2])
results = [False] * q
for a, b, c, tp in edges:
    ra = find(a)
    rb = find(b)
    
    if ra != rb:
        if tp == -1:
            union(ra, rb)
        else:
            results[tp] = True

for v in results:
    if v:
        print('Yes')
    else:
        print('No')
```

### https://atcoder.jp/contests/abc215/tasks/abc215_e E - Chain Contestant

```python
# 일단 포기
mod = 998244353

n = int(input())
get_value = lambda x: ord(x) - 65
li = list(map(get_value, input()))

totals = [0] * n
tpc = 2
get_earns = [[0] * tpc for _ in range(n)]

totals[0] = 1
get_earns[0] = [1] * tpc

for i in range(1, n):
    temp = 0
    for j in range(i):
        if li[i] == li[j]:
            temp += totals[j]
        else:
            temp += get_earns[j][li[i]]    
    totals[i] = temp + 1
    
    for tp in range(tpc):
        temp = totals[i]
        
        for j in range(i):
            if li[j] == tp:
                if li[i] != tp:
                    temp -= get_earns[j][li[i]]
                
        get_earns[i][tp] = temp
        
r = sum(totals) % mod

print(r)
```

### https://atcoder.jp/contests/abc154/tasks/abc154_e E - Almost Everywhere Zero

```python
def get_value(s, k):
    s = s.lstrip('0')
    if not s:
        return 0
    
    total = 0
    if k == 1:
        total += (len(s) - 1) * 9
        total += int(s[0])
    else:
        t1 = get_value(s[1:], k - 1)
        t2 = (int(s[0]) - 1) * get_value('9' * (len(s) - 1), k - 1)
        t3 = get_value('9' * (len(s) - 1), k)
        
        total = t1 + t2 + t3
    
    return total
   
s = input()
k = int(input())

r = get_value(s, k)

print(r)
```

### https://atcoder.jp/contests/abc137/tasks/abc137_d D - Summer Vacation

```python
from heapq import heappush, heappop
import sys
input = lambda: sys.stdin.readline().rstrip()

n, m = map(int, input().split())
stack = []
for _ in range(n):
    a, b = map(int, input().split())
    
    a = m - a
    if a < 0:
        continue
    stack.append((a, b))
    
stack.sort(key=lambda x: -x[0])

heap = []
for cur in range(m):
    while stack and stack[-1][0] == cur:
        heappush(heap, stack.pop()[1])
    while len(heap) > cur + 1:
        heappop(heap)
        
r = sum(heap)

print(r)
```

```python
from collections import Counter
from collections import defaultdict as dd
import sys
input = lambda: sys.stdin.readline().rstrip()

def recur(f, t):
    if dp[f][t] != -1:
        return dp[f][t]
    if t < f:
        return 0
    if f == t:
        dp[f][t] = d[(f, t)]
        return dp[f][t]
    
    temp = d[(f, t)]
    dp[f][t] = temp + recur(f + 1, t) + recur(f, t - 1) - recur(f + 1, t - 1)
    
    return dp[f][t]

n, m, q = map(int, input().split())

get_value = lambda x: int(x) - 1
li = [tuple(map(get_value, input().split())) for _ in range(m)]

ct = Counter(li)
d = dd(int)
d.update(ct)

dp = [[-1] * n for _ in range(n)]
recur(0, n - 1)

for _ in range(q):
    p, q = map(get_value, input().split())
    
    print(dp[p][q])
```

```python
from heapq import heappush, heappop
import sys
input = lambda: sys.stdin.readline().rstrip()

inf = float('inf')

def dij(start):
    dists = [inf] * (n + 1)
    dists[start] = 0
    
    min_back = inf
    heap = [(0, start)]
    while heap:
        dist, value = heappop(heap)
        
        if dists[value] < dist:
            continue
        for v, d in li[value]:
            ndist = dist + d
            if v == start:
                min_back = min(min_back, ndist)
                continue
            if ndist < dists[v]:
                dists[v] = ndist
                heappush(heap, (ndist, v))
                
    return min_back

n, m = map(int, input().split())
li = [[] for _ in range(n + 1)]
for _ in range(m):
    a, b, c = map(int, input().split())
    li[a].append((b, c))

for start in range(1, n + 1):
    r = dij(start)
    
    if r == inf:
        r = -1
    
    print(r)
```
