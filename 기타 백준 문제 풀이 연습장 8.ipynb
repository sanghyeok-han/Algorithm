```python
from heapq import heappush, heappop
import sys
input = lambda: sys.stdin.readline().rstrip()

n = int(input())

heap = []
for _ in range(n):
    temp = list(map(int, input().split()))
    for v in temp:
        heappush(heap, v)
        if len(heap) > n:
            heappop(heap)
            
r = heap[0]

print(r)
```

```python
def prime_factor(num):
    r = []
    for mod in range(2, int(num ** 0.5) + 1):
         while True:
            if num % mod == 0:
                num //= mod
                r.append(mod)
            else:
                break
    
    if num != 1:
        r.append(num)
    
    if not r:
        r.append(1)
        
    return r

def get_s_num(num):
    if num == 0:
        s = '.,'
    elif num < 0:
        s = ',' * abs(num)
    else:
        r = []
        used = prime_factor(num)
        for v in used:
            r.append('.' * v)
            
        s = ' '.join(r)
        
    return s
```
