```python
# 누적합 미사용
from collections import deque

n, k = map(int, input().split())
li = list(map(int, input().split()))

total = 0
deq = deque()
for i in range(n - k + 1):
    if deq and deq[0] == i:
        deq.popleft()
    
    li[i] ^= len(deq) % 2
    
    if li[i] == 1:
        li[i] = 0
        total += 1
        deq.append(i + k)
        
ip = True
for i in range(n - k + 1, n):
    if deq and deq[0] == i:
        deq.popleft()
    
    li[i] ^= len(deq) % 2
    if li[i] == 1:
        ip = False
        break
        
if not ip:
    print('Insomnia')
else:
    print(total)
```
