### 단절점 11266

```python
import sys
sys.setrecursionlimit(int(1e5) + 100)
# input = lambda: sys.stdin.readline().rstrip()

def dfs(value, is_root):
    global next_num
    
    child_count = 0
    min_visited = vt[value]
    for v in li[value]:
        if vt[v] == 0:
            vt[v] = next_num
            next_num += 1
            temp = dfs(v, False)
         
            if not is_root and temp >= vt[value]:
                is_ans[value] = True
            
            min_visited = min(min_visited, temp)
            child_count += 1
        else:
            min_visited = min(min_visited, vt[v])
    
    if is_root and child_count >= 2:
        is_ans[value] = True
        
    return min_visited
        
n, m = map(int, input().split())

li = [[] for _ in range(n + 1)]
for _ in range(m):
    a, b = map(int, input().split())
    
    li[a].append(b)
    li[b].append(a)
    
vt = [0] * (n + 1)
next_num = 1
is_ans = [False] * (n + 1)
for i in range(1, n + 1):
    if vt[i] != 0:
        continue
    vt[i] = next_num
    next_num += 1
    
    dfs(i, True)

r = []
for i in range(1, n + 1):
    if is_ans[i]:
        r.append(i)
        
print(len(r))
print(*r)
```

### 단절선 11400

```python
import sys
input = lambda: sys.stdin.readline().rstrip()

## PYRIVAL BOOTSTRAP
# https://github.com/cheran-senthil/PyRival/blob/master/pyrival/misc/bootstrap.py
# This decorator allows for recursion without actually doing recursion
## @bootstrap, yield when getting and returning value in recursive functions, end of functions
from types import GeneratorType

def bootstrap(f, stack=[]):
    def wrappedfunc(*args, **kwargs):
        if stack:
            return f(*args, **kwargs)
        else:
            to = f(*args, **kwargs)
            while True:
                if type(to) is GeneratorType:
                    stack.append(to)
                    to = next(to)
                else:
                    stack.pop()
                    if not stack:
                        break
                    to = stack[-1].send(to)
            return to
    return wrappedfunc

@bootstrap
def dfs(value, to_evade):
    global next_num
    
    min_vt = vt[value]
    for v in li[value]:
        if vt[v] != 0:
            if v != to_evade:
                min_vt = min(min_vt, vt[v])
        else:
            vt[v] = next_num
            next_num += 1
            
            temp = yield dfs(v, value)
            
            if temp > vt[value]:
                a, b = value, v
                if a > b:
                    a, b = b, a
                r.append((a, b))
            
            min_vt = min(min_vt, temp)
            
    yield min_vt

n, m = map(int, input().split())
li = [[] for _ in range(n + 1)]
for _ in range(m):
    a, b = map(int, input().split())
    
    li[a].append(b)
    li[b].append(a)
    
next_num = 1
r = []
vt = [0] * (n + 1)
for i in range(1, n + 1):
    if vt[i]:
        continue
    vt[i] = next_num
    next_num += 1
    dfs(i, -1)

r.sort()

print(len(r))
for v in r:
    print(*v)
```

https://www.acmicpc.net/problem/11266 단절점 (다시 풀음)

```python
import sys
sys.setrecursionlimit(100100)
# input = lambda: sys.stdin.readline().rstrip()

def dfs(value, parent):
    global nth
    
    vt[value] = True
    vt_nth[value] = nth
    can_vt_mins[value] = nth
    nth += 1
    
    can_vt_child_c = 0
    
    for v in li[value]:
        if v == parent:
            continue
        if vt[v]:
            can_vt_mins[value] = min(can_vt_mins[value], vt_nth[v])
            continue
        can_vt_child_c += 1
        
        dfs(v, value)
        
        can_vt_mins[value] = min(can_vt_mins[value], can_vt_mins[v])
        
        if can_vt_mins[v] >= vt_nth[value]:
            is_aps[value] = True
        
    can_vt_child_cs[value] = can_vt_child_c

n, e = map(int, input().split())
li = [[] for _ in range(n + 1)]

for _ in range(e):
    u, v = map(int, input().split())
    li[u].append(v)
    li[v].append(u)
    
vt = [False] * (n + 1)
vt_nth = [-1] * (n + 1)
can_vt_mins = [n + 1] * (n + 1)
can_vt_child_cs = [0] * (n + 1)
is_aps = [False] * (n + 1) # is articulation point

nth = 1
for i in range(1, n + 1):
    if vt[i]:
        continue
    dfs(i, 0)
        
    is_aps[i] = False
    if can_vt_child_cs[i] >= 2:
        is_aps[i] = True

result = [i for i in range(1, n + 1) if is_aps[i]]
    
print(len(result))
print(*result)
```

https://www.acmicpc.net/problem/6672 Electricity

```python
import io, os, sys
sys.setrecursionlimit(11000)
input = io.BytesIO(os.read(0, os.fstat(0).st_size)).readline

def dfs(value, parent):
    global nth
    
    vt[value] = True
    vt_nth[value] = nth
    can_vt_mins[value] = nth
    nth += 1
    
    divided_c = 0
    
    for v in li[value]:
        if v == parent:
            continue
        if vt[v]:
            can_vt_mins[value] = min(can_vt_mins[value], vt_nth[v])
            continue
        dfs(v, value)
        
        can_vt_mins[value] = min(can_vt_mins[value], can_vt_mins[v])
        
        if can_vt_mins[v] >= vt_nth[value]:
            divided_c += 1
            
    divided_cs[value] = divided_c

    
while True:
    n, e = map(int, input().split())
    if n == e == 0:
        break
    
    li = [[] for _ in range(n + 1)]

    for _ in range(e):
        u, v = map(int, input().split())
        li[u].append(v)
        li[v].append(u)

    vt = [False] * n
    vt_nth = [-1] * n
    can_vt_mins = [n + 1] * n
    divided_cs = [0] * n

    component_c = 0
    nth = 1
    for i in range(n):
        if vt[i]:
            continue
        component_c += 1

        dfs(i, 0)

        divided_cs[i] -= 1

    # 루트가 아닌 노드들은 +1을 해줌
    divided_cs = [v + 1 for v in divided_cs]

    result = max(divided_cs) + component_c - 1

    print(result)
```
