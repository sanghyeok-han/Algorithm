```python
from math import log2, ceil
import io, os, sys
input = io.BytesIO(os.read(0,os.fstat(0).st_size)).readline

def get_ari_sum(a, n, d):
    return n * (2 * a + (n - 1) * d) // 2

def construct(node, s, e):
    if s == e:
        segtree[node] = li[s]
        return segtree[node]
    mid = (s + e) // 2
    
    t1 = construct(node * 2, s, mid)
    t2 = construct(node * 2 + 1, mid + 1, e)
    
    segtree[node] = t1 + t2
    
    return segtree[node]

def query(node, s, e, pending_a, pending_d):
    segtree[node] += get_ari_sum(pending_a, e - s + 1, pending_d)
    pendings[node][0] += pending_a
    pendings[node][1] += pending_d
    
    if e < qs or s > qe:
        return 0
    if qs <= s <= e <= qe:
        return segtree[node]
    else:
        pending_a, pending_d = pendings[node]
        pendings[node] = [0, 0]
        
        mid = (s + e) // 2
        
        t1 = query(node * 2, s, mid, pending_a, pending_d)
        t2 = query(node * 2 + 1, mid + 1, e, pending_a + (mid + 1 - s) * pending_d, pending_d)
        
        return t1 + t2
    
def update_range(node, s, e, to_add_a):
    if ts <= s <= e <= te:
        
        pendings[node][0] += to_add_a
        pendings[node][1] += 1
        
        to_assign = get_ari_sum(to_add_a, e - s + 1, 1)
        
        segtree[node] += to_assign
        return to_assign, e - s + 1
    elif e < ts or s > te:
        return 0, 0
    else:
        mid = (s + e) // 2
        
        if s != e:   
            t1, vt_1 = update_range(node * 2, s, mid, to_add_a)
            t2, vt_2 = update_range(node * 2 + 1, mid + 1, e, to_add_a + vt_1)
            
            segtree[node] += t1 + t2
            
            return t1 + t2, vt_1 + vt_2

n = int(input())
li = list(map(int, input().split()))

size = 2 ** (ceil(log2(n)) + 1) + 1
segtree = [0] * size
pendings = [[0] * 2 for _ in range(size)]

construct(1, 0, n - 1)

q = int(input())
for _ in range(q):
    tp, *qu = map(int, input().split())
    
    if tp == 1:
        l, r = qu
        
        ts, te = l - 1, r - 1
        to_add_a = 1
        update_range(1, 0, n - 1, to_add_a)
    else:
        x = qu[0]
        qs = qe = x - 1
        
        result = query(1, 0, n - 1, 0, 0)
        
        print(result)
```
