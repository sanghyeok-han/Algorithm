```python
n = int(input())
li = list(map(int, input().split()))

cs = [0] * 1002
for v in li:
    cs[v] += 1
    
r = []
for i in range(1001):
    if not cs[i]:
        continue
    if cs[i + 1]:
        target = -1
        for j in range(i + 2, 1001):
            if cs[j] >= 1:
                target = j
                break
        if target != -1:
            r.extend([i] * cs[i])
            cs[i] = 0
            r.append(target)
            cs[target] -= 1
        else:
            r.extend([i + 1] * cs[i + 1])
            cs[i + 1] = 0
            r.extend([i] * cs[i])
            cs[i] = 0
    else:
        r.extend([i] * cs[i])
        cs[i] = 0
    
print(*r)
```

### 카드 넣기 1426

```python
def check(card_i, can_i):
    used_can_indices.add(can_i)
    used_card_indices.add(card_i)
    
    unused_cans = [v for i, v in enumerate(cans) if i not in used_can_indices]
    unused_cards = [v[0] for i, v in enumerate(cards) if i not in used_card_indices]
    unused_cans.sort()
    unused_cards.sort()
    
    ip = True
    for i in range(len(unused_cans)):
        if unused_cards[i] < unused_cans[i]:
            ip = False
            break
            
    used_can_indices.discard(can_i)
    used_card_indices.discard(card_i)
    
    if ip:
        return True
    else:
        return False

n = int(input())
cards = []
for _ in range(n):
    a, b = input().split()
    a = int(a)
    cards.append((a, b))
    
cans = list(map(int, input().split()))

ncans = sorted(cans)
ncards = sorted([v[0] for v in cards])

ip = True
for i in range(n):
    if ncards[i] < ncans[i]:
        ip = False
        break
        
if not ip:
    print(-1)
else:
    cards.sort(key=lambda x: (x[1], x[0]))
    
    r = [''] * n
    used_card_indices = set()
    used_can_indices = set()
    for card_i in range(n):
        value, char = cards[card_i]
        
        for can_i in range(n):
            if can_i in used_can_indices:
                continue
            if value >= cans[can_i]:
                if check(card_i, can_i):
                    r[can_i] = char
                    used_card_indices.add(card_i)
                    used_can_indices.add(can_i)
                    break
        
    print(''.join(r))
```

```python
# 복잡하지만 틀렸음
from itertools import product

dx = [-1, -1, -1, 0, 0, 0, 1, 1, 1]
dy = [-1, 0, 1, -1, 0, 1, -1, 0, 1]

def rotate(li):
    global n
    global m
    
    nli = [[0] * n for _ in range(m)]
    for i in range(m):
        for j in range(n):
            nli[i][j] = li[n - 1 - j][i]
            
    n, m = m, n
    li = nli.copy()
    return li

def change(x, y):
    for d in range(9):
        nx = x + dx[d]
        ny = y + dy[d]
        
        if nx < 0 or nx >= n or ny < 0 or ny >= m:
            continue
        li[nx][ny] ^= 1

def dfs(x, y, count):
    global min_count
    
    if x == n - 1 and y == m - 1:
        is_changed = False
        if n >= 2 and m >= 2:
            if not li[x - 1][y - 1]:
                change(x, y)
                is_changed = True
        elif n == 1 and m == 1:
            if not li[x][y]:
                change(x, y)
                is_changed = True
        elif n >= 2:
            if not li[x - 1][y]:
                change(x, y)
                is_changed = True
        elif m >= 2:
            if not li[x][y - 1]:
                change(x, y)
                is_changed = True
        
        if n >= 2 and li[x - 1][y] and 0 not in li[x]:
            min_count = min(min_count, count + is_changed)
        if n == 1 and 0 not in li[x]:
            min_count = min(min_count, count + is_changed)
        if m == 1:
            ip = True
            for v in li:
                if not v:
                    ip = False
                    break
            if ip:
                min_count = min(min_count, count + is_changed)
        
        if is_changed:
            change(x, y)
        return

    is_changed = False
    
    if y == 0:
        if m == 1:
            if not li[x - 1][y]:                
                change(x, y)
                is_changed = True
                dfs(x + 1, y, count + 1)
            else:
                dfs(x + 1, y, count)
        else:
            dfs(x, y + 1, count)
            change(x, y)
            is_changed = True
            dfs(x, y + 1, count + 1)
    else:
        if not li[x - 1][y - 1]:
            change(x, y)
            is_changed = True
            if y == m - 1:
                if li[x - 1][y]:
                    dfs(x + 1, 0, count + 1)
            else:
                dfs(x, y + 1, count + 1)
        else:
            if y == m - 1:
                if li[x - 1][y]:
                    dfs(x + 1, 0, count)
            else:
                dfs(x, y + 1, count)
                
    if is_changed:
        change(x, y)

def dfs2(x, y, count):
    global min_count
    
    if x == n - 1 and y == 0:
        is_changed = False
        if n >= 2 and m >= 2:
            if not li[x - 1][y + 1]:
                change(x, y)
                is_changed = True
        elif n == 1 and m == 1:
            if not li[x][y]:
                change(x, y)
                is_changed = True
        elif n >= 2:
            if not li[x - 1][y]:
                change(x, y)
                is_changed = True
        elif m >= 2:
            if not li[x][y + 1]:
                change(x, y)
                is_changed = True
        
        if n >= 2 and li[x - 1][y] and 0 not in li[x]:
            min_count = min(min_count, count + is_changed)
        if n == 1 and 0 not in li[x]:
            min_count = min(min_count, count + is_changed)
        if m == 1:
            ip = True
            for v in li:
                if not v:
                    ip = False
                    break
            if ip:
                min_count = min(min_count, count + is_changed)
        
        if is_changed:
            change(x, y)
        return

    is_changed = False
    
    if y == m - 1:
        if m == 1:
            if not li[x - 1][y]:                
                change(x, y)
                is_changed = True
                dfs2(x + 1, y, count + 1)
            else:
                dfs2(x + 1, y, count)
        else:
            dfs2(x, y - 1, count)
            change(x, y)
            is_changed = True
            dfs2(x, y - 1, count + 1)
    else:
        if not li[x - 1][y + 1]:
            change(x, y)
            is_changed = True
            if y == 0:
                if li[x - 1][y]:
                    dfs2(x + 1, 0, count + 1)
            else:
                dfs2(x, y - 1, count + 1)
        else:
            if y == 0:
                if li[x - 1][y]:
                    dfs2(x + 1, 0, count)
            else:
                dfs2(x, y - 1, count)
                
    if is_changed:
        change(x, y)
            

n, m = map(int, input().split())
li = []
for _ in range(n):
    temp = input()
    temp_li = []
    for v in temp:
        if v == '*':
            temp_li.append(1)
        else:
            temp_li.append(0)
            
    li.append(temp_li)

min_count = float('inf')

for _ in range(4):
    for case in product(*[range(2)] * m):
        count = 0
        for j in range(m):
            if case[j] == 1:
                count += 1
                change(0, j)

        if n >= 2:
            dfs(1, 0, count)
            dfs2(1, m - 1, count)
        else:
            dfs(0, m - 1, count)
            dfs2(0, 0, count)

        for j in range(m):
            if case[j] == 1:
                change(0, j)
    li = rotate(li)

if min_count == float('inf'):
    print(-1)
else:
    print(min_count)
```
