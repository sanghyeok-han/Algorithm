```python
def ccw(x1, y1, x2, y2, x3, y3):
    temp = x1 * y2 + x2 * y3 + x3 * y1 - x2 * y1 - x3 * y2 - x1 * y3
    
    if temp < 0:
        return -1
    elif temp > 0:
        return 1
    else:
        return 0

x1, y1 = map(int, input().split())
x2, y2 = map(int, input().split())
x3, y3 = map(int, input().split())

r = ccw(x1, y1, x2, y2, x3, y3)

print(r)
```

https://www.acmicpc.net/problem/17386 선분 교차 1 (다시 풀음)

```python
# 틀렸음
def ccw(x1, y1, x2, y2, x3, y3):
    temp = x1 * y2 + x2 * y3 + x3 * y1 - x2 * y1 - x3 * y2 - x1 * y3
    
    if temp < 0:
        return -1
    elif temp > 0:
        return 1
    else:
        return 0

def lsi(sx1, sy1, ex1, ey1, sx2, sy2, ex2, ey2):
    t1 = ccw(sx2, sy2, ex2, ey2, sx1, sy1)
    t2 = ccw(sx2, sy2, ex2, ey2, ex1, ex1)
    t3 = ccw(sx1, sy1, ex1, ey1, sx2, sy2)
    t4 = ccw(sx1, sy1, ex1, ey1, ex2, ey2)
    
    if t1 * t2 == 0 and t3 * t4 == 0:
        if (sx1, sy1) > (ex1, ey1):
            sx1, sy1, ex1, ey1 = sx1, sy1, ex1, ey1
        if (sx2, sy2) > (ex2, ey2):
            sx2, sy2, ex2, ey2 = sx2, sy2, ex2, ey2
        
        if (sx2, sy2) <= (ex1, ey1) and (sx1, sy1) <= (ex2, ey2):
            return True
        else:
            return False
    
    if not (t1 * t2 <= 0 and t3 * t4 <= 0):
        return False
    else:
        return True

sx1, sy1, ex1, ey1 = map(int, input().split())
sx2, sy2, ex2, ey2 = map(int, input().split())

r = int(lsi(sx1, sy1, ex1, ey1, sx2, sy2, ex2, ey2))

print(r)
```

https://www.acmicpc.net/problem/3679 단순 다각형

```python
import functools
import io, os, sys
# input = io.BytesIO(os.read(0,os.fstat(0).st_size)).readline
            
def ccw(x1, y1, x2, y2, x3, y3):
    temp = x1 * y2 + x2 * y3 + x3 * y1 - x2 * y1 - x3 * y2 - x1 * y3
    
    if temp < 0:
        return -1
    elif temp > 0:
        return 1
    else:
        return 0
    
def compare(p1, p2):
    x1, y1, _ = p1
    x2, y2, _ = p2
    
    r = ccw(x1, y1, cri_x, cri_y, x2, y2)
    if r == 1:
        return 1
    elif r == -1:
        return -1
    else:
        if y1 < y2:
            return -1
        elif y1 > y2:
            return 1
        if x1 < x2:
            return -1
        else:
            return 1

for _ in range(int(input())):
    n, *li = map(int, input().split())

    nli = []
    for i in range(0, n * 2, 2):
        nli.append([li[i], li[i + 1], i // 2])

    nli.sort(key=lambda x: (-x[1], -x[0]))

    cri_x, cri_y, cri_index = nli.pop()
    n -= 1

    nli.sort(key=functools.cmp_to_key(compare))

    is_same = False
    temps = []
    while len(nli) >= 2:
        x1, y1 = nli[-2][0], nli[-2][1]
        x2, y2 = nli[-1][0], nli[-1][1]
        
        r = ccw(cri_x, cri_y, x1, y1, x2, y2)
        if r == 0:
            is_same = True
            temps.append(nli.pop())
        else:
            break
    
    if is_same:
        temps.append(nli.pop())
        
    temps.sort(key=lambda x: -x[1])
    nli.extend(temps)
    
    result = [cri_index]
    for v in nli:
        result.append(v[2])

    print(*result)
```

https://www.acmicpc.net/problem/4225 쓰레기 슈트

```python
import functools
import io, os, sys
input = io.BytesIO(os.read(0,os.fstat(0).st_size)).readline

inf = float('inf')

def get_area(a, b, c):
    x1, y1 = a
    x2, y2 = b
    x3, y3 = c
    
    t1 = x1 * y2 + x2 * y3 + x3 * y1
    t2 = x2 * y1 + x3 * y2 + x1 * y3
    
    area = abs(t1 - t2) / 2
    
    return area

def get_dist(a, b):
    x1, y1 = a
    x2, y2 = b
    
    return ((x1 - x2) ** 2 + (y1 - y2) ** 2) ** 0.5

def ccw(x1, y1, x2, y2, x3, y3):
    temp = x1 * y2 + x2 * y3 + x3 * y1 - x2 * y1 - x3 * y2 - x1 * y3
    
    if temp < 0:
        return -1
    elif temp > 0:
        return 1
    else:
        return 0
    
def compare(p1, p2):
    x1, y1 = p1
    x2, y2 = p2
    
    r = ccw(x1, y1, cri_x, cri_y, x2, y2)
    if r == 1:
        return 1
    elif r == -1:
        return -1
    else:
        if y1 < y2:
            return -1
        elif y1 > y2:
            return 1
        if x1 < x2:
            return -1
        else:
            return 1

tn = 1
while True:
    n = int(input())
    if n == 0:
        break
    
    li = []
    for _ in range(n):
        x, y = map(int, input().split())

        li.append((x * 10000000, y * 10000000))

    li.sort(key=lambda x: (-x[1], -x[0]))
    cri_x, cri_y = li.pop()
    n -= 1

    li.sort(key=functools.cmp_to_key(compare))

    stack = [[cri_x, cri_y], [li[0][0], li[0][1]]]
    for i in range(1, n):
        while len(stack) >= 2:
            x1, y1 = stack[-2]
            x2, y2 = stack[-1]
            x3, y3 = li[i]

            r = ccw(x1, y1, x2, y2, x3, y3)
            if r != 1:
                stack.pop()
            else:
                break

        stack.append(li[i])

    hull = stack
    hull_len = len(hull)
    min_r = inf

    for i in range(hull_len):
        ai = i
        bi = i + 1
        if bi == hull_len:
            bi = 0

        length = get_dist(hull[ai], hull[bi])

        max_dist = 0
        for ci in range(hull_len):
            if ci == ai or ci == bi:
                continue

            area = get_area(hull[ai], hull[bi], hull[ci])

            dist = area * 2 / length
            max_dist = max(max_dist, dist)

        min_r = min(min_r, max_dist)

    if min_r % 100000:
        min_r += 100000
    min_r //= 100000
    min_r /= 100
    result = '{:.2f}'.format(min_r)

    print(f'Case {tn}:', result)
    
    tn += 1
```
