```python
n, m = map(int, input().split())

if n == 1:
    r = [[0] * m for _ in range(n)]
    r[0][0] = 1
    diff = 1
    for j in range(1, m):
        r[0][j] = r[0][j - 1] + diff
        diff += 1
elif m == 1:
    r = [[0] * m for _ in range(n)]
    r[0][0] = 1
    diff = 1
    for i in range(1, n):
        r[i][0] = r[i - 1][0] + diff
        diff += 1
else:
    rows = [[0] * m for _ in range(n - 1)]
    cols = [[0] * (m - 1) for _ in range(n)]

    x, y = 1, 0

    d = 0 # 0: 위쪽, 오른쪽, 1: 왼쪽, 아래쪽
    cur_move = 0 # 0: row 이동, 1: col 이동

    diff = 1
    end = 2 * n * m - n - m
    while diff <= end:
        if d == 0 and cur_move == 0:
            nx, ny = x - 1, y
            
            if nx < 0:
                nx += 1
                ny += 1

                if ny < m:
                    x, y = nx, ny
                    d = 1
                    cur_move = 1
                else:
                    ny -= 1
                    x, y = nx, ny
                    d = 1
                    cur_move = 0
                continue
                
            x, y = nx, ny
            rows[x][y] = diff
            diff += 1
            cur_move = 1
        elif d == 0 and cur_move == 1:
            nx, ny = x, y + 1
            
            if ny >= m:
                ny -= 1
                nx += 1
                x, y = nx, ny
                d = 1
                cur_move = 0
                continue
                
            x, y = nx, ny
            cols[x][y - 1] = diff
            diff += 1
            cur_move = 0
        elif d == 1 and cur_move == 0:
            nx, ny = x + 1, y
            
            if nx >= n:
                nx -= 1
                ny += 1
                x, y = nx, ny
                d = 0
                cur_move = 1
                continue
                
            x, y = nx, ny
            rows[nx - 1][ny] = diff
            diff += 1
            cur_move = 1
        elif d == 1 and cur_move == 1:        
            nx, ny = x, y - 1
            
            if ny < 0:
                ny += 1
                nx += 1

                if nx < n:
                    x, y = nx, ny
                    d = 0
                    cur_move = 0
                else:
                    nx -= 1
                    d = 0
                    cur_move = 1
                continue

            x, y = nx, ny
            cols[x][y] = diff
            diff += 1
            d = 1
            cur_move = 0

    r = [[0] * m for _ in range(n)]
    r[0][0] = 1
    r[1][0] = r[0][0] + rows[0][0]
    r[0][1] = r[0][0] + cols[0][0]

    cases = [[1, 1], [1, -1], [-1, 1], [-1, -1]]
    for i in range(n - 1):
        for j in range(m - 1):
            if r[i + 1][j + 1]:
                continue
            if r[i + 1][j] and r[i][j + 1]:
                r[i + 1][j + 1] = r[i + 1][j] + cols[i + 1][j]
                continue
            if r[i + 1][j]:
                for case in cases:
                    c1 = r[i][j] + cols[i][j] * case[0]
                    c2 = r[i + 1][j] + cols[i + 1][j] * case[1]

                    if abs(c1 - c2) == rows[i][j + 1]:
                        r[i][j + 1] = c1
                        r[i + 1][j + 1] = c2
                        break
                continue
            if r[i][j + 1]:
                for case in cases:
                    c1 = r[i][j] + rows[i][j] * case[0]
                    c2 = r[i][j + 1] + rows[i][j + 1] * case[1]

                    if abs(c1 - c2) == cols[i + 1][j]:
                        r[i + 1][j] = c1
                        r[i + 1][j + 1] = c2
                        break            

for v in r:
    print(*v)
```
