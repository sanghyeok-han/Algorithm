```python
from collections import defaultdict, deque
import re
import sys
input = sys.stdin.readline

while True:
    n = int(input())
    if n == 0:
        break
    
    pre = defaultdict(list)
    post = defaultdict(list)
    indegree = defaultdict(int)
    total_member = defaultdict(set)

    for i in range(n):
        value = re.findall("\w+", input())
        academy = value[0]

        if i == 0:
            target = academy

        for v in value[1:]:
            pre[academy].append(v)
            post[v].append(academy)
            total_member[academy].add(v)


    groups = set(total_member.keys())
    for key, value in total_member.items():
        indegree[key] = len(value & groups)


    queue = deque()
    for key, value in indegree.items():
        if value == 0:
            queue.append(key)

    while queue:
        academy = queue.popleft()

        for po in post[academy]:
            total_member[po].remove(academy)
            total_member[po] = total_member[po] | total_member[academy]
            indegree[po] -= 1

            if indegree[po] == 0:
                queue.append(po)

    print(len(total_member[target]))
```

```python
from collections import deque


T = int(input())
for _ in range(T):
    k, m, p = map(int, input().split())

    indegree = [0] * (m + 1)
    pre = [[] for _ in range(m + 1)]
    post = [[] for _ in range(m + 1)]
    strahler = [1] * (m + 1)

    for _ in range(p):
        a, b = map(int, input().split())
        pre[b].append(a)
        post[a].append(b)
        indegree[b] += 1

    queue = deque()
    for i in range(1, m + 1):
        if indegree[i] == 0:
            queue.append(i)

    while queue:
        value = queue.popleft()

        max_value = 0
        max_count = 0
        for v in pre[value]:
            if strahler[v] > max_value:
                max_value = strahler[v]
                max_count = 1
            elif strahler[v] == max_value:
                max_count += 1
        if max_count > 1:
            max_value += 1

        if max_value == 0:
            max_value = 1

        strahler[value] = max_value

        for v in post[value]:
            indegree[v] -= 1
            if indegree[v] == 0:
                queue.append(v)

    print(k, strahler[m])
```

```python
from collections import deque
import sys
# input = sys.stdin.readline

n, m = map(int, input().split())

pre = [set() for _ in range(n + 1)]
post = [set() for _ in range(n + 1)]

for _ in range(m):
    li = list(map(int, input().split()))
    if li[0] <= 1:
        continue
    
    del li[0]
    for i in range(len(li) - 1):
        pre[li[i + 1]].add(li[i])
        post[li[i]].add(li[i + 1])
    
indegree = [len(v) for v in pre]
queue = deque()

for i in range(1, n + 1):
    if indegree[i] == 0:
        queue.append(i)
    

result_sequence = [] 

while queue:
    value = queue.popleft()
    result_sequence.append(value)
    
    for v in post[value]:
        indegree[v] -= 1
        if indegree[v] == 0:
            queue.append(v)
    
if len(result_sequence) == n:
    print("\n".join(map(str, result_sequence)))
else:
    print(0)
```

```python
from collections import deque
import sys
#input = sys.stdin.readline

n = int(input())
m = int(input())

pre = [[] for _ in range(n + 1)]
post = [[] for _ in range(n + 1)]
indegree = [0] * (n + 1)
for _ in range(m):
    a, b, num = list(map(int, input().split()))
    pre[a].append((b, num))
    post[b].append(a)
    indegree[a] += 1

    
d = dict()
queue = deque()
for i in range(1, n + 1):
    if indegree[i] == 0:
        d.update({i: 0})
        queue.append(i)
    
total_needed = [d.copy() for _ in range(n + 1)]

for key in d:
    total_needed[key][key] = 1

while queue:
    value = queue.popleft()

    for pr, num in pre[value]:
        for base, base_count in total_needed[pr].items():
            total_needed[value][base] += base_count * num
    
    for v in post[value]:
        indegree[v] -= 1
        if indegree[v] == 0:
            queue.append(v)
            
result_li = list(total_needed[n].items())
result_li.sort()

for v, count in result_li:
    print(v, count)
```

### Cow Jogging 6172
- 다익스트라, 위상 정렬

```python
# 틀렸음, 일단 포기
from collections import deque
import sys
# input = sys.stdin.readline

n, m, k = map(int, input().split())

pre = [[] for _ in range(n + 1)]
post = [[] for _ in range(n + 1)]
indegree = [0] * (n + 1)
total_distance = [[] for _ in range(n + 1)]

for _ in range(m):
    x, y, d = map(int, input().split())
    post[x].append(y)
    pre[y].append((x, d))
    indegree[y] += 1
    
    
if n == 1:
    result_li = [0]
else:
    queue = deque()
    queue.append(n)

    while queue:
        value = queue.popleft()

        if value == n:
            total_case = [0]
        else:
            total_case = []

        for pr_num, pr_d in pre[value]:
            total_case.extend(list(map(lambda x: x + pr_d, total_distance[pr_num])))
        total_case.sort()
        total_case = total_case[:k]

        total_distance[value] = total_case

        for po_num in post[value]:
            indegree[po_num] -= 1
            if indegree[po_num] == 0:
                queue.append(po_num)


    result_li = total_distance[1]
    
if result_li:
    print("\n".join(map(str, result_li)))
remain = k - len(result_li)

while remain >= 1:
    print(-1)
    remain -= 1
```

### 임계경로 1948 (다시 풀음)

```python
from collections import deque
import sys
input = lambda: sys.stdin.readline().rstrip()

n = int(input())
m = int(input())

pres = [[] for _ in range(n + 1)]
posts = [[] for _ in range(n + 1)]
indegrees = [0] * (n + 1)
for _ in range(m):
    a, b, t = map(int, input().split())
    
    posts[a].append(b)
    pres[b].append([a, t, 0])
    indegrees[b] += 1
    
s, e = map(int, input().split())
    
queue = deque()
queue.append(s)
max_spents = [0] * (n + 1)
while queue:
    value = queue.popleft()
    
    temp_max = 0
    temp_max_edges = []
    for i in range(len(pres[value])):
        pre, t, _ = pres[value][i]
        spent = max_spents[pre] + t
        if spent > temp_max:
            temp_max = spent
            temp_max_edges = [i]
        elif spent == temp_max:
            temp_max_edges.append(i)
            
    max_spents[value] = temp_max
    
    for index in temp_max_edges:
        pres[value][index][2] = 1
            
    for post in posts[value]:
        indegrees[post] -= 1
        if not indegrees[post]:
            queue.append(post)

vt = [False] * (n + 1)
vt[e] = True
queue = deque()
queue.append(e)
count = 0
while queue:
    value = queue.popleft()
    
    for i in range(len(pres[value])):
        if pres[value][i][2]:
            count += 1
            pre = pres[value][i][0]
            if vt[pre]:
                continue
            vt[pre] = True
            queue.append(pre)
            
print(max_spents[e])
print(count)
```

```python
각각이 시작점일 경우를 모두 확인해보고 하나라도 되는 것이 있으면 YES를 출력함
->
양방향인 경우는 없는 취급하고 확인함
```

```python
from collections import deque

n = int(input())
li = [input() for _ in range(n)]

st = set()
for i in range(n):
    for j in range(n):
        if li[i][j] == 'Y':
            if (j, i) in st:
                st.discard((j, i))
            else:
                st.add((i, j))
    
posts = [[] for _ in range(n)]
indegree = [0] * n
for s, e in st:
    indegree[e] += 1
    posts[s].append(e)
    
queue = deque()
for i in range(n):
    if indegree[i] == 0:
        queue.append(i)

vt_c = 0
while queue:
    value = queue.popleft()
    
    vt_c += 1
    for v in posts[value]:
        indegree[v] -= 1
        if indegree[v] == 0:
            queue.append(v)
            
if vt_c == n:
    print('YES')
else:
    print('NO')
```
