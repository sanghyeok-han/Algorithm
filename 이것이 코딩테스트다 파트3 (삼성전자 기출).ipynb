### 청소년 상어

```python
# 기존 코드
from collections import deque
from copy import deepcopy

space = [[0] * 4 for _ in range(4)] # 각 자리별 물고기 번호 관리
current_state = [[] for _ in range(17)] # 각 물고기별 상태 관리 [x, y, direction]
for i in range(4):
    li = list(map(int, input().split()))
    
    for j in range(0, 8, 2):
        current_state[li[j]] = [i, j // 2, li[j + 1]]
        space[i][j // 2] = li[j]

init_score = space[0][0]
max_score = 0

fish_info = current_state[space[0][0]]
# 상어는 current_state에서는 0으로 관리하고 space에서는 -1로 관리
current_state[0] = [0, 0, fish_info[2]]
current_state[space[0][0]] = []
space[0][0] = -1

def get_next_coord(x, y, d):
    if d == 1:
        return x - 1, y
    elif d == 2:
        return x - 1, y - 1
    elif d == 3:
        return x, y - 1
    elif d == 4:
        return x + 1, y - 1
    elif d == 5:
        return x + 1, y
    elif d == 6:
        return x + 1, y + 1
    elif d == 7:
        return x, y + 1
    else:
        return x - 1, y + 1
    
def get_next_direction(d):
    if d == 8:
        return 1
    else:
        return d + 1
    
def move_fish(space, current_state):
    for fish_num in range(1, 17):
        # 해당 물고기가 존재하는 경우
        if current_state[fish_num]:
            # 각 물고기마다 유효한 자리를 찾음
            is_possible = False
            for _ in range(8):
                x, y, d = current_state[fish_num]
                
                nx, ny = get_next_coord(x, y, d)
                # 이동할 자리가 유효하지 않은 경우 반시계 방향으로 45도 회전함
                if nx < 0 or 4 <= nx or ny < 0 or 4 <= ny or space[nx][ny] == -1:
                    current_state[fish_num][2] = get_next_direction(d)
                else:
                    is_possible = True
                    break
            
            if is_possible:
                # 물고기가 이동하는 칸이 비어있는 경우
                if space[nx][ny] == 0:
                    space[nx][ny] = fish_num
                    space[x][y] = 0
                    current_state[fish_num][0], current_state[fish_num][1] = nx, ny
                # 물고기가 이동하는 칸에 다른 물고기가 있는 경우
                else:
                    already_occupied_fish_num = space[nx][ny]

                    space[x][y], space[nx][ny] = space[nx][ny], space[x][y]
                    current_state[fish_num][0], current_state[already_occupied_fish_num][0] = current_state[already_occupied_fish_num][0], current_state[fish_num][0]
                    current_state[fish_num][1], current_state[already_occupied_fish_num][1] = current_state[already_occupied_fish_num][1], current_state[fish_num][1]
#         print(1)
#         for v in space:
#             print(v)
    return space, current_state          

queue = deque()
queue.append((deepcopy(space), deepcopy(current_state), init_score))

while queue:
    space, current_state, score = queue.popleft()
    
#     print(0)
#     for v in space:
#         print(v)
    space, current_state = move_fish(space, current_state)
    
    x, y, d = current_state[0] # 상어의 정보
    
    # 상어를 이동할 때마다 먹을 수 있으면 먹을지 말지 선택 가능
    nx, ny = x, y
    while True:
        nx, ny = get_next_coord(nx, ny, d)
        if nx < 0 or nx >= 4 or ny < 0 or ny >= 4:
            if score > max_score:
                max_score = score
            break
        
        if space[nx][ny] == 0:
            if score > max_score:
                max_score = score
            continue
        
        # 먹는 경우
        new_space = deepcopy(space)
        new_current_state = deepcopy(current_state)
        new_score = score + new_space[nx][ny]
        
        new_current_state[0] = [nx, ny, new_current_state[new_space[nx][ny]][2]]
        new_current_state[new_space[nx][ny]] = []
        
        new_space[x][y] = 0
        new_space[nx][ny] = -1
        
        queue.append((new_space, new_current_state, new_score))
        
print(max_score)   
```

```python
# 다시 풀음, while queue에서 move 안한거 디버깅 오래 걸렸음
from collections import deque
from copy import deepcopy

# ↑, ↖, ←, ↙, ↓, ↘, →, ↗
dx = [-1, -1, 0, 1, 1, 1, 0, -1]
dy = [0, -1, -1, -1, 0, 1, 1, 1]

spaces = [[0] * 4 for _ in range(4)] # 상어: 0, 빈칸: -1
infos = [[] for _ in range(17)] # [x, y, direction], 빈 정보: [-1, -1, -1]
for i in range(4):
    temp = list(map(int, input().split()))
    for j in range(0, 8, 2):
        num, d = temp[j], temp[j + 1] - 1
        infos[num].extend([i, j // 2, d])
        spaces[i][j // 2] = num

def move(spaces, infos):
    for num in range(1, 17):
        # 해당 번호의 물고기가 없는 경우
        if infos[num][0] == -1:
            continue
        x, y, d = infos[num]
        while True:
            nx, ny = x + dx[d % 8], y + dy[d % 8]
            # 갈 수 없는 곳이거나 상어가 있는 곳이 아닌 경우
            if 0 <= nx < 4 and 0 <= ny < 4 and spaces[nx][ny] != 0:  
                # 이동할 칸이 비어있는 경우
                if spaces[nx][ny] == -1:
                    infos[num] = [nx, ny, d]
                # 이동할 칸에 다른 물고기가 있는 경우
                else:
                    other_fish_num = spaces[nx][ny]
                    temp = infos[other_fish_num]
                    infos[other_fish_num] = [x, y, temp[2]]
                    infos[num] = [nx, ny, d]
                
                spaces[nx][ny], spaces[x][y] = spaces[x][y], spaces[nx][ny]
                break
            else:
                d += 1
                
    return spaces, infos
                
# 초기 상어 이동
init_eaten_fish_num = spaces[0][0]
init_score = init_eaten_fish_num
spaces[0][0] = 0
infos[0] = [0, 0, infos[init_eaten_fish_num][2]]
infos[init_eaten_fish_num] = [-1, -1, -1]
# 초기 물고기 이동
spaces, infos = move(spaces, infos)

queue = deque()
queue.append((spaces, infos, init_score)) # (spaces, infos, score)
max_score = 0
while queue:
    spaces, infos, score = queue.popleft()
    max_score = max(max_score, score)
    
    x, y, d = infos[0]
    spaces[x][y] = -1
    while True:
        nx, ny = x + dx[d % 8], y + dy[d % 8]
        # 상어가 공간을 벗어나게 되는 경우 break
        if nx < 0 or nx >= 4 or ny < 0 or ny >= 4:
            break
        # 가고자 하는 칸이 빈 칸이 아닌 경우
        if spaces[nx][ny] != -1:
            nspaces = deepcopy(spaces)
            ninfos = deepcopy(infos)
            eaten_fish_num = nspaces[nx][ny]
            
            nscore = score + eaten_fish_num
            ninfos[0] = [nx, ny, ninfos[eaten_fish_num][2]]
            ninfos[eaten_fish_num] = [-1, -1, -1]
            nspaces[nx][ny] = 0
            
            nspaces, ninfos = move(nspaces, ninfos)
            queue.append((nspaces, ninfos, nscore))
        
        x, y = nx, ny
        
print(max_score)
```

### 어른 상어

```python
# 기존 코드
import sys
input = sys.stdin.readline

# 0, 1, 2, 3는 각각 위, 아래, 왼쪽, 오른쪽

n, m, k = map(int, input().rstrip().split())
board = [list(map(int, input().rstrip().split())) for _ in range(n)]
init_directions = [0] + list(map(lambda x: int(x) - 1, input().rstrip().split()))
direction_priority_info = [None] + [[list(map(lambda x: int(x) - 1, input().rstrip().split())) for _ in range(4)] for _ in range(m)]
### direction_priority_info[상어 번호][현재 방향][방향 우선순위]

shark_info = [[] for _ in range(m + 1)]

for i in range(n):
    for j in range(n):
        if board[i][j] >= 1:
            shark_info[board[i][j]] = [i, j, init_directions[board[i][j]]] # [x, y, direction]

scent = [[[0, 0] for _ in range(n)] for _ in range(n)] # [상어 번호, count]


def spread_scent(count):
    for shark_num, (x, y, d) in enumerate(shark_info[1:], 1):
        if x == -1:
            continue
            
        scent[x][y] = [shark_num, count]

        
def yield_next_coord(x, y):
    yield x - 1, y
    yield x + 1, y
    yield x, y - 1
    yield x, y + 1

    
def get_prior_coord(shark_num, x, y, direction, count):
    no_scent_coords = []
    self_scent_coords = []
    for nd, (nx, ny) in enumerate(yield_next_coord(x, y)):
        if 0 <= nx < n and 0 <= ny < n:
            scent_shark_num, scent_count = scent[nx][ny]
            if scent_count >= 1 and count - scent_count < k:
                if scent_shark_num == shark_num:
                    self_scent_coords.append((nx, ny, nd))
            else:
                no_scent_coords.append((nx, ny, nd))
    
    selected = ()
    highest_ranking = float('inf')
    if no_scent_coords: 
        for nx, ny, nd in no_scent_coords:
            ranking = direction_priority_info[shark_num][direction].index(nd)
                
            if ranking < highest_ranking:
                highest_ranking = ranking
                selected = (nx, ny, nd)
    else:
        for nx, ny, nd in self_scent_coords:
            ranking = direction_priority_info[shark_num][direction].index(nd)
            if ranking < highest_ranking:
                highest_ranking = ranking
                selected = (nx, ny, nd)

    return selected
    
    
def move_shark(board, count, shark_count):
    # 작은 번호의 상어가 이미 있으면 그 상어는 추가 안함
    new_board = [[0] * n for _ in range(n)]
    
    for shark_num, (x, y, direction) in enumerate(shark_info[1:], 1):
        if x == -1:
            continue
        
        nx, ny, nd = get_prior_coord(shark_num, x, y, direction, count)
        
        if new_board[nx][ny] != 0:
            shark_count -= 1
            shark_info[shark_num] = [-1, -1, -1]
            continue
            
        shark_info[shark_num] = [nx, ny, nd]
        new_board[nx][ny] = shark_num
        
    return new_board, shark_count


result = -1
count = 1
shark_count = m
while True:
    spread_scent(count)
    board, shark_count = move_shark(board, count, shark_count)
    
    if shark_count == 1:
        result = count
        break
    
    count += 1
    if count > 1000:
        break
        
print(result)
```
